## 5.1 LTI پاسخ فرکانسی سیستم‌های

پاسخ فرکانسی $ H(e^{j\omega}) $ یک سیستم LTI به عنوان بهره مختلط (مقدار ویژه) که سیستم به ورودی نمایی مختلط $ e^{j\omega t} $ اعمال می‌کند، تعریف شده است (بخش 2.6 را ببینید). 
علاوه بر این، همان‌طور که در بخش 2.9.6 بحث شد، از آنجا که تبدیل فوریه یک دنباله، تجزیه‌ای به صورت ترکیب خطی از نمایی‌های مختلط را نشان می‌دهد، تبدیل‌های فوریه ورودی و خروجی سیستم به‌صورت زیر به هم مرتبط هستند:

$$
Y(e^{j\omega}) = H(e^{j\omega}) X(e^{j\omega}) \quad (5.3)
$$

که در آن $ X(e^{j\omega}) $ و $ Y(e^{j\omega}) $ به ترتیب تبدیل فوریه ورودی و خروجی سیستم هستند.

### 5.1.1 فاز پاسخ فرکانسی و تأخیر گروهی

پاسخ فرکانسی به طور کلی در هر فرکانس یک عدد مختلط است. وقتی پاسخ فرکانسی به فرم قطبی بیان شود، دامنه و فاز تبدیل‌های فوریه ورودی و خروجی سیستم به صورت زیر به هم مرتبط هستند:

$$
|Y(e^{j\omega})| = |H(e^{j\omega})| \cdot |X(e^{j\omega})| \quad (5.4a)
$$

$$
\angle Y(e^{j\omega}) = \angle H(e^{j\omega}) + \angle X(e^{j\omega}) \quad (5.4b)
$$

که در آن $ |H(e^{j\omega})| $ نشان‌دهنده پاسخ دامنه یا بهره سیستم است و $ \angle H(e^{j\omega}) $ نشان‌دهنده پاسخ فاز یا شیفت فاز سیستم است.

اثرات دامنه و فاز بیان‌شده در معادلات بالا می‌توانند مطلوب باشند، اگر سیگنال ورودی به شکل مفیدی اصلاح شود، یا نامطلوب باشند، اگر سیگنال ورودی به صورت مخربی تغییر کند. در حالت دوم، معمولاً به اثرات یک سیستم LTI روی یک سیگنال به عنوان اعوجاج دامنه و اعوجاج فاز اشاره می‌کنیم.

فاز هر عدد مختلط به‌طور یکتا تعریف نشده است، زیرا هر مضرب صحیحی از $ 2\pi $ را می‌توان بدون تأثیرگذاری بر عدد اصلی به آن افزود. وقتی فاز به‌صورت عددی با استفاده از زیرروال تابع تانژانت معکوس محاسبه می‌شود، مقدار اصلی آن معمولاً به‌دست می‌آید. مقدار اصلی فاز $ H(e^{j\omega}) $ را به‌صورت $ \text{ARG}[H(e^{j\omega})] $ نشان می‌دهیم، که در بازه زیر قرار دارد:

$$
-\pi < \text{ARG}[H(e^{j\omega})] \leq \pi \quad (5.5)
$$


هر زاویه‌ی دیگری که مقدار مختلط صحیح تابع $ H(e^{j\omega}) $ را بدهد، می‌تواند به صورت مقدار اصلی زیر نمایش داده شود:

$$
\angle H(e^{j\omega}) = \text{ARG}[H(e^{j\omega})] + 2\pi r(\omega) \quad (5.6)
$$

که در آن $ r(\omega) $ یک عدد صحیح مثبت یا منفی است که می‌تواند در هر مقدار $ \omega $ متفاوت باشد. در بسیاری از موارد، مقدار اصلی دارای ناپیوستگی‌هایی به اندازه $ 2\pi $ رادیان در تابع $ \omega $ است. این موضوع در شکل زیر نشان داده شده است:



````python
import numpy as np
import matplotlib.pyplot as plt

omega = np.linspace(-np.pi, np.pi, 400)
phase_continuous = -2 * np.sin(omega) - 2 * omega
phase_principal = np.mod(phase_continuous + np.pi, 2*np.pi) - np.pi
r_values = np.round((phase_continuous - phase_principal) / (2 * np.pi))

fig, axs = plt.subplots(3, 1, figsize=(6, 8))

axs[0].plot(omega, phase_continuous, label='arg[H(e^{j\omega})]')
axs[0].set_ylabel('arg[H(e^{j\omega})]')
axs[0].set_xticks([-np.pi, 0, np.pi])
axs[0].set_xticklabels(['-π', '0', 'π'])
axs[0].grid()

axs[1].plot(omega, phase_principal, label='ARG[H(e^{j\omega})]', color='r')
axs[1].set_ylabel('ARG[H(e^{j\omega})]')
axs[1].set_xticks([-np.pi, 0, np.pi])
axs[1].set_xticklabels(['-π', '0', 'π'])
axs[1].grid()

axs[2].step(omega, r_values, label='r(\omega)', color='g', where='mid')
axs[2].set_ylabel('r(\omega)')
axs[2].set_xlabel('ω')
axs[2].set_xticks([-np.pi, 0, np.pi])
axs[2].set_xticklabels(['-π', '0', 'π'])
axs[2].grid()

plt.tight_layout()
plt.show()
````

هر زاویه‌ی دیگری که مقدار مختلط صحیح تابع $ H(e^{j\omega}) $ را بدهد، می‌تواند به صورت مقدار اصلی زیر نمایش داده شود:

$$
\angle H(e^{j\omega}) = \text{ARG}[H(e^{j\omega})] + 2\pi r(\omega) \quad (5.6)
$$

که در آن $ r(\omega) $ یک عدد صحیح مثبت یا منفی است که می‌تواند در هر مقدار $ \omega $ متفاوت باشد. 
ما به طور کلی در این متن از نماد زاویه در سمت چپ معادله‌ی بالا برای نمایش فاز نامعین استفاده خواهیم کرد، زیرا مقدار $ r(\omega) $ تا حدودی اختیاری است.

در بسیاری از موارد، مقدار اصلی دارای ناپیوستگی‌هایی به اندازه $ 2\pi $ رادیان در تابع $ \omega $ است. این موضوع در شکل 5.1 نشان داده شده است که یک تابع فاز پیوسته $ \arg[H(e^{j\omega})] $ و مقدار اصلی آن $ \text{ARG}[H(e^{j\omega})] $ را در دامنه‌ای مشخص نمایش می‌دهد.

یک نمایش مفید دیگر از فاز، از طریق تاخیر گروهی $ \tau(\omega) $ تعریف می‌شود به صورت:

$$
\tau(\omega) = \text{grd}[H(e^{j\omega})] = - \frac{d}{d\omega} \text{ARG}[H(e^{j\omega})]. \quad (5.7)
$$

شایان ذکر است که از آنجا که مشتق $ \arg[H(e^{j\omega})] $ و $ \text{ARG}[H(e^{j\omega})] $ فقط در نقاط ناپیوستگی اختلاف دارند، تاخیر گروهی را می‌توان با مشتق‌گیری از مقدار اصلی به دست آورد، به جز در ناپیوستگی‌ها.

به طور مشابه، می‌توان تاخیر گروهی را بر حسب فاز نامعین به شکل زیر بیان کرد:

$$
\text{grd}[H(e^{j\omega})] = - \frac{d}{d\omega} \angle H(e^{j\omega}). \quad (5.8)
$$

که در آن ضربه‌های ناشی از ناپیوستگی‌های به اندازه‌ی $ 2\pi $ در $ \angle H(e^{j\omega}) $ نادیده گرفته می‌شوند.

برای درک اثر فاز و به‌ویژه تأخیر گروهی یک سیستم خطی، ابتدا سیستم تأخیر ایده‌آل را در نظر می‌گیریم. پاسخ ضربه‌ای این سیستم به‌صورت زیر است:

$$
 h_{\text{id}}[n] = \delta[n - n_d]. \quad (5.9)
$$

و پاسخ فرکانسی آن برابر است با:

$$
H_{\text{id}}(e^{j\omega}) = e^{-j\omega n_d}. \quad (5.10)
$$

یا به صورت زیر بیان می‌شود:

$$
|H_{\text{id}}(e^{j\omega})| = 1, \quad (5.11a)
$$

$$
\angle H_{\text{id}}(e^{j\omega}) = -\omega n_d, \quad |\omega| < \pi. \quad (5.11b)
$$

که در آن، تناوبیت $ 2\pi $ در $ \omega $ مشاهده می‌شود. از معادله‌ی بالا نتیجه می‌گیریم که تاخیر زمانی (یا پیش‌افتادن اگر $ n_d < 0 $ باشد) با شیب فاز در ارتباط است.

0.54 - 0.46 \cos \left( \frac{2\pi n}{M} \right), & 0 \leq n \leq M \\
0, & \text{در غیر این صورت}
\end{cases}
\]

که در آن \( M = 60 \) است. این توالی به عنوان پنجره Hamming نیز شناخته می‌شود.

### ترکیب ورودی کامل که در شکل 5.5(a) نمایش داده شده:

\[
x[n] = x_3[n] + x_1[n - M - 1] + x_2[n - 2M - 2]
\]

یعنی، پالس با بیش‌ترین فرکانس اول می‌آید، سپس کم‌ترین، و در آخر پالس فرکانس میانی. فرکانس‌ها به صورت زیر هستند:

\[
\omega_1 = 0.2\pi, \quad \omega_2 = 0.4\pi, \quad \omega_3 = 0.8\pi

مطابق با قضیه پنجره‌سازی تبدیل فوریه، پاسخ DTFT یک سینوس محدودشده حاصل کانولوشن DTFT آن سینوس با DTFT پنجره است.

## صفحه 281

پاسخ فاز سیستم خطی و نامتغیر با زمان H(e^{jω}) اغلب توسط تابع فاز θ(ω) تعریف می‌شود:

$$
H(e^{jω}) = |H(e^{jω})|e^{jθ(ω)} \tag{5.18}
$$

تابع θ(ω) اغلب یک تابع غیرفرد است، حتی اگر |H(e^{jω})| زوج باشد. از آنجا که فاز به صورت زاویه در نظر گرفته می‌شود، تابع θ(ω) فقط تا مضربی از 2π مشخص است.

برای تحلیل سیستم‌ها، اغلب فاز "بازشده" استفاده می‌شود که به کمک عملیات ریاضی پیوستگی نسبی در θ(ω) ایجاد می‌کند.

در شکل 5.3، بخش (a) مقدار اصلی فاز را نشان می‌دهد و بخش (b) فاز بازشده را. همانطور که مشاهده می‌شود، فاز اصلی دارای پرش‌هایی در نقاط مختلف ω است که در فاز بازشده حذف شده‌اند.

اگر تابع فاز θ(ω) خطی باشد، به صورت:

$$
θ(ω) = -ωn_0
$$

در این صورت، پاسخ ضربی تبدیل فوریه برابر است با:

$$
H(e^{jω}) = e^{-jωn_0} \tag{5.19}
$$

که نشان‌دهنده یک تأخیر زمانی n₀ نمونه‌ای است. این نوع سیستم‌ها سیگنال ورودی را بدون اعوجاج تغییر شکل نمی‌دهند بلکه تنها آن را به تأخیر می‌اندازند.


````python

import numpy as np
import matplotlib.pyplot as plt

# Create signal x[n]
M = 60
def w(n, M):
    if 0 <= n <= M:
        return 0.54 - 0.46 * np.cos(2 * np.pi * n / M)
    return 0

# Frequencies
omega1 = 0.2 * np.pi
omega2 = 0.4 * np.pi
omega3 = 0.8 * np.pi
n = np.arange(0, 300)
x1 = np.array([np.cos(omega1 * (i - 150)) * w(i - 150 + 30, M) for i in n])
x2 = np.array([np.cos(omega2 * (i - 90)) * w(i - 90 + 30, M) for i in n])
x3 = np.array([np.cos(omega3 * (i - 30)) * w(i - 30 + 30, M) for i in n])
x = x3 + x1 + x2

# Plot waveform
plt.figure(figsize=(10, 4))
plt.plot(n, x)
plt.title("(a) Waveform of signal x[n]")
plt.xlabel("Sample number (n)")
plt.ylabel("x[n]")
plt.grid(True)
plt.tight_layout()
plt.show()

# DTFT using zero-padding
from scipy.fft import fft, fftshift
X = fftshift(fft(x, 2048))
X_mag = np.abs(X)
w = np.linspace(-np.pi, np.pi, len(X_mag))

# Plot DTFT magnitude
plt.figure(figsize=(10, 4))
plt.plot(w, X_mag)
plt.title("(b) Magnitude of DTFT of x[n]")
plt.xlabel("ω")
plt.ylabel("|X(e^{jω})|")
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftshift

# تنظیمات اولیه
M = 60
n = np.arange(0, 3*M)
w = 0.54 - 0.46 * np.cos(2*np.pi*np.arange(0, M)/M)

# فرکانس‌ها
w1 = 0.2 * np.pi
w2 = 0.4 * np.pi
w3 = 0.8 * np.pi

# تولید سیگنال‌های سینوسی ضرب در پنجره
x1 = np.zeros_like(n)
x2 = np.zeros_like(n)
x3 = np.zeros_like(n)

x1[0:M] = np.cos(w3*np.arange(0, M)) * w
x2[M:2*M] = np.cos(w2*np.arange(0, M)) * w
x3[2*M:3*M] = np.cos(w1*np.arange(0, M)) * w

# سیگنال ترکیبی
x = x3 + x1 + x2

# تعریف پاسخ فرکانسی سیستم H(e^{jω})
omega = np.linspace(-np.pi, np.pi, 1024)
H = np.exp(1j * (4 * np.sin(omega))) * np.exp(-1j * omega**2)

# اعمال سیستم به سیگنال ورودی
X = fftshift(fft(x, 1024))
Y = X * H
y = np.real(np.fft.ifft(np.fft.ifftshift(Y)))

# نمایش شکل موج y[n]
plt.figure(figsize=(10, 4))
plt.plot(n, y[:len(n)])
plt.title("Waveform of output signal y[n]")
plt.xlabel("Sample number (n)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()

# محاسبه و نمایش طیف DTFT خروجی
Y_mag = np.abs(Y)

plt.figure(figsize=(10, 4))
plt.plot(omega, Y_mag)
plt.title("Magnitude of DTFT of y[n]")
plt.xlabel("ω")
plt.ylabel("|Y(e^{jω})|")
plt.grid(True)
plt.tight_layout()
plt.show()
````

همان‌طور که در بخش 3.5 دیدیم، با اعمال تبدیل z به دو طرف معادله (5.19) و استفاده از خاصیت خطی بودن (بخش 3.4.1) و خاصیت انتقال زمانی (بخش 3.4.2)، نتیجه می‌گیریم که برای سیستمی که ورودی و خروجی آن یک معادله تفاضلی از فرم معادله (5.19) را ارضا می‌کند، تابع سیستم دارای فرم جبری زیر خواهد بود:

$$
H(z) = \frac{Y(z)}{X(z)} = \frac{\sum_{k=0}^{M} b_k z^{-k}}{\sum_{k=0}^{N} a_k z^{-k}} \tag{5.20}
$$

در معادله (5.20)، \( H(z) \) به فرم نسبت چندجمله‌ای‌هایی بر حسب \( z^{-1} \) است، زیرا معادله (5.19) از دو ترکیب خطی از ترم‌های تاخیر یافته تشکیل شده است.
اگرچه معادله (5.20) را می‌توان به‌گونه‌ای بازنویسی کرد که چندجمله‌ای‌ها بر حسب توان‌های \( z \) نوشته شوند به‌جای \( z^{-1} \)، اما معمولاً این کار انجام نمی‌شود.
همچنین، اغلب راحت‌تر است که معادله (5.20) به فرم فاکتورگیری شده زیر نوشته شود:

$$
H(z) = \left( \frac{b_0}{a_0} \right) \frac{\prod_{k=1}^{M} (1 - c_k z^{-1})}{\prod_{k=1}^{N} (1 - d_k z^{-1})} \tag{5.21}
$$

هر یک از فاکتورهای \( (1 - c_k z^{-1}) \) در صورت کسر یک صفر در \( z = c_k \) و یک صفر در \( z = 0 \) تولید می‌کند.
به‌طور مشابه، هر یک از فاکتورهای \( (1 - d_k z^{-1}) \) در مخرج کسر یک صفر در \( z = 0 \) و یک قطب در \( z = d_k \) تولید می‌کند.

یک رابطه مستقیم بین تابع سیستم و معادله تفاضلی متناظر آن وجود دارد.
به‌طور خاص، چندجمله‌ای صورت در معادله (5.20) دارای ضرایب و ساختار جبری یکسان با طرف راست معادله (5.19) است
(ترم‌هایی به فرم \( b_k x[n-k] \))، در حالی که چندجمله‌ای مخرج در معادله (5.20) دارای ضرایب و ساختار جبری یکسان با طرف چپ معادله (5.19) است
(ترم‌هایی به فرم \( a_k y[n-k] \)). بنابراین، با داشتن یا تابع سیستم به فرم معادله (5.20) یا معادله تفاضلی به فرم (5.19)، به راحتی می‌توان دیگری را به‌دست آورد.

این مطلب در مثال زیر نشان داده شده است:

## 5.2.1 پایداری و علیت


برای به‌دست آوردن معادله (5.20) از معادله (5.19)، فرض کردیم که سیستم خطی و تغییرناپذیر با زمان است تا بتوان از معادله (5.2) استفاده کرد،
اما هیچ فرضی درباره پایداری یا علیت نکردیم. به‌صورت متناظر، از معادله تفاضلی می‌توان فرم جبری تابع سیستم را به‌دست آورد، اما ناحیه همگرایی مشخص نمی‌شود.

به‌طور خاص، ناحیه همگرایی تابع \( H(z) \) از فرآیند استخراج معادله (5.20) به‌دست نمی‌آید، زیرا تنها چیزی که برای اعتبار معادله (5.20) نیاز است این است که \( X(z) \) و \( Y(z) \) دارای ناحیه‌های همگرایی مشترک باشند.
این موضوع با واقعیت مطرح‌شده در فصل 2 نیز سازگار است، که معادله تفاضلی پاسخ ضربه‌ای یک سیستم خطی و تغییرناپذیر با زمان را به‌طور یکتا مشخص نمی‌کند.

برای تابع سیستم به فرم (5.20) یا (5.21)، انتخاب‌های مختلفی برای ناحیه همگرایی وجود دارد. برای یک نسبت از چندجمله‌ای‌ها، هر انتخاب متفاوت برای ناحیه همگرایی به پاسخ ضربه‌ای متفاوتی منجر می‌شود، اما همگی به معادله تفاضلی یکسانی منجر خواهند شد.

اگر فرض کنیم سیستم علی باشد، نتیجه می‌گیریم که \( h[n] \) باید دنباله‌ای از سمت راست باشد (یعنی برای \( n < 0 \) صفر باشد)،
و بنابراین ناحیه همگرایی تابع \( H(z) \) باید بیرونی‌ترین ناحیه باشد.

از طرف دیگر، اگر فرض کنیم سیستم پایدار باشد، از بحث در بخش 2.4 می‌دانیم که پاسخ ضربه‌ای باید مجموع‌پذیر مطلق باشد، یعنی:

$$
\sum_{n=-\infty}^{\infty} |h[n]| < \infty \tag{5.25}
$$

از آنجا که رابطه (5.25) معادل شرط زیر است:

$$
\sum_{n=-\infty}^{\infty} |h[n]z^{-n}| < \infty \tag{5.26}
$$

برای \( |z| = 1 \)، شرط پایداری معادل این است که ناحیه همگرایی تابع \( H(z) \) شامل دایره واحد باشد.
تعیین ناحیه همگرایی متناظر با تابع سیستم که از معادله تفاضلی به‌دست آمده در مثال زیر نشان داده شده است:

## Example 5.2 - تعیین ناحیه همگرایی


در نظر بگیرید که سیستم خطی و تغییرناپذیر با زمان دارای ورودی و خروجی مرتبط با معادله تفاضلی زیر است:

$$
y[n] - \frac{5}{2}y[n - 1] + y[n - 2] = x[n] \tag{5.27}
$$

از بحث‌های پیشین، فرم جبری تابع سیستم \( H(z) \) به صورت زیر به‌دست می‌آید:

$$
H(z) = \frac{1}{1 - \frac{5}{2}z^{-1} + z^{-2}} = \frac{1}{(1 - \frac{1}{2}z^{-1})(1 - 2z^{-1})} \tag{5.28}
$$

نقاط صفر-قطب متناظر برای \( H(z) \) در شکل 5.7 نشان داده شده‌اند. سه انتخاب ممکن برای ناحیه همگرایی وجود دارد.

|z| > 2

## شکل 5.7 - نمودار قطب و صفر برای مثال 5.2


همان‌طور که مثال 5.2 نشان می‌دهد، علیت و پایداری الزاما با هم سازگار نیستند. برای یک سیستم خطی و تغییرناپذیر با زمان که ورودی و خروجی آن یک معادله تفاضلی از فرم معادله (5.19) را ارضا می‌کنند، برای اینکه هم علی و هم پایدار باشد، ناحیه همگرایی تابع سیستم باید بیرونی‌ترین قطب را شامل شود و دایره واحد را نیز در بر گیرد.

واضح است که این شرایط مستلزم آن است که تمام قطب‌های تابع سیستم درون دایره واحد قرار داشته باشند.


## 5.2.2 سیستم‌های معکوس


برای یک سیستم خطی و تغییرناپذیر با زمان با تابع سیستم \( H(z) \)، سیستم معکوس متناظر سیستمی است با تابع سیستم \( H_i(z) \) به‌طوری‌که اگر آن را با \( H(z) \) به‌صورت سری ترکیب کنیم، تابع سیستم کلی برابر با واحد خواهد شد. یعنی:

$$
G(z) = H(z) H_i(z) = 1 \\ \tag{5.29}
$$

از این رابطه نتیجه می‌گیریم:

$$
H_i(z) = \frac{1}{H(z)} \tag{5.30}
$$

معادل زمانی این رابطه به‌صورت:

$$
g[n] = h[n] * h_i[n] = \delta[n] \tag{5.31}
$$

از معادله (5.30)، پاسخ فرکانسی سیستم معکوس (اگر وجود داشته باشد) به شکل زیر خواهد بود:

$$
H_i(e^{j\omega}) = \frac{1}{H(e^{j\omega})} \tag{5.32}
$$


## مثال ۵.۳: سیستم معکوس برای سیستم مرتبه اول

یعنی \( H_i(e^{j\omega}) \) عکس \( H(e^{j\omega}) \) است. به‌طور معادل، اندازه لگاریتمی، فاز، و تأخیر گروهی سیستم معکوس منفی مقادیر متناظر برای سیستم اصلی هستند. همهٔ سیستم‌ها معکوس ندارند. برای مثال، فیلتر پایین‌گذر ایده‌آل معکوس ندارد. هیچ راهی برای بازیابی مؤلفه‌های فرکانسی که توسط چنین فیلتری صفر شده‌اند، وجود ندارد.

بسیاری از سیستم‌ها دارای معکوس هستند و کلاس سیستم‌هایی با تابع تبدیل گویا، یک مثال بسیار مفید و جالب را فراهم می‌کنند. به طور خاص، در نظر بگیرید:

$$
H(z) = \left( \frac{b_0}{a_0} \right) \frac{ \prod_{k=1}^M \left(1 - c_k z^{-1}\right) }{ \prod_{k=1}^N \left(1 - d_k z^{-1} \right) }\tag{5.33}
$$

که دارای صفرهایی در \( z = c_k \) و قطب‌هایی در \( z = d_k \) است، به‌علاوه صفرها یا قطب‌های ممکن در \( z = 0 \) و \( z = \infty \). در این‌صورت:

$$
H_i(z) = \left( \frac{a_0}{b_0} \right) \frac{ \prod_{k=1}^N \left(1 - d_k z^{-1}\right) }{ \prod_{k=1}^M \left(1 - c_k z^{-1}\right) }\tag{5.34}
$$

یعنی، قطب‌های \( H_i(z) \) همان صفرهای \( H(z) \) هستند و بالعکس. اکنون این پرسش مطرح می‌شود که چه ناحیه‌ای از همگرایی (ROC) باید برای \( H_i(z) \) در نظر گرفت. پاسخ این پرسش توسط قضیهٔ هم‌پیچی (convolution theorem) که در رابطهٔ (5.31) بیان شده، ارائه می‌شود. برای برقرار بودن رابطهٔ (5.31)، ناحیهٔ همگرایی \( H(z) \) و \( H_i(z) \) باید هم‌پوشانی داشته باشند. اگر \( H(z) \) علی باشد، ناحیهٔ همگرایی آن به صورت زیر است:

$$
|z| > \max_k |d_k|\tag{5.35}
$$

بنابراین، هر ناحیهٔ همگرایی برای \( H_i(z) \) که با ناحیهٔ فوق هم‌پوشانی داشته باشد، یک ناحیهٔ همگرایی معتبر برای \( H_i(z) \) است. مثال‌های ۵.۳ و ۵.۴ برخی از این حالات را نشان می‌دهند.

---

### مثال ۵.۳: سیستم معکوس برای سیستم مرتبه اول

فرض کنید:

$$
H(z) = \frac{1 - 0.5z^{-1}}{1 - 0.9z^{-1}}
$$

با ناحیهٔ همگرایی:

$$
|z| > 0.9
$$

در این صورت:

$$
H_i(z) = \frac{1 - 0.9z^{-1}}{1 - 0.5z^{-1}}
$$

از آن‌جا که \( H_i(z) \) فقط یک قطب دارد، تنها دو حالت برای ناحیهٔ همگرایی آن وجود دارد و تنها گزینه‌ای که با \( |z| > 0.9 \) هم‌پوشانی دارد، \( |z| > 0.5 \) است. بنابراین، پاسخ ضربه‌ای سیستم معکوس به صورت زیر است:

$$
h_i[n] = (0.5)^n u[n] - 0.9 (0.5)^{n-1} u[n - 1]
$$

در این حالت، سیستم معکوس هم علی است و هم پایدار.


## مثال ۵.۴ – وارون برای سیستمی با صفر در ناحیه همگرایی

فرض کنید تابع تبدیل به‌صورت زیر باشد:

$$
\text{(۵٫۳۴)} \quad H(z) = \frac{z^{-1} - ۰٫۵}{۱ - ۰٫۹z^{-1}}, \quad |z| > ۰٫۹
$$

تابع وارون آن برابر است با:

$$
\text{(۵٫۳۵)} \quad H_i(z) = \frac{۱ - ۰٫۹z^{-1}}{z^{-1} - ۰٫۵} = \frac{-۲ + ۱٫۸z^{-1}}{۱ - ۲z^{-1}}
$$

همان‌طور که پیش‌تر گفته شد، دو ناحیه‌ی همگرایی ممکن برای این تابع جبری وجود دارد:

$$
|z| < ۲ \quad \text{و} \quad |z| > ۲
$$

اما در این حالت، هر دو ناحیه با ناحیه اصلی یعنی $$|z| > ۰٫۹$$ هم‌پوشانی دارند، پس هر دو وارون معتبر هستند.

پاسخ ضربه‌ای متناظر برای ناحیه‌ی همگرایی $$|z| < ۲$$ برابر است با:

$$
h_{i1}[n] = ۲ (۲)^n u[-n - ۱] - ۱٫۸ (۲)^n u[-n]
$$

و برای ناحیه‌ی همگرایی $$|z| > ۲$$ خواهیم داشت:

$$
h_{i2}[n] = -۲ (۲)^n u[n] + ۱٫۸ (۲)^n u[n - ۱]
$$

می‌بینیم که $$h_{i1}[n]$$ پایدار ولی غیرعلّی است، درحالی‌که $$h_{i2}[n]$$ ناپایدار ولی علّی است. به‌طور نظری، اگر یکی از این دو سیستم وارون با $$H(z)$$ به صورت زنجیره‌ای ترکیب شود، نتیجه سیستم همانی خواهد بود.

### تعمیم:

اگر $$H(z)$$ یک سیستم علّی با صفرهایی در $$c_k$$ برای $$k = ۱, ۲, \ldots, M$$ باشد، آنگاه سیستم وارون آن علّی خواهد بود اگر و تنها اگر ناحیه همگرایی زیر برای آن انتخاب شود:

$$
|z| > \max_k |c_k|
$$

اگر بخواهیم سیستم وارون علاوه بر علّی بودن، پایدار هم باشد، آنگاه ناحیه همگرایی آن باید دایره واحد را شامل شود:

$$
\max_k |c_k| < ۱
$$

یعنی همه‌ی صفرهای تابع تبدیل باید درون دایره واحد قرار داشته باشند.

پس یک سیستم زمان‌ناور خطی، پایدار و علّی، دارای وارون پایدار و علّی خواهد بود اگر و تنها اگر هم قطب‌ها و هم صفرهای آن درون دایره‌ی واحد قرار داشته باشند.

به این نوع سیستم‌ها، **سیستم‌های حداقل فاز** گفته می‌شود و در بخش ۵٫۶ به‌تفصیل بررسی خواهند شد.

## ۵٫۲٫۳ پاسخ ضربه‌ای برای توابع گویای سیستم

بحث بسط کسر جزئی (ارائه‌شده در بخش ۳٫۳٫۲) را می‌توان برای تابع تبدیل $$H(z)$$ به‌کار برد تا یک بیان کلی برای پاسخ ضربه‌ای سیستمی که تابع تبدیل آن به شکل زیر باشد، به‌دست آورد:

$$
\text{(۵٫۳۶)} \quad H(z) = \sum_{r=۰}^{M-N} B_r z^{-r} + \sum_{k=۱}^{N} \frac{A_k}{۱ - d_k z^{-1}}
$$

در اینجا جمله‌های مجموع اول از تقسیم چندجمله‌ای صورت بر مخرج به‌دست می‌آیند و تنها زمانی ظاهر می‌شوند که درجه صورت $$M$$ بزرگ‌تر یا مساوی با درجه مخرج $$N$$ باشد (یعنی $$M \geq N$$).

ضرایب $$A_k$$ و جمله دوم از معادله (۳٫۴۳) به‌دست می‌آیند. اگر تابع تبدیل $$H(z)$$ دارای قطب چندگانه باشد یا...



## ادامه‌ی ۵٫۲٫۳ — پاسخ ضربه‌ای برای توابع گویای سیستم

بسط کسر جزئی تابع تبدیل، مطابق رابطه‌ی زیر خواهد بود:

$$
h[n] = \sum_{r=0}^{M-N} B_r \delta[n - r] + \sum_{k=1}^{N} A_k d_k^n u[n] \qquad \text{(5.37)}
$$

جمع اول تنها زمانی وجود دارد که $$M \geq N$$ باشد.

---

### دسته‌بندی سیستم‌های خطی نامتغیر با زمان (LTI)

برای تحلیل سیستم‌های خطی نامتغیر با زمان، تقسیم آن‌ها به دو دسته مفید است:

### ۱. سیستم‌هایی که پاسخ ضربه‌ای آن‌ها بی‌نهایت است (IIR)

در این دسته، حداقل یک جمله‌ی ناصفر از صورت تابع تبدیل توسط صفر مخرج حذف نشده است. در این حالت، $$h[n]$$ حداقل شامل یک جمله از نوع $$A_k (d_k)^n u[n]$$ خواهد بود. بنابراین، $$h[n]$$ طول محدود ندارد (یعنی خارج از یک بازه‌ی محدود صفر نمی‌شود) و به آن **سیستم با پاسخ ضربه‌ای نامتناهی (IIR)** می‌گویند.

---

### ۲. سیستم‌هایی که فقط صفر دارند و هیچ قطبی ندارند (FIR)

در این دسته، تابع تبدیل تنها یک چندجمله‌ای از $$z^{-1}$$ است و هیچ قطبی ندارد به جز در $$z = 0$$. بنابراین بسط کسر جزئی ممکن نیست و خواهیم داشت:

$$
H(z) = \sum_{k=0}^{M} b_k z^{-k} \qquad \text{(5.38)}
$$

(فرض می‌کنیم $$a_0 = 1$$ بدون از دست رفتن کلیت.)

از رابطه‌ی بالا، پاسخ ضربه‌ای مستقیماً به‌صورت زیر به‌دست می‌آید:

$$
h[n] = \sum_{k=0}^{M} b_k \delta[n - k] =
\begin{cases}
b_n, & 0 \leq n \leq M \\
0, & \text{در غیر این صورت}
\end{cases}
\qquad \text{(5.39)}
$$

در این حالت، پاسخ ضربه‌ای دارای طول محدود است (خارج از بازه‌ای مشخص صفر می‌شود). بنابراین این سیستم‌ها، **سیستم‌های با پاسخ ضربه‌ای متناهی (FIR)** نامیده می‌شوند.

برای این سیستم‌ها، معادله‌ی تفاضلی معادل با جمع هم‌نهشتی (کانولوشن) است:

$$
y[n] = \sum_{k=0}^{M} b_k x[n - k] \qquad \text{(5.40)}
$$

---

## مثال ۵٫۵ — یک سیستم FIR ساده

فرض کنید پاسخ ضربه‌ای یک سیستم، بریده‌شده‌ای از پاسخ ضربه‌ای سیستم IIR زیر باشد:

$$
G(z) = \frac{1}{1 - a z^{-1}}, \quad |z| > |a|
$$

در این صورت، پاسخ ضربه‌ای برابر است با:

$$
h[n] =
\begin{cases}
a^n, & 0 \leq n \leq M \\
0, & \text{در غیر این صورت}
\end{cases}
$$

و تابع تبدیل متناظر برابر خواهد بود با:

$$
H(z) = \sum_{n=0}^{M} a^n z^{-n} = \frac{1 - a^{M+1} z^{-(M+1)}}{1 - a z^{-1}} \qquad \text{(5.41)}
$$


از آنجا که صفرهای صورت در مکان‌های صفحه‌ی z به‌صورت زیر قرار دارند:

$$
z_k = a e^{j \frac{2\pi k}{M+1}}, \qquad k = 0, 1, \dots, M \qquad \text{(5.42)}
$$

که در آن، $$a$$ عددی حقیقی و مثبت فرض شده است. قطب در $$z = a$$ توسط صفر $$z_0$$ حذف شده است.

نمودار صفر-قطب برای حالت $$M = 7$$ در شکل زیر نمایش داده شده است:

---

### شکل ۵٫۸ — نمودار صفر و قطب برای مثال ۵٫۵


````python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
M = 7
a = 0.8  # فرض شده عددی حقیقی و مثبت

# محاسبه صفرها
zeros = a * np.exp(1j * 2 * np.pi * np.arange(0, M+1) / (M+1))

# قطب هفتم‌مرتبه در نقطه a
poles = [a] * (M+1)

# رسم نمودار
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(np.real(zeros), np.imag(zeros), 'o', label='zeros')
ax.plot(np.real(poles), np.imag(poles), 'x', label='poles')

# دایره واحد
unit_circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
ax.add_artist(unit_circle)

# تنظیمات نمودار
ax.axhline(0, color='black', lw=0.5)
ax.axvline(0, color='black', lw=0.5)
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal')
# ax.set_title("شکل ۵٫۸ — نمودار صفر و قطب برای مثال ۵٫۵", fontsize=12)
ax.legend()
plt.grid(True)
plt.show()
````

معادله‌ی تفاضلی بین ورودی و خروجی سیستم خطی نامتغیر با زمان برابر است با کانولوشن گسسته:

$$
y[n] = \sum_{k=0}^{M} a^k x[n - k] \qquad \text{(5.43)}
$$

با این حال، معادله‌ی (۵٫۴۱) همچنین بیان می‌کند که ورودی و خروجی معادله‌ی تفاضلی زیر را نیز ارضا می‌کنند:

$$
y[n] - a y[n - 1] = x[n] - a^{M+1} x[n - M - 1] \qquad \text{(5.44)}
$$

این دو معادله‌ی تفاضلی معادل، از دو فرم معادل تابع تبدیل $$H(z)$$ در رابطه‌ی (۵٫۴۱) نتیجه شده‌اند.

## ۵٫۳ — پاسخ فرکانسی برای توابع گویای سیستم

اگر یک سیستم پایدار خطی نامتغیر با زمان دارای تابع انتقال گویای $$H(z)$$ باشد، یعنی اگر ورودی و خروجی آن معادله‌ی تفاضلی از نوع رابطه‌ی (۵٫۱۹) را ارضا کنند، آنگاه پاسخ فرکانسی آن (یعنی تابع انتقال ارزیابی‌شده روی دایره‌ی واحد) به‌صورت زیر خواهد بود:

$$
H(e^{j\omega}) =
\frac{
\sum_{k=0}^{M} b_k e^{-j\omega k}
}{
\sum_{k=0}^{N} a_k e^{-j\omega k}
}
\qquad \text{(5.45)}
$$


تابع $$H(e^{j\omega})$$ نسبت چندجمله‌ای‌هایی بر حسب $$e^{-j\omega}$$ است.  
برای تعیین اندازه، فاز و تأخیر گروهی مربوط به پاسخ فرکانسی چنین سیستم‌هایی، بیان تابع $$H(e^{j\omega})$$ برحسب صفرها و قطب‌های $$H(z)$$ مفید است.

چنین بیانی از جایگذاری $$z = e^{j\omega}$$ در معادله‌ی (۵٫۲۱) به‌دست می‌آید:

$$
H(e^{j\omega}) = \left( \frac{b_0}{a_0} \right)
\cdot
\frac{
\prod_{k=1}^{M} (1 - c_k e^{-j\omega})
}{
\prod_{k=1}^{N} (1 - d_k e^{-j\omega})
}
\qquad \text{(5.46)}
$$

از رابطه‌ی (۵٫۴۶) نتیجه می‌شود که:

$$
|H(e^{j\omega})| = \left| \frac{b_0}{a_0} \right|
\cdot
\frac{
\prod_{k=1}^{M} |1 - c_k e^{-j\omega}|
}{
\prod_{k=1}^{N} |1 - d_k e^{-j\omega}|
}
\qquad \text{(5.47)}
$$

در نتیجه، تابع اندازه به توان دو برابر است با:

$$
|H(e^{j\omega})|^2 = H(e^{j\omega}) \cdot H^*(e^{j\omega}) =
\left( \frac{b_0}{a_0} \right)^2
\cdot
\frac{
\prod_{k=1}^{M} (1 - c_k e^{-j\omega})(1 - c_k^* e^{j\omega})
}{
\prod_{k=1}^{N} (1 - d_k e^{-j\omega})(1 - d_k^* e^{j\omega})
}
\qquad \text{(5.48)}
$$

از معادله‌ی (۵٫۴۷) نتیجه می‌شود که $$|H(e^{j\omega})|$$ برابر است با حاصل‌ضرب اندازه‌ی تمام عوامل صفر تابع $$H(z)$$ روی دایره واحد، تقسیم بر حاصل‌ضرب اندازه‌ی تمام عوامل قطب روی دایره واحد.

اگر بهره به دسی‌بل بیان شود، داریم:

$$
\text{Gain in dB} = 20 \log_{10} |H(e^{j\omega})| \qquad \text{(5.49)}
$$

بنابراین:

$$
\text{Gain in dB} =
20 \log_{10} \left| \frac{b_0}{a_0} \right|
+ \sum_{k=1}^{M} 20 \log_{10} |1 - c_k e^{-j\omega}|
- \sum_{k=1}^{N} 20 \log_{10} |1 - d_k e^{-j\omega}|
\qquad \text{(5.50)}
$$

پاسخ فازی برای تابع سیستم گویای زیر به‌صورت زیر است:

$$
\arg \left[ H(e^{j\omega}) \right] =
\arg \left[ \frac{b_0}{a_0} \right]
+ \sum_{k=1}^{M} \arg \left[ 1 - c_k e^{-j\omega} \right]
- \sum_{k=1}^{N} \arg \left[ 1 - d_k e^{-j\omega} \right]
\qquad \text{(5.51)}
$$

که در آن $$\arg[\cdot]$$ نمایانگر فاز پیوسته (بدون پیچش) است.

تأخیر گروهی متناظر برای تابع سیستم گویای $$H(e^{j\omega})$$ برابر است با:

$$
\text{grd}[H(e^{j\omega})] =
\sum_{k=1}^{N} \frac{d}{d\omega} \arg \left(1 - d_k e^{-j\omega} \right)
-
\sum_{k=1}^{M} \frac{d}{d\omega} \arg \left(1 - c_k e^{-j\omega} \right)
\qquad \text{(5.52)}
$$


یک عبارت معادل برای تأخیر گروهی به‌صورت زیر است:

$$
\text{grd}[H(e^{j\omega})] =
\sum_{k=1}^{N}
\frac{
|d_k|^2 - \text{Re}(d_k e^{-j\omega})
}{
1 + |d_k|^2 - 2\text{Re}(d_k e^{-j\omega})
}
-
\sum_{k=1}^{M}
\frac{
|c_k|^2 - \text{Re}(c_k e^{-j\omega})
}{
1 + |c_k|^2 - 2\text{Re}(c_k e^{-j\omega})
}
\qquad \text{(5.53)}
$$

در معادله‌ی (۵٫۵۱)، فاز هر جمله ذاتاً مبهم است، یعنی می‌توان هر مقدار مضربی از $$2\pi$$ به آن اضافه کرد بدون اینکه مقدار عدد مختلط تغییر کند. اما تأخیر گروهی با مشتق فاز پیوسته تعریف می‌شود.

معادلات (۵٫۵۰)، (۵٫۵۱) و (۵٫۵۳) به ترتیب اندازه (برحسب دسی‌بل)، فاز و تأخیر گروهی را نشان می‌دهند و هر سه از مجموع سهم‌های قطب‌ها و صفرهای تابع سیستم به‌دست می‌آیند. بنابراین، برای درک بهتر پاسخ فرکانسی سیستم، بررسی تأثیر مکان قطب‌ها و صفرهای سیستم (مخصوصاً پایدار از مرتبه بالاتر) روی پاسخ فرکانسی مفید است.

## ۵٫۳٫۱ — پاسخ فرکانسی سیستم‌های مرتبه اول

در این بخش، ویژگی‌های یک عامل تکی از فرم $$(1 - r e^{j\theta} e^{-j\omega})$$ را بررسی می‌کنیم، که در آن $$r$$ شعاع و $$\theta$$ زاویه قطب یا صفر در صفحه‌ی z است.

این عامل می‌تواند نماینده‌ی یک قطب یا صفر منفرد روی شعاع $$r$$ و زاویه $$\theta$$ باشد.

توان دوم اندازه‌ی چنین عبارتی به‌صورت زیر است:

$$
|1 - r e^{j\theta} e^{-j\omega}|^2 =
(1 - r e^{j(\theta - \omega)})(1 - r e^{-j(\theta - \omega)}) =
1 + r^2 - 2r \cos(\omega - \theta)
\qquad \text{(5.54)}
$$

بهره (gain) برحسب دسی‌بل با این عامل برابر است با:

$$
(\pm) 20 \log_{10} |1 - r e^{j\theta} e^{-j\omega}| =
(\pm) 10 \log_{10} \left[ 1 + r^2 - 2r \cos(\omega - \theta) \right]
\qquad \text{(5.55)}
$$

علامت مثبت در صورتی استفاده می‌شود که عامل یک صفر باشد، و علامت منفی در صورتی که یک قطب باشد.

مشارکت این عامل در فاز به‌صورت زیر است:

$$
(\pm) \arg(1 - r e^{j\theta} e^{-j\omega}) =
(\pm) \arctan \left( \frac{r \sin(\omega - \theta)}{1 - r \cos(\omega - \theta)} \right)
\qquad \text{(5.56)}
$$

با مشتق‌گیری از معادله‌ی بالا (به‌جز نقاط گسستگی)، مشارکت عامل در تأخیر گروهی برابر است با:

$$
(\pm) \left[ -\text{grd}(1 - r e^{j\theta} e^{-j\omega}) \right] =
(\pm) \frac{r^2 - r \cos(\omega - \theta)}{1 + r^2 - 2r \cos(\omega - \theta)} =
(\pm) \frac{r \cos(\omega - \theta)}{|1 - r e^{j\theta} e^{-j\omega}|^2}
\qquad \text{(5.57)}
$$

در اینجا نیز، علامت مثبت برای صفر و منفی برای قطب به‌کار می‌رود.

معادلات (۵٫۵۴) تا (۵٫۵۷) تابع‌هایی متناوب نسبت به $$\omega$$ با دوره‌ی $$2\pi$$ هستند.

**شکل 5.9(a)** نمودار رابطه‌ی (۵٫۵۵) را به‌صورت تابعی از $$\omega$$ برای مقادیر مختلف $$\theta$$ در $$r = 0.9$$ نشان می‌دهد.

**شکل 5.9(b)** نیز فاز معادله‌ی (۵٫۵۶) را به‌عنوان تابعی از $$\omega$$ برای $$r = 0.9$$ و مقادیر مختلفی از $$\theta$$ رسم کرده است. توجه کنید که نقطه‌ی ماکزیمم همیشه در $$\omega = \theta$$ رخ می‌دهد و برای $$r$$ ثابت، این تابع با تغییر $$\theta$$ فقط جابه‌جا می‌شود.

**شکل 5.9(c)** تأخیر گروهی رابطه‌ی (۵٫۵۷) را نشان می‌دهد. توجه کنید که مقدار بالای مثبت این تابع برای $$\omega = \theta$$ به افت ناگهانی تأخیر در همان نقطه منجر می‌شود.

در استخراج ویژگی‌های پاسخ فرکانسی از نمودارهای صفر–قطب (چه گسسته چه پیوسته)، سیستم‌های حداقل–فاز به‌ویژه اهمیت زیادی دارند، و بردارهای فاز و تأخیرشان به‌طور مستقیم قابل تحلیل هستند.


````python

import numpy as np
import matplotlib.pyplot as plt

# تعریف پارامترها
r = 0.9
theta_vals = [0, np.pi/2, np.pi]
omega = np.linspace(0, 2*np.pi, 1000)

# آماده‌سازی شکل
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

for theta in theta_vals:
    # H(e^jw) = 1 - r*e^{j(theta)} * e^{-j*omega} = 1 - r * e^{j(theta - omega)}
    z = 1 - r * np.exp(1j * (theta - omega))
    H = z

    # (a) Log Magnitude (dB)
    axs[0].plot(omega, 20 * np.log10(np.abs(H)), label=r'$\theta = {:.0f}$'.format(theta))
    
    # (b) Phase
    axs[1].plot(omega, np.angle(H), label=r'$\theta = {:.0f}$'.format(theta))
    
    # (c) Group Delay: derivative of unwrapped phase
    phase_unwrapped = np.unwrap(np.angle(H))
    dphi = -np.gradient(phase_unwrapped, omega)  # منفی مشتق فاز
    axs[2].plot(omega, dphi, label=r'$\theta = {:.0f}$'.format(theta))

# تنظیمات subplotها
titles = ['(a) Log Magnitude (dB)', '(b) Phase (radians)', '(c) Group Delay (samples)']
ylabels = ['dB', 'Radians', 'Samples']
for ax, title, ylabel in zip(axs, titles, ylabels):
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.set_xlim(0, 2*np.pi)
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
    ax.grid(True)

axs[2].set_xlabel('Radian frequency (ω)')
axs[0].legend()

plt.tight_layout()
plt.show()
````

### شکل ۵٫۱۰  $$r < 1$$

````python

import numpy as np
import matplotlib.pyplot as plt

# پارامترها
r = 0.7
theta = np.pi / 3
omega = np.pi / 2

# نقاط کلیدی
zero = r * np.exp(1j * theta)
ejw = np.exp(1j * omega)

v1 = ejw
v2 = zero
v3 = v1 - v2

# رسم نمودار
fig, ax = plt.subplots(figsize=(6,6))

# محورهای مختلط
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# دایره واحد
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
ax.add_artist(circle)

# بردارها
ax.arrow(0, 0, v1.real, v1.imag, head_width=0.05, length_includes_head=True, color='black', label=r'$v_1$')
ax.arrow(0, 0, v2.real, v2.imag, head_width=0.05, length_includes_head=True, color='black', linestyle='--', label=r'$v_2$')
ax.arrow(v2.real, v2.imag, v3.real, v3.imag, head_width=0.05, length_includes_head=True, color='black', linestyle='-.', label=r'$v_3 = v_1 - v_2$')

# برچسب‌ها
ax.text(v1.real + 0.05, v1.imag, '$e^{j\\omega}$')
ax.text(v2.real + 0.05, v2.imag, '$r e^{j\\theta}$')
ax.text(0.3, 0.2, r'$\phi_1$', fontsize=12)
ax.text(0.7, 0.7, r'$\phi_3$', fontsize=12)
ax.text(0.3, -0.1, r'$\theta$', fontsize=12)

# تنظیمات شکل
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal')
# ax.set_title("شکل ۵٫۱۰ — بردارهای صفحه‌ی z برای تابع مرتبه اول", fontsize=12)
ax.grid(True)
plt.show()
````

نمایش برداری قطب و صفرها در صفحه مختلط معمولاً بسیار مفید است. در این نمایش، هر قطب یا صفر را می‌توان با برداری در صفحه‌ی z از موقعیت آن تا نقطه‌ای روی دایره واحد نشان داد.

برای یک تابع سیستم مرتبه اول از فرم زیر:

$$
H(z) = (1 - r e^{j\theta} z^{-1}) = \frac{z - r e^{j\theta}}{z}, \qquad r < 1 \qquad \text{(5.58)}
$$

الگوی صفر–قطب در شکل زیر نشان داده شده است:


در این شکل، بردارهای $$v_1$$، $$v_2$$ و $$v_3 = v_1 - v_2$$ نشان داده شده‌اند که به‌ترتیب نمایانگر اعداد مختلط $$e^{j\omega}$$، $$r e^{j\theta}$$ و $$(e^{j\omega} - r e^{j\theta})$$ هستند.

بر حسب این بردارها، اندازه‌ی عدد مختلط:

$$
\frac{e^{j\omega} - r e^{j\theta}}{e^{j\omega}}
$$

برابر است با نسبت اندازه‌ی بردارهای $$v_3$$ به $$v_1$$، یعنی:

$$
|1 - r e^{j\theta} e^{-j\omega}| = \left| \frac{e^{j\omega} - r e^{j\theta}}{e^{j\omega}} \right| = \frac{|v_3|}{|v_1|} \qquad \text{(5.59)}
$$

و چون $$|v_1| = 1$$، معادله (۵٫۵۹) به‌سادگی برابر با $$|v_3|$$ است.

فاز متناظر با این عبارت برابر است با:

$$
\angle(1 - r e^{j\theta} e^{-j\omega}) = \angle(e^{j\omega} - r e^{j\theta}) - \angle(e^{j\omega}) = \angle(v_3) - \angle(v_1)
= \phi_3 - \phi_1 = \phi_3 - \omega \qquad \text{(5.60)}
$$

بنابراین، مشارکت یک عامل تکی از نوع $$(1 - r e^{j\theta} z^{-1})$$ در تابع اندازه در فرکانس $$\omega$$ برابر است با طول بردار از صفر تا نقطه $$z = e^{j\omega}$$ روی دایره واحد. این بردار هنگامی که $$\omega = \theta$$ است حداقل طول را دارد و این همان دلیل افت تیز تابع اندازه در نمودار شکل ۵٫۹(a) است.

بردار $$v_1$$ از قطب در $$z = 0$$ تا $$e^{j\omega}$$ همیشه طولی برابر با ۱ دارد و روی تابع اندازه اثری ندارد.

معادله‌ی (۵٫۶۰) نشان می‌دهد که تابع فاز برابر است با اختلاف زاویه‌ی بردار از صفر در $$r e^{j\theta}$$ تا $$e^{j\omega}$$ و بردار از قطب در $$0$$ تا $$e^{j\omega}$$.

وابستگی کاهش پاسخ فرکانسی به مکان صفر از عامل $$(1 - r e^{j\theta} e^{-j\omega})$$ برحسب شعاع $$r$$، در شکل ۵٫۱۱ برای $$\theta = \pi$$ و مقادیر مختلفی از $$r$$ نشان داده شده است.

دور از $$\omega = \theta$$، شیب تابع فاز منفی است. از آنجا که تأخیر گروهی برابر منفی مشتق تابع فاز است، تأخیر گروهی در نزدیکی $$\omega = \theta$$ منفی است و با نزدیک شدن $$r$$ به ۱ به‌طور ناگهانی افزایش می‌یابد (که منجر به یک ضربه می‌شود – در شکل نشان داده نشده).

## ۵٫۳٫۲ — مثال‌هایی با چند قطب و صفر

در این بخش، با استفاده از مباحث بخش ۵٫۳٫۱، پاسخ فرکانسی سیستم‌هایی با توابع انتقال گویای پیچیده‌تر را بررسی می‌کنیم.

---

### مثال ۵٫۶ — سیستم مرتبه دوم IIR

تابع انتقال مرتبه دوم زیر را در نظر بگیرید:

$$
H(z) = \frac{1}{(1 - r e^{j\theta} z^{-1})(1 - r e^{-j\theta} z^{-1})}
= \frac{1}{1 - 2r \cos\theta z^{-1} + r^2 z^{-2}}
\qquad \text{(5.61)}
$$

معادله‌ی تفاضلی بین ورودی و خروجی این سیستم برابر است با:

$$
y[n] - 2r \cos\theta \, y[n-1] + r^2 y[n-2] = x[n]
$$

با استفاده از تکنیک بسط کسر جزئی، می‌توان نشان داد که پاسخ ضربه‌ای یک سیستم علّی با این تابع انتقال برابر است با:

$$
h[n] = \frac{r^n \sin[(\theta)(n+1)]}{\sin\theta} \, u[n] \qquad \text{(5.62)}
$$

تابع انتقال رابطه‌ی (۵٫۶۱) دارای یک قطب در $$z = r e^{j\theta}$$ و یک مزدوج آن در $$z = r e^{-j\theta}$$ است و دو صفر نیز در $$z = 0$$ دارد.

---

### شکل ۵٫۱۲ — نمودار صفر–قطب برای مثال ۵٫۶



````python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
r = 0.8
theta = np.pi / 3
pole1 = r * np.exp(1j * theta)
pole2 = r * np.exp(-1j * theta)
zero1 = 0
zero2 = 0

# دایره واحد
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')

# بردارها
omega = np.pi / 2
ejw = np.exp(1j * omega)
v1 = ejw
v2 = pole1
v3 = v1 - v2

# رسم نمودار
fig, ax = plt.subplots(figsize=(6,6))
ax.add_artist(circle)
ax.plot(np.real(pole1), np.imag(pole1), 'x', markersize=10, label='Pole 1')
ax.plot(np.real(pole2), np.imag(pole2), 'x', markersize=10, label='Pole 2')
ax.plot(np.real(zero1), np.imag(zero1), 'o', label='Zero (x2)')

# بردارها v1، v2، v3
ax.arrow(0, 0, v1.real, v1.imag, head_width=0.04, color='black')
ax.arrow(0, 0, v2.real, v2.imag, head_width=0.04, color='black', linestyle='--')
ax.arrow(v2.real, v2.imag, v3.real, v3.imag, head_width=0.04, color='black', linestyle=':')

# تنظیمات
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_aspect('equal')
ax.set_title("Figure 5.12 — Pole-zero plot for Example 5.6")
ax.grid(True)
plt.legend()
plt.show()
````

براساس بحث بخش ۵٫۳٫۱، خواهیم داشت:

$$
20 \log_{10} |H(e^{j\omega})| =
-10 \log_{10} \left[ 1 + r^2 - 2r \cos(\omega - \theta) \right]
-10 \log_{10} \left[ 1 + r^2 - 2r \cos(\omega + \theta) \right]
\qquad \text{(5.63a)}
$$

$$
\angle H(e^{j\omega}) =
- \arctan \left( \frac{r \sin(\omega - \theta)}{1 - r \cos(\omega - \theta)} \right)
- \arctan \left( \frac{r \sin(\omega + \theta)}{1 - r \cos(\omega + \theta)} \right)
\qquad \text{(5.63b)}
$$

$$
\text{grd}H(e^{j\omega}) =
\frac{r^2 - r \cos(\omega - \theta)}{1 + r^2 - 2r \cos(\omega - \theta)}
+
\frac{r^2 - r \cos(\omega + \theta)}{1 + r^2 - 2r \cos(\omega + \theta)}
\qquad \text{(5.63c)}
$$

این توابع در شکل ۵٫۱۳ برای $$r = 0.9$$ و $$\theta = \pi / 4$$ رسم شده‌اند.

شکل ۵٫۱۲ بردارهای صفر و قطب $$v_1$$، $$v_2$$ و $$v_3$$ را نشان می‌دهد. تابع اندازه برابر است با حاصل‌ضرب طول بردارهای صفر (که در این مثال برابر یک هستند)، تقسیم بر حاصل‌ضرب طول بردارهای قطب. بنابراین:

$$
|H(e^{j\omega})| = \frac{|v_3|^2}{|v_1| \cdot |v_2|} = \frac{1}{|v_1| \cdot |v_2|} \qquad \text{(5.64)}
$$

وقتی $$\omega \approx \theta$$، طول بردار $$v_1 = e^{j\omega} - r e^{j\theta}$$ کوچک می‌شود و نسبت به تغییرات $$\omega$$ حساس است. اما طول $$v_2 = e^{j\omega} - r e^{-j\theta}$$ تقریباً ثابت باقی می‌ماند.

در نتیجه، قطب با زاویه $$\theta$$ تأثیر عمده‌ای بر پاسخ فرکانسی اطراف $$\omega = \theta$$ دارد (همان‌طور که در شکل ۵٫۱۳ مشخص است). به‌طور متقارن، قطب با زاویه $$-\theta$$ پاسخ را در اطراف $$\omega = -\theta$$ کنترل می‌کند.

---

### شکل ۵٫۱۳ — پاسخ فرکانسی برای زوج مزدوج از قطب‌ها (مطابق مثال ۵٫۶)، برای $$r = 0.9$$ و $$\theta = \pi / 4$$


````python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
r = 0.9
theta = np.pi / 4
omega = np.linspace(0, 2*np.pi, 1000)

# محاسبه |H(e^jω)| بر اساس معادله (5.63a)
mag_squared = (1 + r**2 - 2 * r * np.cos(omega - theta)) * (1 + r**2 - 2 * r * np.cos(omega + theta))
mag_dB = -10 * np.log10(mag_squared)

# رسم نمودار
plt.figure(figsize=(8, 4))
plt.plot(omega, mag_dB, color='black')
plt.title('Figure 5.13 — Frequency response for complex-conjugate pole pair')
plt.ylabel('dB')
plt.xlabel('Radian frequency (ω)')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
plt.grid(True)
plt.ylim(-10, 20)
plt.show()
````

(b) و (c) از شکل ۵٫۱۳ نیز به‌ترتیب فاز و تأخیر گروهی مربوط به زوج مزدوج از قطب‌ها را برای $$r = 0.9$$ و $$\theta = \frac{\pi}{4}$$ نمایش می‌دهند.

---

### شکل ۵٫۱۳ (ادامه) — پاسخ فرکانسی برای زوج مزدوج از قطب‌ها (مطابق مثال ۵٫۶)، برای $$r = 0.9$$ و $$\theta = \frac{\pi}{4}$$  
(b) فاز  
(c) تأخیر گروهی

````python



import numpy as np
import matplotlib.pyplot as plt

# پارامترها
r = 0.9
theta = np.pi / 4
omega = np.linspace(0, 2*np.pi, 1000)

# عبارت‌های فاز و تاخیر گروهی (معادلات 5.63b و 5.63c)
# فاز
arg1 = np.arctan2(r * np.sin(omega - theta), 1 - r * np.cos(omega - theta))
arg2 = np.arctan2(r * np.sin(omega + theta), 1 - r * np.cos(omega + theta))
phase = -arg1 - arg2

# تأخیر گروهی
gd1 = (r**2 - r * np.cos(omega - theta)) / (1 + r**2 - 2 * r * np.cos(omega - theta))
gd2 = (r**2 - r * np.cos(omega + theta)) / (1 + r**2 - 2 * r * np.cos(omega + theta))
group_delay = gd1 + gd2

# رسم شکل
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# (b) Phase
axs[0].plot(omega, phase, color='black')
axs[0].set_title('(b) Phase')
axs[0].set_ylabel('Radians')
axs[0].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[0].set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
axs[0].grid(True)

# (c) Group Delay
axs[1].plot(omega, group_delay, color='black')
axs[1].set_title('(c) Group Delay')
axs[1].set_ylabel('Samples')
axs[1].set_xlabel('Radian frequency (ω)')
axs[1].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[1].set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
axs[1].grid(True)

plt.tight_layout()
plt.show()
````

## Example 5.7 —  سیستم FIR مرتبه دوم

در این مثال، یک سیستم پاسخ ضربه‌ای محدود (FIR) را در نظر می‌گیریم که پاسخ ضربه‌ای آن به‌صورت زیر است:

$$
h[n] = \delta[n] - 2r \cos\theta \, \delta[n-1] + r^2 \delta[n-2] \qquad \text{(5.65)}
$$

تابع تبدیل متناظر این سیستم برابر است با:

$$
H(z) = 1 - 2r \cos\theta \, z^{-1} + r^2 z^{-2} \qquad \text{(5.66)}
$$

که این تابع، معکوس تابع تبدیل ارائه‌شده در مثال ۵٫۶ است.

بنابراین، نمودارهای پاسخ فرکانسی برای این سیستم FIR، دقیقاً برابر با منفیِ نمودارهای شکل ۵٫۱۳ خواهند بود.

همچنین توجه شود که در معکوس تابع تبدیل، مکان قطب‌ها و صفرها با یکدیگر جابه‌جا می‌شوند.


## مثال ۵٫۸ — سیستم IIR مرتبه سوم

در این مثال، یک فیلتر پایین‌گذر (lowpass) را بررسی می‌کنیم که با استفاده از یکی از روش‌های تقریبی طراحی شده (که در فصل ۷ توضیح داده خواهد شد).

تابع سیستم به صورت زیر در نظر گرفته می‌شود:

$$
H(z) = \frac{0.056634(1 + z^{-1})(1 - 1.0166z^{-1} + z^{-2})}{
(1 - 0.683z^{-1})(1 - 1.4461z^{-1} + 0.7957z^{-2})}
\qquad \text{(5.67)}
$$

و فرض شده که این سیستم پایدار است.

مکان صفرهای این تابع سیستم به صورت زیر هستند:

| شعاع (Radius) | زاویه (Angle)              |
|---------------|-----------------------------|
| 1             | $$\pi$$ رادیان              |
| 1             | $$\pm 1.0376$$ رادیان  (59.45°) |

مکان قطب‌های این سیستم به صورت زیر است:

| شعاع (Radius) | زاویه (Angle)               |
|---------------|------------------------------|
| 0.683         | 0                            |
| 0.892         | $$\pm 0.6257$$ رادیان (35.85°) |

نمودار صفر–قطب برای این سیستم در شکل زیر نمایش داده شده است:


### شکل ۵٫۱۴ — نمودار صفر–قطب برای فیلتر پایین‌گذر مثال ۵٫۸



````python

import numpy as np
import matplotlib.pyplot as plt

# تعریف صفرها (بر اساس زاویه و شعاع)
zeros = [
    np.exp(1j * np.pi),  # زاویه π
    np.exp(1j * 1.0376),  # زاویه +1.0376 rad
    np.exp(-1j * 1.0376)  # زاویه -1.0376 rad
]

# تعریف قطب‌ها
poles = [
    0.683 * np.exp(1j * 0),           # قطب روی محور حقیقی
    0.892 * np.exp(1j * 0.6257),      # زاویه +0.6257 rad
    0.892 * np.exp(-1j * 0.6257)      # زاویه -0.6257 rad
]

# رسم نمودار
fig, ax = plt.subplots(figsize=(6, 6))

# دایره واحد
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
ax.add_artist(circle)

# محورهای مختلط
ax.axhline(0, color='black', lw=0.5)
ax.axvline(0, color='black', lw=0.5)

# رسم صفرها
for z in zeros:
    ax.plot(np.real(z), np.imag(z), 'o', markersize=8, label='Zero' if z == zeros[0] else "")

# رسم قطب‌ها
for p in poles:
    ax.plot(np.real(p), np.imag(p), 'x', markersize=8, label='Pole' if p == poles[0] else "")

# تنظیمات
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.grid(True)
ax.set_title("Figure 5.14 — Pole-zero plot for Example 5.8", fontsize=12)
ax.legend()
plt.show()
````

شکل ۵٫۱۵، نمودارهای اندازه لگاریتمی، فاز و تأخیر گروهی این سیستم را نشان می‌دهد. تأثیر صفرهایی که روی دایره واحد در زوایای $$\omega = \pm 1.0376$$ و $$\pi$$ قرار دارند، کاملاً مشهود است.

اما قطب‌ها به‌گونه‌ای قرار گرفته‌اند که به‌جای اینکه باعث افزایش شدید در فرکانس‌های نزدیک زاویه خودشان شوند، پاسخ کلی اندازه را در بازه‌ی $$\omega = 0$$ تا $$\omega = 0.2\pi$$ نزدیک به $$0$$ دسی‌بل نگه می‌دارند (و به‌صورت متقارن در بازه $$\omega = 1.8\pi$$ تا $$2\pi$$ نیز همین اتفاق می‌افتد).

سپس پاسخ به‌طور ناگهانی کاهش یافته و کمتر از $$-25$$ دسی‌بل باقی می‌ماند، در بازه‌ای از حدود $$\omega = 0.3\pi$$ تا $$1.7\pi$$.

همان‌طور که از این مثال مشاهده می‌شود، تقریب‌های سودمند برای پاسخ‌های فیلتر انتخابی فرکانس را می‌توان با استفاده از قطب‌ها برای ساختن پاسخ اندازه و استفاده از صفرها برای حذف آن به‌دست آورد.

### شکل ۵٫۱۵ — Frequency response for the lowpass filter of Example 5.8  
(a) Log magnitude (b) Phase (c) Group delay

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz, group_delay

# ضرایب صورت و مخرج تابع H(z) از معادله (5.67)
b = 0.056634 * np.convolve([1, 1], [1, -1.0166, 1])  # (1 + z^-1)(1 - 1.0166 z^-1 + z^-2)
a = np.convolve([1, -0.683], [1, -1.4461, 0.7957])   # (1 - 0.683 z^-1)(1 - 1.4461 z^-1 + 0.7957 z^-2)

# پاسخ فرکانسی
w, h = freqz(b, a, worN=1024)

# فاز بدون پیچش و تأخیر گروهی
phase = np.unwrap(np.angle(h))
_, gd = group_delay((b, a), w)

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

# (a) Log magnitude
axs[0].plot(w, 20 * np.log10(np.abs(h)), color='black')
axs[0].set_title('(a) Log magnitude')
axs[0].set_ylabel('dB')

# (b) Phase
axs[1].plot(w, phase, color='black')
axs[1].set_title('(b) Phase')
axs[1].set_ylabel('Radians')

# (c) Group delay
axs[2].plot(w, gd, color='black')
axs[2].set_title('(c) Group delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian frequency (ω)')

# تنظیمات مشترک
for ax in axs:
    ax.set_xlim(0, np.pi * 2)
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
    ax.grid(True)

plt.tight_layout()
plt.show()
````

در این مثال، شاهد هر دو نوع گسستگی در نمودار تابع فاز هستیم.

در $$\omega \approx 0.22\pi$$، یک گسستگی به‌اندازه‌ی $$2\pi$$ وجود دارد که ناشی از استفاده از فاز اصلی در رسم است.

در $$\omega = \pm 1.0376$$ و $$\omega = \pi$$، گسستگی‌ها ناشی از صفرهایی هستند که روی دایره واحد قرار گرفته‌اند.

## ۵٫۴ — رابطه بین اندازه و فاز

به‌طور کلی، دانستن اندازه پاسخ فرکانسی یک سیستم خطی نامتغیر با زمان (LTI)، اطلاعاتی درباره‌ی فاز ارائه نمی‌دهد و بالعکس.

اما برای سیستم‌هایی که با معادلات تفاضلی با ضرایب ثابت توصیف می‌شوند (یعنی سیستم‌های دارای تابع انتقال گویا)، نوعی محدودیت بین اندازه و فاز برقرار است.

به‌طور خاص، اگر اندازه‌ی پاسخ فرکانسی و تعداد قطب‌ها و صفرها معلوم باشد، تنها تعداد محدودی انتخاب برای فاز ممکن است.

به‌طور مشابه، اگر تعداد قطب‌ها و صفرها و فاز معلوم باشد، تنها تعداد محدودی انتخاب برای اندازه وجود دارد (تا یک ضریب مقیاس).

علاوه بر این، تحت شرطی به‌نام فاز مینیمم (minimum-phase)، پاسخ اندازه، پاسخ فاز را به‌صورت یکتا تعیین می‌کند و بالعکس.

---

برای بررسی انتخاب‌های ممکن برای تابع سیستم، با فرض معلوم بودن توان دوم اندازه‌ی پاسخ فرکانسی سیستم، رابطه‌ی زیر را در نظر می‌گیریم:

$$
|H(e^{j\omega})|^2 =
H(e^{j\omega}) H^*(e^{j\omega}) =
H(z) H^*(1/z^*) \big|_{z = e^{j\omega}} \qquad \text{(5.68)}
$$

اگر تابع $$H(z)$$ را به فرم گویای زیر در نظر بگیریم (مطابق رابطه‌ی ۵٫۲۱):

$$
H(z) = \left( \frac{b_0}{a_0} \right)
\cdot
\frac{
\prod_{k=1}^{M} (1 - c_k z^{-1})
}{
\prod_{k=1}^{N} (1 - d_k z^{-1})
}
\qquad \text{(5.69)}
$$

آنگاه خواهیم داشت:

$$
H^*\left( \frac{1}{z^*} \right) =
\left( \frac{b_0}{a_0} \right)
\cdot
\frac{
\prod_{k=1}^{M} (1 - c_k^* z)
}{
\prod_{k=1}^{N} (1 - d_k^* z)
}
\qquad \text{(5.70)}
$$

که در آن فرض شده است که $$a_0$$ و $$b_0$$ اعداد حقیقی هستند.

بنابراین، رابطه‌ی (۵٫۶۸) بیان می‌کند که توان دوم اندازه‌ی پاسخ فرکانسی برابر است با مقدار تابع $$H(z) H^*(1/z^*)$$ روی دایره‌ی واحد.


با استفاده از این رابطه:

$$
4(1 + 0.5 z^{-1})(1 + 0.5 z) = (1 + z^{-1})(1 + 2z) \qquad \text{(5.77)}
$$

می‌توان نتیجه گرفت که:

$$
C_1(z) = C_2(z)
$$

نمودار صفر–قطب برای $$C_1(z)$$ و $$C_2(z)$$ (که کاملاً یکسان هستند) در شکل ۵٫۱۶(c) نمایش داده شده است.
### شکل ۵٫۱۶ — نمودار صفر–قطب برای دو تابع سیستم و تابع مشترک مربع اندازه‌ی آن‌ها  
(a) \( H_1(z) \)  (b) \( H_2(z) \)  (c) \( C_1(z), C_2(z) \)

````python

import numpy as np
import matplotlib.pyplot as plt

def plot_pz(ax, zeros, poles, title):
    # دایره واحد
    unit_circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
    ax.add_artist(unit_circle)

    # محورهای مختلط
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

    # صفرها
    for z in zeros:
        ax.plot(np.real(z), np.imag(z), 'o', markersize=8)

    # قطب‌ها
    for p in poles:
        ax.plot(np.real(p), np.imag(p), 'x', markersize=8)

    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.set_title(title)
    ax.grid(True)

# تعریف سیستم‌ها
# (a) H1(z)
zeros_a = [-0.5, -2]
poles_a = [-0.5j, 0.5j]

# (b) H2(z)
zeros_b = [-2, -0.5]
poles_b = [-0.5j, 0.5j]

# (c) C1(z) = C2(z)
zeros_c = [-0.5, -0.5, -2, -2]
poles_c = [-0.5j, 0.5j, -0.5j, 0.5j]

# رسم نمودارها
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
plot_pz(axs[0], zeros_a, poles_a, '(a) $H_1(z)$')
plot_pz(axs[1], zeros_b, poles_b, '(b) $H_2(z)$')
plot_pz(axs[2], zeros_c, poles_c, '(c) $C_1(z)$ and $C_2(z)$')
plt.tight_layout()
plt.show()
````


توابع سیستم $$H_1(z)$$ و $$H_2(z)$$ در مثال ۵٫۹ تنها در مکان یکی از صفرهایشان تفاوت دارند.

در این مثال، عامل $$2(1 + 0.5 z^{-1}) = (z^{-1} + 2)$$ دقیقاً همان سهم را در مربع اندازه پاسخ فرکانسی دارد که عامل $$(1 + 2z^{-1})$$ دارد.

در نتیجه:

$$
|H_1(e^{j\omega})|^2 = |H_2(e^{j\omega})|^2
$$

اما فاز پاسخ این دو تابع با یکدیگر متفاوت است.


## مثال ۵٫۱۰ — تعیین $$H(z)$$ از روی $$C(z)$$

فرض کنید نمودار صفر–قطب تابع $$C(z)$$ (در شکل ۵٫۱۷) داده شده باشد و هدف ما تعیین قطب‌ها و صفرهایی باشد که به $$H(z)$$ مربوط می‌شوند.

زوج مزدوج قطب‌ها و صفرهایی که یکی از اعضای هر جفت به $$H(z)$$ و دیگری به $$H^*(1/z^*)$$ تعلق دارد به‌صورت زیر تعریف می‌شوند:

**زوج قطب ۱**:  $$(p_1, p_4)$$  
**زوج قطب ۲**:  $$(p_2, p_5)$$  
**زوج قطب ۳**:  $$(p_3, p_6)$$  
**زوج صفر ۱**:  $$(z_1, z_4)$$  
**زوج صفر ۲**:  $$(z_2, z_5)$$  
**زوج صفر ۳**:  $$(z_3, z_6)$$

با دانستن اینکه $$H(z)$$ به یک سیستم علّی و پایدار تعلق دارد، باید از هر جفت قطب، آن عضوی را انتخاب کنیم که درون دایره واحد قرار دارد؛ یعنی $$p_1$$، $$p_2$$، و $$p_3$$.

اما برای صفرها چنین محدودیتی وجود ندارد.

اگر فرض کنیم که ضرایب $$a_k$$ و $$b_k$$ در روابط (۵٫۱۹) و (۵٫۲۰) حقیقی هستند، آنگاه صفرها (و قطب‌ها) یا حقیقی‌اند، یا به‌صورت زوج مزدوج مختلط ظاهر می‌شوند. در نتیجه، صفرهایی که باید به $$H(z)$$ نسبت دهیم، به یکی از دو صورت زیر خواهند بود:

$$z_3$$ 
یا
$$z_6$$
و
$$(z_1, z_2)$$
یا
$$(z_4, z_5)$$

بنابراین، چهار انتخاب ممکن برای سیستم پایدار و علّی $$H(z)$$ با سه قطب و سه صفر وجود دارد که پاسخ فرکانسی $$C(z)$$ برای همه‌ی آن‌ها یکسان است (نمودار صفر–قطب $$C(z)$$ در شکل ۵٫۱۷ نشان داده شده است).

اگر شرط حقیقی بودن ضرایب $$a_k$$ و $$b_k$$ را حذف کنیم، تعداد انتخاب‌ها حتی بیشتر خواهد بود.

اگر هیچ محدودیتی روی تعداد قطب‌ها و صفرهای $$H(z)$$ وجود نداشته باشد، تعداد انتخاب‌های ممکن بی‌نهایت خواهد بود.

برای دیدن این موضوع، فرض کنید $$H(z)$$ دارای ضریبی از فرم زیر باشد:

$$
\frac{z^{-1} - a^*}{1 - a z^{-1}}
$$

در این صورت:

$$
H(z) = H_1(z) \cdot \frac{z^{-1} - a^*}{1 - a z^{-1}} \qquad \text{(5.78)}
$$

عبارت بالا یک **عامل تمام‌عبوری (all-pass)** است، چون اندازه آن روی دایره واحد برابر ۱ است.  
(جزئیات بیشتر درباره‌ی این عوامل در بخش ۵٫۵ آمده است.)

به راحتی می‌توان تأیید کرد که:

$$
C(z) = H(z) H^*(1/z^*) = H_1(z) \cdot \frac{z^{-1} - a^*}{1 - a z^{-1}} \cdot H_1^*(1/z^*) \cdot \frac{1 - a^* z}{z^{-1} - a}
\qquad \text{(5.79)}
$$

تمام عوامل all-pass در $$C(z)$$ حذف می‌شوند و در نمودار صفر–قطب آن دیده نمی‌شوند.

در نتیجه، اگر تعداد قطب‌ها و صفرهای $$H(z)$$ مشخص نشده باشد، برای یک $$C(z)$$ داده شده، می‌توان $$H(z)$$ را با هر تعداد دلخواه عامل تمام‌عبوری که قطب آن‌ها داخل دایره واحد است (یعنی $$|a| < 1$$) ترکیب کرد.


### شکل ۵٫۱۷ — نمودار صفر–قطب برای تابع مربع اندازه در مثال ۵٫۱۰


````python


import numpy as np
import matplotlib.pyplot as plt

# نقاط قطب‌ها و صفرها از شکل (مختصاتی تقریبی)
poles = [0.6 + 0.3j, 0.6 - 0.3j, -0.8]        # p1, p2, p3
poles_conj = [np.conj(p) for p in poles]     # p4, p5, p6

zeros = [0.6 + 0.7j, 0.6 - 0.7j, -1.0]        # z1, z2, z3
zeros_conj = [np.conj(z) for z in zeros]     # z4, z5, z6

# ترکیب همه قطب‌ها و صفرها
all_poles = poles + poles_conj
all_zeros = zeros + zeros_conj

# رسم نمودار
fig, ax = plt.subplots(figsize=(6, 6))

# دایره واحد
unit_circle = plt.Circle((0, 0), 1, color='gray', linestyle='--', fill=False)
ax.add_artist(unit_circle)

# محورهای مختلط
ax.axhline(0, color='black', lw=0.5)
ax.axvline(0, color='black', lw=0.5)

# رسم قطب‌ها (×)
for p in all_poles:
    ax.plot(np.real(p), np.imag(p), 'x', markersize=8, color='black')

# رسم صفرها (○)
for z in all_zeros:
    ax.plot(np.real(z), np.imag(z), 'o', markersize=8, color='black', fillstyle='none')

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.set_title("Figure 5.17 — Pole-zero plot for the magnitude-squared function in Example 5.10")
ax.grid(True)
plt.show()
````



## ۵٫۵ — سیستم‌های تمام‌عبور (All-Pass)

همان‌طور که در بحث مثال ۵٫۱۰ نیز مطرح شد، یک تابع سیستم پایدار از فرم:

$$
H_{ap}(z) = \frac{z^{-1} - a^*}{1 - a z^{-1}} \qquad \text{(5.80)}
$$

دارای اندازه پاسخ فرکانسی مستقل از فرکانس $$\omega$$ است.

این موضوع با بازنویسی تابع روی دایره واحد به‌صورت زیر قابل مشاهده است:

$$
H_{ap}(e^{j\omega}) = \frac{e^{-j\omega} - a^*}{1 - a e^{-j\omega}} =
\frac{e^{-j\omega} (1 - a^* e^{j\omega})}{1 - a e^{-j\omega}} \qquad \text{(5.81)}
$$

در معادله (۵٫۸۱)، عبارت $$e^{-j\omega}$$ دارای اندازه یک است و صورت و مخرج نیز مزدوج مختلط یکدیگرند و بنابراین اندازه مساوی دارند.  
در نتیجه:

$$
|H_{ap}(e^{j\omega})| = 1
$$

سیستمی که در آن اندازه پاسخ فرکانسی عدد ثابتی باشد، «سیستم تمام‌عبور» (All-Pass System) نامیده می‌شود، چون تمام مؤلفه‌های فرکانسی ورودی را با بهره‌ی ثابت یا تضعیف ثابتی عبور می‌دهد.³

---

³ در برخی منابع، سیستم تمام‌عبور سیستمی تعریف می‌شود که بهره آن دقیقاً برابر یک باشد.  
در این کتاب، منظور از سیستم تمام‌عبور سیستمی است که بهره آن مستقل از فرکانس باشد، ولی لزوماً یک نباشد.


````python
### شکل ۵٫۱۸ — نمودار صفر–قطب معمولی برای یک سیستم تمام‌عبور


import numpy as np
import matplotlib.pyplot as plt

# قطب‌های سیستم
real_poles = [-1/3, -3/4]  # d1, d2
complex_pole = 0.8 * np.exp(1j * np.pi / 4)  # e1
complex_pole_conj = np.conj(complex_pole)

# صفرهای متناظر با قطب‌ها (مزدوج وارون)
real_zeros = [1/d for d in real_poles]
complex_zero = 1/complex_pole_conj
complex_zero_conj = 1/complex_pole

# رسم نمودار
fig, ax = plt.subplots(figsize=(6, 6))

# دایره واحد
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
ax.add_artist(circle)

# محور مختلط
ax.axhline(0, color='black', lw=0.5)
ax.axvline(0, color='black', lw=0.5)

# قطب‌ها (×)
for p in real_poles + [complex_pole, complex_pole_conj]:
    ax.plot(np.real(p), np.imag(p), 'x', markersize=8, color='black')

# صفرها (○)
for z in real_zeros + [complex_zero, complex_zero_conj]:
    ax.plot(np.real(z), np.imag(z), 'o', markersize=8, color='black', fillstyle='none')

# تنظیمات
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_aspect('equal')
ax.set_title("Figure 5.18 — Typical pole-zero plot for an all-pass system")
ax.grid(True)
plt.show()
````

## ادامه‌ی بخش ۵٫۵ — سیستم‌های تمام‌عبور

هر تابع سیستم تمام‌عبور کلی را می‌توان به‌صورت ضربی از عوامل مرتبه اول و دوم به شکل زیر نوشت:

$$
H_{ap}(z) =
A \cdot \prod_{k=1}^{M_r} \frac{z^{-1} - d_k}{1 - d_k z^{-1}}
\cdot
\prod_{k=1}^{M_c} \frac{(z^{-1} - e_k^*)(z^{-1} - e_k)}{(1 - e_k z^{-1})(1 - e_k^* z^{-1})}
\qquad \text{(5.82)}
$$

که در آن:

- \( A \) یک عدد حقیقی مثبت است.
- \( d_k \) قطب‌های حقیقی هستند.
- \( e_k \) قطب‌های مختلط هستند (به‌همراه مزدوج آن‌ها).

برای سیستم‌های علی و پایدار تمام‌عبور، باید داشته باشیم:

$$
|d_k| < 1 \quad \text{و} \quad |e_k| < 1
$$

در شکل ۵٫۱۸، نمودار صفر–قطب یک سیستم تمام‌عبور را نشان می‌دهد که دارای:

- \( M_r = 2 \): دو قطب حقیقی  
- \( M_c = 1 \): یک زوج قطب مزدوج مختلط

توجه کنید که به‌ازای هر قطب، صفر متناظر مزدوج وارون  وجود دارد.


پاسخ فرکانسی یک سیستم تمام‌عبور کلی را می‌توان به‌صورت ترکیبی از پاسخ‌های فرکانسی سیستم‌های تمام‌عبور مرتبه اول (مشابه معادله 5.80) بیان کرد.

برای یک سیستم تمام‌عبور علّی، هرکدام از این عوامل شامل یک قطب درون دایره واحد و یک صفر در مکان مزدوج وارون همان قطب هستند.

همان‌طور که پیش‌تر نشان داده شد، اندازهٔ پاسخ فرکانسی چنین سیستمی برابر ۱ است.  
در نتیجه، اندازه لگاریتمی در واحد دسی‌بل برابر با صفر خواهد بود.

اگر پارامتر \( a \) را در فرم قطبی به‌صورت \( a = r e^{j\theta} \) بنویسیم، آنگاه فاز پاسخ برای معادلهٔ (5.80) به‌صورت زیر است:

$$
\angle \left[
\frac{e^{-j\omega} - r e^{-j\theta}}{1 - r e^{j\theta} e^{-j\omega}}
\right]
= -\omega - 2 \arctan \left[
\frac{r \sin(\omega - \theta)}{1 - r \cos(\omega - \theta)}
\right]
\qquad \text{(5.83)}
$$

به‌طور مشابه، فاز یک سیستم تمام‌عبور مرتبه دوم که قطب‌هایی در \( z = r e^{j\theta} \) و \( z = r e^{-j\theta} \) دارد، به صورت زیر خواهد بود:

$$
\angle \left[
\frac{(e^{-j\omega} - r e^{-j\theta})(e^{-j\omega} - r e^{j\theta})}
{(1 - r e^{j\theta} e^{-j\omega})(1 - r e^{-j\theta} e^{-j\omega})}
\right]
= -2\omega - 2 \arctan \left[
\frac{r \sin(\omega - \theta)}{1 - r \cos(\omega - \theta)}
\right]
- 2 \arctan \left[
\frac{r \sin(\omega + \theta)}{1 - r \cos(\omega + \theta)}
\right]
\qquad \text{(5.84)}


## مثال ۵٫۱۱ — سیستم‌های تمام‌عبور مرتبه اول و دوم

شکل ۵٫۱۹ نمودارهای مربوط به مقدار لگاریتمی، فاز، و تأخیر گروهی برای دو سیستم تمام‌عبور مرتبه اول را نشان می‌دهد.

یکی از این سیستم‌ها دارای قطبی در \( z = 0.9 \) (یعنی \( \theta = 0 \)، \( r = 0.9 \))  
و دیگری دارای قطبی در \( z = -0.9 \) است.


````python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات
omega = np.linspace(0, 2*np.pi, 1000)
r1 = 0.9
r2 = -0.9

# تعریف تابع سیستم تمام‌عبور مرتبه اول H(z) = (z^-1 - a*) / (1 - a z^-1)
def H_ap(omega, a):
    ejw = np.exp(1j * omega)
    num = np.conj(a) - ejw
    den = 1 - a * ejw
    return num / den

# محاسبه پاسخ‌ها
H1 = H_ap(omega, r1)
H2 = H_ap(omega, r2)

# Log magnitude (در dB)
mag1 = 20 * np.log10(np.abs(H1))
mag2 = 20 * np.log10(np.abs(H2))

# Phase (unwrap یا principal)
phase1 = np.angle(H1)
phase2 = np.angle(H2)

# Group delay
def group_delay(H):
    dphi = np.unwrap(np.angle(H))
    gd = -np.gradient(dphi, omega)
    return gd

gd1 = group_delay(H1)
gd2 = group_delay(H2)

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

# (a) log magnitude
axs[0].plot(omega, mag1, label='z = 0.9', color='black')
axs[0].plot(omega, mag2, '--', label='z = -0.9', color='black')
axs[0].set_title('(a) Log magnitude')
axs[0].set_ylabel('dB')
axs[0].legend()

# (b) phase
axs[1].plot(omega, phase1, label='z = 0.9', color='black')
axs[1].plot(omega, phase2, '--', label='z = -0.9', color='black')
axs[1].set_title('(b) Phase')
axs[1].set_ylabel('Radians')
axs[1].legend()

# (c) group delay
axs[2].plot(omega, gd1, label='z = 0.9', color='black')
axs[2].plot(omega, gd2, '--', label='z = -0.9', color='black')
axs[2].set_title('(c) Group delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian frequency (ω)')
axs[2].legend()

# تنظیمات عمومی
for ax in axs:
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
    ax.grid(True)

plt.tight_layout()
plt.show()
````

## شکل ۵٫۲۰ — پاسخ فرکانسی سیستم تمام‌عبور مرتبه دوم

در هر دو سیستم، شعاع قطب‌ها برابر با \( r = 0.9 \) است.  
شکل ۵٫۲۰ تابع‌های لگاریتم اندازه، فاز، و تأخیر گروهی را برای یک سیستم تمام‌عبور مرتبه دوم با قطب‌های:

\[
    $$
z = 0.9 e^{j\pi/4} \quad \text{و} \quad z = 0.9 e^{-j\pi/4}
    $$
\]

نشان می‌دهد.



````python
import numpy as np
import matplotlib.pyplot as plt

# تنظیم فرکانس و قطب‌ها
omega = np.linspace(0, 2*np.pi, 1000)
r = 0.9
theta = np.pi / 4

# قطب‌های مختلط مزدوج
p1 = r * np.exp(1j * theta)
p2 = np.conj(p1)

# تابع سیستم تمام‌عبور مرتبه دوم
def H_ap_2nd_order(omega, p1, p2):
    ejw = np.exp(1j * omega)
    num = (np.conj(p1) - ejw) * (np.conj(p2) - ejw)
    den = (1 - p1 * ejw) * (1 - p2 * ejw)
    return num / den

# محاسبه پاسخ‌ها
H = H_ap_2nd_order(omega, p1, p2)
mag = 20 * np.log10(np.abs(H))
phase = np.angle(H)

# تأخیر گروهی
def group_delay(H, omega):
    dphi = np.unwrap(np.angle(H))
    gd = -np.gradient(dphi, omega)
    return gd

gd = group_delay(H, omega)

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

# (a) log magnitude
axs[0].plot(omega, mag, color='black')
axs[0].set_title('(a) Log magnitude')
axs[0].set_ylabel('dB')

# (b) phase
axs[1].plot(omega, phase, color='black')
axs[1].set_title('(b) Phase (principal value)')
axs[1].set_ylabel('Radians')

# (c) group delay
axs[2].plot(omega, gd, color='black')
axs[2].set_title('(c) Group delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian frequency (ω)')

# تنظیم محور افقی
for ax in axs:
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
    ax.grid(True)

plt.tight_layout()
plt.show()
````

## ادامه‌ی مثال ۵٫۱۱ — خاصیت کلی سیستم‌های تمام‌عبور علّی

مثال ۵٫۱۱ ویژگی عمومی سیستم‌های تمام‌عبور علّی را نشان می‌دهد.  
در شکل ۵٫۱۹(b)، مشاهده می‌کنیم که فاز سیستم در بازه \( 0 < omega < \pi \) **غیرفزاینده** است.

به‌طور مشابه، در شکل ۵٫۲۰(b)، اگر پرش \( 2\pi \) ناشی از فاز اصلی  حذف شود، منحنی فاز پیوسته نیز در بازه \( 0 < \omega < \pi \) نزولی خواهد بود.

از آنجا که تابع سیستم تمام‌عبور کلی در رابطه (5.82) تنها حاصل‌ضرب عواملی از نوع مرتبه اول و دوم است، نتیجه می‌گیریم که فاز بدون‌پیچش تابع سیستم (یعنی \( \arg[H_{ap}(e^{j\omega})] \)) برای سیستم تمام‌عبور علّی همیشه در بازه \( 0 < \omega < \pi \) **منفی یا صفر** است.

ممکن است برای مقدار اصلی فاز این موضوع صادق نباشد (مثلاً در شکل ۵٫۲۱ دیده می‌شود)، ولی برای فاز پیوسته قطعاً صادق است.

### تأیید رابطه با تأخیر گروهی

تأخیر گروهی عامل ساده‌ی مرتبه اول تمام‌عبور از معادله (5.80) برابر است با مشتق منفی فاز آن از رابطه (5.83).  
با کمی جبر، داریم:

$$
\text{grd} \left[
\frac{e^{-j\omega} - r e^{-j\theta}}{1 - r e^{j\theta} e^{-j\omega}}
\right] =
\frac{1 - r^2}{1 + r^2 - 2r \cos(\omega - \theta)} =
\frac{1 - r^2}{|1 - r e^{j\theta} e^{-j\omega}|^2}
\qquad \text{(5.85)}
$$

چون برای سیستم علّی و پایدار تمام‌عبور داریم \( r < 1 \)، پس از رابطه (5.85) نتیجه می‌گیریم که **تأخیر گروهی مثبت است.**

چون تأخیر گروهی حاصل‌جمع این عوامل است، نتیجه می‌گیریم که تأخیر گروهی برای سیستم تمام‌عبور مرتبه بالا نیز **مثبت خواهد بود**.

این در شکل‌های ۵٫۱۹(c)، ۵٫۲۰(c)، و ۵٫۲۱(c) برای سیستم‌های تمام‌عبور مرتبه ۱، ۲ و ۳ نشان داده شده است.

---

### نتیجه‌گیری: فاز نزولی و تأخیر گروهی مثبت

مثبت بودن تأخیر گروهی سیستم تمام‌عبور علّی به‌طور مستقیم منجر به این نتیجه می‌شود که:

$$
\arg[H_{ap}(e^{j\omega})] = - \int_0^\omega \text{grd}[H_{ap}(e^{j\phi})] d\phi + \arg[H_{ap}(e^{j0})]
\qquad \text{(5.86)}
$$

برای بازه \( 0 \leq \omega \leq \pi \).

از رابطه (5.82) نیز داریم:

$$
H_{ap}(e^{j0}) =
A \cdot \prod_{k=1}^{M_r} \frac{1 - d_k}{1 - d_k} \cdot
\prod_{k=1}^{M_c} \frac{|1 - e_k|^2}{|1 - e_k|^2} = A
\qquad \text{(5.87)}
$$

در نتیجه:

$$
\arg[H_{ap}(e^{j0})] = 0
$$

و چون طبق رابطهٔ (5.85) داریم:

$$
\text{grd}[H_{ap}(e^{j\omega})] \geq 0
\qquad \text{(5.88)}
$$

از رابطه (5.86) نتیجه می‌شود که:

$$
\arg[H_{ap}(e^{j\omega})] \leq 0 \qquad \text{for } 0 \leq \omega \leq \pi
\qquad \text{(5.89)}
$$

- طراحی فیلترهای فاز مینیمم (فصل ۵٫۶)
- تبدیل فیلترهای پایین‌گذر به فیلترهای دلخواه با پهنای قابل تنظیم



````python
import numpy as np
import matplotlib.pyplot as plt

# برد فرکانسی
omega = np.linspace(0, 2*np.pi, 1000)
ejw = np.exp(1j * omega)

# قطب‌ها طبق شکل 5.18 (تقریبی)
real_poles = [-1/3, -3/4]
complex_pole = 0.8 * np.exp(1j * np.pi / 4)
complex_pole_conj = np.conj(complex_pole)

# صفرهای متناظر (conjugate reciprocal)
real_zeros = [1/d for d in real_poles]
complex_zero = 1 / complex_pole_conj
complex_zero_conj = 1 / complex_pole

# تابع سیستم H(z)
def H_allpass(ejw):
    # حقیقی‌ها
    Hr = np.ones_like(ejw, dtype=complex)
    for p, z in zip(real_poles, real_zeros):
        Hr *= (z - ejw) / (1 - p * ejw)
    # مختلط مزدوج
    Hc = ((complex_zero - ejw)*(complex_zero_conj - ejw)) / ((1 - complex_pole * ejw)*(1 - complex_pole_conj * ejw))
    return Hr * Hc

# محاسبه پاسخ
H = H_allpass(ejw)
mag = 20 * np.log10(np.abs(H))
phase = np.angle(H)
group_delay = -np.gradient(np.unwrap(phase), omega)

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

# (a) Log magnitude
axs[0].plot(omega, mag, color='black')
axs[0].set_title('(a) Log magnitude')
axs[0].set_ylabel('dB')

# (b) Phase
axs[1].plot(omega, phase, color='black')
axs[1].set_title('(b) Phase (principal value)')
axs[1].set_ylabel('Radians')

# (c) Group delay
axs[2].plot(omega, group_delay, color='black')
axs[2].set_title('(c) Group delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian frequency (ω)')

# تنظیم محور x
for ax in axs:
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
    ax.grid(True)

plt.tight_layout()
plt.show()
````
