import

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft
```

signal

```python
fs = 500
t = np.linspace(0, 2, 2 * fs, endpoint=False)
signal = np.sin(2 * np.pi * 10 * t) + np.sin(2 * np.pi * 50 * t) + 0.5 * np.random.normal(size=t.shape)
```

stft

```python
f, time, Zxx = stft(signal, fs=fs, nperseg=128)
```

```python
plt.figure(figsize=(12, 6))
plt.pcolormesh(time, f, np.abs(Zxx), shading='gouraud')
plt.title('STFT')
plt.ylabel('frequency(Hz)')
plt.xlabel('time(s)')
plt.colorbar(label='power')
plt.grid()
plt.show()
```
