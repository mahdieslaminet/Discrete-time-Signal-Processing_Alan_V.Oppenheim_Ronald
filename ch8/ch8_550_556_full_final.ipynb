### ۲-۶-۸ تداخل بین نمادها در سیستم‌های PAM

در سیستم‌های مدولاسیون دامنه‌ی پالس TDM توصیف شده در بالا، برای جدا کردن کانال‌ها می‌توان از شکل موج مالتی‌پلکس شده در زمان‌های مناسب نمونه‌برداری کرد. سیگنال مالتی‌پلکس زمانی شکل ۷-۸ را در نظر بگیرید که از PAM سه سیگنال X1(t) ,X2(t) ,X3(t)  تشکیل شده است.

اگر در زمان‌های مناسب، مثلاً وسط هر پالس، از \( y(t) \) نمونه بگیریم، می‌توانیم نمونه‌های سه سیگنال را از هم جدا کنیم.

کنیم. یعنی:

$$
\begin{aligned}
y(t) &= A x_1(t), && t = 0, \pm 3T_1, \pm 6T_1, \dots \\
y(t) &= A x_2(t), && t = T_1, \pm 2T_1, \pm 5T_1, \dots \\
y(t) &= A x_3(t), && t = 2T_1, \pm 4T_1, \pm 7T_1, \dots
\end{aligned}
$$
(27-8)



که در آن \( T_1 \) فاصله بین نمادها، و در اینجا برابر است با:

$$
T_1 = \frac{T}{3}
$$

و \( A \) یک ضریب ثابت است. به بیان دیگر، با نمونه‌برداری مناسب از سیگنال PAM مالتی‌پلکس زمانی دریافت‌شده، می‌توان نمونه‌های X1(t),X2(t),X3(t) را به دست آورد.

راه‌برد بیان‌شده در بالا با این فرض کارایی دارد که پالس‌های ارسالی در حین انتقال از کانال مخابراتی تغییر نیابند. ولی در انتقال از هر محیط واقعی، پالس‌ها به خاطر اثرهای فیلترینگ و نویز، دچار اعوجاج می‌شوند. نویز اضافه‌شده در کانال نیز در دامنه‌ی نمادها وارد می‌شود. اگر فیلتر کردن تنها برای اعمال کردن پاسخ فرکانسی کانال باشد ولی باعث درهم‌ریختن پالس‌ها شود، به نحوی که پالس‌ها در هم بیفتند، آنگاه "تداخل بین نمادها (ISI)" رخ می‌دهد.

درهم‌ریزی زمانی پالس‌های ایده‌آل شکل بالا می‌تواند از محدودیت پهنای باند یا پخش فاز ناشی از تأخیر گروهی (مانند آنچه در بخش ۲-۶-۴ مورد بحث قرار گرفت) ناشی شود.

اگر تداخل بین نمادها تنها از محدودیت پهنای باند کانال ناشی شده باشد، بهترین سعی در این شرایط آن است که پاسخ فرکانسی کانال از اعوجاج خالی باشد. در واقع اگر پاسخ فرکانسی کانال به صورت زیر باشد:

$$
H(j\omega) =
\begin{cases}
1, & 0 < \omega < w \\
0, & |\omega| > w
\end{cases}
$$

و تابع چگالی طیفی پالس به‌صورت \( |P(j\omega)| = 0 \) برای \( |\omega| > w \) باشد، آنگاه استفاده از سیگنال PAM بدون اعوجاج ممکن خواهد بود. ولی در این صورت دیگر پالس‌ها نامحدود می‌شوند (چون پالس‌های شکل ۲-۷۷ باند محدود نبودند). ما می‌خواهیم شکلی از پالس را داشته باشیم که عبور صفر مضرب‌های صحیح در زمان‌های بعدی رخ دهد، طوری که معادله (۲-۷۸) همچنان برقرار باشد.

همچنین اگر بخواهیم تداخل بین نمادها پیش نیاید، باید از تابع نمونه‌ی پالس زیر استفاده کنیم:

$$
p(t) = \frac{T_1 \sin\left( \frac{\pi t}{T_1} \right)}{\pi t}
$$

### ❓ سوال تستی – صفحه ۵۵۲
**سؤال:**  
افزایش طول پنجره چه اثری بر پاسخ فرکانسی دارد؟  
الف) پهنای lobe اصلی را افزایش می‌دهد  
ب) دامنه فرکانس را کم می‌کند  
ج) lobe اصلی را باریک‌تر می‌کند  
د) دامنه نویز را زیاد می‌کند  


```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف محور زمان
t = np.linspace(-6, 10, 1000)

# تابع پالس با دامنه متغیر
def pulse(t, t0, A):
    return A * np.exp(-((t - t0) ** 2) / 2)

# زمان و دامنه پالس‌ها برای سه کانال
pulses = [
    {"t0": 0, "A": 0.9, "label": "Channel 1"},
    {"t0": 3, "A": 0.6, "label": "Channel 2"},  # کوتاه‌ترین
    {"t0": 6, "A": 1.1, "label": "Channel 3"}   # بلندترین
]

# رسم نمودار
plt.figure(figsize=(10, 4))
for p in pulses:
    y = pulse(t, p["t0"], p["A"])
    plt.plot(t, y, label=p["label"], linewidth=2)
    plt.axvline(p["t0"], linestyle='--', color='gray')

# توضیحات نمونه‌برداری و تداخل
plt.text(0, 1.05, "Sampling time\nfor channel 1", ha='center')
plt.text(3, 0.75, "Sampling time\nfor channel 2", ha='center')
plt.text(6, 1.25, "Sampling time\nfor channel 3", ha='center')
plt.text(3, 0.5, "← Intersymbol\ninterference", ha='left')

# تنظیمات نمودار
plt.title("Figure 8-28: Intersymbol Interference with Varying Amplitudes")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.ylim(-0.2, 1.4)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

الف) کاهش آن‌ها به کاهش leakage کمک می‌کند  
ب) هیچ تأثیری بر تحلیل فرکانس ندارند  
ج) همیشه بزرگ‌تر از lobe اصلی هستند  
د) باعث حذف سیگنال می‌شوند  


<div dir="rtl" style="text-align: right; font-family: sans-serif; line-height: 1.8">

که به همراه طیفش در شکل ۸-۲۹ نشان داده شده است، چون این پالس در مضارب صحیح فاصله بین نمادها صفر است، چنانچه در شکل ۸-۲۸ می‌توان دید، در این زمان‌ها تداخل بین نمادها وجود ندارد. یعنی اگر از سیگنال دریافتی در $t = kT_1$ نمونه‌برداری کنیم، اثر پالس‌های دیگر در آن نمونه ظاهر نمی‌شود زیرا:

$$
p(t - mT_1) = 0 \quad \text{برای} \quad m \ne 0
$$

یعنی برابر صفر است. البته برای پرهیز از تداخل بین نمادها، باید زمان‌های نمونه‌برداری با دقت فراوانی تعیین شوند تا نمونه‌برداری در عبور از صفرهای دیگر پالس‌های دیگر انجام شود.

پالس **sinc** تنها یکی از پالس‌های محدود در باند متعددی است که در $ \pm T_1, \pm 2T_1, \dots $ عبور از صفر دارد. به طور کلی، پالس $p(t)$ را با طیف زیر در نظر بگیرید:

$$
P(j\omega) =
\begin{cases}
1 + P_1(j\omega), & |\omega| \leq \frac{\pi}{T_1} \\
P_1(j\omega), & \frac{\pi}{T_1} < |\omega| \leq \frac{\gamma \pi}{T_1} \\
0, & \text{غیر این صورت}
\end{cases}
$$

حول $\omega = \frac{\pi}{T_1}$، تابع $P_1(j\omega)$ باید تقارن فرد داشته باشد، به نحوی که:

$$
P_1\left(-j\omega + j\frac{\pi}{T_1} \right) = - P_1\left(j\omega + j\frac{\pi}{T_1} \right), \quad 0 \leq \omega \leq \frac{\pi}{T_1}
$$

(رابطه ۸-۲۹)

شکل ۸-۳۱ چنین طیفی را نشان می‌دهد. به ازای $P_1(j\omega) = 0$، تابع $P(t)$ همان تابع سینک است.

در مسئله ۸-۴۲ نشان می‌دهیم که به ازای هر $P_1(j\omega)$ که شرایط (۸-۲۸) و (۸-۲۹) را در $t = \pm T_1, \pm 2T_1, \dots$ صفر کند، یک پالس متفاوت می‌توان داشت.

گرچه سیگنال‌هایی که معادلات (۸-۲۸) و (۸-۲۹) را ارضا کنند، امکان غلبه بر مشکل محدودیت پهنای باند کانال را فراهم می‌کنند، ولی اعوجاج‌های دیگری نیز ممکن است بروز کنند که گزینش پالس با شکل متفاوتی را بطلبد.

</div>


<div dir="rtl" style="text-align: right; font-family: sans-serif; line-height: 1.8">

گرچه سیگنال‌هایی که معادلات (۸-۲۸) و (۸-۲۹) را ارضا می‌کنند، امکان غلبه بر مشکل محدودیت پهنای باند کانال را فراهم می‌کنند، ولی اعوجاج‌های دیگری نیز ممکن است بروز کند که گزینش پالس با شکل متفاوتی را بطلبد.

شکل ۸-۲۹، پالس سینک را همراه با طیف آن نشان می‌دهد. تابع زمان آن به‌صورت زیر تعریف می‌شود:

$$
p(t) = \frac{\sin\left(\frac{\pi t}{T_1}\right)}{\frac{\pi t}{T_1}} = \text{sinc}\left(\frac{t}{T_1}\right)
$$

این تابع در زمان‌های $t = \pm T_1, \pm 2T_1, \pm 3T_1, \dots$ برابر صفر است و فقط در $t = 0$ مقدار آن برابر ۱ است. این خاصیت باعث می‌شود که در سیستم‌های نمونه‌برداری، تداخلی بین نمادها وجود نداشته باشد.

طیف این تابع در حوزه فرکانس به‌صورت تابع مستطیلی است:

$$
P(j\omega) =
\begin{cases}
1, & |\omega| \leq \frac{\pi}{T_1} \\
0, & \text{otherwise}
\end{cases}
$$

این یعنی پالس سینک دارای پهنای باند محدود برابر با $\frac{2\pi}{T_1}$ است و یک پاسخ ایده‌آل برای عبور باند پایه می‌سازد.

### ❓ سوال تستی – صفحه ۵۵۴
**سؤال:**  
استفاده از پنجره در DFT چه کاربردی دارد؟

```python
import numpy as np
import matplotlib.pyplot as plt

# Time and sinc pulse
T1 = 1
t = np.linspace(-5 * T1, 5 * T1, 1000)
p_t = np.sinc(t / T1)

# Frequency and spectrum
omega = np.linspace(-4 * np.pi / T1, 4 * np.pi / T1, 1000)
P_omega = np.where(np.abs(omega) <= np.pi / T1, 1, 0)

# Plot both time and frequency domain
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Time-domain pulse p(t)
axs[0].plot(t, p_t, color='black')
axs[0].set_title("Figure 8-29: A sinc pulse", loc='center')
axs[0].set_xlabel("t")
axs[0].set_ylabel("p(t)")
axs[0].grid(True)
axs[0].axvline(0, color='gray', linestyle='--')
axs[0].axhline(0, color='black', linewidth=0.5)
axs[0].annotate('1', xy=(0, 1), xytext=(0.2, 1.1))
axs[0].annotate('$-T_1$', xy=(-T1, -0.1), xytext=(-T1, -0.2))
axs[0].annotate('$2T_1$', xy=(2 * T1, 0.1), xytext=(2 * T1, 0.2))
axs[0].annotate('$3T_1$', xy=(3 * T1, 0.05), xytext=(3 * T1, 0.15))

# Frequency-domain P(jw)
axs[1].plot(omega, P_omega, color='black')
axs[1].set_title("Spectrum of sinc pulse: $P(j\\omega)$", loc='center')
axs[1].set_xlabel("$\\omega$")
axs[1].set_ylabel("$P(j\\omega)$")
axs[1].grid(True)
axs[1].axvline(0, color='gray', linestyle='--')
axs[1].axhline(0, color='black', linewidth=0.5)
axs[1].annotate(r'$-\pi/T_1$', xy=(-np.pi / T1, 0.5), xytext=(-np.pi / T1, 0.6))
axs[1].annotate(r'$\pi/T_1$', xy=(np.pi / T1, 0.5), xytext=(np.pi / T1, 0.6))
axs[1].annotate(r'$T_1$', xy=(0, 1), xytext=(0.2, 1.1))

plt.tight_layout()
plt.show()
```

📊 **توضیح نمودار:** این کد یک نمودار برای نمایش ویژگی‌های طیف یا پنجره سیگنال رسم می‌کند. توجه کنید که با استفاده از `plt.plot` یا `plt.stem` معمولاً نمایی از دامنه یا شکل پنجره ارائه می‌شود.

```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف محور زمان
T1 = 1
t = np.linspace(-6 * T1, 6 * T1, 2000)

# تابع سینک با دامنه متغیر
def sinc_shifted(t, shift, amplitude=1):
    return amplitude * np.sinc((t - shift) / T1)

# موقعیت‌ها و دامنه‌های متفاوت
pulses = [
    {"shift": -2*T1, "amplitude": 1.1, "label": "Channel 1"},  # بلندترین
    {"shift": 0,      "amplitude": 0.6, "label": "Channel 2"},  # کوتاه‌ترین
    {"shift": 2*T1,   "amplitude": 0.9, "label": "Channel 3"},
]

# رسم نمودار
plt.figure(figsize=(10, 5))
for p in pulses:
    plt.plot(t, sinc_shifted(t, p["shift"], p["amplitude"]),
             label=f'Pulse for {p["label"]}', color='black')

# علامت‌گذاری زمان نمونه‌برداری
for p in pulses:
    plt.axvline(p["shift"], color='gray', linestyle='--')
    plt.text(p["shift"], -0.4, f'Sampling time\nfor {p["label"]}', ha='center')

# تنظیمات نمودار
plt.title("Figure 8-30: Avoiding Intersymbol Interference using Sinc Pulses (Variable Amplitudes)")
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.ylim(-0.6, 1.4)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters (Adjust based on image details)
T1 = 1  # Replace with actual T₁ from the image if specified
omega = np.linspace(-np.pi/T1, np.pi/T1, 1000)  # Range: -π/T₁ to π/T₁

# Define P₁(jω) (Example: First-order low-pass filter)
# Replace this with the actual formula from the image if different
P1 = 1j * omega / (1 + 1j * omega)

# Calculate P(jω) = 1 / (1 + P₁(jω))
P = 1 / (1 + P1)

# Find critical points
max_magnitude = np.max(np.abs(P))  # Maximum of |P(jω)|
omega_max = omega[np.argmax(np.abs(P))]  # Frequency at max magnitude

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(omega, np.abs(P), label='$|P(j\omega)|$', color='blue', linewidth=2)

# Highlight key features
plt.axvline(x=np.pi/T1, color='red', linestyle='--', label='$\pm \pi/T_1$')
plt.axvline(x=-np.pi/T1, color='red', linestyle='--')
plt.axhline(y=1, color='gray', linestyle=':', label='Reference (1.0)')
plt.scatter(omega_max, max_magnitude, color='darkorange', zorder=5,
            label=f'Max: {max_magnitude:.2f} at $\omega = {omega_max:.2f}$')

# Annotate axes with ω/T₁
plt.xticks(
    ticks=[-np.pi/T1, -np.pi/(2*T1), 0, np.pi/(2*T1), np.pi/T1],
    labels=['$-\pi/T_1$', '$-\pi/2T_1$', '0', '$\pi/2T_1$', '$\pi/T_1$']
)

# Add labels and title
plt.title('Frequency Response: $P(j\omega) = \\frac{1}{1 + P_1(j\omega)}$', fontsize=14)
plt.xlabel('Frequency ($\omega$) [rad/s]', fontsize=12)
plt.ylabel('Magnitude $|P(j\omega)|$', fontsize=12)
plt.grid(True, which='both', linestyle='--', alpha=0.7)
plt.legend(loc='upper right', frameon=True)
plt.xlim(-np.pi/T1, np.pi/T1)
plt.ylim(0, 1.2 * max_magnitude)
plt.tight_layout()
plt.show()
```

**۷-۸ مدولاسیون فرکانس سینوسی**

در بخش‌های پیش چند سیستم مدولاسیون دامنه را بررسی کردیم که در آنها سیگنال مدوله‌کننده، برای تغییر دامنه حامل سیگنال بر یا پایین به کار می‌رفت. دیدیم که می‌توان با استفاده از تغییر دامنه یا حوزه فرکانسی، پردازش کرد. در فصل‌های قبل این سیستم را به دقت تحلیل کرد. در یک دسته بزرگ‌تر مدولاسیون، نوع دیگر موسوم به مدولاسیون فرکانس (FM) سیگنال مدوله‌کننده برای کنترل فرکانس حامل سینوسی به کار می‌رود.

این نوع مدولاسیون نسبت به مدولاسیون دامنه مزایایی دارد. چنانچه شکل ۱۰-۸ نشان می‌دهد در مدولاسیون فرکانس، سیگنال مدوله‌کننده نسبت مستقیم دارد، و اندازه آن سیگنال می‌تواند اثر خود را بر تغییر فرکانس در طی زمان حامل همیشه باقی بماند. بنابراین سیستم FM تقریباً همیشه با فیلتر پایین‌تر نیز کار کند. این به این معناست که برای تغییرات دامنه ناشی از اغتشاشات یا موجود در کانال از خطا جلوگیری می‌شود. سیستم FM اغلب در برابر نویز حذف‌کردن قوی‌تر است. به این خاطر در بخش عمومی و کاربردهای متعدد دیگری گفتیم که در AM ضعیف است، در حالی که FM بهتر از آن است. چون AM نویز را به طور مستقیم به دامنه اضافه می‌کند، معمولاً از نواحی بالا‌تر از برای سیستم‌های مخابراتی استفاده می‌شود.

در مدولاسیون فرکانس، فرکانس معمولاً از نواحی پایین تا بالا تغییر می‌کند. همان‌طور که در تحلیل سیستم‌های مدولاسیون نشان دادیم، با تغییر فرکانس، گسترش طیفی در سیگنال رخ می‌دهد و در نتیجه سیگنال تحلیل فرکانس‌اش امکان‌پذیر نمی‌شود. به طبیعت و طرز کار این مدولاسیون توجه خوبی پیدا کرده‌ایم.

از مفهوم عمومی مدولاسیون زاویه شروع می‌کنیم. یک حامل سینوسی بیان‌شده به‌صورت زیر در نظر بگیرید:

$$
c(t) = A \cos(\omega_c t + \theta_c) = A \cos(\theta(t)) \tag{8-30}
$$

که در آن $ \theta(t) = \omega_c t + \theta_c $ فاز حامل است.  
مفهوم کلی مدولاسیون زاویه به‌کار بردن یک سیگنال مدوله‌کننده برای تغییر $ \theta(t) $ است.  
یک شکل انجام این‌کار که گاهی به‌کار می‌رود استفاده از سیگنال مدوله‌کننده $ x(t) $ برای تغییر کل $ \theta(t) $ به‌نحوی که سیگنال مدوله‌شده به‌صورت زیر درآید:

$$
y(t) = A \cos(\omega_c t + \theta_0 + k_p x(t)) \tag{8-31}
$$

در این حالت تابعی از زمان $ \theta(t) $ به شکل زیر است:

$$
\theta(t) = \theta_0 + k_p x(t) \tag{8-32}
$$

مثلاً اگر $ x(t) $ ثابت باشد، فاز ثابت و با دامنه $ k_p x(t) $ تغییر یافته است.  
مدولاسیون زاویه معادله 8-31 را مدولاسیون فاز می‌نامند.  
شکل دیگر مدولاسیون زاویه، تغییر خطی مشتق زاویه با سیگنال مدوله‌کننده است:

$$
y(t) = A \cos(\theta(t)) \tag{8-33}
$$

که در آن:

$$
\frac{d\theta(t)}{dt} = \omega_c + k_f x(t) \tag{8-34}
$$

به‌ازای $ x(t) $ ثابت، $ x(t) $ یک سیگنال سینوسی است که فرکانس آن با اندازه‌ای متناسب با دامنه $ \omega_c $ حامل تفاوت دارد.  
به همین خاطر این نوع مدولاسیون زاویه را معمولاً مدولاسیون فرکانسی می‌نامند.

گرچه مدولاسیون فاز و مدولاسیون فرکانس صوری متفاوتی از مدولاسیون زاویه دارند، ولی رابطه نزدیکی با هم دارند.  
با توجه به معادلات (8-31) و (8-33)، برای مدولاسیون فاز:

$$
\frac{d\theta(t)}{dt} = \omega_c + k_p \frac{dx(t)}{dt} \tag{8-35}
$$

بنابراین مقایسه معادله‌های (8-34) و (8-35) نشان می‌دهد که مدولاسیون فاز با $ x(t) $ شبیه مدولاسیون فرکانس با $ \frac{dx(t)}{dt} $ است،  
به‌نحوی که مدولاسیون فرکانسی با $ x(t) $ شبیه مدولاسیون زاویه با انتگرال $ x(t) $ است.

شکل سیگنال‌های 8-28 الف و ب نمونه‌هایی از مدولاسیون فاز و فرکانس را نشان می‌دهند.  
در هر دو شکل سیگنال مدوله‌کننده $ x(t) = t $ است (یعنی یک سیگنال شیب‌دار که در ۰ جا گذاشت ولی به‌طور خطی زیاد می‌شود).  
در شکل 8-28-ج مدولاسیون فرکانس با سیگنال مدوله‌کننده $ x(t) = \cos(2\pi f_m t) $ نمایش داده شده است.

این‌چنین اگر $ x(t) = r(t) $، ناظر در شکل‌های 8-28 (الف) و (ج) کاملاً واضح است.

در مدولاسیون فرکانس با سیگنال با تابعی که فرکانس آن تغییر می‌کند،  
نتیجه این است که اگر مقدار دیگری از فرکانس در لحظه $ t $ انتخاب شود،  
آن‌گاه مقدار لحظه‌ای از سیگنال مدوله‌کننده فرکانس لحظه‌ای سیگنال مدوله‌شده را تعیین می‌کند.

به همین دلیل، شکل 8-28 (ب) فرکانس را به‌طور خطی با زمان تغییر می‌دهد.  
فرکانس متغیر با زمان – یا به عبارتی مفهوم **فرکانس لحظه‌ای** – بهترین نمایش عددی فرکانس لحظه‌ای برای سیگنال سینوسی است:

$$
y(t) = A \cos(\theta(t)) \tag{8-36}
$$

</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# زمان و پارامترها
t = np.linspace(0, 5, 1000)
A = 1
fc = 3  # فرکانس پایه حامل
k = 20  # ضریب مدولاسیون برای وضوح بهتر

# شکل (الف): PM با x(t) = t
x_a = t
theta_a = 2 * np.pi * fc * t + k * x_a
y_a = A * np.cos(theta_a)

# شکل (ب): FM با x(t) = t
x_b = t
theta_b = 2 * np.pi * fc * t + k * np.cumsum(x_b) * (t[1] - t[0])
y_b = A * np.cos(theta_b)

# شکل (ج): FM با x(t) پله‌ای (مشتق شیب)
x_c = np.where(t >= 2, 1, 0)
theta_c = 2 * np.pi * fc * t + k * np.cumsum(x_c) * (t[1] - t[0])
y_c = A * np.cos(theta_c)

# رسم نمودارها
fig, axs = plt.subplots(3, 2, figsize=(12, 9))

# رنگ‌ها
colors = ['tab:blue', 'tab:green', 'tab:orange']

# (الف)
axs[0, 0].plot(t, x_a, color=colors[0])
axs[0, 0].set_title("x(t) - PM with ramp input", fontsize=10)
axs[0, 1].plot(t, y_a, color=colors[0])
axs[0, 1].set_title("y(t) - PM", fontsize=10)

# (ب)
axs[1, 0].plot(t, x_b, color=colors[1])
axs[1, 0].set_title("x(t) - FM with ramp input", fontsize=10)
axs[1, 1].plot(t, y_b, color=colors[1])
axs[1, 1].set_title("y(t) - FM (ramp)", fontsize=10)

# (ج)
axs[2, 0].plot(t, x_c, color=colors[2])
axs[2, 0].set_title("x(t) - FM with step input", fontsize=10)
axs[2, 1].plot(t, y_c, color=colors[2])
axs[2, 1].set_title("y(t) - FM (step)", fontsize=10)

# تنظیم نهایی
for ax in axs.flat:
    ax.set_xlabel("t")
    ax.grid(True)

plt.tight_layout()
plt.show()
```

**شکل ۸-۳۲** مدولاسیون فاز، مدولاسیون فرکانس و رابطه آن‌ها:  
(الف) مدولاسیون فاز با سیگنال مدوله‌کننده شبه $u(t)$  
(ب) مدولاسیون فرکانس با سیگنال مدوله‌کننده شبه $u(t)$  
(ج) مدولاسیون فرکانس با سیگنال مدوله‌کننده به گونه‌ای که $\theta(t)$ متغیر پیوسته است.

به صورت زیر تعریف می‌شود:

$$
\omega_i(t) = \frac{d\theta(t)}{dt} \tag{۸-۲۷}
$$

بنابراین اگر $\theta(t)$ واقعاً سینوسی باشد (یعنی $\theta(t) = (\omega_c t + \theta_0)$)، $\omega_i$ فرکانس لحظه‌ای است.  
برای مدولاسیون فاز بیان شده در معادله‌های (۸-31) و (۸-32)، فرکانس لحظه‌ای از $\omega_c$ و برای مدولاسیون فرکانس بیان شده در معادله‌های (۸-34) و (۸-32)، فرکانس لحظه‌ای از $\omega_c + k_f x(t)$ است.

چون مدولاسیون فرکانس و مدولاسیون فاز، به اساسی به هم مربوط می‌شوند، نتیجه مباحث را تنها بر حسب مدولاسیون فرکانس بیان می‌کنیم؛ یعنی به‌جای بررسی تأثیر سیگنال مدوله‌کننده $x(t)$ بر طیف سیگنال فاز مدوله، حالت را با سیگنال‌های مدوله‌کننده ساده در نظر می‌گیریم تا آن‌قدر ساده که بتوان اصول اساسی مدولاسیون فرکانس را بیان کنیم.
