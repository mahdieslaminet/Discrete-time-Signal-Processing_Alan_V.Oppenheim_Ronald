```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter
```

signal + noise

```python
np.random.seed(42)
fs = 500
t = np.linspace(0,1,fs,endpoint=False)
signal = np.sin(2 * np.pi * 10 * t) + 0.5 * np.random.normal(size=t.shape)
```

Butterworth filter

```python
def butter_lowpass_filter(data, cutoff, fs, order):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    filtered_data = lfilter(b, a, data)
    return filtered_data
```

```python
cutoff_freq = 15
order = 4
filtered_signal = butter_lowpass_filter(signal, cutoff_freq, fs, order)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, signal, label="signal with noise", alpha=0.5)
plt.plot(t, filtered_signal, label="filter", linewidth=2)
plt.xlabel("time(s)")
plt.ylabel("amplitude")
plt.legend()
plt.title("Butterworth")
plt.grid()
plt.show()
```
