```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# تابع طراحی فیلتر پایین‌گذر
def lowpass_filter(data, cutoff, fs, order=5):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = lfilter(b, a, data)
    return y

# پارامترها
fs = 1000  # فرکانس نمونه‌برداری
t = np.linspace(0, 1, fs, endpoint=False)  # زمان
frequency = 50  # فرکانس سیگنال اصلی
noise = np.random.normal(0, 0.5, t.shape)  # نویز
signal = np.sin(2 * np.pi * frequency * t) + noise  # سیگنال اصلی با نویز
cutoff = 100  # فرکانس برش فیلتر پایین‌گذر

# اعمال فیلتر
filtered_signal = lowpass_filter(signal, cutoff, fs)

# رسم نمودار
plt.figure(figsize=(12, 6))
plt.plot(t, signal, label="Original Signal with Noise", alpha=0.75)
plt.plot(t, filtered_signal, label="Filtered Signal", linewidth=2)
plt.xlabel("Time [seconds]")
plt.ylabel("Amplitude")
plt.title("Lowpass Filter Example")
plt.legend()
plt.grid()
plt.show()
```
