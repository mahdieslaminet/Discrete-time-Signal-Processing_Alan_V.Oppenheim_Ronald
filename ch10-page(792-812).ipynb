```python
import numpy as np
import matplotlib.pyplot as plt

Omega = np.linspace(-10, 10, 1000)

Omega_0 = 3

S_x = 0.5 * np.exp(-((Omega - Omega_0) ** 2)) + 0.5 * np.exp(-((Omega + Omega_0) ** 2)) + 0.2 / (1 + Omega**2)

flat_region = (np.abs(Omega) < Omega_0 * 0.8)
S_x[flat_region] = np.max(S_x[flat_region]) * 0.9

plt.figure(figsize=(6, 3))
plt.plot(Omega, S_x, 'k', linewidth=2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xticks([-Omega_0, 0, Omega_0], [r'$-\Omega_0$', '0', r'$\Omega_0$'])  # Fixed xticks
plt.yticks([])
plt.xlabel(r'$\Omega$ (Frequency)')  # Fixed xlabel
plt.ylabel(r'$S_x(\Omega)$')  # Fixed ylabel
plt.title("Frequency Spectrum with Flatter Region")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define frequency range
Omega = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Define cutoff frequency
Omega_c = np.pi / 2  # Equivalent to π/2

# Define the frequency response
H_aa = np.zeros_like(Omega)

# Flat region (where H_aa = 1)
H_aa[np.abs(Omega) <= Omega_c] = 1

# Smooth transition on edges
transition_width = np.pi / 4  # Controls smoothness of transition
smooth_transition = (np.abs(Omega) > Omega_c) & (np.abs(Omega) < Omega_c + transition_width)

# Apply a cosine-based roll-off for smooth edges
H_aa[smooth_transition] = 0.5 * (1 + np.cos(np.pi * (np.abs(Omega[smooth_transition]) - Omega_c) / transition_width))

# Plot the frequency response
plt.figure(figsize=(6, 3))
plt.plot(Omega, H_aa, 'k', linewidth=2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Fixed xticks (using LaTeX for π and Ω)
plt.xticks([-Omega_c, 0, Omega_c], [r'$-\pi/2$', r'$0$', r'$\pi/2$'])

# Set y-ticks
plt.yticks([0, 1])

# Labels with LaTeX formatting
plt.xlabel(r'$\Omega$ (Frequency)')
plt.ylabel(r'$H_{aa}(\Omega)$')
plt.title("Smooth Band-Limited Frequency Response")

plt.grid(True, linestyle='--', alpha=0.5)  # Optional: Add grid for better readability
plt.show()
```

10.2 تجزیه و تحلیل DFT سیگنال های سینوسی
10.2 تجزیه و تحلیل DFT سیگنال های سینوسی DTFT یک سیگنال سینوسی $A cos(w_{0}n +\phi )$(موجود برای همه n) یک جفت تکانه در $+w_{0}$ و $-w_{0}$ است (تکرار دوره ای با دوره $2\pi$). در تجزیه و تحلیل سیگنال‌های سینوسی با استفاده از DFT، نمونه‌برداری پنجره‌ای و طیفی (حوزه فرکانس) تأثیرات مهمی دارد. همانطور که در بخش 10.2.1 خواهیم دید، پنجره سازی تکانه های نمایش فوریه را لکه دار یا گسترده می کند، بنابراین، فرکانس دقیق کمتر تعریف شده است. پنجره سازی همچنین توانایی تشخیص سیگنال های سینوسی که از نظر فرکانس نزدیک به هم هستند را کاهش می دهد. نمونه برداری طیفی ذاتی در DFT به طور بالقوه یک تصویر اشتباه یا نادرست از طیف واقعی سیگنال سینوسی ارائه می دهد. این اثر در بخش 10.2.3 مورد بحث قرار گرفته است.

10.2.3 اثر نمونه برداری طیفی
همانطور که قبلا ذکر شد، DFT دنباله پنجره‌دار u[n] نمونه‌هایی از آن را ارائه می‌دهد
$V(e^{jw})$در N فرکانس های زمان گسسته با فاصله مساوی (k = 2xk/N، k=0،1، ...،
N - 1.  اینها معادل فرکانسهای زمان پیوسته  $\Omega_{k} = (2\pi k)/(NT)$هستند.
برای k = 0,1 .... , N/2 (با فرض اینکه N زوج باشد). شاخص‌های k = N/2+1، ...، N -1
با فرکانس های منفی زمان پیوسته -2 (N - k)/(NT) مطابقت دارد. طیفی
نمونه برداری، همانطور که توسط DFT تحمیل شده است، گاهی اوقات می تواند نتایج گمراه کننده ای ایجاد کند. این اثر
به بهترین شکل با مثال نشان داده شده است.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import iv  # تابع بسل

# پارامترها
L = 10  # مقدار L
alpha = 5.09  # مقدار alpha
pi_alpha = 16  # مقدار pi * alpha
x = np.linspace(-L/2, L/2, 1000)  # مقادیر x در بازه [-L/2, L/2]

# تعریف تابع w0(x)
w0_x = np.zeros_like(x)
for i, xi in enumerate(x):
    if abs(xi) <= L/2:
        w0_x[i] = (1/L) * (iv(0, pi_alpha * np.sqrt(1 - (2 * xi / L)**2))) / iv(0, pi_alpha)
    else:
        w0_x[i] = 0

# رسم نمودار
plt.figure(figsize=(10, 6))
plt.plot(x, w0_x, label=r'$w_0(x)$', color='b')
plt.title(r'Plot of $w_0(x)$ with $\pi \alpha = 16$ and $\alpha = 5.09$')
plt.xlabel(r'$x$')
plt.ylabel(r'$w_0(x)$')
plt.grid(True)
plt.legend()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# پارامترهای پنجره کایزر
N = 64  # طول پنجره
beta = 8  # پارامتر شکل‌دهی (beta)

# ایجاد پنجره کایزر
window = np.kaiser(N, beta)

# محاسبه تبدیل فوریه
fft_result = np.fft.fft(window, n=2048)  # n تعداد نقاط برای FFT
fft_magnitude = np.abs(fft_result)  # دامنه تبدیل فوریه
fft_magnitude_db = 20 * np.log10(fft_magnitude)  # تبدیل به دسی‌بل

# فرکانس‌ها برای محور افقی
frequencies = np.fft.fftfreq(len(fft_result))

# رسم نمودار
plt.plot(frequencies[:len(frequencies)//2], fft_magnitude_db[:len(fft_result)//2])
plt.title('Fourier Transform of Kaiser Window')
plt.xlabel('Frequency (Normalized)')
plt.ylabel('Magnitude (dB)')
plt.grid(True)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# پارامترها
N = 64  # تعداد نمونه‌ها
n = np.arange(N)  # محور زمان گسسته

# فرکانس‌ها
w1 = 2 * np.pi / 8  # فرکانس اول
w0 = 2 * np.pi / 16  # فرکانس دوم

# سیگنال‌های کسینوسی
cos_w1 = np.cos(w1 * n)
cos_w0 = np.cos(w0 * n)

# تبدیل فوریه گسسته (DFT)
dft_w1 = np.fft.fft(cos_w1)
dft_w0 = np.fft.fft(cos_w0)

# فرکانس‌های متناظر با نمونه‌های DFT
frequencies = np.fft.fftfreq(N)

# رسم سیگنال‌های کسینوسی
plt.figure(figsize=(12, 8))

# سیگنال کسینوسی با فرکانس w1
plt.subplot(2, 2, 1)
plt.stem(n, cos_w1, basefmt=" ")
plt.title(f'Cosine Signal with $w_1 = 2\pi/8$')
plt.xlabel('n')
plt.ylabel('Amplitude')

# سیگنال کسینوسی با فرکانس w0
plt.subplot(2, 2, 2)
plt.stem(n, cos_w0, basefmt=" ")
plt.title(f'Cosine Signal with $w_0 = 2\pi/16$')
plt.xlabel('n')
plt.ylabel('Amplitude')

# تبدیل فوریه برای سیگنال با فرکانس w1
plt.subplot(2, 2, 3)
plt.stem(frequencies, np.abs(dft_w1), basefmt=" ")
plt.title('DFT of Cosine Signal with $w_1 = 2\pi/8$')
plt.xlabel('Frequency (k)')
plt.ylabel('Magnitude')

# تبدیل فوریه برای سیگنال با فرکانس w0
plt.subplot(2, 2, 4)
plt.stem(frequencies, np.abs(dft_w0), basefmt=" ")
plt.title('DFT of Cosine Signal with $w_0 = 2\pi/16$')
plt.xlabel('Frequency (k)')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()
```
