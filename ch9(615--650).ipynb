   <h2>مثال9.29 </h2> 
<div dir="rtl">
حال یک سیستم LTI علی را با تابع تبدیل زیر در نظر بگیرید:

$$H(s) = \frac{s+2}{s+3} = \left(\frac{1}{s+3}\right)(s+2) \quad(9.164) $$
<div dir="rtl">
همانطور که معادله (9.164) تداعی می‌کند، می‌توان این سیستم را اتصال سری دو سیستم با توابع تبدیل $\frac{1}{s+3}$ و $(s+2)$ در نظر گرفت، و آن را به صورت شکل 9.33 (الف) نمایش داد. در این شکل برای نمایش $\frac{1}{s+3}$ از نمودار جعبه‌ای شکل ۹-۳۳ (الف) را به کار برده‌ایم.
<div dir="rtl">
برای نمایش سیستم معادله (9.164) می‌توان نمایش جعبه‌ای دیگری نیز به دست آورد. با توجه به خواص خطی بودن و مشتق‌گیری تبدیل لاپلاس می‌دانیم که $(sY(s))$ و $(Y(s))$ شکل 9.33 (الف) به صورت زیر به هم مرتبط‌اند:

$$y(t) = \frac{dz(t)}{dt} + 2z(t)$$
<div dir="rtl">
ولی ورودی $e(t)$ و انتگرالگیر دقیقاً مشتق خروجی $z(t)$ است، پس:
 <div dir="rtl">   
که نمایش جعبه‌ای متفاوت شکل 9.33 (ب) را به دست می‌دهد. توجه کنید که در نمایش جعبه‌ای شکل ۹-۳۳ (الف) به مشتق $z(t)$ نیاز داریم، زیرا:
<div dir="rtl">
$$y(t) = \frac{dz(t)}{dt} + 2z(t)$$
<div dir="rtl">
ولی در نمایش جعبه‌ای شکل 9.33(ب) از هیچ سیگنالی مشتق‌گیری نشده است.

    <h2>مثال 9.30 </h2>
<div dir="rtl">
حال یک سیستم مرتبه دوم LTI با تابع تبدیل زیر در نظر می‌گیریم:

$$H(s) = \frac{1}{(s+1)(s+2)} = \frac{1}{s^2 + 3s + 2} \quad(9.165) $$
<div dir="rtl">
ورودی $x(t)$ و خروجی $y(t)$ این سیستم از معادله دیفرانسیل زیر تبعیت می‌کنند:

$$\frac{d^2y(t)}{dt^2} + 3\frac{dy(t)}{dt} + 2y(t) = x(t) \quad(9.166)$$
<div dir="rtl">
با استفاده از ایده‌هایی مشابه آنچه در مثال‌های قبل به کار گرفته شد، نمایش جعبه‌ای شکل 9.34 (الف) را به دست می‌آوریم. چون ورودی انتگرالگیر، مشتق خروجی انتگرالگیر است، سیگنال‌های این نمودار جعبه‌ای عبارت‌اند از:

$$f(t) = \frac{dy(t)}{dt}$$
$$e(t) = \frac{df(t)}{dt} = \frac{d^2y(t)}{dt^2}$$

$$\frac{d^2y(t)}{dt^2} = -3 \frac{dy(t)}{dt} - 2y(t) + x(t)$$
<div dir="rtl">
یا

$$e(t) = -3f(t) - 2y(t) + x(t)$$
<div dir="rtl">
که دقیقاً معادله نمایش یافته در شکل 9.34 (الف) است.
 <div dir="rtl">   
نمایش جعبه‌ای این شکل را گاهی نمایش مستقیم می‌نامند، زیرا ضرایب ظاهر شده در نمودار همان ضرایب تابع تبدیل یا معادله دیفرانسیل هستند. نمایش‌های جعبه‌ای دیگری نیز وجود دارند که اهمیت عملی فراوانی دارند و با کمی عملیات جبری بر تابع تبدیل به دست می‌آیند. معادله (9.165) را می‌توان به شکل زیر بازنویسی کرد:

$$H(s) = \left(\frac{1}{s+1}\right)\left(\frac{1}{s+2}\right)$$
 <div dir="rtl"> 
که نمایش این سیستم به صورت ترکیب سری دو سیستم مرتبه اول را تداعی می‌کند. نمایش سری $H(s)$ در شکل 9.34 (ب) نشان داده شده است.
<div dir="rtl">
به صورتی دیگر می‌توان با بسط $H(s)$ به کسرهای جزئی زیر به دست آورد:

$$H(s) = \frac{1}{s+1} - \frac{1}{s+2}$$
<div dir="rtl">
که نمایش موازی شکل 9.34 (ج) را به دست می‌دهد.

    <h2>مثال 9.31 </h2>
 <div dir="rtl">   
    به عنوان آخرین مثال تابع سیستم زیر را در نظر بگیرید:

$$H(s) = \frac{2s^2 + 4s - 6}{s^2 + 3s + 2} \quad(9.167)$$
 <div dir="rtl">
باز هم می‌توانیم به کمک عملیات جبری $H(s)$ را به شکل‌های مختلفی بنویسیم، که هر کدام نمایش جعبه‌ای متفاوتی را تداعی می‌کنند. می‌توانیم بنویسیم:

$$H(s) = \left(\frac{1}{s^2 + 3s + 2}\right)(2s^2 + 4s - 6)$$
 <div dir="rtl">
که $H(s)$ را به صورت ترکیب سری شکل 9.35 (الف) و سیستمی با تابع تبدیل $2s^2 + 4s - 6$ نمایش می‌دهد. ولی دقیقاً به همان روش به کار رفته در مثال 9.29 می‌توان مشتق‌های لازم برای ساختن سیستم دوم را از سیگنال‌های ورودی انتگرالگیرهای سیستم اول «اخذ کرد». جزئیات کار در مسئله 9.36 بررسی شده است، و شکل 9.35 نمودار جعبه‌ای حاصل را نشان می‌دهد. باز هم در نمایش جعبه‌ای ضرایب ظاهر شده را می‌توان مستقیماً از ضرایب تابع تبدیل (9.167) به دست آورد.
    <div dir="rtl">
   همچنین می‌توانیم $H(s)$ را به شکل زیر بنویسیم:

$$H(s) = \left(\frac{2(s-2)}{s+2}\right)\left(\frac{s+3}{s+1}\right)\quad(9.168)$$
<div dir="rtl">
یا

$$H(s) = 2 + \frac{6}{s+2} - \frac{8}{s+1}\quad(9.169)$$
<div dir="rtl">
معادله اول نمایش سری و معادله دوم نمایش موازی را تداعی می‌کند. این نمایش‌ها نیز در مسئله 9.36بررسی شده‌اند. 
<div dir="rtl">   
    روش‌های بیان شده برای ساختن نمایش جعبه‌ای سیستم‌های LTI علی توصیف شده با معادلات دیفرانسیل و توابع تبدیل گویا را می‌توان مستقیماً به سیستم‌های مرتبه بالاتر تعمیم داد. این کار بسیار پر انعطاف است. مثلاً با جابجا کردن صورت‌های معادله (9.168) می‌توان نوشت:

$$H(s) = \left(\frac{s+3}{s+2}\right)\left(\frac{2(s-1)}{s+1}\right)$$
<div dir="rtl">
که نمایش سری متفاوتی به دست می‌دهد. همچنین در مسئله 9.38 خواهیم دید که یک تابع تبدیل مرتبه چهار را می‌توان به صورت حاصلضرب دو تابع تبدیل مرتبه دوم نوشت، و هر کدام را می‌توان به روش‌های مختلف ساخت (روش مستقیم، سری یا موازی)؛ همچنین می‌توان آن را به صورت مجموع چند جمله مرتبه پایین نوشت و چند نمایش متفاوت دیگر به دست آورد. به این ترتیب سیستم‌های ساده مرتبه پایین را می‌توان به عنوان سنگ بناهای ساخت سیستم‌های مرتبه بالا به کار برد.

 <h2>9.9 تبدیل لاپلاس یک طرفه</h2>
 <div dir="rtl">   
 در بخش‌های قبل این فصل با آنچه تبدیل لاپلاس دو طرفه نام دارد کار کردیم. در این بخش تبدیل لاپلاس یکطرفه نسبتاً متفاوتی را معرفی و بررسی می‌کنیم، تبدیل لاپلاس یکطرفه که در تحلیل سیستم‌های علی، و مخصوصاً سیستمهایی که ابتدائاً ساکن نیستند (یعنی دارای شرایط اولیه ثابت)، نقش مهمی ایفا می‌کند.
<div dir="rtl">  
تبدیل لاپلاس یکطرفه سیگنال $x(t)$ به صورت زیر تعریف می‌شود:

$$X(s) = \mathcal{L}\{x(t)\} \triangleq \int_{0^-}^{\infty} x(t) e^{-st} dt \quad (9.170)$$
<div dir="rtl">  
حد پایین انتگرال، یعنی $0^-$ نشان می‌دهد که باید ضربه‌ها و توابع تکین مرتبه بالاتر موجود در $t=0$ را در فاصله انتگرالگیری منظور کرد. برای تبدیل لاپلاس یکطرفه نیز یک نماد اختصاری به صورت زیر وجود دارد:

$$x(t) \stackrel{\mathcal{L}}{\longleftrightarrow} X(s) = \mathcal{L}\{x(t)\} \quad (9.171)$$   
<div dir="rtl">     
مقایسه معادلات (9.170) و (9.3) نشان می‌دهد که تعریف تبدیل لاپلاس دو طرفه و یکطرفه تنها در حد پایین انتگرال تفاوت دارند. تبدیل دو طرفه به تمام سیگنال از $t = -\infty$ تا $t = +\infty$ بستگی دارد، ولی تبدیل لاپلاس یکطرفه تنها به مقدار سیگنال در $t = ۰^-$ تا $t = +\infty$ بستگی دارد. پس اگر دو سیگنال در $t < ۰$ یکسان باشند، ولی در $t \ge ۰$ با هم فرق بکنند، تبدیل لاپلاس دو طرفه متفاوتی دارند ولی تبدیل لاپلاس یکطرفه‌اش هر دو یکی است. همچنین اگر سیگنالی در $t < ۰$ متحد با صفر باشد، تبدیل لاپلاس یکطرفه و دو طرفه‌اش یکی است.
<div dir="rtl">  
چون تبدیل یکطرفه $\mathcal{L}\{x(t)\}$ مشابه تبدیل دو طرفه سیگنالی است که در $t < ۰$ با صفر برابر است و در $t \ge ۰$ با $x(t)$ برابر است، بسیاری از مفاهیم و نتایج به دست آمده برای تبدیل دو طرفه را می‌توان مستقیماً به حالت یکطرفه تعمیم داد. مثلاً با توجه به خاصیت 4 بیان شده در بخش 9.2 می‌بینیم که ROC معادله (9.170) همیشه یک    
راست-نیمصفحه است. عکس تبدیل لاپلاس یکطرفه نیز به همان صورت تبدیل لاپلاس دو طرفه حساب می‌شود، با این قید که ROC تبدیل یکطرفه همیشه یک راست-نیمصفحه است.
<div dir="rtl"> 
برای تشریح تبدیل لاپلاس یکطرفه چند مثال در نظر می‌گیریم.

$$\mathcal{X}(s) = \frac{۱}{(s+a)^n}, \quad \text{Re}\{s\} > -a \quad (9.173)$$

<h2>مثال 9.33</h2>
حال سیگنال زیر را در نظر بگیرید:

$$x(t) = e^{-a(t+۱)} u(t+۱) \quad (9.174)$$
<div dir="rtl">
تبدیل لاپلاس دو طرفه $\mathcal{X}(s)$ را می‌توان با توجه به مثال 9.1 و خاصیت جابه‌جایی زمانی (بخش 9.5.2) به دست آورد:

$$\mathcal{X}(s) = \frac{e^s}{s+a}, \quad \text{Re}\{s\} > -a \quad (9.175)$$
<div dir="rtl">
ولی تبدیل لاپلاس یکطرفه عبارت است از:
<div dir="rtl">
$$\begin{aligned}
\mathcal{X}(s) &= \int_{۰^-}^{\infty} e^{-a(t+۱)} u(t+۱) e^{-st} dt \quad (9.176) \\
&= \int_{۰}^{\infty} e^{-at - a} e^{-st} dt \\
&= e^{-a} \int_{۰}^{\infty} e^{-(s+a)t} dt \\
&= e^{-a} \left[ -\frac{e^{-(s+a)t}}{s+a} \right]_{۰}^{\infty} \\
&= e^{-a} \frac{۱}{s+a}, \quad \text{Re}\{s\} > -a
\end{aligned}$$
<div dir="rtl">
پس در این مثال تبدیلهای یکطرفه و دو طرفه آشکارا تفاوت دارند. در واقع می‌توان $\mathcal{L}_+(x(t))$ را تبدیل دو طرفه $x_+(t)$ در نظر گرفت، قبلاً هم گفتیم که تبدیل دو طرفه $x_+(t)$ با تبدیل دو طرفه سیگنالی که در $t < ۰$ با $۰$ برابر و در $t \ge ۰$ با $x(t)$ است، یکی است.

<h2>مثال 9.35</h2>
<div dir="rtl"> 
تبدیل لاپلاس یکطرفه زیر را در نظر بگیرید:

$$\mathcal{X}(s) = \frac{1}{(s+1)(s+2)} \quad (9.179)$$
<div dir="rtl">
در مثال 9.9 عکس تبدیل دو طرفه تابعی مشابه معادله (9.179) را با چند ROC در نظر گرفتیم. برای تبدیل یکطرفه ROC باید یک راست-نیمصفحه، در سمت راست راست‌ترین قطب $\mathcal{X}(s)$ باشد؛ پس در این مثال ROC تمام نقاط دارای $\text{Re}\{s\} > -1$ است. می‌توانیم عکس این تبدیل را دقیقاً مانند مثال 9.9 حساب کنیم:

$$x(t) = [e^{-t} - e^{-2t}]u(t), \quad t > 0^- \quad (9.180)$$
<div dir="rtl">
در این معادله بر یکطرفه بودن تبدیل لاپلاس، با نشان دادن این که اطلاعات به دست آمده در مورد سیگنال تنها در $t > 0^-$ معتبر است، تأکید کرده‌ایم.

<h2>مثال 9.36</h2>
<div dir="rtl"> 
تبدیل لاپلاس یکطرفه زیر را در نظر بگیرید:

$$\mathcal{X}(s) = \frac{s^2 - 3}{s+2} \quad (9.181)$$
<div dir="rtl"> 
چون مرتبه صورت $\mathcal{X}(s)$ کوچکتر از مرتبه مخرج نیست، آن را به صورت زیر بسط می‌دهیم:

$$\mathcal{X}(s) = A + Bs + \frac{C}{s+2} \quad (9.182)$$
<div dir="rtl"> 
با برابر قرار دادن معادلات (9.181) و (9.182) و حذف مخرج به دست می‌آوریم:

$$s^2 - 3 = (A + Bs)(s+2) + C \quad (9.183)$$
<div dir="rtl"> 
با برابر قرار دادن ضریب جملات هم توان $s$ به دست می‌آوریم:

$$\mathcal{X}(s) = -2 + s + \frac{1}{s+2} \quad (9.184)$$
<div dir="rtl"> 
ناحیه همگرایی $\text{Re}\{s\} > -2$ است. با گرفتن عکس تبدیل لاپلاس جملات خواهیم داشت:

$$x(t) = -2\delta(t) + u_1(t) + e^{-2t}u(t), \quad t > ۰^- \quad (9.185)$$

<h2>خواص تبدیل لاپلاس یکطرفه</h2>
 <div dir="rtl">   
تبدیل لاپلاس یکطرفه نیز مانند تبدیل لاپلاس دو طرفه خواصی دارد که بسیاری از آنها مشابه خواص متناظر تبدیل لاپلاس دو طرفه است، ولی برخی از آنها تفاوتهایی قابل ملاحظه دارند. این خواص در جدول 9.3 خلاصه شده‌اند. ستونی برای مشخص کردن ROC تبدیل لاپلاس یکطرفه سیگنالها در نظر
نگرفته ایم،زیرا ROC هر تبدیل لاپلاس یک طرفه ای یک راست-نیم صفحه است. برای مثال ROC هر تبدیل لاپلاس یکطرفه گویایی سمت راست راست ترین قطب است.
مقایسه جدول ۹-۳ و جدول ۹-۱ تبدیل دو طرفه نشان می‌دهد که خواص خطی بودن، جابه‌جایی در حوزه $s$، تغییر مقیاس زمانی، مزدوج‌گیری و مشتق‌گیری در حوزه زمان با همتاهای دو طرفه شان مشابه اند. همچنین قضایای مقدار اولیه و مقدار نهایی بیان شده در بخش ۹-۲-۵ برای تبدیل لاپلاس یکطرفه نیز صادق‌اند. اشتقاق این خواص همانند همتاهای دو طرفه شان صورت می‌گیرد.
خاصیت کانولوشن تبدیل یکطرفه نیز کاملاً مشابۀ خاصیت کانولوشن تبدیل دو طرفه است. این خاصیت می‌گوید که اگر
 <div dir="rtl">  
     
$$ x_1(t) = x_2(t) = 0\quad t < 0$$
  <div dir="rtl">
     آنگاه:
      
$$x_1(t) * x_2(t) \stackrel{\mathcal{L}}{\longleftrightarrow} \mathcal{X}_1(s) \mathcal{X}_2(s) \quad (9.178)$$
<div dir="rtl">
معادله (9.187) بدون واسطه از خاصیت کانولوشن دو طرفه ناشی می‌شود. زیرا تحت شرایط معادله (9.186) تبدیلهای یکطرفه و دو طرفه $x_1(t)$ و $x_2(t)$ یکسان‌اند. پس ابزارهای تحلیلی و جبر تابع تبدیل که بر اساس این خاصیت تبدیل دو طرفه پی‌ریزی شدند، بدون تغییر برای تبدیلهای یکطرفه هم قابل اعمال‌اند، به شرطی که با سیستم‌های علی کار کنیم (که برایشان تبدیل یکطرفه و دو طرفه پاسخ ضربه یکسان است) و ورودیها در $t < ۰$ متحد با صفر باشند. خاصیت انتگرالگیری ذکر شده در جدول 9.3 نمونه‌ای از این مورد است: اگر در $t < ۰$، $x(t) = ۰$ آنگاه

$$\int_{0^-}^{t} x(\tau) d\tau = x(t) * u(t) \stackrel{\mathcal{L}}{\longleftrightarrow} \mathcal{X}(s) \{u(S)\} = \frac{1}{s} \mathcal{X}(s) \quad (9.188)$$
<div dir="rtl">
    به عنوان مورد دوم مثال زیر رو در نظر بگیرید.
<div dir="rtl">
<h2>مثال 9.37</h2>
<div dir="rtl"> 
یک سیستم LTI علی با معادله دیفرانسیل زیر و شرط سکون ابتدایی توصیف شده است.
$$\frac{d^2y(t)}{dt^2} + 3\frac{dy(t)}{dt} + 2y(t) = x(t) \quad (9.189)$$
<div dir="rtl"> 
با استفاده از معادله (9.189) تابع تبدیل این سیستم را به صورت زیر می‌یابیم:

$$\mathcal{H}(s) = \frac{1}{s^2 + 3s + 2} \quad (9.190)$$
<div dir="rtl">
ورودی این سیستم را $x(t) = au(t)$ فرض کنید. در این حالت تبدیل لاپلاس یکطرفه (و دو طرفه) خروجی $y(t)$ عبارت است از:

$$\begin{aligned}
\mathcal{Y}(s) &= \mathcal{H}(s) \mathcal{X}(s) = \frac{a}{s(s+1)(s+2)} \\
&= \frac{a/2}{s} - \frac{a}{s+1} + \frac{a/2}{s+2} \quad (9.191)
\end{aligned}$$
<div dir="rtl">
با استفاده از مثال 9.3 به دست می‌آوریم:

$$y(t) = a\left[\frac{1}{2} - e^{-t} + \frac{1}{2}e^{-2t}\right]u(t) \quad (9.192)$$    
<div dir="rtl">
توجه به این نکته مهم است که خاصیت کانولوشن تبدیل لاپلاس یکطرفه تنها هنگامی قابل اعمال است که سیگنالهای $x_1(t)$ و $x_2(t)$ معادله (9.187) در $t < ۰$ صفر باشند. تبدیل لاپلاس دو طرفه $x_1(t) * x_2(t)$ همیشه با حاصلضرب تبدیلهای دو طرفه $\mathcal{X}_1(t)$ و $\mathcal{X}_2(t)$ برابر است، ولی تبدیل لاپلاس یکطرفه $x_1(t) * x_2(t)$ با حاصلضرب تبدیلهای لاپلاس یکطرفه $\mathcal{X}_1(t)$ و $\mathcal{X}_2(t)$ برابر نیست، مگر این که $x_1(t)$ و $x_2(t)$ در $t < ۰$ صفر باشند. (مسئله 9.39 را ببینید.)
با یک تفاوت مهم خواص تبدیل لاپلاس یکطرفه و دو طرفه خاصیت مشتق‌گیری است. سیگنال $x(t)$ با
تبدیل لاپلاس یکطرفه $\mathcal{Y}(s)$ را در نظر بگیرید. با انتگرالگیری جزء به جزء، تبدیل لاپلاس یکطرفه $\frac{dx(t)}{dt}$ را به صورت زیر می‌یابیم:
$$\int_{0^-}^{\infty} \frac{dy(t)}{dt} e^{-st} dt = x(t) e^{-st} \Big|_{0^-}^{\infty} + s \int_{0^-}^{\infty} x(t) e^{-st} dt$$
$$= s \mathscr{X}(s) - x(0^-)$$


<div dir="rtl">
به نحوی مشابه، با اعمال دوباره این قضیه تبدیل لاپلاس یکطرفه $\frac{d^2x(t)}{dt^2}$ را به دست می‌آوریم:

$$s^2\mathcal{X}(s) - sx(0^-) - x'(0^-) \quad (9.194)$$
<div dir="rtl">
که در آن $x'(0^-)$ مشتق $x(t)$ در $t = 0^-$ را نشان می‌دهد. می‌توانیم این فرایند را ادامه دهیم و تبدیل فوریه مشتق‌های بالاتر را به دست آوریم.
<div dir="rtl">
 <h2> 9.9.3 حل معادلات دیفرانسیل به کمک تبدیل لاپلاس یک طرفه </h2>
<div dir="rtl">
<h2>مثال 9.38</h2>
<div dir="rtl">  
سیستم توصیف شده با معادله دیفرانسیل (9.189) و شرایط اولیه زیر را در نظر بگیرید:

$$y(0^-) = \beta, \quad y'(0^-) = \gamma \quad (9.195)$$
<div dir="rtl">
فرض کنید $x(t) = au(t)$. از دو طرف معادله (9.189) تبدیل لاپلاس یکطرفه می‌گیریم و به دست می‌آوریم:

$$s^2\mathcal{Y}(s) - \beta s - \gamma + 3[s\mathcal{Y}(s) - \beta] + 2\mathcal{Y}(s) = \frac{a}{s} \quad (9.196)$$
<div dir="rtl">
یا

$$\mathcal{Y}(s) = \frac{\beta(s+3)}{(s+1)(s+2)} + \frac{\gamma}{(s+1)(s+2)} + \frac{a}{s(s+1)(s+2)} \quad (9.197)$$ 
<div dir="rtl">
که در آن $\mathcal{Y}(s)$ تبدیل لاپلاس یکطرفه $y(t)$ است. با توجه به مثال 9.37 و معادلات (9.191) آن، می‌بینیم که جمله آخر طرف راست معادله (9.197) پاسخ سیستم به ازای شرایط اولیه صفر است ($y(0^-) = 0$ و $y'(0^-) = 0$). پس جمله آخر پاسخ سیستم LTI علی توصیف شده با معادله (9.189) و شرط سکون ابتدایی است. این پاسخ را غالباً پاسخ حالت صفر می‌نامند، که پاسخ به ازای شرایط اولیه صفر (توصیف شده در معادله (9.195)) است.
برای دو جمله اول سمت راست معادله (9.197) نیز تعبیر مشابهی وجود دارد. این جملات تبدیل یکطرفه پاسخ سیستم به ازای صفر بودن ورودی ($a = 0$) را نشان می‌دهند. این پاسخ را معمولاً پاسخ ورودی صفر می‌نامند. توجه کنید که پاسخ ورودی صفر، تابعی خطی از مقدار شرایط اولیه است (مثلاً با دو برابر شدن $\beta$ و $\gamma$، پاسخ ورودی صفر نیز دو برابر می‌شود). معادله (9.197) حقیقت مهم دیگری راجع به جواب معادلات دیفرانسیل خطی با ضرایب ثابت به ازای شرایط اولیه غیر صفر نشان می‌دهد، و آن این که پاسخ کل برهم‌نهش پاسخهای حالت صفر و ورودی صفر است. پاسخ حالت صفر پاسخی است که با صفر کردن شرایط اولیه به دست
می‌آید، یعنی پاسخ سیستم LTI توصیف شده با معادله دیفرانسیل و شرط سکون ابتدایی پاسخ ورودی صفر پاسخ به شرایط اولیه، هنگام صفر بودن ورودی است. مسائل 9.40، 9.45 و 9.46 مثال‌های دیگری از این وضعیت هستند.
سرانجام به ازای هر مقدار $\beta$ و $\gamma$ می‌توان $\mathcal{Y}(s)$ معادله (9.197) را به کسرهای جزئی بسط داد و با گرفتن عکس تبدیل لاپلاس $y(t)$ را یافت. برای مثال به ازای $a = 2$، $\beta = 3$ و $\gamma = -5$ بسط به کسرهای جزئی معادله (9.197) به دست می‌دهد:

$$\mathcal{Y}(s) = \frac{1}{s} - \frac{1}{s+2} + \frac{3}{s+2} \quad (9.198)$$
<div dir="rtl">
با استفاده از مثال 9.32 به دست می‌آوریم:

$$y(t) = [-1 - e^{-t} + 3e^{-2t}]u(t), \quad t > 0^- \quad (9.199)$$
<div dir="rtl">
<h2>خلاصه </h2>
در این فصل تبدیل لاپلاس را، که می‌توان آن را تعمیمی از تبدیل فوریه دانست، مورد مطالعه قرار دادیم. تبدیل لاپلاس مخصوصاً ابزار تحلیلی مفیدی برای تحلیل و مطالعه سیستم‌های LTI است. سیستم‌های LTI، از جمله سیستم‌های توصیف شده با معادلات دیفرانسیل خطی با ضرایب ثابت را می‌توان به خاطر خواص تبدیل لاپلاس در حوزه $s$ و به کمک عملیات جبری توصیف و تحلیل کرد. به علاوه عملیات جبری بر روی تابع تبدیل سیستم ابزار مناسبی برای تحلیل اتصال سیستم‌ها و یافتن نمایش جعبه‌ای سیستم‌های LTI توصیف شده با معادلات دیفرانسیل است.
برای سیگنال‌ها و سیستم‌هایی که تبدیل لاپلاس گویا دارند، غالباً تبدیل لاپلاس را با مشخص کردن محل قطب‌ها و صفرها و ناحیه همگرایی نشان می‌دهند. از روی نمودار قطب-صفر می‌توان تبدیل فوریه را به روش هندسی به دست آورد. با دانستن محل قطب‌ها و ناحیه همگرایی به راحتی می‌توان علی بودن، پایداری و دیگر مشخصه‌های سیستم را استنتاج کرد.
در این فصل عمدتاً تبدیل لاپلاس دو طرفه مورد توجه قرار داشت. تبدیل یکطرفه که نوع نسبتاً متفاوتی از تبدیل لاپلاس است نیز معرفی شد. در واقع تبدیل یکطرفه را می‌توان به عنوان تبدیل دو طرفه سیگنالی که مقادیر آن قبل از $t = 0^-$ صفر شده‌اند، در نظر گرفت. این نوع تبدیل لاپلاس در به دست آوردن پاسخ معادلات دیفرانسیل خطی دارای ضرایب ثابت با شرایط اولیه غیر صفر، بسیار به کار می‌رود.

```python
import matplotlib.pyplot as plt

def plot_complex_plane(b):
    """
    نمودار صفحه مختلط با محورهای Re و Im و خطوط محدوده ROC را رسم می‌کند.
    
    ورودی:
    b: مقدار b برای تعیین محدوده ROC
    """
    
    # تنظیمات نمودار
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # محورها
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    
    # محدوده ROC (ناحیه خاکستری)
    ax.axvspan(-b, b, facecolor='brown', alpha=0.3, edgecolor='brown', linestyle='--')
    
    # علامت ضربدر برای قطب‌ها
    ax.plot([-b, b], [0, 0], 'x', markersize=10, color='black')
    
    # برچسب‌ها
    ax.text(-b, -0.2, '-b', ha='center', va='top')
    ax.text(b, -0.2, 'b', ha='center', va='top')
    ax.text(0, 1.2, 'Im', ha='center', va='bottom')
    ax.text(1.2, 0, 'Re', ha='left', va='center')
    ax.text(0.8, 1, 'صفحه -S', ha='left', va='bottom', rotation=90)
    
    # محدوده نمایش محورها
    ax.set_xlim([-b - 2, b + 2])
    ax.set_ylim([-1, 1.5])
    
    # حذف تیک‌های محورها
    ax.set_xticks([])
    ax.set_yticks([])
    
    # نمایش نمودار
    plt.show()

# مثال استفاده
b_value = 3  # مقدار b را می‌توانید تغییر دهید
plot_complex_plane(b_value)
```

```python
# Define the figure and axes
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs = axs.flatten()

# Titles for each subplot
titles = [
    "(a)",
    "(b)",
    "(c)",
    "(d)"
]

# Define the real and imaginary axes limits
real_limits = (-5, 5)
imag_limits = (-5, 5)

# Positions for poles and ROC lines
poles = [0]  # All poles at Re = 0

# Plot configurations
for i, ax in enumerate(axs):
    # Draw axes
    ax.axhline(y=0, color='k', linewidth=0.8)
    ax.axvline(x=0, color='k', linewidth=0.8)

    # Add poles
    for pole in poles:
        ax.plot(pole, 0, 'kx', markersize=10, label="Pole")

    # Add ROC regions based on the subplot
    if i == 0:  # (a) No ROC lines
        pass
    elif i == 1:  # (b) ROC to the right
        ax.axvline(x=0, color='k', linestyle='--', linewidth=1)
        ax.fill_betweenx(imag_limits, 0, 5, color='#d3d3d3', alpha=0.7)
    elif i == 2:  # (c) ROC to the left
        ax.axvline(x=0, color='k', linestyle='--', linewidth=1)
        ax.fill_betweenx(imag_limits, -5, 0, color='#d3d3d3', alpha=0.7)
    elif i == 3:  # (d) ROC between two lines
        ax.axvline(x=-1, color='k', linestyle='--', linewidth=1)
        ax.axvline(x=1, color='k', linestyle='--', linewidth=1)
        ax.fill_betweenx(imag_limits, -1, 1, color='#d3d3d3', alpha=0.7)

    # Set limits, title, and labels
    ax.set_xlim(real_limits)
    ax.set_ylim(imag_limits)
    ax.set_title(titles[i])
    ax.set_xlabel("Re")
    ax.set_ylabel("Im")

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the frequency range
omega = np.logspace(-1, 2, 500)

# Define the magnitude response
magnitude = 20 * np.log10(1 / np.sqrt(1 + omega**2))

# Define the phase response
phase = -np.arctan(omega)

# Create a new figure
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plot the magnitude response
axs[0].semilogx(omega, magnitude)
axs[0].set_title('Bode Magnitude Plot')
axs[0].set_xlabel('Normalized Frequency (ω)')
axs[0].set_ylabel('Magnitude (dB)')
axs[0].grid(True)

# Plot the phase response
axs[1].semilogx(omega, phase)
axs[1].set_title('Bode Phase Plot')
axs[1].set_xlabel('Normalized Frequency (ω)')
axs[1].set_ylabel('Phase (radians)')
axs[1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامتر سیستم
tau = 1

# محور فرکانس لگاریتمی
omega = np.logspace(-1, 2, 500)  # از 0.1 تا 100 (در واحد 1/τ)

# تابع انتقال: H(jw) = 1 / (1 + jwτ)
H = 1 / (1j * omega * tau + 1)

# محاسبه فاز به درجه
phase = np.angle(H, deg=True)

# تقریب مایل‌بین فاز: سه ناحیه
def phase_asymptotic(w):
    phi = np.zeros_like(w)
    for i, omega_i in enumerate(w):
        if omega_i < 0.1 / tau:
            phi[i] = 0
        elif omega_i > 10 / tau:
            phi[i] = -90
        else:
            # ناحیه میانی
            phi[i] = -45 * np.log10(omega_i * tau) - 45
    return phi

phase_approx = phase_asymptotic(omega)

# رسم نمودار
plt.figure(figsize=(10, 6))
plt.semilogx(omega / (1/tau), phase, label='پاسخ فاز واقعی', linewidth=2)
plt.semilogx(omega / (1/tau), phase_approx, '--', label='تقریب مایل‌بین', linewidth=2)

# خطوط کمکی
plt.axvline(x=1, color='gray', linestyle='--')
plt.axhline(y=-45, color='gray', linestyle='--')
plt.text(1.2, -40, '-45°', fontsize=12)
plt.text(3, -90, 'تقریب مایل‌بین', fontsize=12)

# تنظیمات نمودار
plt.xlabel('ω')
plt.ylabel('فاز (درجه)')
plt.xticks([0.1, 1, 10, 100], ['0.1/τ', '1/τ', '10/τ', '100/τ'])
plt.yticks(np.arange(-90, 10, 30))
plt.title('Phase response of the first-order system')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# مقدار ωc
wc = 1  # فرکانس قطع

# مختصات قطب‌ها
poles = [-wc, wc]

# ایجاد نمودار
fig, ax = plt.subplots()

# رسم محورهای مختلط
ax.axhline(0, color='black')  # محور حقیقی
ax.axvline(0, color='black')  # محور موهومی

# رسم دایره با شعاع ωc
theta = np.linspace(0, 2*np.pi, 300)
x_circle = wc * np.cos(theta)
y_circle = wc * np.sin(theta)
ax.plot(x_circle, y_circle, 'k--', linewidth=2)

# رسم قطب‌ها روی محور حقیقی
for p in poles:
    ax.plot(p, 0, 'x', markersize=10, color='red')

# برچسب‌گذاری
ax.text(wc, -0.1, r'$\omega_c$', ha='center', va='top', fontsize=12)
ax.text(-wc, -0.1, r'$-\omega_c$', ha='center', va='top', fontsize=12)
ax.text(0.5, 0.5, "N=1", fontsize=12)

# تنظیمات نمایش
ax.set_xlim(-1.5*wc, 1.5*wc)
ax.set_ylim(-1.5*wc, 1.5*wc)
ax.set_aspect('equal')
ax.set_xlabel("Re")
ax.set_ylabel("Im")
ax.set_title("Poles on Real Axis (N=1)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
n = 2
wc = 1  # شعاع دایره (فرکانس قطع)

# محاسبه قطب‌ها (2n قطب روی دایره)
poles = [wc * np.exp(1j * np.pi * (2*k + 1) / (2*n)) for k in range(2*n)]

# ایجاد نمودار
fig, ax = plt.subplots()

# رسم دایره
theta = np.linspace(0, 2*np.pi, 400)
x_circle = wc * np.cos(theta)
y_circle = wc * np.sin(theta)
ax.plot(x_circle, y_circle, 'k--', linewidth=2)

# رسم محورهای مختلط
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# رسم قطب‌ها
for p in poles:
    ax.plot(p.real, p.imag, 'x', markersize=10, color='blue')

# رسم شعاع از مبدأ تا یک قطب (زاویه π/4)
angle = np.pi / 4
ax.plot([0, wc*np.cos(angle)], [0, wc*np.sin(angle)], 'k-', linewidth=1)

# برچسب‌ها
ax.text(wc, -0.1, r'$\omega_c$', ha='center', va='top', fontsize=12)
ax.text(wc*np.cos(angle)/2, wc*np.sin(angle)/2, r'$\frac{\pi}{4}$', fontsize=12)
ax.text(0.6, 1.1, "N=2", fontsize=12)

# تنظیمات نمودار
ax.set_aspect('equal')
ax.set_xlim(-1.5*wc, 1.5*wc)
ax.set_ylim(-1.5*wc, 1.5*wc)
ax.set_xlabel("Re", fontsize=12)
ax.set_ylabel("Im", fontsize=12)
ax.set_title("Butterworth Poles (n=2)", fontsize=14)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_butterworth_poles(n, wc=1):
    # محاسبه قطب‌ها
    poles = [wc * np.exp(1j * np.pi * (2*k + 1) / (2*n)) for k in range(2*n)]

    # رسم نمودار
    fig, ax = plt.subplots()

    # رسم دایره با شعاع ωc
    theta = np.linspace(0, 2*np.pi, 400)
    x_circle = wc * np.cos(theta)
    y_circle = wc * np.sin(theta)
    ax.plot(x_circle, y_circle, 'k--', linewidth=2)

    # محورهای مختلط
    ax.axhline(0, color='black')
    ax.axvline(0, color='black')

    # رسم قطب‌ها
    for p in poles:
        ax.plot(p.real, p.imag, 'x', markersize=10, color='green')

    # برچسب‌ها
    ax.text(wc, -0.1, r'$\omega_c$', ha='center', va='top', fontsize=12)
    ax.text(0.5, 1.2, f"N={n}", fontsize=12)

    # تنظیمات
    ax.set_aspect('equal')
    ax.set_xlim(-1.5*wc, 1.5*wc)
    ax.set_ylim(-1.5*wc, 1.5*wc)
    ax.set_xlabel("Re")
    ax.set_ylabel("Im")
    ax.set_title(f"Butterworth Filter Poles (n={n})")
    plt.grid(True)
    plt.show()

# رسم قطب‌ها برای n=3
plot_butterworth_poles(n=3)

# رسم قطب‌ها برای n=6
plot_butterworth_poles(n=6)
```

شکل 9.28 
محل قطب های فیلتر باتروث 

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
wc = 1  # فرکانس قطع

# ایجاد شکل
fig, ax = plt.subplots(figsize=(6, 6))

# رسم نیم‌دایره در سمت چپ صفحه مختلط (نیم‌دایره در جهت ساعت‌گرد)
theta = np.linspace(np.pi/2, 3*np.pi/2, 200)
x = wc * np.cos(theta)
y = wc * np.sin(theta)
ax.plot(x, y, 'k--', linewidth=2)

# رسم محورهای مختلط
ax.axhline(0, color='black')  # محور Re
ax.axvline(0, color='black')  # محور Im
ax.text(1.05, 0, 'Re', fontsize=14)
ax.text(0, 1.05, 'Im', fontsize=14)

# محل قطب
ax.plot([-wc], [0], 'x', markersize=10, label='Pole')

# برچسب‌ها
ax.text(-wc - 0.3, -0.15, r'$-\omega_c$', fontsize=14)
ax.text(0.05, wc, r'$j\omega_c$', fontsize=14)
ax.text(0.05, -wc, r'$-j\omega_c$', fontsize=14)
ax.text(0.6, 0.6, 'N=1', fontsize=14)

# تنظیمات نهایی
ax.set_xlim(-1.5*wc, 1.5*wc)
ax.set_ylim(-1.5*wc, 1.5*wc)
ax.set_aspect('equal')
ax.axis('off')
plt.title('Butterworth Filter Pole (n=1)', fontsize=14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_butterworth_left_half(n, wc=1):
    poles = []
    for k in range(1, n + 1):
        theta = (2 * k - 1) * np.pi / (2 * n)
        pole = wc * np.exp(1j * np.pi - 1j * theta)  # فقط نیم‌صفحه چپ
        if pole.real < 0:
            poles.append(pole)

    fig, ax = plt.subplots(figsize=(6, 6))

    # رسم نیم‌دایره چپ
    theta = np.linspace(np.pi / 2, 3 * np.pi / 2, 300)
    x = wc * np.cos(theta)
    y = wc * np.sin(theta)
    ax.plot(x, y, 'k--', linewidth=1.5)

    # رسم قطب‌ها (در ربع دوم و سوم)
    for p in poles:
        ax.plot(p.real, p.imag, 'x', markersize=10, color='red')
        ax.plot(p.real, -p.imag, 'x', markersize=10, color='red')  # قرینه روی نیم‌دایره چپ

    # محورهای مختلط
    ax.axhline(0, color='black')
    ax.axvline(0, color='black')

    # برچسب‌ها
    ax.text(-wc - 0.2, 0.05, r'$-\omega_c$', fontsize=12)
    ax.text(0.05, wc + 0.1, r'$j\omega_c$', fontsize=12)
    ax.text(1.05 * wc, 0, 'Re', fontsize=12)
    ax.text(0, 1.05 * wc, 'Im', fontsize=12)
    ax.text(0.6 * wc, -0.6 * wc, 'N=2', fontsize=12)

    ax.set_xlim(-1.5 * wc, 1.5 * wc)
    ax.set_ylim(-1.5 * wc, 1.5 * wc)
    ax.set_aspect('equal')
    ax.grid(True)
    plt.show()

# اجرا برای n = 2
plot_butterworth_left_half(n=2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_butterworth_n3(wc=1):
    fig, ax = plt.subplots(figsize=(6, 6))

    # رسم نیم‌دایره چپ روی دایره واحد
    theta = np.linspace(np.pi / 2, 3 * np.pi / 2, 300)
    x = wc * np.cos(theta)
    y = wc * np.sin(theta)
    ax.plot(x, y, 'k--', linewidth=1.5)

    # پارامترها
    N = 3
    poles = []
    for k in range(1, N + 1):
        angle = np.pi * (2 * k - 1) / (2 * N)
        pole = wc * np.exp(1j * angle)
        if pole.real < 0:
            poles.append(pole)

    # رسم قطب‌های مختلط مزدوج
    for p in poles:
        ax.plot(p.real, p.imag, 'x', markersize=10, color='green')
        ax.plot(p.real, -p.imag, 'x', markersize=10, color='green')

    # رسم قطب واقعی خالص در زاویه π → (−ω_c, 0)
    ax.plot(-wc, 0, 'x', markersize=10, color='green')

    # محورهای مختلط
    ax.axhline(0, color='black')
    ax.axvline(0, color='black')

    # برچسب‌ها
    ax.text(-wc - 0.2, 0.05, r'$-\omega_c$', fontsize=12)
    ax.text(0.05, wc + 0.1, r'$j\omega_c$', fontsize=12)
    ax.text(1.05 * wc, 0, 'Re', fontsize=12)
    ax.text(0, 1.05 * wc, 'Im', fontsize=12)
    ax.text(0.6 * wc, -0.6 * wc, f'N=3', fontsize=12)

    ax.set_xlim(-1.5 * wc, 1.5 * wc)
    ax.set_ylim(-1.5 * wc, 1.5 * wc)
    ax.set_aspect('equal')
    ax.grid(True)
    plt.show()

# اجرای تابع
plot_butterworth_n3()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(8, 4))

# ورودی x(t)
ax.text(0.05, 0.5, r'$x(t)$', fontsize=14, verticalalignment='center')
ax.arrow(0.15, 0.5, 0.1, 0, head_width=0.03, head_length=0.03, fc='k', ec='k')

# انشعاب به دو مسیر
ax.plot([0.25, 0.25], [0.3, 0.7], 'k--', linewidth=1)

# مسیر بالا: H1
ax.add_patch(patches.Rectangle((0.3, 0.6), 0.2, 0.2, edgecolor='black', facecolor='none'))
ax.text(0.32, 0.72, r'$h_1(t)$', fontsize=12)
ax.text(0.32, 0.66, r'$H_1(s)$', fontsize=12)
ax.arrow(0.25, 0.7, 0.05, 0, head_width=0.02, head_length=0.02, fc='k', ec='k')  # ورودی H1
ax.arrow(0.5, 0.7, 0.1, 0, head_width=0.02, head_length=0.02, fc='k', ec='k')   # خروجی H1

# مسیر پایین: H2
ax.add_patch(patches.Rectangle((0.3, 0.2), 0.2, 0.2, edgecolor='black', facecolor='none'))
ax.text(0.32, 0.32, r'$h_2(t)$', fontsize=12)
ax.text(0.32, 0.26, r'$H_2(s)$', fontsize=12)
ax.arrow(0.25, 0.3, 0.05, 0, head_width=0.02, head_length=0.02, fc='k', ec='k')  # ورودی H2
ax.arrow(0.5, 0.3, 0.1, 0, head_width=0.02, head_length=0.02, fc='k', ec='k')   # خروجی H2

# جمع‌کننده خروجی‌ها
ax.plot([0.6, 0.6], [0.3, 0.7], 'k')  # خط عمودی جمع‌کننده
ax.plot(0.6, 0.5, 'o', markersize=10, markeredgecolor='k', markerfacecolor='white')  # جمع‌کننده

# خروجی y(t)
ax.arrow(0.6, 0.5, 0.15, 0, head_width=0.03, head_length=0.03, fc='k', ec='k')
ax.text(0.78, 0.5, r'$y(t)$', fontsize=14, verticalalignment='center')

# تنظیمات نهایی
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')
ax.set_title("Parallel connection", fontsize=14)
plt.show()
```
