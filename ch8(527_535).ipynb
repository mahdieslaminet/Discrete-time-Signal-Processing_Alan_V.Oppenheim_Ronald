سیستم‌های مخابراتی در دنیای جدید
سیستم‌های مخابراتی در دنیای جدید نقشی کلیدی در انتقال اطلاعات بین افراد، سیستم‌ها، و کامپیوترها بازی می‌کند. به طور کلی در تمام سیستم‌های مخابراتی اطلاعات ابتدا توسط فرستنده یا مدولاتور پردازش شده، به صورتی مناسب برای ارسال بر روی کانال مخابراتی در می‌آید. سپس در گیرنده سیگنال با پردازش مناسبی بازیابی می‌شود. این پردازش به دلایلی چند لازم است. اول این که در بسیاری از موارد برای کانال مخابراتی گستره فرکانسی خاصی وجود دارد که انتقال سیگنال در آن به خوبی صورت می‌گیرد، و در خارج آن گستره انتقال نا ممکن است یا باعث افت شدید کیفیت مخابره می‌شود. مثلاً سیگنال‌های گستره شنیداری 10 تا 20 کیلوهرتز در جو به سرعت تضعیف می‌شوند، ولی سیگنال‌های گستره‌های فرکانسی بالاتر تا مسافتهای بسیار طولانی منتشر می‌شوند. پس در انتقال سیگنال‌های شنیداری چون گفتار و موسیقی در کانالی که انتقال در آن اساساً به صورت انتشار در جو صورت می‌گیرد، فرستنده ابتدا سیگنال را با پردازش مناسب به سیگنالی با فرکانس بالاتر تبدیل می‌کند.
بسیاری از مفاهیم و روش‌های پی ریزی شده در فصول اولیه این کتاب نقش محوری در تحلیل و طراحی سیستم‌های مخابراتی دارند. مانند هر مفهوم دیگری که کاربردهای گسترده و مهمی دارد، مسائل جزیی فراوانی نیز وجود دارد که در موردشان کتاب‌های بسیار خوبی نوشته شده است و ما تعدادی از آن‌ها را در کتابنامه آخر کتاب درج کرده‌ایم. گرچه تحلیل کامل و مفصل سیستم‌های مخابراتی فراتر از حد این کتاب است، ولی با اطلاعات پایه‌ای کسب شده در فصول قبلی آماده‌ایم که اصول پایه‌ای و مباحث اصلی طراحی و تحلیل این سیستم‌ها آشنا شدیم.
به طور کلی فرایند گنجاندن سیگنال حاوی اطلاعات در سیگنالی دیگر را مدولاسیون می‌نامند. اخذ سیگنال حاوی اطلاعات دمدولاسیون نام دارد. چنانچه خواهیم دید روش‌های مدولاسیون علاوه بر آن که امکان انتقال سیگنال را در کانال‌های مخابراتی مختلف فراهم می‌کنند، امکانات دیگری نیز برای سیستم‌های مخابراتی فراهم می‌سازند. برای مثال مدولاسیون امکان انتقال چند سیگنال را به طور همزمان بر روی یک کانال فراهم می‌سازد.

<div dir="rtl">
گنجاندن اطلاعات را بر روی سیگنالی که انتشار موثرتری دارد فراهم می‌کند، این امکان را نیز می‌دهد چند سیگنال دارای طیف همپوشان از طریق یک کانال انتقال یابند، این مفهوم را مالتی پلکس می‌نامند.
روش‌های مختلفی برای مدولاسیون وجود دارد و به کار می‌رود، و در این فصل چند روش مهمتر را بررسی می‌کنیم. یک دسته بزرگ از روش‌های مدولاسیون بر مدولاسیون دامنه یا AM پایه گذاری شده‌اند، در این نوع مدولاسیون سیگنالی که قصد انتقالش را داریم دامنه سیگنال دیگری را تغییر می‌دهد. یک شکل بسیار متداول مدولاسیون دامنه مدولاسیون دامنه سینوسی است که آن را در بخش‌های 8.1 تا 8.4، به همراه مفهوم مرتبط مالتی پلکس فرکانس به تفصیل بررسی می‌کنیم. دیگر دسته مهم سیستم‌های PAM با مدولاسیون دامنه سیگنال پالسی سر و کار دارند، در بخش‌های 8-5 و 8-6 این شکل مدولاسیون را به همراه مفهوم مالتی پلکس زمانی بررسی می‌کنیم. در بخش 8.7 مدولاسیون متفاوتی موسوم به مدولاسیون فرکانس سینوسی را بررسی می‌کنیم که در آن سیگنال حاوی اطلاعات برای تغییر فرکانس یک سیگنال سینوسی به کار می‌رود.
تا بخش 8.7 تمام تاکید بر سیگنال‌های پیوسته در زمان است، زیرا اکثر محیط‌های انتقال را بهترست سیستم‌های پیوسته در زمان به حساب آورد. ولی برای سیگنال‌های گسسته در زمان نیز می‌توان روش‌های مشابهی پی ریخت و در نظر گرفتن مفاهیم مدولاسیون چنین سیگنال‌هایی اهمیت عملی فراوانی دارد. در بخش 8.8 برخی ایده‌های مربوط به مخابرات سیگنال‌های گسسته در زمان را در نظر خواهیم گرفت.

    <H2> 8.1 مدولاسیون دامنه نمایی مختلط و سینوسی</H2>
<div dir="rtl">
مفهوم مدولاسیون دامنه سینوسی اساس بسیاری از سیستم‌های مخابراتی است. در این مدولاسیون دامنه یک سیگنال نمایی مختلط یا یک سیگنال سینوسی (c(t در سیگنال حاوی اطلاعات (x(t ضرب (مدوله) می‌شود. سیگنال (x(t را سیگنال مدوله کننده و سیگنال (c(t را سیگنال حامل می‌نامند. پس سیگنال مدوله شده (y(t حاصلضرب دو سیگنال است.

<div dir="rtl">
$$y(t) = x(t) c(t)$$


<div dir="rtl">
همانطور که در بخش8.1 گفتیم هدف اصلی در مدولاسیون ایجاد سیگنالی است که گستره فرکانسی‌اش برای مخابره بر روی کانال مورد استفاده مناسب باشد. مثلاً در سیستم‌های تلفن انتقال راه دور توسط امواج میکروویو یا ماهواره‌ای صورت می‌گیرد. گستره فرکانسی صدای افراد ۲۰۰Hz تا ۴kHz است، ولی دستگاه‌های میکروویو سیگنال‌هایی در گستره ۳۰۰MHz تا ۳۰۰GHz را منتقل می‌کنند و ماهواره‌ها در گستره فرکانسی چند صد مگاهرتز تا ۴۰GHz کار می‌کنند. پس برای انتقال روی این کانال‌ها اطلاعات سیگنال گفتار باید به این فرکانس‌های بالا منتقل شود. همانطور که در این بخش می‌بینیم مدولاسیون دامنه سینوسی این انتقال فرکانسی را به روشی بسیار ساده صورت می‌دهد.

    <H2> 8.1.1 مدولاسیون دامنه با حامل نمایی مختلط</H2>    
<div dir="rtl">
مدولاسیون دامنه سینوسی دو شکل متداول دارد، در یکی سیگنال حامل نمایی مختلطی به شکل زیرست:
   $$c(t) = e^{j(\omega_c t + \theta_c)}  \qquad(8.1)$$
   <div dir="rtl">
   و در دیگری سیگنال حامل سینوسی و به صورت زیر است
   $$c(t) = \cos(\omega_c t + \theta_c)\qquad (8.2) $$  
   
  <div dir="rtl">
   در هر دو صورت فرکانس $\omega_c$ را فرکانس حامل می‌نامند. ابتدا شکل اول را در نظر می‌گیریم و برای سهولت فرض می‌کنیم $\theta_c = 0$ تا سیگنال مدوله شده $y(t)$ به صورت زیر باشد:
  $$y(t) = x(t) e^{j\omega_c t}\qquad (8.3) $$   


<div dir="rtl">
طبق خاصیت ضرب (بخش4.5) و با نشان دادن تبدیل فوریه $x(t)$، $y(t)$ و $c(t)$ با $X(j\omega)$، $Y(j\omega)$ و $C(j\omega)$ داریم:
<div dir="rtl">
\begin{equation}
Y(j\omega) = \frac{1}{2\pi} X(j\omega) * C(j\omega) \tag{8.4}
\end{equation}
<div dir="rtl">
اگر $c(t)$ سیگنال نمایی مختلط معادله (8.1) باشد، آنگاه:
<div dir="rtl">
\begin{equation}
C(j\omega) = 2\pi\delta(\omega - \omega_c) \tag{8.5}
\end{equation}
<div dir="rtl">
پس:
<div dir="rtl">
\begin{equation}
Y(j\omega) = X(j\omega - j\omega_c) \tag{8.6}
\end{equation}
<div dir="rtl">
بنابراین طیف خروجی مدوله شده $y(t)$ همان طیف ورودی است که به اندازه فرکانس حامل $\omega_c$ انتقال جابجا شده است. مثلاً اگر مطابق شکل 8.1 (الف) $X(j\omega)$ محدود باند و بالاترین فرکانس آن $\omega_M$ باشد (یعنی پهنای باند آن $2\omega_M$ باشد)، طیف خروجی $Y(j\omega)$ به صورت شکل 8.1 (ج) است.
<div dir="rtl">
از معادله (8.3) پیداست که می‌توان $x(t)$ را با ضرب $y(t)$ در نمایی مختلط $e^{-j\omega_c t}$ از سیگنال مدوله شده بدست آورد.
<div dir="rtl">
<div dir="rtl">
\begin{equation}
x(t) = y(t) e^{-j\omega_c t} \tag{8.7}
\end{equation}
<div dir="rtl">
این کار در حوزه فرکانس معادل انتقال طیف سیگنال مدوله شده به عقب و در محل اولیه خودش است. فرایند بازیابی سیگنال اصلی از سیگنال مدوله شده را دمدولاسیون می‌نامند. این موضوع را در بخش 8.2 دنبال خواهیم کرد.
<div dir="rtl">
چون $e^{j\omega_c t}$ سیگنال مختلطی است، معادله (8.3) را می‌توان به صورت زیر بازنویسی کرد:
<div dir="rtl">
\begin{equation}
y(t) = x(t) \cos \omega_c t + jx(t) \sin \omega_c t \tag{8.7}
\end{equation}
<div dir="rtl">
ساختن معادله (8.7) یا (8.8) با $x(t)$ حقیقی دو ضرب کننده و دو سیگنال حامل سینوسی با اختلاف فاز $\frac{\pi}{2}$ به ازای $c(t)$ معادله (8.1) نشان داده شده است. در بخش 8.4 کاربرد نمونه‌ای را معرفی می‌کنیم که در آن به کاربردن سیستمی مانند شکل 8.2، با دو حامل سینوسی دارای اختلاف فاز $\frac{\pi}{2}$، مزایایی به همراه دارد.

    <h2> 8.1.2 مدولاسیون دامنه با حامل سینوسی

<div dir="rtl">
  در وضعیتهای بسیاری استفاده از حامل سینوسی به صورت معادله (8.2) ساده‌تر و به همان کارآمدی حامل نمایی مختلط است. در واقع استفاده از حامل سینوسی معادل این است که در شکل 8.2 تنها یکی از بخش‌های حقیقی یا موهومی خروجی را نگه داریم. شکل 8.3 سیستمی با حامل سینوسی را نشان می‌دهد.
اثر مدولاسیون دامنه با حامل سینوسی به صورت معادله (8.2) را می‌توان با روشی مشابه روش بالا تحلیل کرد. بار دیگر برای سهولت فرض می‌کنیم $\theta_c = 0$. در این صورت طیف سیگنال حامل عبارت است از:
<div dir="rtl">    
از:
<div dir="rtl">
\begin{equation}
C(j\omega) = \pi [\delta(\omega - \omega_c) + \delta(\omega + \omega_c)] \tag{8.9}
\end{equation}
<div dir="rtl">
و بنابراین طبق معادله (7.4) داریم:
<div dir="rtl">
\begin{equation}
Y(j\omega) = \frac{1}{2} [X(j\omega - j\omega_c) + X(j\omega + j\omega_c)] \tag{8.10}
\end{equation}    
    
    
    

<div dir="rtl">
به ازای $X(j\omega)$ شکل 8.1 (الف)، طیف $y(t)$ به صورت شکل 8.1 (ج) است. توجه کنید که در این حالت هم:
حول $\omega_c$ و $-\omega_c$ گونه‌ای از طیف سیگنال اصلی وجود دارد. پس تنها به شرطی می‌توان $x(t)$ را از $y(t)$ بازیابی کرد که $\omega_c < \omega_M$، زیرا در غیر این صورت دو طیف روی هم می‌افتند. در مورد حامل نمایی مختلط که تنها یک نمونه طیف حول $\omega_c$ قرار می‌گیرد چنین نیست. در واقع چنانچه در بخش 8.1.1 دیدیم در مدولاسیون دامنه با حامل نمایی مختلط همیشه می‌توان $x(t)$ را از $y(t)$ به ازای هر مقدار $\omega_c$ بازیافت، به این منظور باید طیف را با ضرب در $e^{-j\omega_c t}$ مطابق معادله (8.7) به محل اصلی‌اش برگرداند. ولی برای حامل سینوسی، چنانچه شکل 8.4 نشان می‌دهد، به ازای $\omega_c < \omega_M$ دو نمونه طیف $X(j\omega)$ روی هم می‌افتند. شکل 8.5 $Y(j\omega)$ را به ازای $\omega_c = \omega_M / 2$ نشان می‌دهد. واضح است که تکرار طیف $x(t)$ در $Y(j\omega)$ مشاهده نمی‌شود. بنابراین نمی‌توان $x(t)$ را از $y(t)$ به دست آورد.

    <h2> دمدولاسیون AM سینوسی</h2>
<div dir="rtl">
 در گیرنده، سیگنال حاوی اطلاعات $x(t)$ توسط فرایند دمدولاسیون بازیابی می شود. در این بخش فرایند دمدولاسیون را برای مدولاسیون دامنه سینوسی معرفی شده در بخش پیش بررسی می کنیم. برای دمدولاسیون دو روش متداول وجود دارد که هر کدام مزایا و معایب خودشان را دارند. روش اول را در بخش 8.2.1 مورد بحث قرار می دهیم. این فرایند را **دمدولاسیون همزمان** می نامند و در آن فرستنده و گیرنده از لحاظ فاز همزمان می شوند. در بخش 8.2.2 روش دیگری مرسوم به **دمدولاسیون ناهمزمان** را معرفی می کنیم.   

<div dir="rtl">
<h2> 8.2.1 دمدولاسیون همزمان</h2>
<div dir="rtl">
دمدولاسیون سیگنالی که با حامل سینوسی مدوله شده است به ازای $\omega_M < \omega_c$ به روشی سر راست صورت می‌گیرد. سیگنال زیر را در نظر بگیرید:
<div dir="rtl">
\begin{equation}
y(t) = x(t) \cos \omega_c t \tag{8.11}
\end{equation}
<div dir="rtl">
چنان که از مثال 4.23 بر می‌آید، با مدولاسیون y(t) با همان حامل سینوسی و گذراندن سیگنال حاصل از فیلتر پایین گذر می‌توان سیگنال اصلی را به دست آورد. برای پی بردن به این مطلب فرض کنید:
<div dir="rtl">
\begin{equation}
w(t) = y(t) \cos \omega_c t \tag{8.12}
\end{equation}
<div dir="rtl">
شکل 8.6 طیف $y(t)$ و $w(t)$ را نشان می‌دهد و می‌بینیم که با به کار بردن یک فیلتر پایین گذر ایده‌آل با بهره 2 و فرکانس قطع بزرگتر از $\omega_M$ و کوچکتر از $2\omega_c - \omega_M$، می‌توان $x(t)$ را از $w(t)$ بازیافت. پاسخ فرکانسی فیلتر پایین گذر در شکل 8.6 (ج) با خط چین نشان داده شده است.

<div dir="rtl">
 مبنای استفاده از معادله8.12 و فیلتر پایین گذر برای دمدولاسیون y(t) را می توان به روش جبری زیر نشان داد.از معادله های 8.11 و 8.12 نتیجه می شود که:
$$w(t) = x(t) \cos^2 \omega_c t$$

<div dir="rtl">
    با استفاده از اتحاد مثلثاتی زیر
   $$\cos^2 \omega_c t = \frac{1}{2} + \frac{1}{2} \cos 2\omega_c t $$
   
  <div dir="rtl">
می توان w(t)را به این صورت نوشت
  <div dir="rtl">  
$$w(t) = \frac{1}{2} x(t) + \frac{1}{2} x(t) \cos 2\omega_c t  \qquad(8.13) $$   
     



<div dir="rtl">
بنابراین $w(t)$ مجموع دو جمله است، یکی نصف سیگنال اصلی و دیگری هم نصف سیگنال اصلی که توسط حاملی با فرکانس $2\omega_c$ مدوله شده باشد. هر دو جمله در طیف شکل (8.6 (ج)) مشخص شده اند. به کار بردن فیلتر پایین گذر برای $w(t)$ با نگه داشتن جمله اول سمت راست معادله (8.13) و حذف جمله دوم آن، متناظر است.
<div dir="rtl">
شکل 8.7 سیستم مدولاسیون و دمدولاسیون دامنه با حامل نمایی مختلط و شکل 8.8 سیستم مدولاسیون و دمدولاسیون با حامل سینوسی را نشان می‌دهد. در این شکلها حالت کلی تری را نشان داده‌ایم که در آن هر حامل نمایی مختلط و سینوسی دارای فاز $\theta_c$ هستند. اصلاح تحلیل پیش برای منظور کردن $\theta_c$ سر راست است و در مسئله 8.21 بررسی می‌شود.
<div dir="rtl">
در سیستم‌های شکل‌های8.7 و 8.8 فرض شده است که سیگنال دمدولاسیون از لحاظ فاز با سیگنال مدوله کننده همزمان است؛ به همین دلیل این فرایند را دمدولاسیون همزمان می‌نامند. ولی اکنون فرض کنید مدولاتور و دمدولاتور همزمان نیستند. برای حامل نمایی مختلط اگر فاز حامل مدوله کننده $\theta_c$ و فاز حامل دمدوله کننده $\theta_c$ باشد، داریم:
 <div dir="rtl">   
$$
\begin{aligned}
y(t) &= e^{j(\omega_c t + \theta_c)} x(t) \qquad(8.14)\\
w(t) &= e^{j(\omega_c t + \phi_c)} y(t) \qquad(8.15)
\end{aligned}
$$

<div dir="rtl">
    و در نتیجه
    $$w(t) = e^{j(\theta_c - \phi_c)} x(t) \qquad(8.16) $$
 <div dir="rtl">   
    پس به ازای $\theta_c \neq \phi_c$، $w(t)$ یک ضریب مختلط دارد. برای حالت خاصی که $x(t)$ مثبت است، $|w(t)| = x(t)$. و بنابراین $x(t)$ را می‌توان با گرفتن اندازه سیگنال دمدوله شده به دست آورد.
 <div dir="rtl">
برای حامل سینوسی فاز حاملهای مدوله کننده و مدوله کننده را مطابق شکل 9.8 به ترتیب با $\theta_c$ و $\phi_c$ نشان می‌دهیم. اکنون ورودی فیلتر پایین گذر عبارت است از:
  <div dir="rtl">   
 $$w(t) = x(t) \cos(\omega_c t + \theta_c) \cos(\omega_c t + \phi_c) \quad (8.17)$$    
 <div dir="rtl">
<div dir="rtl">   
  و با استفاده از اتحاد مثلثاتی
  <div dir="rtl">
  $$\cos(\omega_c t + \theta_c) \cos(\omega_c t + \phi_c) = \frac{1}{2} \cos(\theta_c - \phi_c) + \frac{1}{2} \cos(2\omega_c t + \theta_c + \phi_c) \quad (8.18)$$
  
  <div dir="rtl">    
     بدست می اوریم:
 $$w(t) = \frac{1}{2} \cos(\theta_c - \phi_c) x(t) + \frac{1}{2} x(t) \cos(2\omega_c t + \theta_c + \phi_c) \quad (8.19)$$   
    
  <div dir="rtl">    
     و در این صورت خروجی فیلتر پایین گذر حاصل ضرب $x(t)$ و ضریب دامنه $\cos(\theta_c - \phi_c)$ است. اگر نوسانسازهای مدوله کننده و دمدوله کننده همفاز باشند، $\theta_c = \phi_c$، و خروجی فیلتر پایین گذر $x(t)$ است. برعکس، اگر اختلاف فاز نوسانسازها $\pi/2$ باشد، خروجی صفر است. به طور کلی برای ماکزیمم بودن سیگنال خروجی، نوسانسازها باید همفاز باشند. مهمتر این است که رابطه فاز دو نوسانساز باید ثابت بماند تا ضریب دامنه $\cos(\theta_c - \phi_c)$ تغییر نکند. این امر مستلزم همزمانی دقیق نوسانسازهای مدوله کننده و دمدوله کننده است، امری که در عمل، مخصوصاً هنگامی که سیستمهای مدوله کننده و دمدوله کننده دور از هم قرار دارند، کاری مشکل است. اثرهای این همزمانی و نیاز به همزمانی فرکانسی مدولاتور و دمدولاتور، علاوه بر همزمانی فاز، در مسئله 8.23 به تفصیل بررسی خواهد شد.
