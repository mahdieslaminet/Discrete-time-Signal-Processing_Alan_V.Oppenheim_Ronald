   <h2>مثال9.29 </h2> 
<div dir="rtl">
حال یک سیستم LTI علی را با تابع تبدیل زیر در نظر بگیرید:

$$H(s) = \frac{s+2}{s+3} = \left(\frac{1}{s+3}\right)(s+2) \quad(9.164) $$
<div dir="rtl">
همانطور که معادله (9.164) تداعی می‌کند، می‌توان این سیستم را اتصال سری دو سیستم با توابع تبدیل $\frac{1}{s+3}$ و $(s+2)$ در نظر گرفت، و آن را به صورت شکل 9.33 (الف) نمایش داد. در این شکل برای نمایش $\frac{1}{s+3}$ از نمودار جعبه‌ای شکل ۹-۳۳ (الف) را به کار برده‌ایم.
<div dir="rtl">
برای نمایش سیستم معادله (9.164) می‌توان نمایش جعبه‌ای دیگری نیز به دست آورد. با توجه به خواص خطی بودن و مشتق‌گیری تبدیل لاپلاس می‌دانیم که $(sY(s))$ و $(Y(s))$ شکل 9.33 (الف) به صورت زیر به هم مرتبط‌اند:

$$y(t) = \frac{dz(t)}{dt} + 2z(t)$$
<div dir="rtl">
ولی ورودی $e(t)$ و انتگرالگیر دقیقاً مشتق خروجی $z(t)$ است، پس:
    
که نمایش جعبه‌ای متفاوت شکل 9.33 (ب) را به دست می‌دهد. توجه کنید که در نمایش جعبه‌ای شکل ۹-۳۳ (الف) به مشتق $z(t)$ نیاز داریم، زیرا:
<div dir="rtl">
$$y(t) = \frac{dz(t)}{dt} + 2z(t)$$
<div dir="rtl">
ولی در نمایش جعبه‌ای شکل 9.33(ب) از هیچ سیگنالی مشتق‌گیری نشده است.    
    

<div dir="rtl">
    <h2>مثال 9.30 </h2>
<div dir="rtl">
حال یک سیستم مرتبه دوم LTI با تابع تبدیل زیر در نظر می‌گیریم:

$$H(s) = \frac{1}{(s+1)(s+2)} = \frac{1}{s^2 + 3s + 2} \quad(9.165) $$
<div dir="rtl">
ورودی $x(t)$ و خروجی $y(t)$ این سیستم از معادله دیفرانسیل زیر تبعیت می‌کنند:

$$\frac{d^2y(t)}{dt^2} + 3\frac{dy(t)}{dt} + 2y(t) = x(t) \quad(9.166)$$
<div dir="rtl">
با استفاده از ایده‌هایی مشابه آنچه در مثال‌های قبل به کار گرفته شد، نمایش جعبه‌ای شکل 9.34 (الف) را به دست می‌آوریم. چون ورودی انتگرالگیر، مشتق خروجی انتگرالگیر است، سیگنال‌های این نمودار جعبه‌ای عبارت‌اند از:

$$f(t) = \frac{dy(t)}{dt}$$
$$e(t) = \frac{df(t)}{dt} = \frac{d^2y(t)}{dt^2}$$

$$\frac{d^2y(t)}{dt^2} = -3 \frac{dy(t)}{dt} - 2y(t) + x(t)$$
<div dir="rtl">
یا

$$e(t) = -3f(t) - 2y(t) + x(t)$$
<div dir="rtl">
که دقیقاً معادله نمایش یافته در شکل 9.34 (الف) است.
 <div dir="rtl">   
نمایش جعبه‌ای این شکل را گاهی نمایش مستقیم می‌نامند، زیرا ضرایب ظاهر شده در نمودار همان ضرایب تابع تبدیل یا معادله دیفرانسیل هستند. نمایش‌های جعبه‌ای دیگری نیز وجود دارند که اهمیت عملی فراوانی دارند و با کمی عملیات جبری بر تابع تبدیل به دست می‌آیند. معادله (9.165) را می‌توان به شکل زیر بازنویسی کرد:

$$H(s) = \left(\frac{1}{s+1}\right)\left(\frac{1}{s+2}\right)$$
 <div dir="rtl"> 
که نمایش این سیستم به صورت ترکیب سری دو سیستم مرتبه اول را تداعی می‌کند. نمایش سری $H(s)$ در شکل 9.34 (ب) نشان داده شده است.
<div dir="rtl">
به صورتی دیگر می‌توان با بسط $H(s)$ به کسرهای جزئی زیر به دست آورد:

$$H(s) = \frac{1}{s+1} - \frac{1}{s+2}$$
<div dir="rtl">
که نمایش موازی شکل 9.34 (ج) را به دست می‌دهد.



<div dir="rtl">
    <h2>مثال 9.31 </h2>
 <div dir="rtl">   
    به عنوان آخرین مثال تابع سیستم زیر را در نظر بگیرید:

$$H(s) = \frac{2s^2 + 4s - 6}{s^2 + 3s + 2} \quad(9.167)$$
 <div dir="rtl">
باز هم می‌توانیم به کمک عملیات جبری $H(s)$ را به شکل‌های مختلفی بنویسیم، که هر کدام نمایش جعبه‌ای متفاوتی را تداعی می‌کنند. می‌توانیم بنویسیم:

$$H(s) = \left(\frac{1}{s^2 + 3s + 2}\right)(2s^2 + 4s - 6)$$
 <div dir="rtl">
که $H(s)$ را به صورت ترکیب سری شکل 9.35 (الف) و سیستمی با تابع تبدیل $2s^2 + 4s - 6$ نمایش می‌دهد. ولی دقیقاً به همان روش به کار رفته در مثال 9.29 می‌توان مشتق‌های لازم برای ساختن سیستم دوم را از سیگنال‌های ورودی انتگرالگیرهای سیستم اول «اخذ کرد». جزئیات کار در مسئله 9.36 بررسی شده است، و شکل 9.35 نمودار جعبه‌ای حاصل را نشان می‌دهد. باز هم در نمایش جعبه‌ای ضرایب ظاهر شده را می‌توان مستقیماً از ضرایب تابع تبدیل (9.167) به دست آورد.
    <div dir="rtl">
   همچنین می‌توانیم $H(s)$ را به شکل زیر بنویسیم:

$$H(s) = \left(\frac{2(s-2)}{s+2}\right)\left(\frac{s+3}{s+1}\right)\quad(9.168)$$
<div dir="rtl">
یا

$$H(s) = 2 + \frac{6}{s+2} - \frac{8}{s+1}\quad(9.169)$$
<div dir="rtl">
معادله اول نمایش سری و معادله دوم نمایش موازی را تداعی می‌کند. این نمایش‌ها نیز در مسئله 9.36بررسی شده‌اند. 
<div dir="rtl">   
    روش‌های بیان شده برای ساختن نمایش جعبه‌ای سیستم‌های LTI علی توصیف شده با معادلات دیفرانسیل و توابع تبدیل گویا را می‌توان مستقیماً به سیستم‌های مرتبه بالاتر تعمیم داد. این کار بسیار پر انعطاف است. مثلاً با جابجا کردن صورت‌های معادله (9.168) می‌توان نوشت:

$$H(s) = \left(\frac{s+3}{s+2}\right)\left(\frac{2(s-1)}{s+1}\right)$$
<div dir="rtl">
که نمایش سری متفاوتی به دست می‌دهد. همچنین در مسئله 9.38 خواهیم دید که یک تابع تبدیل مرتبه چهار را می‌توان به صورت حاصلضرب دو تابع تبدیل مرتبه دوم نوشت، و هر کدام را می‌توان به روش‌های مختلف ساخت (روش مستقیم، سری یا موازی)؛ همچنین می‌توان آن را به صورت مجموع چند جمله مرتبه پایین نوشت و چند نمایش متفاوت دیگر به دست آورد. به این ترتیب سیستم‌های ساده مرتبه پایین را می‌توان به عنوان سنگ بناهای ساخت سیستم‌های مرتبه بالا به کار برد.
