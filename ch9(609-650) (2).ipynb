```python
import numpy as np
import matplotlib.pyplot as plt

# ูพุงุฑุงูุชุฑ ุณุณุชู
tau = 1

# ูุญูุฑ ูุฑฺฉุงูุณ ูฺฏุงุฑุชู
omega = np.logspace(-1, 2, 500)  # ุงุฒ 0.1 ุชุง 100 (ุฏุฑ ูุงุญุฏ 1/ฯ)

# ุชุงุจุน ุงูุชูุงู: H(jw) = 1 / (1 + jwฯ)
H = 1 / (1j * omega * tau + 1)

# ูุญุงุณุจู ูุงุฒ ุจู ุฏุฑุฌู
phase = np.angle(H, deg=True)

# ุชูุฑุจ ูุงูโุจู ูุงุฒ: ุณู ูุงุญู
def phase_asymptotic(w):
    phi = np.zeros_like(w)
    for i, omega_i in enumerate(w):
        if omega_i < 0.1 / tau:
            phi[i] = 0
        elif omega_i > 10 / tau:
            phi[i] = -90
        else:
            # ูุงุญู ูุงู
            phi[i] = -45 * np.log10(omega_i * tau) - 45
    return phi

phase_approx = phase_asymptotic(omega)

# ุฑุณู ูููุฏุงุฑ
plt.figure(figsize=(10, 6))
plt.semilogx(omega / (1/tau), phase, label='ูพุงุณุฎ ูุงุฒ ูุงูุน', linewidth=2)
plt.semilogx(omega / (1/tau), phase_approx, '--', label='ุชูุฑุจ ูุงูโุจู', linewidth=2)

# ุฎุทูุท ฺฉูฺฉ
plt.axvline(x=1, color='gray', linestyle='--')
plt.axhline(y=-45, color='gray', linestyle='--')
plt.text(1.2, -40, '-45ยฐ', fontsize=12)
plt.text(3, -90, 'ุชูุฑุจ ูุงูโุจู', fontsize=12)

# ุชูุธูุงุช ูููุฏุงุฑ
plt.xlabel('ฯ')
plt.ylabel('ูุงุฒ (ุฏุฑุฌู)')
plt.xticks([0.1, 1, 10, 100], ['0.1/ฯ', '1/ฯ', '10/ฯ', '100/ฯ'])
plt.yticks(np.arange(-90, 10, 30))
plt.title('Phase response of the first-order system')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ูพุงุฑุงูุชุฑูุง
n = 2
wc = 1  # ุดุนุงุน ุฏุงุฑู (ูุฑฺฉุงูุณ ูุทุน)

# ูุญุงุณุจู ูุทุจโูุง (2n ูุทุจ ุฑู ุฏุงุฑู)
poles = [wc * np.exp(1j * np.pi * (2*k + 1) / (2*n)) for k in range(2*n)]

# ุงุฌุงุฏ ูููุฏุงุฑ
fig, ax = plt.subplots()

# ุฑุณู ุฏุงุฑู
theta = np.linspace(0, 2*np.pi, 400)
x_circle = wc * np.cos(theta)
y_circle = wc * np.sin(theta)
ax.plot(x_circle, y_circle, 'k--', linewidth=2)

# ุฑุณู ูุญูุฑูุง ูุฎุชูุท
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# ุฑุณู ูุทุจโูุง
for p in poles:
    ax.plot(p.real, p.imag, 'x', markersize=10, color='blue')

# ุฑุณู ุดุนุงุน ุงุฒ ูุจุฏุฃ ุชุง ฺฉ ูุทุจ (ุฒุงูู ฯ/4)
angle = np.pi / 4
ax.plot([0, wc*np.cos(angle)], [0, wc*np.sin(angle)], 'k-', linewidth=1)

# ุจุฑฺุณุจโูุง
ax.text(wc, -0.1, r'$\omega_c$', ha='center', va='top', fontsize=12)
ax.text(wc*np.cos(angle)/2, wc*np.sin(angle)/2, r'$\frac{\pi}{4}$', fontsize=12)
ax.text(0.6, 1.1, "N=2", fontsize=12)

# ุชูุธูุงุช ูููุฏุงุฑ
ax.set_aspect('equal')
ax.set_xlim(-1.5*wc, 1.5*wc)
ax.set_ylim(-1.5*wc, 1.5*wc)
ax.set_xlabel("Re", fontsize=12)
ax.set_ylabel("Im", fontsize=12)
ax.set_title("Butterworth Poles (n=2)", fontsize=14)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_butterworth_poles(n, wc=1):
    # ูุญุงุณุจู ูุทุจโูุง
    poles = [wc * np.exp(1j * np.pi * (2*k + 1) / (2*n)) for k in range(2*n)]

    # ุฑุณู ูููุฏุงุฑ
    fig, ax = plt.subplots()

    # ุฑุณู ุฏุงุฑู ุจุง ุดุนุงุน ฯc
    theta = np.linspace(0, 2*np.pi, 400)
    x_circle = wc * np.cos(theta)
    y_circle = wc * np.sin(theta)
    ax.plot(x_circle, y_circle, 'k--', linewidth=2)

    # ูุญูุฑูุง ูุฎุชูุท
    ax.axhline(0, color='black')
    ax.axvline(0, color='black')

    # ุฑุณู ูุทุจโูุง
    for p in poles:
        ax.plot(p.real, p.imag, 'x', markersize=10, color='green')

    # ุจุฑฺุณุจโูุง
    ax.text(wc, -0.1, r'$\omega_c$', ha='center', va='top', fontsize=12)
    ax.text(0.5, 1.2, f"N={n}", fontsize=12)

    # ุชูุธูุงุช
    ax.set_aspect('equal')
    ax.set_xlim(-1.5*wc, 1.5*wc)
    ax.set_ylim(-1.5*wc, 1.5*wc)
    ax.set_xlabel("Re")
    ax.set_ylabel("Im")
    ax.set_title(f"Butterworth Filter Poles (n={n})")
    plt.grid(True)
    plt.show()

# ุฑุณู ูุทุจโูุง ุจุฑุง n=3
plot_butterworth_poles(n=3)

# ุฑุณู ูุทุจโูุง ุจุฑุง n=6
plot_butterworth_poles(n=6)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ูพุงุฑุงูุชุฑูุง
wc = 1  # ูุฑฺฉุงูุณ ูุทุน

# ุงุฌุงุฏ ุดฺฉู
fig, ax = plt.subplots(figsize=(6, 6))

# ุฑุณู ููโุฏุงุฑู ุฏุฑ ุณูุช ฺูพ ุตูุญู ูุฎุชูุท (ููโุฏุงุฑู ุฏุฑ ุฌูุช ุณุงุนุชโฺฏุฑุฏ)
theta = np.linspace(np.pi/2, 3*np.pi/2, 200)
x = wc * np.cos(theta)
y = wc * np.sin(theta)
ax.plot(x, y, 'k--', linewidth=2)

# ุฑุณู ูุญูุฑูุง ูุฎุชูุท
ax.axhline(0, color='black')  # ูุญูุฑ Re
ax.axvline(0, color='black')  # ูุญูุฑ Im
ax.text(1.05, 0, 'Re', fontsize=14)
ax.text(0, 1.05, 'Im', fontsize=14)

# ูุญู ูุทุจ
ax.plot([-wc], [0], 'x', markersize=10, label='Pole')

# ุจุฑฺุณุจโูุง
ax.text(-wc - 0.3, -0.15, r'$-\omega_c$', fontsize=14)
ax.text(0.05, wc, r'$j\omega_c$', fontsize=14)
ax.text(0.05, -wc, r'$-j\omega_c$', fontsize=14)
ax.text(0.6, 0.6, 'N=1', fontsize=14)

# ุชูุธูุงุช ููุง
ax.set_xlim(-1.5*wc, 1.5*wc)
ax.set_ylim(-1.5*wc, 1.5*wc)
ax.set_aspect('equal')
ax.axis('off')
plt.title('Butterworth Filter Pole (n=1)', fontsize=14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_butterworth_left_half(n, wc=1):
    poles = []
    for k in range(1, n + 1):
        theta = (2 * k - 1) * np.pi / (2 * n)
        pole = wc * np.exp(1j * np.pi - 1j * theta)  # ููุท ููโุตูุญู ฺูพ
        if pole.real < 0:
            poles.append(pole)

    fig, ax = plt.subplots(figsize=(6, 6))

    # ุฑุณู ููโุฏุงุฑู ฺูพ
    theta = np.linspace(np.pi / 2, 3 * np.pi / 2, 300)
    x = wc * np.cos(theta)
    y = wc * np.sin(theta)
    ax.plot(x, y, 'k--', linewidth=1.5)

    # ุฑุณู ูุทุจโูุง (ุฏุฑ ุฑุจุน ุฏูู ู ุณูู)
    for p in poles:
        ax.plot(p.real, p.imag, 'x', markersize=10, color='red')
        ax.plot(p.real, -p.imag, 'x', markersize=10, color='red')  # ูุฑูู ุฑู ููโุฏุงุฑู ฺูพ

    # ูุญูุฑูุง ูุฎุชูุท
    ax.axhline(0, color='black')
    ax.axvline(0, color='black')

    # ุจุฑฺุณุจโูุง
    ax.text(-wc - 0.2, 0.05, r'$-\omega_c$', fontsize=12)
    ax.text(0.05, wc + 0.1, r'$j\omega_c$', fontsize=12)
    ax.text(1.05 * wc, 0, 'Re', fontsize=12)
    ax.text(0, 1.05 * wc, 'Im', fontsize=12)
    ax.text(0.6 * wc, -0.6 * wc, 'N=2', fontsize=12)

    ax.set_xlim(-1.5 * wc, 1.5 * wc)
    ax.set_ylim(-1.5 * wc, 1.5 * wc)
    ax.set_aspect('equal')
    ax.grid(True)
    plt.show()

# ุงุฌุฑุง ุจุฑุง n = 2
plot_butterworth_left_half(n=2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_butterworth_n3(wc=1):
    fig, ax = plt.subplots(figsize=(6, 6))

    # ุฑุณู ููโุฏุงุฑู ฺูพ ุฑู ุฏุงุฑู ูุงุญุฏ
    theta = np.linspace(np.pi / 2, 3 * np.pi / 2, 300)
    x = wc * np.cos(theta)
    y = wc * np.sin(theta)
    ax.plot(x, y, 'k--', linewidth=1.5)

    # ูพุงุฑุงูุชุฑูุง
    N = 3
    poles = []
    for k in range(1, N + 1):
        angle = np.pi * (2 * k - 1) / (2 * N)
        pole = wc * np.exp(1j * angle)
        if pole.real < 0:
            poles.append(pole)

    # ุฑุณู ูุทุจโูุง ูุฎุชูุท ูุฒุฏูุฌ
    for p in poles:
        ax.plot(p.real, p.imag, 'x', markersize=10, color='green')
        ax.plot(p.real, -p.imag, 'x', markersize=10, color='green')

    # ุฑุณู ูุทุจ ูุงูุน ุฎุงูุต ุฏุฑ ุฒุงูู ฯ โ (โฯ_c, 0)
    ax.plot(-wc, 0, 'x', markersize=10, color='green')

    # ูุญูุฑูุง ูุฎุชูุท
    ax.axhline(0, color='black')
    ax.axvline(0, color='black')

    # ุจุฑฺุณุจโูุง
    ax.text(-wc - 0.2, 0.05, r'$-\omega_c$', fontsize=12)
    ax.text(0.05, wc + 0.1, r'$j\omega_c$', fontsize=12)
    ax.text(1.05 * wc, 0, 'Re', fontsize=12)
    ax.text(0, 1.05 * wc, 'Im', fontsize=12)
    ax.text(0.6 * wc, -0.6 * wc, f'N=3', fontsize=12)

    ax.set_xlim(-1.5 * wc, 1.5 * wc)
    ax.set_ylim(-1.5 * wc, 1.5 * wc)
    ax.set_aspect('equal')
    ax.grid(True)
    plt.show()

# ุงุฌุฑุง ุชุงุจุน
plot_butterworth_n3()
```
