   <h2>مثال9.29 </h2> 
<div dir="rtl">
حال یک سیستم LTI علی را با تابع تبدیل زیر در نظر بگیرید:

$$H(s) = \frac{s+2}{s+3} = \left(\frac{1}{s+3}\right)(s+2) \quad(9.164) $$
<div dir="rtl">
همانطور که معادله (9.164) تداعی می‌کند، می‌توان این سیستم را اتصال سری دو سیستم با توابع تبدیل $\frac{1}{s+3}$ و $(s+2)$ در نظر گرفت، و آن را به صورت شکل 9.33 (الف) نمایش داد. در این شکل برای نمایش $\frac{1}{s+3}$ از نمودار جعبه‌ای شکل ۹-۳۳ (الف) را به کار برده‌ایم.
<div dir="rtl">
برای نمایش سیستم معادله (9.164) می‌توان نمایش جعبه‌ای دیگری نیز به دست آورد. با توجه به خواص خطی بودن و مشتق‌گیری تبدیل لاپلاس می‌دانیم که $(sY(s))$ و $(Y(s))$ شکل 9.33 (الف) به صورت زیر به هم مرتبط‌اند:

$$y(t) = \frac{dz(t)}{dt} + 2z(t)$$
<div dir="rtl">
ولی ورودی $e(t)$ و انتگرالگیر دقیقاً مشتق خروجی $z(t)$ است، پس:
    
که نمایش جعبه‌ای متفاوت شکل 9.33 (ب) را به دست می‌دهد. توجه کنید که در نمایش جعبه‌ای شکل ۹-۳۳ (الف) به مشتق $z(t)$ نیاز داریم، زیرا:
<div dir="rtl">
$$y(t) = \frac{dz(t)}{dt} + 2z(t)$$
<div dir="rtl">
ولی در نمایش جعبه‌ای شکل 9.33(ب) از هیچ سیگنالی مشتق‌گیری نشده است.

    <h2>مثال 9.30 </h2>
<div dir="rtl">
حال یک سیستم مرتبه دوم LTI با تابع تبدیل زیر در نظر می‌گیریم:

$$H(s) = \frac{1}{(s+1)(s+2)} = \frac{1}{s^2 + 3s + 2} \quad(9.165) $$
<div dir="rtl">
ورودی $x(t)$ و خروجی $y(t)$ این سیستم از معادله دیفرانسیل زیر تبعیت می‌کنند:

$$\frac{d^2y(t)}{dt^2} + 3\frac{dy(t)}{dt} + 2y(t) = x(t) \quad(9.166)$$
<div dir="rtl">
با استفاده از ایده‌هایی مشابه آنچه در مثال‌های قبل به کار گرفته شد، نمایش جعبه‌ای شکل 9.34 (الف) را به دست می‌آوریم. چون ورودی انتگرالگیر، مشتق خروجی انتگرالگیر است، سیگنال‌های این نمودار جعبه‌ای عبارت‌اند از:

$$f(t) = \frac{dy(t)}{dt}$$
$$e(t) = \frac{df(t)}{dt} = \frac{d^2y(t)}{dt^2}$$

$$\frac{d^2y(t)}{dt^2} = -3 \frac{dy(t)}{dt} - 2y(t) + x(t)$$
<div dir="rtl">
یا

$$e(t) = -3f(t) - 2y(t) + x(t)$$
<div dir="rtl">
که دقیقاً معادله نمایش یافته در شکل 9.34 (الف) است.
 <div dir="rtl">   
نمایش جعبه‌ای این شکل را گاهی نمایش مستقیم می‌نامند، زیرا ضرایب ظاهر شده در نمودار همان ضرایب تابع تبدیل یا معادله دیفرانسیل هستند. نمایش‌های جعبه‌ای دیگری نیز وجود دارند که اهمیت عملی فراوانی دارند و با کمی عملیات جبری بر تابع تبدیل به دست می‌آیند. معادله (9.165) را می‌توان به شکل زیر بازنویسی کرد:

$$H(s) = \left(\frac{1}{s+1}\right)\left(\frac{1}{s+2}\right)$$
 <div dir="rtl"> 
که نمایش این سیستم به صورت ترکیب سری دو سیستم مرتبه اول را تداعی می‌کند. نمایش سری $H(s)$ در شکل 9.34 (ب) نشان داده شده است.
<div dir="rtl">
به صورتی دیگر می‌توان با بسط $H(s)$ به کسرهای جزئی زیر به دست آورد:

$$H(s) = \frac{1}{s+1} - \frac{1}{s+2}$$
<div dir="rtl">
که نمایش موازی شکل 9.34 (ج) را به دست می‌دهد.

    <h2>مثال 9.31 </h2>
 <div dir="rtl">   
    به عنوان آخرین مثال تابع سیستم زیر را در نظر بگیرید:

$$H(s) = \frac{2s^2 + 4s - 6}{s^2 + 3s + 2} \quad(9.167)$$
 <div dir="rtl">
باز هم می‌توانیم به کمک عملیات جبری $H(s)$ را به شکل‌های مختلفی بنویسیم، که هر کدام نمایش جعبه‌ای متفاوتی را تداعی می‌کنند. می‌توانیم بنویسیم:

$$H(s) = \left(\frac{1}{s^2 + 3s + 2}\right)(2s^2 + 4s - 6)$$
 <div dir="rtl">
که $H(s)$ را به صورت ترکیب سری شکل 9.35 (الف) و سیستمی با تابع تبدیل $2s^2 + 4s - 6$ نمایش می‌دهد. ولی دقیقاً به همان روش به کار رفته در مثال 9.29 می‌توان مشتق‌های لازم برای ساختن سیستم دوم را از سیگنال‌های ورودی انتگرالگیرهای سیستم اول «اخذ کرد». جزئیات کار در مسئله 9.36 بررسی شده است، و شکل 9.35 نمودار جعبه‌ای حاصل را نشان می‌دهد. باز هم در نمایش جعبه‌ای ضرایب ظاهر شده را می‌توان مستقیماً از ضرایب تابع تبدیل (9.167) به دست آورد.
    <div dir="rtl">
   همچنین می‌توانیم $H(s)$ را به شکل زیر بنویسیم:

$$H(s) = \left(\frac{2(s-2)}{s+2}\right)\left(\frac{s+3}{s+1}\right)\quad(9.168)$$
<div dir="rtl">
یا

$$H(s) = 2 + \frac{6}{s+2} - \frac{8}{s+1}\quad(9.169)$$
<div dir="rtl">
معادله اول نمایش سری و معادله دوم نمایش موازی را تداعی می‌کند. این نمایش‌ها نیز در مسئله 9.36بررسی شده‌اند. 
<div dir="rtl">   
    روش‌های بیان شده برای ساختن نمایش جعبه‌ای سیستم‌های LTI علی توصیف شده با معادلات دیفرانسیل و توابع تبدیل گویا را می‌توان مستقیماً به سیستم‌های مرتبه بالاتر تعمیم داد. این کار بسیار پر انعطاف است. مثلاً با جابجا کردن صورت‌های معادله (9.168) می‌توان نوشت:

$$H(s) = \left(\frac{s+3}{s+2}\right)\left(\frac{2(s-1)}{s+1}\right)$$
<div dir="rtl">
که نمایش سری متفاوتی به دست می‌دهد. همچنین در مسئله 9.38 خواهیم دید که یک تابع تبدیل مرتبه چهار را می‌توان به صورت حاصلضرب دو تابع تبدیل مرتبه دوم نوشت، و هر کدام را می‌توان به روش‌های مختلف ساخت (روش مستقیم، سری یا موازی)؛ همچنین می‌توان آن را به صورت مجموع چند جمله مرتبه پایین نوشت و چند نمایش متفاوت دیگر به دست آورد. به این ترتیب سیستم‌های ساده مرتبه پایین را می‌توان به عنوان سنگ بناهای ساخت سیستم‌های مرتبه بالا به کار برد.

 <h2>9.9 تبدیل لاپلاس یک طرفه</h2>
 <div dir="rtl">   
 در بخش‌های قبل این فصل با آنچه تبدیل لاپلاس دو طرفه نام دارد کار کردیم. در این بخش تبدیل لاپلاس یکطرفه نسبتاً متفاوتی را معرفی و بررسی می‌کنیم، تبدیل لاپلاس یکطرفه که در تحلیل سیستم‌های علی، و مخصوصاً سیستمهایی که ابتدائاً ساکن نیستند (یعنی دارای شرایط اولیه ثابت)، نقش مهمی ایفا می‌کند.
<div dir="rtl">  
تبدیل لاپلاس یکطرفه سیگنال $x(t)$ به صورت زیر تعریف می‌شود:

$$X(s) = \mathcal{L}\{x(t)\} \triangleq \int_{0^-}^{\infty} x(t) e^{-st} dt \quad (9.170)$$
<div dir="rtl">  
حد پایین انتگرال، یعنی $0^-$ نشان می‌دهد که باید ضربه‌ها و توابع تکین مرتبه بالاتر موجود در $t=0$ را در فاصله انتگرالگیری منظور کرد. برای تبدیل لاپلاس یکطرفه نیز یک نماد اختصاری به صورت زیر وجود دارد:

$$x(t) \stackrel{\mathcal{L}}{\longleftrightarrow} X(s) = \mathcal{L}\{x(t)\} \quad (9.171)$$   
<div dir="rtl">     
مقایسه معادلات (9.170) و (9.3) نشان می‌دهد که تعریف تبدیل لاپلاس دو طرفه و یکطرفه تنها در حد پایین انتگرال تفاوت دارند. تبدیل دو طرفه به تمام سیگنال از $t = -\infty$ تا $t = +\infty$ بستگی دارد، ولی تبدیل لاپلاس یکطرفه تنها به مقدار سیگنال در $t = ۰^-$ تا $t = +\infty$ بستگی دارد. پس اگر دو سیگنال در $t < ۰$ یکسان باشند، ولی در $t \ge ۰$ با هم فرق بکنند، تبدیل لاپلاس دو طرفه متفاوتی دارند ولی تبدیل لاپلاس یکطرفه‌اش هر دو یکی است. همچنین اگر سیگنالی در $t < ۰$ متحد با صفر باشد، تبدیل لاپلاس یکطرفه و دو طرفه‌اش یکی است.
<div dir="rtl">  
چون تبدیل یکطرفه $\mathcal{L}\{x(t)\}$ مشابه تبدیل دو طرفه سیگنالی است که در $t < ۰$ با صفر برابر است و در $t \ge ۰$ با $x(t)$ برابر است، بسیاری از مفاهیم و نتایج به دست آمده برای تبدیل دو طرفه را می‌توان مستقیماً به حالت یکطرفه تعمیم داد. مثلاً با توجه به خاصیت 4 بیان شده در بخش 9.2 می‌بینیم که ROC معادله (9.170) همیشه یک    
راست-نیمصفحه است. عکس تبدیل لاپلاس یکطرفه نیز به همان صورت تبدیل لاپلاس دو طرفه حساب می‌شود، با این قید که ROC تبدیل یکطرفه همیشه یک راست-نیمصفحه است.
<div dir="rtl"> 
برای تشریح تبدیل لاپلاس یکطرفه چند مثال در نظر می‌گیریم.

حال سیگنال زیر را در نظر بگیرید:

$$x(t) = e^{-a(t+۱)} u(t+۱) \quad (9.174)$$
<div dir="rtl">
تبدیل لاپلاس دو طرفه $\mathcal{X}(s)$ را می‌توان با توجه به مثال 9.1 و خاصیت جابه‌جایی زمانی (بخش 9.5.2) به دست آورد:

$$\mathcal{X}(s) = \frac{e^s}{s+a}, \quad \text{Re}\{s\} > -a \quad (9.175)$$
<div dir="rtl">
ولی تبدیل لاپلاس یکطرفه عبارت است از:
<div dir="rtl">
$$\begin{aligned}
\mathcal{X}(s) &= \int_{۰^-}^{\infty} e^{-a(t+۱)} u(t+۱) e^{-st} dt \quad (9.176) \\
&= \int_{۰}^{\infty} e^{-at - a} e^{-st} dt \\
&= e^{-a} \int_{۰}^{\infty} e^{-(s+a)t} dt \\
&= e^{-a} \left[ -\frac{e^{-(s+a)t}}{s+a} \right]_{۰}^{\infty} \\
&= e^{-a} \frac{۱}{s+a}, \quad \text{Re}\{s\} > -a
\end{aligned}$$
<div dir="rtl">
پس در این مثال تبدیلهای یکطرفه و دو طرفه آشکارا تفاوت دارند. در واقع می‌توان $\mathcal{L}_+(x(t))$ را تبدیل دو طرفه $x_+(t)$ در نظر گرفت، قبلاً هم گفتیم که تبدیل دو طرفه $x_+(t)$ با تبدیل دو طرفه سیگنالی که در $t < ۰$ با $۰$ برابر و در $t \ge ۰$ با $x(t)$ است، یکی است.

<h2>مثال 9.35</h2>
<div dir="rtl"> 
تبدیل لاپلاس یکطرفه زیر را در نظر بگیرید:

$$\mathcal{X}(s) = \frac{1}{(s+1)(s+2)} \quad (9.179)$$
<div dir="rtl">
در مثال 9.9 عکس تبدیل دو طرفه تابعی مشابه معادله (9.179) را با چند ROC در نظر گرفتیم. برای تبدیل یکطرفه ROC باید یک راست-نیمصفحه، در سمت راست راست‌ترین قطب $\mathcal{X}(s)$ باشد؛ پس در این مثال ROC تمام نقاط دارای $\text{Re}\{s\} > -1$ است. می‌توانیم عکس این تبدیل را دقیقاً مانند مثال 9.9 حساب کنیم:

$$x(t) = [e^{-t} - e^{-2t}]u(t), \quad t > 0^- \quad (9.180)$$
<div dir="rtl">
در این معادله بر یکطرفه بودن تبدیل لاپلاس، با نشان دادن این که اطلاعات به دست آمده در مورد سیگنال تنها در $t > 0^-$ معتبر است، تأکید کرده‌ایم.

<h2>مثال 9.36</h2>
<div dir="rtl"> 
تبدیل لاپلاس یکطرفه زیر را در نظر بگیرید:

$$\mathcal{X}(s) = \frac{s^2 - 3}{s+2} \quad (9.181)$$
<div dir="rtl"> 
چون مرتبه صورت $\mathcal{X}(s)$ کوچکتر از مرتبه مخرج نیست، آن را به صورت زیر بسط می‌دهیم:

$$\mathcal{X}(s) = A + Bs + \frac{C}{s+2} \quad (9.182)$$
<div dir="rtl"> 
با برابر قرار دادن معادلات (9.181) و (9.182) و حذف مخرج به دست می‌آوریم:

$$s^2 - 3 = (A + Bs)(s+2) + C \quad (9.183)$$
<div dir="rtl"> 
با برابر قرار دادن ضریب جملات هم توان $s$ به دست می‌آوریم:

$$\mathcal{X}(s) = -2 + s + \frac{1}{s+2} \quad (9.184)$$
<div dir="rtl"> 
ناحیه همگرایی $\text{Re}\{s\} > -2$ است. با گرفتن عکس تبدیل لاپلاس جملات خواهیم داشت:

$$x(t) = -2\delta(t) + u_1(t) + e^{-2t}u(t), \quad t > ۰^- \quad (9.185)$$

<h2>خواص تبدیل لاپلاس یکطرفه</h2>
 <div dir="rtl">   
تبدیل لاپلاس یکطرفه نیز مانند تبدیل لاپلاس دو طرفه خواصی دارد که بسیاری از آنها مشابه خواص متناظر تبدیل لاپلاس دو طرفه است، ولی برخی از آنها تفاوتهایی قابل ملاحظه دارند. این خواص در جدول 9.3 خلاصه شده‌اند. ستونی برای مشخص کردن ROC تبدیل لاپلاس یکطرفه سیگنالها در نظر
نگرفته ایم،زیرا ROC هر تبدیل لاپلاس یک طرفه ای یک راست-نیم صفحه است. برای مثال ROC هر تبدیل لاپلاس یکطرفه گویایی سمت راست راست ترین قطب است.
مقایسه جدول ۹-۳ و جدول ۹-۱ تبدیل دو طرفه نشان می‌دهد که خواص خطی بودن، جابه‌جایی در حوزه $s$، تغییر مقیاس زمانی، مزدوج‌گیری و مشتق‌گیری در حوزه زمان با همتاهای دو طرفه شان مشابه اند. همچنین قضایای مقدار اولیه و مقدار نهایی بیان شده در بخش ۹-۲-۵ برای تبدیل لاپلاس یکطرفه نیز صادق‌اند. اشتقاق این خواص همانند همتاهای دو طرفه شان صورت می‌گیرد.
خاصیت کانولوشن تبدیل یکطرفه نیز کاملاً مشابۀ خاصیت کانولوشن تبدیل دو طرفه است. این خاصیت می‌گوید که اگر
 <div dir="rtl">  
     
$$ x_1(t) = x_2(t) = 0\quad t < 0$$
  <div dir="rtl">
     آنگاه:
      
$$x_1(t) * x_2(t) \stackrel{\mathcal{L}}{\longleftrightarrow} \mathcal{X}_1(s) \mathcal{X}_2(s) \quad (9.178)$$

معادله (9.187) بدون واسطه از خاصیت کانولوشن دو طرفه ناشی می‌شود. زیرا تحت شرایط معادله (9.186) تبدیلهای یکطرفه و دو طرفه $x_1(t)$ و $x_2(t)$ یکسان‌اند. پس ابزارهای تحلیلی و جبر تابع تبدیل که بر اساس این خاصیت تبدیل دو طرفه پی‌ریزی شدند، بدون تغییر برای تبدیلهای یکطرفه هم قابل اعمال‌اند، به شرطی که با سیستم‌های علی کار کنیم (که برایشان تبدیل یکطرفه و دو طرفه پاسخ ضربه یکسان است) و ورودیها در $t < ۰$ متحد با صفر باشند. خاصیت انتگرالگیری ذکر شده در جدول 9.3 نمونه‌ای از این مورد است: اگر در $t < ۰$، $x(t) = ۰$ آنگاه

$$\int_{0^-}^{t} x(\tau) d\tau = x(t) * u(t) \stackrel{\mathcal{L}}{\longleftrightarrow} \mathcal{X}(s) \{u(S)\} = \frac{1}{s} \mathcal{X}(s) \quad (9.188)$$
