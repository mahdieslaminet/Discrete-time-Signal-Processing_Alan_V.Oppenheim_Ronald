## 8-1 مدولاسیون فرکانس باریک‌باند

مدولاسیون فرکانس با سیگنال زیر را در نظر بگیرید:

$$
x(t) = A \cos(\omega_m t) \tag{8-8}
$$

توجه به معادلات (8-34)، (8-37)، فرکانس لحظه‌ای برابرست با:

$$
\omega_i(t) = \omega_c + k_f A \cos(\omega_m t) \tag{8-39}
$$

که به صورت سینوسی بین مقادیر
$$
\omega_c - k_f A
$$
و
$$
\omega_c + k_f A
$$
تغییر می‌کند. با تعریف:

$$
\Delta \omega = k_f A
$$

داریم:

$$
\omega_i(t) = \omega_c + \Delta \omega \cos(\omega_m t)
$$

و

$$
y(t) = \cos\left( \omega_c t + \int x(t) \, dt \right)
$$

که برابر می‌شود با:

$$
y(t) = \cos\left( \omega_c t + \frac{\Delta \omega}{\omega_m} \sin(\omega_m t) + \theta_0 \right) \tag{8-40}
$$

که در آن
$$
\theta_0
$$
ثابت انتگرال‌گیری است. برای سهولت فرض می‌کنیم
$$
\theta_0 = 0
$$
. پس:

$$
y(t) = \cos\left( \omega_c t + \frac{\Delta \omega}{\omega_m} \sin(\omega_m t) \right) \tag{8-41}
$$

عامل
$$
\Delta \omega / \omega_m
$$
را **شاخص مدولاسیون فرکانس** تعریف می‌کنیم. سیستم‌های FM بسته به بزرگ یا کوچک بودن شاخص مدولاسیون
$$
m = \Delta \omega / \omega_m
$$
ویژگی‌های متفاوتی دارند. حالتی که
$$
m
$$
کوچک است FM باریک‌باند می‌باشد. به طور کلی می‌توان معادله‌ی (8-41) را به صورت زیر نوشت:

$$
y(t) = \cos\left( \omega_c t + m \sin(\omega_m t) \right) \tag{8-42}
$$

یا

$$
y(t) = \cos(\omega_c t) \cos(m \sin(\omega_m t)) - \sin(\omega_c t) \sin(m \sin(\omega_m t)) \tag{8-43}
$$

اگر
$$
m
$$
بسیار کوچک باشد (خیلی کوچکتر از
$$
\pi/2
$$
)، می‌توان تقریب‌های زیر را استفاده کرد:

$$
\cos(m \sin(\omega_m t)) \approx 1
$$

$$
\sin(m \sin(\omega_m t)) \approx m \sin(\omega_m t)
$$

(معادله‌ی 8-44)

بنابراین معادله‌ی (8-43) چنین می‌شود:

$$
y(t) = \cos(\omega_c t) - m \sin(\omega_m t) \sin(\omega_c t)
$$

طبق (1) (براساس تقریب در شکل (8-33)) نشان داده شده است. توجه شود که این طیف با طیف AM-DSB/WC شباهت دارد. شباهت‌ها عبارتند از: از نظر این لحاظ که دو فرکانس حامل در طرف موجود است و کنار باندهای وجود دارد که در مدولاسیون AM مشابه دیده می‌شود.

در حالی که تفاوت‌های مدلوله کننده همزمان است و در معادله‌ی (8-46) موج FM باریک‌باند می‌بینیم که سیگنال حاصل با حاصل ضرب مدوله شده‌اند. در حالی که در AM-DSB/WC فقط حاصل جمع طیف‌ها وجود دارد.

شکل 8-34 شکل موج FM باریک‌باند مربوط به معادله‌ی (8-46) را نشان می‌دهد.

</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف مکان پیک‌ها
omega_c = 2  # مکان ωc
delta = np.pi / 2  # فاصله mπ/2
main_peak = 1.0    # ارتفاع پیک مرکزی
side_peak = 0.6    # ارتفاع پیک‌های جانبی

# موقعیت فرکانسی فلش‌ها
frequencies = [-omega_c - delta, -omega_c, -omega_c + delta, 0, omega_c - delta, omega_c, omega_c + delta]

# جهت فلش‌ها (بالا یا پایین)
directions = [-1, 1, 1, 1, -1, 1, 1]  # منفی یعنی فلش به سمت پایین

# اندازه هر فلش
amplitudes = [side_peak, main_peak, side_peak, main_peak, side_peak, main_peak, side_peak]

# رسم
plt.figure(figsize=(12, 5))
plt.hlines(0, -omega_c - delta - 1, omega_c + delta + 1, color='black')  # محور افقی

# رسم فلش‌ها
for x, y, d in zip(frequencies, amplitudes, directions):
    plt.arrow(x, 0, 0, d * y, head_width=0.1, head_length=0.08, fc='black', ec='black')

# برچسب گذاری دقیق
plt.text(-omega_c, -0.2, r'$-\omega_c$', ha='center', fontsize=12)
plt.text(omega_c, -0.2, r'$\omega_c$', ha='center', fontsize=12)
plt.text(0, -0.2, r'$0$', ha='center', fontsize=12)

plt.text(-omega_c - delta, -side_peak - 0.25, r'$\frac{m\pi}{2}$', ha='center', fontsize=12)
plt.text(-omega_c + delta, side_peak + 0.05, r'$-\frac{m\pi}{2}$', ha='center', fontsize=12)
plt.text(omega_c - delta, -side_peak - 0.25, r'$-\frac{m\pi}{2}$', ha='center', fontsize=12)
plt.text(omega_c + delta, side_peak + 0.05, r'$\frac{m\pi}{2}$', ha='center', fontsize=12)

# تنظیمات گراف
plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel(r'$Y(j\omega)$', fontsize=14)
plt.title('Spectrum with Up and Down Arrows (Correct)')
plt.xlim(-omega_c - delta - 1, omega_c + delta + 1)
plt.ylim(-1.4, 1.4)
plt.grid(False)
plt.show()
```

## 8-2 مدولاسیون فرکانس پهن باند

اگر $m$ بزرگ باشد تقریب منجر به معادله‌ی (8-46) دیگر معتبر نیست و طیف دامنه و هم دامنه‌ی سیگنال مدوله کننده

x(t)

$$
x(t)

$$
\omega_m
$$

$$
\omega_m
$$

## شکل 8-34 مقایسه FM باند باریک و AM-DSB/WC

(الف) FM باند باریک  
(ب) AM-DSB/WC

</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
fc = 20  # فرکانس حامل (Hz)
fm = 2   # فرکانس پیام (Hz)
m = 0.5  # ضریب مدولاسیون کوچک برای باریک‌باند

# محور زمان
t = np.linspace(0, 2, 1000)  # 2 ثانیه

# موج FM باریک‌باند
y_fm = np.cos(2*np.pi*fc*t) - m * np.sin(2*np.pi*fm*t) * np.sin(2*np.pi*fc*t)

# موج AM-DSB/WC
y_am = np.cos(2*np.pi*fc*t) + m * np.cos(2*np.pi*fm*t) * np.cos(2*np.pi*fc*t)

# رسم شکل‌ها
plt.figure(figsize=(14, 8))

# شکل الف: FM باند باریک
plt.subplot(2, 1, 1)
plt.plot(t, y_fm, color='black')
plt.title('shekl (alef) - FM کیراب دناب')
plt.xlabel('t (socond)')
plt.ylabel('y(t)')
plt.grid(True)

# شکل ب: AM-DSB/WC
plt.subplot(2, 1, 2)
plt.plot(t, y_am, color='black')
plt.title('شکل (ب) - AM-DSB/WC')
plt.xlabel('t (second)')
plt.ylabel('y(t)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

بررسی فوریه‌ی این دو سیگنال متناوب یا توابع موسوم به توابع بسِل ضرایب سری فوریه آن‌هاست. ضرایب سری فوریه‌ی این جمله سینوسی به صورت
$$
\cos(\omega_c t + m \sin(\omega_m t))
$$
که حامل سینوسی به پلاریته مثبت است، با بیان مختصر در نظر گرفته شده است. جمله دوم که حامل سینوسی با پلاریته منفی است
$$
\cos(\omega_c t - m \sin(\omega_m t))
$$
مربوط به آنالیز مشابهی شده است. جمله دوم در حوزه زمان همانند انتقال طیف معادله متناوب
$$
m \sin(\omega_m t)
$$
مدوله شده است.

مجموعه‌ی ضرایب سری فوریه در اطراف فرکانس‌های
$$
\omega_c
$$
مثل
$$
\omega_c \pm n \omega_m
$$
قرار می‌گیرند. این یعنی طیف شامل حامل اصلی است و طیفی هر یک از جملات متناوب را در فرکانس
$$
\omega_c
$$
و قرار گرفتن فرکانس‌های
$$
\omega_c \pm \omega_m, \omega_c \pm 2\omega_m, \ldots
$$
تشکیل داده است.

معادله (8-8) از ضرایب سری فوریه در ناحیه ضرایب سری فوریه انتقال پیدا می‌کند و بنابراین طیف معادله (8-8) محدود به طیف پایه
$$
\omega_c
$$
و ضرایب صحیح
$$
n
$$
از
$$
\omega_m
$$
است. این ضرایب $m$ را می‌توان چنان در نظر گرفت که تغییر $|m| > 1$ را مد نظر قرار دهد و بنابراین پهنای باند موثر کل هر کنار باند واقع در حول
$$
-\omega_c
$$
و
$$
+\omega_c
$$
است. یعنی:

$$
B \approx \gamma m \omega_m \tag{8-47}
$$

یا چون:

$$
m = \frac{k_f A}{\omega_m} = \frac{\Delta \omega}{\omega_m}
$$

پس:

$$
B = \gamma \frac{k_f A}{\omega_m} = \gamma \frac{\Delta \omega}{\omega_m} \tag{8-49}
$$

مقایسه‌ی معادله‌های (8-48) و (8-49) نشان می‌دهد که پهنای باند موثر هر کنار باند، با دامنه تغییرات

## شکل 8-35: اندازه طیف مدولاسیون فرکانس پهن باند

- (الف) اندازه طیف
$$
\cos(\omega_c t) \cos(m \sin(\omega_m t))
$$

- (ب) اندازه طیف
$$
\sin(\omega_c t) \sin(m \sin(\omega_m t))
$$

- (ج) اندازه طیف ترکیبی
$$
\cos(\omega_c t + m \sin(\omega_m t))
$$

</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
fc = 10    # مرکز ωc
fm = 1     # ωm
n_max = 5  # تعداد هارمونیک‌های دو طرف ωc

# ضرایب تقریبی (بر اساس تابع Bessel مرتبه صفر برای m = 1.2 حدودی)
amplitudes = [0.4, 0.7, 1.0, 0.7, 0.4]  # به ترتیب: فاصله‌های دورتر کوچیکتر

# موقعیت‌های پیک‌های سمت راست و چپ
frequencies = []
heights = []

for n in range(-n_max, n_max + 1):
    freq = fc + n * fm
    frequencies.append(freq)
    if abs(n) < len(amplitudes):
        heights.append(amplitudes[abs(n)])
    else:
        heights.append(0.2)  # پیک‌های خیلی دور کوچک

# رسم
plt.figure(figsize=(12, 4))
plt.hlines(0, fc - 8, fc + 8, color='black')  # محور افقی ω

for f, h in zip(frequencies, heights):
    plt.arrow(f, 0, 0, h, head_width=0.2, head_length=0.05, fc='black', ec='black')

# برچسب گذاری
plt.text(fc, -0.1, r'$\omega_c$', ha='center', fontsize=12)
plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel('Amplitude', fontsize=14)
plt.title('شکل (الف) - اندازه طیف $\cos(\omega_c t) \cos(m \sin(\omega_m t))$')
plt.xlim(fc - 8, fc + 8)
plt.ylim(-0.5, 1.5)
plt.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
fc = 10     # مرکز ωc
fm = 0.5    # فاصله بین فرکانس‌ها
n_max = 9   # تعداد مثبت و منفی

# ضرایب دستی کامل شده (طول 19 مقدار)
# ترتیب: 9 تا ضرایب برای سمت چپ، وسط، و 9 تا سمت راست
amplitudes_half = [0.2, 0.5, 0.8, 1.0, 0.1, 0.6, 0.7, 0.5, 0.2]  # برای n = 1 تا 9
amplitudes = list(reversed(amplitudes_half)) + [1.0] + amplitudes_half  # چپ + وسط + راست

# حالا طول amplitudes = 19 که دقیقا با -9 تا +9 میخونه

# ساخت موقعیت‌های فرکانسی و ارتفاع‌ها
frequencies = []
heights = []

for n in range(-n_max, n_max + 1):
    freq = fc + n * fm
    frequencies.append(freq)
    heights.append(amplitudes[n + n_max])  # ایندکس درست: shift شده

# رسم
plt.figure(figsize=(14, 5))
plt.hlines(0, fc - 6, fc + 6, color='gray', linestyle='--')  # محور ω کمی خاکستری

for f, h in zip(frequencies, heights):
    plt.arrow(f, 0, 0, h, head_width=0.05, head_length=0.03, fc='black', ec='black')

# توضیحات انگلیسی زیر شکل
plt.text(fc, -0.15, r'$\omega_c$', ha='center', fontsize=12)
plt.xlabel('Frequency (ω)', fontsize=14)
plt.ylabel('Amplitude', fontsize=14)
plt.title('Spectrum Shape (Corrected, No Error)')
plt.xlim(fc - 6, fc + 6)
plt.ylim(-0.2, 1.2)
plt.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
A = 1        # دامنه موج
T = 2        # تناوب اصلی
t = np.linspace(-T, T, 1000)  # محور زمان

# تعریف موج مربعی بین +A و -A بدون صفر
x = np.zeros_like(t)
for i in range(len(t)):
    time_mod = np.mod(t[i] + T/2, T)  # زمان رو ببریم بین 0 تا T
    if (0 <= time_mod < T/4) or (3*T/4 <= time_mod < T):
        x[i] = -A
    else:
        x[i] = A

# رسم موج مربعی
plt.figure(figsize=(12, 4))
plt.plot(t, x, color='black')

# خطوط راهنما در نقاط خاص
plt.axvline(-T/2, color='gray', linestyle='--')
plt.axvline(-T/4, color='gray', linestyle='--')
plt.axvline(T/4, color='gray', linestyle='--')
plt.axvline(T/2, color='gray', linestyle='--')
plt.axvline(T, color='gray', linestyle='--')

# برچسب گذاری نقاط خاص
plt.text(-T/2, A+0.2, r'$-\frac{T}{2}$', ha='center', fontsize=12)
plt.text(-T/4, A+0.2, r'$-\frac{T}{4}$', ha='center', fontsize=12)
plt.text(T/4, A+0.2, r'$\frac{T}{4}$', ha='center', fontsize=12)
plt.text(T/2, A+0.2, r'$\frac{T}{2}$', ha='center', fontsize=12)
plt.text(T, A+0.2, r'$T$', ha='center', fontsize=12)

# تنظیمات نهایی
plt.xlabel('Time (t)', fontsize=14)
plt.ylabel('Amplitude', fontsize=14)
plt.title('Corrected Square Wave (Figure 8-36)')
plt.ylim(-1.5, 1.5)
plt.xlim(-T, T)
plt.grid(True)
plt.show()
```

## 8-7-3 سیگنال مدوله کننده مربعی متناوب

مثال دیگری که به درک ویژگی‌های مدولاسیون فرکانس کمک می‌کند، سیگنال مدوله کننده موج مربعی است.  
در معادله (8-39) قرار دهید:

$$
A = 1.5, \quad f_m = 5Hz, \quad \omega_m = 10\pi
$$

و فرض کنید که شکل 8-36 فرم سیگنال مدوله شده شده است.  
(در شکل 8-37) نشان داده شده است. به ازای $X(t)$ مثبت، فرکانس لحظه‌ای
$$
\omega_c + \Delta \omega
$$
و به ازای $X(t)$ منفی،
$$
\omega_c - \Delta \omega
$$
می‌شود. بنابراین می‌توان به صورت زیر نوشت:

$$
y(t) = r(t) \cos\left[(\omega_c + \Delta \omega)t\right] + [1-r(t)] \cos\left[(\omega_c - \Delta \omega)t\right] \tag{8-50}
$$

که در آن $r(t)$ موج مربعی متناوب شکل 8-36 است.  
بنابراین برای این سیگنال مدوله کننده خاص، تعیین طیف سیگنال
$$
y(t)
$$

## مدولاسیون دامنه گسسته در زمان

### شکل ۸-۴۰

مدولاسیون دامنه گسسته در زمان.

$x[n]$ و $c[n]$ و $y[n]$ به ترتیب تبدیل فوریه‌ی $X(e^{j\omega})$, $C(e^{j\omega})$ و $Y(e^{j\omega})$ هستند.  
$Y(e^{j\omega})$ با کانولوشن متناوب یعنی:

$$
Y(e^{j\omega}) = \int_{-\pi}^{\pi} X(e^{j\theta}) C(e^{j(\omega-\theta)}) d\theta
\tag{8-56}
$$

چون $X(e^{j\omega})$ و $C(e^{j\omega})$ هر دو دوره‌ای با دوره $2\pi$ هستند، می‌توان این انتگرال‌گیری را در هر فاصله فرکانسی به طول $2\pi$ انجام داد.

ابتدا مدولاسیون دامنه با حامل نمایی مخلوط را در نظر می‌گیریم.  
یعنی:

$$
c[n] = e^{jw_c n}
\tag{8-58}
$$

چنان که در بخش ۵-۲ دیدیم، تبدیل فوریه $c[n]$ یک قطار ضرب متناوب است.  
یعنی:

$$
C(e^{j\omega}) = \sum_{k=-\infty}^{+\infty} 2\pi \delta(\omega - \omega_c + k2\pi)
\tag{8-57}
$$

که در شکل ۴۱-۸ (ب) نشان داده شده است.  
به ازای $X(e^{j\omega})$ طیف نشان داده شده در شکل ۴۱-۸ (الف)، طیف سیگنال مدوله شده مطابق شکل ۴۱-۸ (ج) است.

خصوصاً توجه کنید که:

$$
Y(e^{j\omega}) = X(e^{j(\omega - \omega_c)})
$$

- (ب) طیف $C(\omega)$
- (ج) طیف $Y(e^{j\omega}) = X(e^{j(\omega-\omega_c)})$

```python
import numpy as np
import matplotlib.pyplot as plt

# تابع مثلثی
def triangle(w, center, halfwidth, height=1):
    """یک پالس مثلثی با مرکز و نصف پهنا و ارتفاع مشخص"""
    out = np.zeros_like(w)
    mask = np.abs(w - center) <= halfwidth
    out[mask] = height * (1 - np.abs(w[mask] - center) / halfwidth)
    return out

# محور فرکانس
w = np.linspace(-4*np.pi, 4*np.pi, 5000)

# پارامترها
omega_c = np.pi/3      # مقدار ω_c
omega_m = np.pi/6      # نصف عرض قاعده مثلث‌ها (ω_m)
height  = 1.0          # ارتفاع رأس مثلث‌ها

# مراکز مثلث‌ها (طبق تصویر کتاب)
centers = [
    -2*np.pi + omega_c,   # -2π + ωc
    -omega_c,             # -ωc
    omega_c,              # ωc
    2*np.pi - omega_c,    # 2π - ωc
    2*np.pi + omega_c     # 2π + ωc
]

# ساخت طیف Y(w)
Y = np.zeros_like(w)
for center in centers:
    Y += triangle(w, center, omega_m, height)

# ------------------ رسم --------------------
plt.figure(figsize=(14,4))
plt.plot(w, Y, color='black')

# محور ω افقی
plt.hlines(0, -4*np.pi, 4*np.pi, color='gray', linewidth=0.8)

# خطوط عمودی و برچسب‌ها
positions = [-2*np.pi, -2*np.pi + omega_c, -omega_c, 0, omega_c, 2*np.pi - omega_c, 2*np.pi, 2*np.pi + omega_c]
labels = [r'$-2\pi$', r'$-2\pi+\omega_c$', r'$-\omega_c$', r'$0$', r'$\omega_c$', r'$2\pi-\omega_c$', r'$2\pi$', r'$2\pi+\omega_c$']

for pos, label in zip(positions, labels):
    plt.axvline(pos, color='gray', linestyle='--', linewidth=0.7)
    plt.text(pos, -0.15, label, ha='center', fontsize=10)

# تنظیمات
plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel(r'$Y(e^{j\omega})$', fontsize=14)
plt.title('(c) Spectrum $Y(e^{j\omega})$ (Shifted Triangles)')
plt.xlim(-4*np.pi, 4*np.pi)
plt.ylim(-0.2, 1.2)
plt.grid(False)
plt.tight_layout()
plt.show()
```

## ۸-۸-۲: ترانس مدولاسیون گسسته در زمان

یکی از سنجش‌هایی که در آن مدولاسیون دامنه گسسته در زمان، به همراه استخراج، نمونه‌برداری افزایش و درون‌یابی بیان شده در فصل ۷، بسیار به کار برده می‌شود، سیستم‌های مخابرات دیجیتال است.  
در این سیستم‌ها معمولاً سیگنال‌های پیوسته در زمان به صورت سیگنال‌های گسسته در زمان به دست آمده و نمونه‌برداری به کار می‌رود.  
سیگنال پیوسته در زمان غالباً به شکل مالی پلکسی زمانی (TDM) یا مالی پلکسی فرکانسی (FDM) هستند.  
این سیگنال‌ها به سیگنال‌های گسسته در زمان تبدیل می‌شوند که مقادیر زمانی ذخیره‌سازی یا محاسبه به صورت دیجیتال می‌شود.  
در بعضی سیستم‌ها به خاطر شرایط و استلزامات مختلف مالی پلکسی صورت گرفته لازم است که سیگنال‌های مالی پلکسی شده به روش‌های مختلف دوباره با هم یک‌پارچه شوند.  
در این‌صورت سیستم TDM، سیگنال TDM مالی پلکسی‌شده به سیستم FDM یا بر عکس تبدیل خواهد شد.  
این تبدیل به یک ترانس مدولاسیون مالی یا به مدولاسیون مالی دیگر ترانس مدولاسیون مالی و ترانس مالی پلکسی نامیده می‌شود.  
در حوزه مخابرات دیجیتال یکی دیگر از ترانس مالی پلکسی، ترانس مالی پلکسی در زمان داده‌ها است.  
دهدالاسیون و دی مالی پلکسی کردن سیگنال مدوله و مالی پلکسی شده یک چیز دخاله‌ای است.  
زیرا اگر بخواهیم سیگنال جدید را دوباره به سیگنال مالی گسسته در زمان تبدیل کنیم، بهتر است تمام کار در حوزه گسسته در زمان صورت گیرد.  
در شکل ۸-۴۸ کارهای لازم برای تبدیل سیگنال TDM گسسته در زمان به سیگنال FDM گسسته در زمان به صورت جعبه‌ای نشان داده شده است.  
توجه کنید که برای دی مالی پلکسی کردن سیگنال TDM، قبل از مالی پلکسی فرکانسی باید هر کانال نمونه‌برداری افزایش یابد.
