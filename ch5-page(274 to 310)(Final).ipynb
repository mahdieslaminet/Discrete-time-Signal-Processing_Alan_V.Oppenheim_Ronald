## ۵.۰ مقدمه

در فصل ۲، نمایش تبدیل فوریه برای سیگنال‌ها و سیستم‌های گسسته در زمان را توسعه دادیم، و در فصل ۳ این نمایش را به تبدیل Z گسترش دادیم.  
در هر دو فصل، تمرکز بر خود تبدیل‌ها و ویژگی‌های آن‌ها بود، و فقط مروری کوتاه بر جزئیات کاربردشان در تحلیل سیستم‌های خطی نامتغیر با زمان (LTI) داشتیم.

در این فصل، نمایش و تحلیل سیستم‌های LTI را با استفاده از تبدیل فوریه و تبدیل Z با جزئیات بیشتری توسعه می‌دهیم. این مطالب، پیش‌زمینه‌ی ضروری برای بحث در فصل ۶ درباره‌ی پیاده‌سازی سیستم‌های LTI و در فصل ۷ برای طراحی چنین سیستم‌هایی است.

همان‌طور که در فصل ۲ مطرح شد، یک سیستم LTI را می‌توان به‌طور کامل در حوزه‌ی زمان، با پاسخ ضربه‌ای آن \( h[n] \) مشخص کرد، به‌طوری که خروجی \( y[n] \) برای ورودی دلخواه \( x[n] \) از رابطه‌ی کانولوشن زیر به‌دست می‌آید:

$$
y[n] = \sum_{k=-\infty}^{\infty} x[k] h[n - k] \tag{5.1}
$$

از طرف دیگر، از آنجا که پاسخ فرکانسی و پاسخ ضربه‌ای از طریق تبدیل فوریه با هم مرتبط‌اند، پاسخ فرکانسی—در صورتی که وجود داشته باشد (یعنی \( H(z) \) ناحیه همگرایی‌ای داشته باشد که شامل \( $z = e^{j\omega}$ \) باشد)—نیز یک نمایش کامل برای سیستم‌های LTI ارائه می‌دهد.

در فصل ۳، تبدیل Z را به‌عنوان تعمیمی از تبدیل فوریه توسعه دادیم. تبدیل Z خروجی یک سیستم LTI را به تبدیل Z ورودی و تبدیل Z پاسخ ضربه‌ای سیستم مربوط می‌سازد:

$$
Y(z) = H(z) X(z) \tag{5.2}
$$

که در آن \( Y(z) \)، \( X(z) \)، و \( H(z) \) به‌ترتیب تبدیل Z سیگنال‌های \( y[n] \)، \( x[n] \)، و \( h[n] \) هستند، و دارای ناحیه‌های همگرایی مناسب‌اند.  
\( H(z) \) معمولاً به‌عنوان تابع انتقال سیستم شناخته می‌شود.

</div>

  <div dir="rtl">

از آنجا که یک دنباله و تبدیل Z آن یک زوج منحصربه‌فرد تشکیل می‌دهند، نتیجه می‌شود که هر سیستم LTI (خطی تغییرناپذیر با زمان) به‌طور کامل توسط تابع سیستم آن مشخص می‌شود، البته با فرض همگرایی.  

هم پاسخ فرکانسی، که معادل مقدار تابع سیستم روی دایره واحد است، و هم تابع سیستم که به‌طور کلی‌تر به‌عنوان تابعی از متغیر مختلط \( z \) تعریف می‌شود، در تحلیل و نمایش سیستم‌های LTI بسیار مفید هستند، زیرا می‌توانیم بسیاری از ویژگی‌های پاسخ سیستم را به‌راحتی از آن‌ها استنتاج کنیم.

</div>

## 5.1 LTI پاسخ فرکانسی سیستم‌های

پاسخ فرکانسی $ H(e^{j\omega}) $ یک سیستم LTI به عنوان بهره مختلط (مقدار ویژه) که سیستم به ورودی نمایی مختلط $ e^{j\omega t} $ اعمال می‌کند، تعریف شده است (بخش 2.6 را ببینید). 
علاوه بر این، همان‌طور که در بخش 2.9.6 بحث شد، از آنجا که تبدیل فوریه یک دنباله، تجزیه‌ای به صورت ترکیب خطی از نمایی‌های مختلط را نشان می‌دهد، تبدیل‌های فوریه ورودی و خروجی سیستم به‌صورت زیر به هم مرتبط هستند:

<div dir="ltr">

$$
Y(e^{j\omega}) = H(e^{j\omega}) X(e^{j\omega}) \quad \tag{5.3}
$$

</div>

که در آن $ X(e^{j\omega}) $ و $ Y(e^{j\omega}) $ به ترتیب تبدیل فوریه ورودی و خروجی سیستم هستند.

### 5.1.1 فاز پاسخ فرکانسی و تأخیر گروهی

پاسخ فرکانسی به طور کلی در هر فرکانس یک عدد مختلط است. وقتی پاسخ فرکانسی به فرم قطبی بیان شود، دامنه و فاز تبدیل‌های فوریه ورودی و خروجی سیستم به صورت زیر به هم مرتبط هستند:

<div dir="ltr">

$$
|Y(e^{j\omega})| = |H(e^{j\omega})| \cdot |X(e^{j\omega})| \quad (5.4a)
$$

$$
\angle Y(e^{j\omega}) = \angle H(e^{j\omega}) + \angle X(e^{j\omega}) \quad (5.4b)
$$

</div>

که در آن $ |H(e^{j\omega})| $ نشان‌دهنده پاسخ دامنه یا بهره سیستم است و $ \angle H(e^{j\omega}) $ نشان‌دهنده پاسخ فاز یا شیفت فاز سیستم است.

اثرات دامنه و فاز بیان‌شده در معادلات بالا می‌توانند مطلوب باشند، اگر سیگنال ورودی به شکل مفیدی اصلاح شود، یا نامطلوب باشند، اگر سیگنال ورودی به صورت مخربی تغییر کند. در حالت دوم، معمولاً به اثرات یک سیستم LTI روی یک سیگنال به عنوان اعوجاج دامنه و اعوجاج فاز اشاره می‌کنیم.

فاز هر عدد مختلط به‌طور یکتا تعریف نشده است، زیرا هر مضرب صحیحی از $ 2\pi $ را می‌توان بدون تأثیرگذاری بر عدد اصلی به آن افزود. وقتی فاز به‌صورت عددی با استفاده از زیرروال تابع تانژانت معکوس محاسبه می‌شود، مقدار اصلی آن معمولاً به‌دست می‌آید. مقدار اصلی فاز $ H(e^{j\omega}) $ را به‌صورت $ \text{ARG}[H(e^{j\omega})] $ نشان می‌دهیم، که در بازه زیر قرار دارد:

<div dir="ltr">

$$
-\pi < \text{ARG}[H(e^{j\omega})] \leq \pi \quad (5.5)
$$

</div>

هر زاویه‌ی دیگری که مقدار مختلط صحیح تابع $ H(e^{j\omega}) $ را بدهد، می‌تواند به صورت مقدار اصلی زیر نمایش داده شود:

<div dir="ltr">

$$
\angle H(e^{j\omega}) = \text{ARG}[H(e^{j\omega})] + 2\pi r(\omega) \quad (5.6)
$$

</div>

که در آن $ r(\omega) $ یک عدد صحیح مثبت یا منفی است که می‌تواند در هر مقدار $ \omega $ متفاوت باشد. در بسیاری از موارد، مقدار اصلی دارای ناپیوستگی‌هایی به اندازه $ 2\pi $ رادیان در تابع $ \omega $ است. این موضوع در شکل زیر نشان داده شده است:

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

omega = np.linspace(-np.pi, np.pi, 400)
phase_continuous = -2 * np.sin(omega) - 2 * omega
phase_principal = np.mod(phase_continuous + np.pi, 2*np.pi) - np.pi
r_values = np.round((phase_continuous - phase_principal) / (2 * np.pi))

fig, axs = plt.subplots(3, 1, figsize=(6, 8))

axs[0].plot(omega, phase_continuous, label='arg[H(e^{j\omega})]')
axs[0].set_ylabel('arg[H(e^{j\omega})]')
axs[0].set_xticks([-np.pi, 0, np.pi])
axs[0].set_xticklabels(['-π', '0', 'π'])
axs[0].grid()

axs[1].plot(omega, phase_principal, label='ARG[H(e^{j\omega})]', color='r')
axs[1].set_ylabel('ARG[H(e^{j\omega})]')
axs[1].set_xticks([-np.pi, 0, np.pi])
axs[1].set_xticklabels(['-π', '0', 'π'])
axs[1].grid()

axs[2].step(omega, r_values, label='r(\omega)', color='g', where='mid')
axs[2].set_ylabel('r(\omega)')
axs[2].set_xlabel('ω')
axs[2].set_xticks([-np.pi, 0, np.pi])
axs[2].set_xticklabels(['-π', '0', 'π'])
axs[2].grid()

plt.tight_layout()
plt.show()
```

<div dir="rtl">


هر زاویه‌ی دیگری که مقدار مختلط صحیح تابع $ H(e^{j\omega}) $ را بدهد، می‌تواند به صورت مقدار اصلی زیر نمایش داده شود:

<div dir="ltr">

$$
\angle H(e^{j\omega}) = \text{ARG}[H(e^{j\omega})] + 2\pi r(\omega) \quad (5.6)
$$

</div>

که در آن $ r(\omega) $ یک عدد صحیح مثبت یا منفی است که می‌تواند در هر مقدار $ \omega $ متفاوت باشد. 
ما به طور کلی در این متن از نماد زاویه در سمت چپ معادله‌ی بالا برای نمایش فاز نامعین استفاده خواهیم کرد، زیرا مقدار $ r(\omega) $ تا حدودی اختیاری است.

در بسیاری از موارد، مقدار اصلی دارای ناپیوستگی‌هایی به اندازه $ 2\pi $ رادیان در تابع $ \omega $ است. این موضوع در شکل 5.1 نشان داده شده است که یک تابع فاز پیوسته $ \arg[H(e^{j\omega})] $ و مقدار اصلی آن $ \text{ARG}[H(e^{j\omega})] $ را در دامنه‌ای مشخص نمایش می‌دهد.

یک نمایش مفید دیگر از فاز، از طریق تاخیر گروهی $ \tau(\omega) $ تعریف می‌شود به صورت:

<div dir="ltr">

$$
\tau(\omega) = \text{grd}[H(e^{j\omega})] = - \frac{d}{d\omega} \text{ARG}[H(e^{j\omega})]. \quad (5.7)
$$

</div>

شایان ذکر است که از آنجا که مشتق $ \arg[H(e^{j\omega})] $ و $ \text{ARG}[H(e^{j\omega})] $ فقط در نقاط ناپیوستگی اختلاف دارند، تاخیر گروهی را می‌توان با مشتق‌گیری از مقدار اصلی به دست آورد، به جز در ناپیوستگی‌ها.

به طور مشابه، می‌توان تاخیر گروهی را بر حسب فاز نامعین به شکل زیر بیان کرد:

<div dir="ltr">

$$
\text{grd}[H(e^{j\omega})] = - \frac{d}{d\omega} \angle H(e^{j\omega}). \quad (5.8)
$$

</div>

که در آن ضربه‌های ناشی از ناپیوستگی‌های به اندازه‌ی $ 2\pi $ در $ H(e^{j\omega}) \angle $ نادیده گرفته می‌شوند.

برای درک اثر فاز و به‌ویژه تأخیر گروهی یک سیستم خطی، ابتدا سیستم تأخیر ایده‌آل را در نظر می‌گیریم. پاسخ ضربه‌ای این سیستم به‌صورت زیر است:

<div dir="ltr">

$$
 h_{\text{id}}[n] = \delta[n - n_d]. \quad (5.9)
$$

</div>

و پاسخ فرکانسی آن برابر است با:

<div dir="ltr">

$$
H_{\text{id}}(e^{j\omega}) = e^{-j\omega n_d}. \quad (5.10)
$$

</div>

یا به صورت زیر بیان می‌شود:

<div dir="ltr">

$$
|H_{\text{id}}(e^{j\omega})| = 1, \quad (5.11a)
$$

$$
\angle H_{\text{id}}(e^{j\omega}) = -\omega n_d, \quad |\omega| < \pi. \quad (5.11b)
$$

</div>

که در آن، تناوبیت $ 2\pi $ در $ \omega $ مشاهده می‌شود. از معادله‌ی بالا نتیجه می‌گیریم که تاخیر زمانی (یا پیش‌افتادن اگر $ n_d < 0 $ باشد) با شیب فاز در ارتباط است.

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات اولیه
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_title("Pole-Zero Plot (Double-order Poles and Zeros)")
ax.set_xlabel("Real Part")
ax.set_ylabel("Imaginary Part")
ax.grid(True)
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)
ax.set_aspect('equal')

# رسم دایره واحد
theta = np.linspace(0, 2 * np.pi, 1000)
unit_circle = np.exp(1j * theta)
ax.plot(unit_circle.real, unit_circle.imag, 'k--', label="Unit Circle")

# تعریف پارامترها
r = 0.95
phi = 0.15 * np.pi
N = 4  # تعداد قطب‌ها و صفرها

# محاسبه قطب‌ها و صفرها مرتبه 2
poles = [r * np.exp(1j * (phi + 2 * np.pi * k / N)) for k in range(N)]
poles += [np.conj(p) for p in poles]  # قرینه‌های مختلط

zeros = [1 / p for p in poles]  # صفرهای متناظر

# رسم قطب‌ها
for p in poles:
    ax.plot(p.real, p.imag, 'x', color='blue', markersize=10)
    ax.text(p.real + 0.02, p.imag + 0.02, '2', fontsize=8, color='blue')

# رسم صفرها
for z in zeros:
    ax.plot(z.real, z.imag, 'o', markerfacecolor='none', markeredgecolor='orange', markersize=10)
    ax.text(z.real + 0.02, z.imag + 0.02, '2', fontsize=8, color='orange')

# محدوده نمایش
ax.set_xlim([-1.2, 1.2])
ax.set_ylim([-1.2, 1.2])

plt.legend(["Unit Circle", "Poles", "Zeros"])
plt.show()
```

## صفحه 281

پاسخ فاز سیستم خطی و نامتغیر با زمان $H(e^{jω})$ اغلب توسط تابع فاز θ(ω) تعریف می‌شود:

<div dir="ltr">

$$
H(e^{jω}) = |H(e^{jω})|e^{jθ(ω)} \tag{5.18}
$$

</div>

تابع θ(ω) اغلب یک تابع غیرفرد است، حتی اگر $|H(e^{jω})|$ زوج باشد. از آنجا که فاز به صورت زاویه در نظر گرفته می‌شود، تابع $θ(ω)$ فقط تا مضربی از 2π مشخص است.

برای تحلیل سیستم‌ها، اغلب فاز "بازشده" استفاده می‌شود که به کمک عملیات ریاضی پیوستگی نسبی در $θ(ω)$ ایجاد می‌کند.

در شکل 5.3، بخش (a) مقدار اصلی فاز را نشان می‌دهد و بخش (b) فاز بازشده را. همانطور که مشاهده می‌شود، فاز اصلی دارای پرش‌هایی در نقاط مختلف ω است که در فاز بازشده حذف شده‌اند.

اگر تابع فاز $θ(ω)$ خطی باشد، به صورت:

<div dir="ltr">

$$
θ(ω) = -ωn_0
$$

</div>

در این صورت، پاسخ ضربی تبدیل فوریه برابر است با:

<div dir="ltr">

$$
H(e^{jω}) = e^{-jωn_0} \tag{5.19}
$$

</div>

که نشان‌دهنده یک تأخیر زمانی n₀ نمونه‌ای است. این نوع سیستم‌ها سیگنال ورودی را بدون اعوجاج تغییر شکل نمی‌دهند بلکه تنها آن را به تأخیر می‌اندازند.

</div>


```python

import numpy as np
import matplotlib.pyplot as plt

# Create signal x[n]
M = 60
def w(n, M):
    if 0 <= n <= M:
        return 0.54 - 0.46 * np.cos(2 * np.pi * n / M)
    return 0

# Frequencies
omega1 = 0.2 * np.pi
omega2 = 0.4 * np.pi
omega3 = 0.8 * np.pi
n = np.arange(0, 300)
x1 = np.array([np.cos(omega1 * (i - 150)) * w(i - 150 + 30, M) for i in n])
x2 = np.array([np.cos(omega2 * (i - 90)) * w(i - 90 + 30, M) for i in n])
x3 = np.array([np.cos(omega3 * (i - 30)) * w(i - 30 + 30, M) for i in n])
x = x3 + x1 + x2

# Plot waveform
plt.figure(figsize=(10, 4))
plt.plot(n, x)
plt.title("(a) Waveform of signal x[n]")
plt.xlabel("Sample number (n)")
plt.ylabel("x[n]")
plt.grid(True)
plt.tight_layout()
plt.show()

# DTFT using zero-padding
from scipy.fft import fft, fftshift
X = fftshift(fft(x, 2048))
X_mag = np.abs(X)
w = np.linspace(-np.pi, np.pi, len(X_mag))

# Plot DTFT magnitude
plt.figure(figsize=(10, 4))
plt.plot(w, X_mag)
plt.title("(b) Magnitude of DTFT of x[n]")
plt.xlabel("ω")
plt.ylabel("|X(e^{jω})|")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftshift

# تنظیمات اولیه
M = 60
n = np.arange(0, 3*M)
w = 0.54 - 0.46 * np.cos(2*np.pi*np.arange(0, M)/M)

# فرکانس‌ها
w1 = 0.2 * np.pi
w2 = 0.4 * np.pi
w3 = 0.8 * np.pi

# تولید سیگنال‌های سینوسی ضرب در پنجره
x1 = np.zeros_like(n)
x2 = np.zeros_like(n)
x3 = np.zeros_like(n)

x1[0:M] = np.cos(w3*np.arange(0, M)) * w
x2[M:2*M] = np.cos(w2*np.arange(0, M)) * w
x3[2*M:3*M] = np.cos(w1*np.arange(0, M)) * w

# سیگنال ترکیبی
x = x3 + x1 + x2

# تعریف پاسخ فرکانسی سیستم H(e^{jω})
omega = np.linspace(-np.pi, np.pi, 1024)
H = np.exp(1j * (4 * np.sin(omega))) * np.exp(-1j * omega**2)

# اعمال سیستم به سیگنال ورودی
X = fftshift(fft(x, 1024))
Y = X * H
y = np.real(np.fft.ifft(np.fft.ifftshift(Y)))

# نمایش شکل موج y[n]
plt.figure(figsize=(10, 4))
plt.plot(n, y[:len(n)])
plt.title("Waveform of output signal y[n]")
plt.xlabel("Sample number (n)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()

# محاسبه و نمایش طیف DTFT خروجی
Y_mag = np.abs(Y)

plt.figure(figsize=(10, 4))
plt.plot(omega, Y_mag)
plt.title("Magnitude of DTFT of y[n]")
plt.xlabel("ω")
plt.ylabel("|Y(e^{jω})|")
plt.grid(True)
plt.tight_layout()
plt.show()
```

همان‌طور که در بخش 3.5 دیدیم، با اعمال تبدیل z به دو طرف معادله (5.19) و استفاده از خاصیت خطی بودن (بخش 3.4.1) و خاصیت انتقال زمانی (بخش 3.4.2)، نتیجه می‌گیریم که برای سیستمی که ورودی و خروجی آن یک معادله تفاضلی از فرم معادله (5.19) را ارضا می‌کند، تابع سیستم دارای فرم جبری زیر خواهد بود:

<div dir="ltr">

$$
H(z) = \frac{Y(z)}{X(z)} = \frac{\sum_{k=0}^{M} b_k z^{-k}}{\sum_{k=0}^{N} a_k z^{-k}} \tag{5.20}
$$

</div>

در معادله (5.20)، \( H(z) \) به فرم نسبت چندجمله‌ای‌هایی بر حسب \( z^{-1} \) است، زیرا معادله (5.19) از دو ترکیب خطی از ترم‌های تاخیر یافته تشکیل شده است.
اگرچه معادله (5.20) را می‌توان به‌گونه‌ای بازنویسی کرد که چندجمله‌ای‌ها بر حسب توان‌های \( z \) نوشته شوند به‌جای \( z^{-1} \)، اما معمولاً این کار انجام نمی‌شود.
همچنین، اغلب راحت‌تر است که معادله (5.20) به فرم فاکتورگیری شده زیر نوشته شود:

<div dir="ltr">

$$
H(z) = \left( \frac{b_0}{a_0} \right) \frac{\prod_{k=1}^{M} (1 - c_k z^{-1})}{\prod_{k=1}^{N} (1 - d_k z^{-1})} \tag{5.21}
$$

</div>

هر یک از فاکتورهای \( (1 - c_k z^{-1}) \) در صورت کسر یک صفر در \( z = c_k \) و یک صفر در \( z = 0 \) تولید می‌کند.
به‌طور مشابه، هر یک از فاکتورهای \( (1 - d_k z^{-1}) \) در مخرج کسر یک صفر در \( z = 0 \) و یک قطب در \( z = d_k \) تولید می‌کند.

یک رابطه مستقیم بین تابع سیستم و معادله تفاضلی متناظر آن وجود دارد.
به‌طور خاص، چندجمله‌ای صورت در معادله (5.20) دارای ضرایب و ساختار جبری یکسان با طرف راست معادله (5.19) است
(ترم‌هایی به فرم \( b_k x[n-k] \))، در حالی که چندجمله‌ای مخرج در معادله (5.20) دارای ضرایب و ساختار جبری یکسان با طرف چپ معادله (5.19) است
(ترم‌هایی به فرم \( a_k y[n-k] \)). بنابراین، با داشتن یا تابع سیستم به فرم معادله (5.20) یا معادله تفاضلی به فرم (5.19)، به راحتی می‌توان دیگری را به‌دست آورد.

## 5.2.1 پایداری و علیت


برای به‌دست آوردن معادله (5.20) از معادله (5.19)، فرض کردیم که سیستم خطی و تغییرناپذیر با زمان است تا بتوان از معادله (5.2) استفاده کرد،
اما هیچ فرضی درباره پایداری یا علیت نکردیم. به‌صورت متناظر، از معادله تفاضلی می‌توان فرم جبری تابع سیستم را به‌دست آورد، اما ناحیه همگرایی مشخص نمی‌شود.

به‌طور خاص، ناحیه همگرایی تابع \( H(z) \) از فرآیند استخراج معادله (5.20) به‌دست نمی‌آید، زیرا تنها چیزی که برای اعتبار معادله (5.20) نیاز است این است که \( X(z) \) و \( Y(z) \) دارای ناحیه‌های همگرایی مشترک باشند.
این موضوع با واقعیت مطرح‌شده در فصل 2 نیز سازگار است، که معادله تفاضلی پاسخ ضربه‌ای یک سیستم خطی و تغییرناپذیر با زمان را به‌طور یکتا مشخص نمی‌کند.

اگر فرض کنیم سیستم علی باشد، نتیجه می‌گیریم که \( h[n] \) باید دنباله‌ای از سمت راست باشد (یعنی برای \( n < 0 \) صفر باشد)،

## 5.2.2 سیستم‌های معکوس


برای یک سیستم خطی و تغییرناپذیر با زمان با تابع سیستم \( H(z) \)، سیستم معکوس متناظر سیستمی است با تابع سیستم \( H_i(z) \) به‌طوری‌که اگر آن را با \( H(z) \) به‌صورت سری ترکیب کنیم، تابع سیستم کلی برابر با واحد خواهد شد. یعنی:

<div dir="ltr">

$$
G(z) = H(z) H_i(z) = 1 \\ \tag{5.29}
$$

</div>

از این رابطه نتیجه می‌گیریم:

<div dir="ltr">

$$
H_i(z) = \frac{1}{H(z)} \tag{5.30}
$$

</div>

معادل زمانی این رابطه به‌صورت:

<div dir="ltr">

$$
g[n] = h[n] * h_i[n] = \delta[n] \tag{5.31}
$$

</div>

از معادله (5.30)، پاسخ فرکانسی سیستم معکوس (اگر وجود داشته باشد) به شکل زیر خواهد بود:

<div dir="ltr">

$$
H_i(e^{j\omega}) = \frac{1}{H(e^{j\omega})} \tag{5.32}
$$
</div>


</div>

<div dir="rtl">

## مثال ۵.۳: سیستم معکوس برای سیستم مرتبه اول

یعنی \( $H_i(e^{j\omega})$ \) عکس \( $H(e^{j\omega})$ \) است. به‌طور معادل، اندازه لگاریتمی، فاز، و تأخیر گروهی سیستم معکوس منفی مقادیر متناظر برای سیستم اصلی هستند. همهٔ سیستم‌ها معکوس ندارند. برای مثال، فیلتر پایین‌گذر ایده‌آل معکوس ندارد. هیچ راهی برای بازیابی مؤلفه‌های فرکانسی که توسط چنین فیلتری صفر شده‌اند، وجود ندارد.

بسیاری از سیستم‌ها دارای معکوس هستند و کلاس سیستم‌هایی با تابع تبدیل گویا، یک مثال بسیار مفید و جالب را فراهم می‌کنند. به طور خاص، در نظر بگیرید:

<div dir="ltr">

$$
H(z) = \left( \frac{b_0}{a_0} \right) \frac{ \prod_{k=1}^M \left(1 - c_k z^{-1}\right) }{ \prod_{k=1}^N \left(1 - d_k z^{-1} \right) }\tag{5.33}
$$

</div>

که دارای صفرهایی در \( $z = c_k$ \) و قطب‌هایی در \( z = d_k \) است، به‌علاوه صفرها یا قطب‌های ممکن در \( z = 0 \) و \( z = \infty \). در این‌صورت:

<div dir="ltr">

$$
H_i(z) = \left( \frac{a_0}{b_0} \right) \frac{ \prod_{k=1}^N \left(1 - d_k z^{-1}\right) }{ \prod_{k=1}^M \left(1 - c_k z^{-1}\right) }\tag{5.34}
$$

</div>

یعنی، قطب‌های \( H_i(z) \) همان صفرهای \( H(z) \) هستند و بالعکس. اکنون این پرسش مطرح می‌شود که چه ناحیه‌ای از همگرایی (ROC) باید برای \( H_i(z) \) در نظر گرفت. پاسخ این پرسش توسط قضیهٔ هم‌پیچی (convolution theorem) که در رابطهٔ (5.31) بیان شده، ارائه می‌شود. برای برقرار بودن رابطهٔ (5.31)، ناحیهٔ همگرایی \( H(z) \) و \( H_i(z) \) باید هم‌پوشانی داشته باشند. اگر \( H(z) \) علی باشد، ناحیهٔ همگرایی آن به صورت زیر است:

<div dir="ltr">

$$
|z| > \max_k |d_k|\tag{5.35}
$$

</div>

بنابراین، هر ناحیهٔ همگرایی برای \( H_i(z) \) که با ناحیهٔ فوق هم‌پوشانی داشته باشد، یک ناحیهٔ همگرایی معتبر برای \( H_i(z) \) است. مثال‌های ۵.۳ و ۵.۴ برخی از این حالات را نشان می‌دهند.



### مثال ۵.۳: سیستم معکوس برای سیستم مرتبه اول

فرض کنید:

<div dir="ltr">

$$
H(z) = \frac{1 - 0.5z^{-1}}{1 - 0.9z^{-1}}
$$

</div>

با ناحیهٔ همگرایی:

<div dir="ltr">

$$
|z| > 0.9
$$

</div>

در این صورت:

<div dir="ltr">

$$
H_i(z) = \frac{1 - 0.9z^{-1}}{1 - 0.5z^{-1}}
$$

</div>

از آن‌جا که \( H_i(z) \) فقط یک قطب دارد، تنها دو حالت برای ناحیهٔ همگرایی آن وجود دارد و تنها گزینه‌ای که با \( |z| > 0.9 \) هم‌پوشانی دارد، \( |z| > 0.5 \) است. بنابراین، پاسخ ضربه‌ای سیستم معکوس به صورت زیر است:

<div dir="ltr">

$$
h_i[n] = (0.5)^n u[n] - 0.9 (0.5)^{n-1} u[n - 1]
$$

</div>

در این حالت، سیستم معکوس هم علی است و هم پایدار.

</div>

## مثال ۵.۴ – وارون برای سیستمی با صفر در ناحیه همگرایی

فرض کنید تابع تبدیل به‌صورت زیر باشد:

<div dir="ltr">

$$
\text{(۵٫۳۴)} \quad H(z) = \frac{z^{-1} - ۰٫۵}{۱ - ۰٫۹z^{-1}}, \quad |z| > ۰٫۹
$$

</div>

تابع وارون آن برابر است با:

<div dir="ltr">

$$
\text{(۵٫۳۵)} \quad H_i(z) = \frac{۱ - ۰٫۹z^{-1}}{z^{-1} - ۰٫۵} = \frac{-۲ + ۱٫۸z^{-1}}{۱ - ۲z^{-1}}
$$

</div>

همان‌طور که پیش‌تر گفته شد، دو ناحیه‌ی همگرایی ممکن برای این تابع جبری وجود دارد:

<div dir="ltr">

$$
|z| < ۲ \quad \text{و} \quad |z| > ۲
$$

</div>


اما در این حالت، هر دو ناحیه با ناحیه اصلی یعنی $|z| > ۰٫۹$ هم‌پوشانی دارند، پس هر دو وارون معتبر هستند.

پاسخ ضربه‌ای متناظر برای ناحیه‌ی همگرایی $|z| < ۲$ برابر است با:

<div dir="ltr">

$$
h_{i1}[n] = ۲ (۲)^n u[-n - ۱] - ۱٫۸ (۲)^n u[-n]
$$

</div>

و برای ناحیه‌ی همگرایی $|z| > ۲$ خواهیم داشت:

<div dir="ltr">

$$
h_{i2}[n] = -۲ (۲)^n u[n] + ۱٫۸ (۲)^n u[n - ۱]
$$

</div>

می‌بینیم که $h_{i1}[n]$ پایدار ولی غیرعلّی است، درحالی‌که $h_{i2}[n]$ ناپایدار ولی علّی است. به‌طور نظری، اگر یکی از این دو سیستم وارون با $H(z)$ به صورت زنجیره‌ای ترکیب شود، نتیجه سیستم همانی خواهد بود.

### تعمیم:

اگر $H(z)$ یک سیستم علّی با صفرهایی در $c_k$ برای $k = ۱, ۲, \ldots, M$ باشد، آنگاه سیستم وارون آن علّی خواهد بود اگر و تنها اگر ناحیه همگرایی زیر برای آن انتخاب شود:

<div dir="ltr">

$$
|z| > \max_k |c_k|
$$

</div>

اگر بخواهیم سیستم وارون علاوه بر علّی بودن، پایدار هم باشد، آنگاه ناحیه همگرایی آن باید دایره واحد را شامل شود:

<div dir="ltr">

$$
\max_k |c_k| < ۱
$$

</div>

یعنی همه‌ی صفرهای تابع تبدیل باید درون دایره واحد قرار داشته باشند.

پس یک سیستم زمان‌ناور خطی، پایدار و علّی، دارای وارون پایدار و علّی خواهد بود اگر و تنها اگر هم قطب‌ها و هم صفرهای آن درون دایره‌ی واحد قرار داشته باشند.

به این نوع سیستم‌ها، **سیستم‌های حداقل فاز** گفته می‌شود و در بخش ۵٫۶ به‌تفصیل بررسی خواهند شد.

## ۵٫۲٫۳ پاسخ ضربه‌ای برای توابع گویای سیستم

بحث بسط کسر جزئی (ارائه‌شده در بخش ۳٫۳٫۲) را می‌توان برای تابع تبدیل $H(z)$ به‌کار برد تا یک بیان کلی برای پاسخ ضربه‌ای سیستمی که تابع تبدیل آن به شکل زیر باشد، به‌دست آورد:

<div dir="ltr">

$$
\text{(۵٫۳۶)} \quad H(z) = \sum_{r=۰}^{M-N} B_r z^{-r} + \sum_{k=۱}^{N} \frac{A_k}{۱ - d_k z^{-1}}
$$

</div>

## ادامه‌ی ۵٫۲٫۳ — پاسخ ضربه‌ای برای توابع گویای سیستم

بسط کسر جزئی تابع تبدیل، مطابق رابطه‌ی زیر خواهد بود:

<div dir="ltr">

$$
h[n] = \sum_{r=0}^{M-N} B_r \delta[n - r] + \sum_{k=1}^{N} A_k d_k^n u[n] \qquad \text{(5.37)}
$$

</div>

جمع اول تنها زمانی وجود دارد که $M \geq N$ باشد.




### دسته‌بندی سیستم‌های خطی نامتغیر با زمان (LTI)

برای تحلیل سیستم‌های خطی نامتغیر با زمان، تقسیم آن‌ها به دو دسته مفید است:

### ۱. سیستم‌هایی که پاسخ ضربه‌ای آن‌ها بی‌نهایت است (IIR)

در این دسته، حداقل یک جمله‌ی ناصفر از صورت تابع تبدیل توسط صفر مخرج حذف نشده است. در این حالت، $h[n]$ حداقل شامل یک جمله از نوع $A_k (d_k)^n u[n]$ خواهد بود. بنابراین، $h[n]$ طول محدود ندارد (یعنی خارج از یک بازه‌ی محدود صفر نمی‌شود) و به آن **سیستم با پاسخ ضربه‌ای نامتناهی (IIR)** می‌گویند.

---

### ۲. سیستم‌هایی که فقط صفر دارند و هیچ قطبی ندارند (FIR)

در این دسته، تابع تبدیل تنها یک چندجمله‌ای از $z^{-1}$ است و هیچ قطبی ندارد به جز در $z = 0$. بنابراین بسط کسر جزئی ممکن نیست و خواهیم داشت:

<div dir="ltr">

$$
H(z) = \sum_{k=0}^{M} b_k z^{-k} \qquad \text{(5.38)}
$$

</div>

(فرض می‌کنیم $a_0 = 1$ بدون از دست رفتن کلیت.)

از رابطه‌ی بالا، پاسخ ضربه‌ای مستقیماً به‌صورت زیر به‌دست می‌آید:

<div dir="ltr">

$$
h[n] = \sum_{k=0}^{M} b_k \delta[n - k] =
\begin{cases}
b_n, & 0 \leq n \leq M \\
0, & \text{در غیر این صورت}
\end{cases}
\qquad \text{(5.39)}
$$

</div>

در این حالت، پاسخ ضربه‌ای دارای طول محدود است (خارج از بازه‌ای مشخص صفر می‌شود). بنابراین این سیستم‌ها، **سیستم‌های با پاسخ ضربه‌ای متناهی (FIR)** نامیده می‌شوند.

برای این سیستم‌ها، معادله‌ی تفاضلی معادل با جمع هم‌نهشتی (کانولوشن) است:

<div dir="ltr">

$$
y[n] = \sum_{k=0}^{M} b_k x[n - k] \qquad \text{(5.40)}
$$

</div>

## مثال ۵٫۵ — یک سیستم FIR ساده

فرض کنید پاسخ ضربه‌ای یک سیستم، بریده‌شده‌ای از پاسخ ضربه‌ای سیستم IIR زیر باشد:

<div dir="ltr">

$$
G(z) = \frac{1}{1 - a z^{-1}}, \quad |z| > |a|
$$

</div>

در این صورت، پاسخ ضربه‌ای برابر است با:

<div dir="ltr">

$$
h[n] =
\begin{cases}
a^n, & 0 \leq n \leq M \\
0, & \text{در غیر این صورت}
\end{cases}
$$

</div>

و تابع تبدیل متناظر برابر خواهد بود با:

<div dir="ltr">

$$
H(z) = \sum_{n=0}^{M} a^n z^{-n} = \frac{1 - a^{M+1} z^{-(M+1)}}{1 - a z^{-1}} \qquad \text{(5.41)}
$$

</div>



</div>

<div dir="rtl">


از آنجا که صفرهای صورت در مکان‌های صفحه‌ی z به‌صورت زیر قرار دارند:

<div dir="ltr">

$$
z_k = a e^{j \frac{2\pi k}{M+1}}, \qquad k = 0, 1, \dots, M \qquad \text{(5.42)}
$$

</div>

که در آن، $a$ عددی حقیقی و مثبت فرض شده است. قطب در $z = a$ توسط صفر $z_0$ حذف شده است.

نمودار صفر-قطب برای حالت $M = 7$ در شکل زیر نمایش داده شده است:

### شکل ۵٫۸ — نمودار صفر و قطب برای مثال ۵٫۵

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
M = 7
a = 0.8  # فرض شده عددی حقیقی و مثبت

# محاسبه صفرها
zeros = a * np.exp(1j * 2 * np.pi * np.arange(0, M+1) / (M+1))

# قطب هفتم‌مرتبه در نقطه a
poles = [a] * (M+1)

# رسم نمودار
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(np.real(zeros), np.imag(zeros), 'o', label='zeros')
ax.plot(np.real(poles), np.imag(poles), 'x', label='poles')

# دایره واحد
unit_circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
ax.add_artist(unit_circle)

# تنظیمات نمودار
ax.axhline(0, color='black', lw=0.5)
ax.axvline(0, color='black', lw=0.5)
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal')
# ax.set_title("شکل ۵٫۸ — نمودار صفر و قطب برای مثال ۵٫۵", fontsize=12)
ax.legend()
plt.grid(True)
plt.show()
```

<div dir="rtl">

معادله‌ی تفاضلی بین ورودی و خروجی سیستم خطی نامتغیر با زمان برابر است با کانولوشن گسسته:

<div dir="ltr">

$$
y[n] = \sum_{k=0}^{M} a^k x[n - k] \qquad \text{(5.43)}
$$

</div>

با این حال، معادله‌ی (۵٫۴۱) همچنین بیان می‌کند که ورودی و خروجی معادله‌ی تفاضلی زیر را نیز ارضا می‌کنند:

<div dir="ltr">

$$
y[n] - a y[n - 1] = x[n] - a^{M+1} x[n - M - 1] \qquad \text{(5.44)}
$$

</div>

این دو معادله‌ی تفاضلی معادل، از دو فرم معادل تابع تبدیل $H(z)$ در رابطه‌ی (۵٫۴۱) نتیجه شده‌اند.

## ۵٫۳ — پاسخ فرکانسی برای توابع گویای سیستم

اگر یک سیستم پایدار خطی نامتغیر با زمان دارای تابع انتقال گویای $H(z)$ باشد، یعنی اگر ورودی و خروجی آن معادله‌ی تفاضلی از نوع رابطه‌ی (۵٫۱۹) را ارضا کنند، آنگاه پاسخ فرکانسی آن (یعنی تابع انتقال ارزیابی‌شده روی دایره‌ی واحد) به‌صورت زیر خواهد بود:

<div dir="ltr">

$$
H(e^{j\omega}) =
\frac{
\sum_{k=0}^{M} b_k e^{-j\omega k}
}{
\sum_{k=0}^{N} a_k e^{-j\omega k}
}
\qquad \text{(5.45)}
$$

</div>

</div>


<div dir="rtl">


تابع $H(e^{j\omega})$ نسبت چندجمله‌ای‌هایی بر حسب $e^{-j\omega}$ است.  
برای تعیین اندازه، فاز و تأخیر گروهی مربوط به پاسخ فرکانسی چنین سیستم‌هایی، بیان تابع $H(e^{j\omega})$ برحسب صفرها و قطب‌های $H(z)$ مفید است.

چنین بیانی از جایگذاری $z = e^{j\omega}$ در معادله‌ی (۵٫۲۱) به‌دست می‌آید:

<div dir="ltr">

$$
H(e^{j\omega}) = \left( \frac{b_0}{a_0} \right)
\cdot
\frac{
\prod_{k=1}^{M} (1 - c_k e^{-j\omega})
}{
\prod_{k=1}^{N} (1 - d_k e^{-j\omega})
}
\qquad \text{(5.46)}
$$

</div>

از رابطه‌ی (۵٫۴۶) نتیجه می‌شود که:

<div dir="ltr">

$$
|H(e^{j\omega})| = \left| \frac{b_0}{a_0} \right|
\cdot
\frac{
\prod_{k=1}^{M} |1 - c_k e^{-j\omega}|
}{
\prod_{k=1}^{N} |1 - d_k e^{-j\omega}|
}
\qquad \text{(5.47)}
$$

</div>

در نتیجه، تابع اندازه به توان دو برابر است با:

<div dir="ltr">

$$
|H(e^{j\omega})|^2 = H(e^{j\omega}) \cdot H^*(e^{j\omega}) =
\left( \frac{b_0}{a_0} \right)^2
\cdot
\frac{
\prod_{k=1}^{M} (1 - c_k e^{-j\omega})(1 - c_k^* e^{j\omega})
}{
\prod_{k=1}^{N} (1 - d_k e^{-j\omega})(1 - d_k^* e^{j\omega})
}
\qquad \text{(5.48)}
$$

</div>

از معادله‌ی (۵٫۴۷) نتیجه می‌شود که $|H(e^{j\omega})|$ برابر است با حاصل‌ضرب اندازه‌ی تمام عوامل صفر تابع $H(z)$ روی دایره واحد، تقسیم بر حاصل‌ضرب اندازه‌ی تمام عوامل قطب روی دایره واحد.

اگر بهره به دسی‌بل بیان شود، داریم:

<div dir="ltr">

$$
\text{Gain in dB} = 20 \log_{10} |H(e^{j\omega})| \qquad \text{(5.49)}
$$

</div>

بنابراین:

<div dir="ltr">

$$
\text{Gain in dB} =
20 \log_{10} \left| \frac{b_0}{a_0} \right|
+ \sum_{k=1}^{M} 20 \log_{10} |1 - c_k e^{-j\omega}|
- \sum_{k=1}^{N} 20 \log_{10} |1 - d_k e^{-j\omega}|
\qquad \text{(5.50)}
$$

</div>

پاسخ فازی برای تابع سیستم گویای زیر به‌صورت زیر است:

<div dir="ltr">

$$
\arg \left[ H(e^{j\omega}) \right] =
\arg \left[ \frac{b_0}{a_0} \right]
+ \sum_{k=1}^{M} \arg \left[ 1 - c_k e^{-j\omega} \right]
- \sum_{k=1}^{N} \arg \left[ 1 - d_k e^{-j\omega} \right]
\qquad \text{(5.51)}
$$

</div>

که در آن $\arg[\cdot]$ نمایانگر فاز پیوسته (بدون پیچش) است.

تأخیر گروهی متناظر برای تابع سیستم گویای $H(e^{j\omega})$ برابر است با:

<div dir="ltr">

$$
\text{grd}[H(e^{j\omega})] =
\sum_{k=1}^{N} \frac{d}{d\omega} \arg \left(1 - d_k e^{-j\omega} \right)
-
\sum_{k=1}^{M} \frac{d}{d\omega} \arg \left(1 - c_k e^{-j\omega} \right)
\qquad \text{(5.52)}
$$

</div>


</div>

<div dir="rtl">


یک عبارت معادل برای تأخیر گروهی به‌صورت زیر است:

<div dir="ltr">

$$
\text{grd}[H(e^{j\omega})] =
\sum_{k=1}^{N}
\frac{
|d_k|^2 - \text{Re}(d_k e^{-j\omega})
}{
1 + |d_k|^2 - 2\text{Re}(d_k e^{-j\omega})
}
-
\sum_{k=1}^{M}
\frac{
|c_k|^2 - \text{Re}(c_k e^{-j\omega})
}{
1 + |c_k|^2 - 2\text{Re}(c_k e^{-j\omega})
}
\qquad \text{(5.53)}
$$

</div>

در معادله‌ی (۵٫۵۱)، فاز هر جمله ذاتاً مبهم است، یعنی می‌توان هر مقدار مضربی از $2\pi$ به آن اضافه کرد بدون اینکه مقدار عدد مختلط تغییر کند. اما تأخیر گروهی با مشتق فاز پیوسته تعریف می‌شود.

معادلات (۵٫۵۰)، (۵٫۵۱) و (۵٫۵۳) به ترتیب اندازه (برحسب دسی‌بل)، فاز و تأخیر گروهی را نشان می‌دهند و هر سه از مجموع سهم‌های قطب‌ها و صفرهای تابع سیستم به‌دست می‌آیند. بنابراین، برای درک بهتر پاسخ فرکانسی سیستم، بررسی تأثیر مکان قطب‌ها و صفرهای سیستم (مخصوصاً پایدار از مرتبه بالاتر) روی پاسخ فرکانسی مفید است.

---

## ۵٫۳٫۱ — پاسخ فرکانسی سیستم‌های مرتبه اول

در این بخش، ویژگی‌های یک عامل تکی از فرم <div dir="ltr"> $$(1 - r e^{j\theta} e^{-j\omega})$$ </div> را بررسی می‌کنیم، که در آن $r$ شعاع و $\theta$ زاویه قطب یا صفر در صفحه‌ی z است.

این عامل می‌تواند نماینده‌ی یک قطب یا صفر منفرد روی شعاع $r$ و زاویه $\theta$ باشد.

توان دوم اندازه‌ی چنین عبارتی به‌صورت زیر است:

<div dir="ltr">

$$
|1 - r e^{j\theta} e^{-j\omega}|^2 =
(1 - r e^{j(\theta - \omega)})(1 - r e^{-j(\theta - \omega)}) =
1 + r^2 - 2r \cos(\omega - \theta)
\qquad \text{(5.54)}
$$

</div>

بهره (gain) برحسب دسی‌بل با این عامل برابر است با:

<div dir="ltr">

$$
(\pm) 20 \log_{10} |1 - r e^{j\theta} e^{-j\omega}| =
(\pm) 10 \log_{10} \left[ 1 + r^2 - 2r \cos(\omega - \theta) \right]
\qquad \text{(5.55)}
$$

</div>

علامت مثبت در صورتی استفاده می‌شود که عامل یک صفر باشد، و علامت منفی در صورتی که یک قطب باشد.

مشارکت این عامل در فاز به‌صورت زیر است:

<div dir="ltr">

$$
(\pm) \arg(1 - r e^{j\theta} e^{-j\omega}) =
(\pm) \arctan \left( \frac{r \sin(\omega - \theta)}{1 - r \cos(\omega - \theta)} \right)
\qquad \text{(5.56)}
$$

</div>

با مشتق‌گیری از معادله‌ی بالا (به‌جز نقاط گسستگی)، مشارکت عامل در تأخیر گروهی برابر است با:

<div dir="ltr">

$$
(\pm) \left[ -\text{grd}(1 - r e^{j\theta} e^{-j\omega}) \right] =
(\pm) \frac{r^2 - r \cos(\omega - \theta)}{1 + r^2 - 2r \cos(\omega - \theta)} =
(\pm) \frac{r \cos(\omega - \theta)}{|1 - r e^{j\theta} e^{-j\omega}|^2}
\qquad \text{(5.57)}
$$

</div>

در اینجا نیز، علامت مثبت برای صفر و منفی برای قطب به‌کار می‌رود.

معادلات (۵٫۵۴) تا (۵٫۵۷) تابع‌هایی متناوب نسبت به $\omega$ با دوره‌ی $2\pi$ هستند.

**شکل 5.9(a)** نمودار رابطه‌ی (۵٫۵۵) را به‌صورت تابعی از $\omega$ برای مقادیر مختلف $\theta$ در $r = 0.9$ نشان می‌دهد.

**شکل 5.9(b)** نیز فاز معادله‌ی (۵٫۵۶) را به‌عنوان تابعی از $\omega$ برای $r = 0.9$ و مقادیر مختلفی از $\theta$ رسم کرده است. توجه کنید که نقطه‌ی ماکزیمم همیشه در $\omega = \theta$ رخ می‌دهد و برای $r$ ثابت، این تابع با تغییر $\theta$ فقط جابه‌جا می‌شود.

**شکل 5.9(c)** تأخیر گروهی رابطه‌ی (۵٫۵۷) را نشان می‌دهد. توجه کنید که مقدار بالای مثبت این تابع برای $\omega = \theta$ به افت ناگهانی تأخیر در همان نقطه منجر می‌شود.

در استخراج ویژگی‌های پاسخ فرکانسی از نمودارهای صفر–قطب (چه گسسته چه پیوسته)، سیستم‌های حداقل–فاز به‌ویژه اهمیت زیادی دارند، و بردارهای فاز و تأخیرشان به‌طور مستقیم قابل تحلیل هستند.

</div>


```python

import numpy as np
import matplotlib.pyplot as plt

# تعریف پارامترها
r = 0.9
theta_vals = [0, np.pi/2, np.pi]
omega = np.linspace(0, 2*np.pi, 1000)

# آماده‌سازی شکل
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

for theta in theta_vals:
    # H(e^jw) = 1 - r*e^{j(theta)} * e^{-j*omega} = 1 - r * e^{j(theta - omega)}
    z = 1 - r * np.exp(1j * (theta - omega))
    H = z

    # (a) Log Magnitude (dB)
    axs[0].plot(omega, 20 * np.log10(np.abs(H)), label=r'$\theta = {:.0f}$'.format(theta))
    
    # (b) Phase
    axs[1].plot(omega, np.angle(H), label=r'$\theta = {:.0f}$'.format(theta))
    
    # (c) Group Delay: derivative of unwrapped phase
    phase_unwrapped = np.unwrap(np.angle(H))
    dphi = -np.gradient(phase_unwrapped, omega)  # منفی مشتق فاز
    axs[2].plot(omega, dphi, label=r'$\theta = {:.0f}$'.format(theta))

# تنظیمات subplotها
titles = ['(a) Log Magnitude (dB)', '(b) Phase (radians)', '(c) Group Delay (samples)']
ylabels = ['dB', 'Radians', 'Samples']
for ax, title, ylabel in zip(axs, titles, ylabels):
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.set_xlim(0, 2*np.pi)
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
    ax.grid(True)

axs[2].set_xlabel('Radian frequency (ω)')
axs[0].legend()

plt.tight_layout()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# پارامترها
r = 0.7
theta = np.pi / 3
omega = np.pi / 2

# نقاط کلیدی
zero = r * np.exp(1j * theta)
ejw = np.exp(1j * omega)

v1 = ejw
v2 = zero
v3 = v1 - v2

# رسم نمودار
fig, ax = plt.subplots(figsize=(6,6))

# محورهای مختلط
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# دایره واحد
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
ax.add_artist(circle)

# بردارها
ax.arrow(0, 0, v1.real, v1.imag, head_width=0.05, length_includes_head=True, color='black', label=r'$v_1$')
ax.arrow(0, 0, v2.real, v2.imag, head_width=0.05, length_includes_head=True, color='black', linestyle='--', label=r'$v_2$')
ax.arrow(v2.real, v2.imag, v3.real, v3.imag, head_width=0.05, length_includes_head=True, color='black', linestyle='-.', label=r'$v_3 = v_1 - v_2$')

# برچسب‌ها
ax.text(v1.real + 0.05, v1.imag, '$e^{j\\omega}$')
ax.text(v2.real + 0.05, v2.imag, '$r e^{j\\theta}$')
ax.text(0.3, 0.2, r'$\phi_1$', fontsize=12)
ax.text(0.7, 0.7, r'$\phi_3$', fontsize=12)
ax.text(0.3, -0.1, r'$\theta$', fontsize=12)

# تنظیمات شکل
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal')
# ax.set_title("شکل ۵٫۱۰ — بردارهای صفحه‌ی z برای تابع مرتبه اول", fontsize=12)
ax.grid(True)
plt.show()
```

<div dir="rtl">

نمایش برداری قطب و صفرها در صفحه مختلط معمولاً بسیار مفید است. در این نمایش، هر قطب یا صفر را می‌توان با برداری در صفحه‌ی z از موقعیت آن تا نقطه‌ای روی دایره واحد نشان داد.

برای یک تابع سیستم مرتبه اول از فرم زیر:

<div dir="ltr">

$$
H(z) = (1 - r e^{j\theta} z^{-1}) = \frac{z - r e^{j\theta}}{z}, \qquad r < 1 \qquad \text{(5.58)}
$$

</div>

الگوی صفر–قطب در شکل زیر نشان داده شده است:


در این شکل، بردارهای $v_1$، $v_2$ و $v_3 = v_1 - v_2$  نشان داده شده‌اند که به‌ترتیب نمایانگر اعداد مختلط $e^{j\omega}$، $r e^{j\theta}$ و $(e^{j\omega} - r e^{j\theta})$ هستند.

بر حسب این بردارها، اندازه‌ی عدد مختلط:

<div dir="ltr">

$$
\frac{e^{j\omega} - r e^{j\theta}}{e^{j\omega}}
$$

</div>

برابر است با نسبت اندازه‌ی بردارهای  $v_3$ به $v_1$ ، یعنی:

<div dir="ltr">

$$
|1 - r e^{j\theta} e^{-j\omega}| = \left| \frac{e^{j\omega} - r e^{j\theta}}{e^{j\omega}} \right| = \frac{|v_3|}{|v_1|} \qquad \text{(5.59)}
$$

</div>

و چون <div dir="ltr"> $$|v_1| = 1$$ </div> ، معادله (۵٫۵۹) به‌سادگی برابر با <div dir="ltr"> $$|v_3|$$ </div>  است.

فاز متناظر با این عبارت برابر است با:

<div dir="ltr">

$$
\angle(1 - r e^{j\theta} e^{-j\omega}) = \angle(e^{j\omega} - r e^{j\theta}) - \angle(e^{j\omega}) = \angle(v_3) - \angle(v_1)
= \phi_3 - \phi_1 = \phi_3 - \omega \qquad \text{(5.60)}
$$

</div>

بنابراین، مشارکت یک عامل تکی از نوع <div dir="ltr"> $$(1 - r e^{j\theta} z^{-1})$$ </div>  در تابع اندازه در فرکانس $$\omega$$ برابر است با طول بردار از صفر تا نقطه <div dir="ltr"> $$z = e^{j\omega}$$ </div> روی دایره واحد. این بردار هنگامی که <div dir="ltr"> $$\omega = \theta$$ </div> است حداقل طول را دارد و این همان دلیل افت تیز تابع اندازه در نمودار شکل ۵٫۹(a) است.

بردار <div dir="ltr"> $$v_1$$ </div>  از قطب در <div dir="ltr"> $$z = 0$$ </div> تا <div dir="ltr"> $$e^{j\omega}$$ </div> همیشه طولی برابر با ۱ دارد و روی تابع اندازه اثری ندارد.

معادله‌ی (۵٫۶۰) نشان می‌دهد که تابع فاز برابر است با اختلاف زاویه‌ی بردار از صفر در <div dir="ltr"> $$r e^{j\theta}$$ </div> تا $e^{j\omega}$ و بردار از قطب در $0$ تا  <div dir="ltr"> $$e^{j\omega}$$ </div>.

وابستگی کاهش پاسخ فرکانسی به مکان صفر از عامل <div dir="ltr"> $$(1 - r e^{j\theta} e^{-j\omega})$$ </div> برحسب شعاع  $r$  ، در شکل ۵٫۱۱ برای $\theta = \pi$  و مقادیر مختلفی از $r$ نشان داده شده است.

</div>

### شکل ۵٫۱۱ — پاسخ فرکانسی برای یک صفر منفرد با $\theta = \pi$ و مقادیر $r = 1, 0.9, 0.7, 0.5$  
(a) اندازه لگاریتمی  
(b) فاز  
(c) تأخیر گروهی برای $r = 0.9, 0.7, 0.5$

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# مقادیر پارامترها
r_values = [0.5, 0.7, 0.9, 1.0]
theta = np.pi
omega = np.linspace(0, 2 * np.pi, 1000)

# آماده‌سازی نمودار
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

line_styles = ['-.', '--', ':', '-']
labels = [r'$r = 0.5$', r'$r = 0.7$', r'$r = 0.9$', r'$r = 1$']

for r, ls, label in zip(r_values, line_styles, labels):
    z = 1 - r * np.exp(1j * (theta - omega))
    H = z

    # (a) Log magnitude
    axs[0].plot(omega, 20 * np.log10(np.abs(H)), linestyle=ls, label=label)

    # (b) Phase
    axs[1].plot(omega, np.angle(H), linestyle=ls, label=label)

# فقط r ≠ 1 برای تأخیر گروهی، چون مشتق فاز برای r=1 ناپیوستگی دارد
for r, ls, label in zip(r_values[:-1], line_styles[:-1], labels[:-1]):
    z = 1 - r * np.exp(1j * (theta - omega))
    phase_unwrapped = np.unwrap(np.angle(z))
    dphi = -np.gradient(phase_unwrapped, omega)
    axs[2].plot(omega, dphi, linestyle=ls, label=label)

# تنظیمات گراف‌ها
titles = ['(a) Log magnitude (dB)', '(b) Phase (radians)', '(c) Group delay (samples)']
ylabels = ['dB', 'Radians', 'Samples']
for ax, title, ylabel in zip(axs, titles, ylabels):
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.set_xlim(0, 2*np.pi)
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
    ax.grid(True)

axs[2].set_xlabel('Radian frequency (ω)')
axs[0].legend()
plt.tight_layout()
plt.show()
```

<div dir="rtl">


توجه کنید که تابع اندازه لگاریتمی رسم‌شده در شکل ۵٫۱۱(a)، هنگامی که $r$ به ۱ نزدیک می‌شود، افت شدیدتری دارد؛ در واقع، اندازه برحسب دسی‌بل به $-\infty$ میل می‌کند وقتی $\omega = \theta$ و $r \to 1$.

تابع فاز در شکل ۵٫۱۱(b) دارای شیب مثبت و بزرگ در $\omega = \theta$ است که وقتی $r \to 1$ بی‌نهایت می‌شود. بنابراین برای $r = 1$، تابع فاز ناپیوسته است و یک پرش به‌اندازه‌ی $$\pi$$ رادیان در $$\omega = \theta$$ دارد.

دور از $\omega = \theta$، شیب تابع فاز منفی است. از آنجا که تأخیر گروهی برابر منفی مشتق تابع فاز است، تأخیر گروهی در نزدیکی $\omega = \theta$ منفی است و با نزدیک شدن $r$ به ۱ به‌طور ناگهانی افزایش می‌یابد (که منجر به یک ضربه می‌شود – در شکل نشان داده نشده).

شکل ۵٫۱۱(c) نشان می‌دهد که با دور شدن از $\omega = \theta$، تأخیر گروهی مثبت شده و نسبتاً تخت باقی می‌ماند.

## ۵٫۳٫۲ — مثال‌هایی با چند قطب و صفر

در این بخش، با استفاده از مباحث بخش ۵٫۳٫۱، پاسخ فرکانسی سیستم‌هایی با توابع انتقال گویای پیچیده‌تر را بررسی می‌کنیم.

---

### مثال ۵٫۶ — سیستم مرتبه دوم IIR

تابع انتقال مرتبه دوم زیر را در نظر بگیرید:

<div dir="ltr">

$$
H(z) = \frac{1}{(1 - r e^{j\theta} z^{-1})(1 - r e^{-j\theta} z^{-1})}
= \frac{1}{1 - 2r \cos\theta z^{-1} + r^2 z^{-2}}
\qquad \text{(5.61)}
$$

</div>

معادله‌ی تفاضلی بین ورودی و خروجی این سیستم برابر است با:

<div dir="ltr">

$$
y[n] - 2r \cos\theta \, y[n-1] + r^2 y[n-2] = x[n]
$$

</div>

با استفاده از تکنیک بسط کسر جزئی، می‌توان نشان داد که پاسخ ضربه‌ای یک سیستم علّی با این تابع انتقال برابر است با:

<div dir="ltr">

$$
h[n] = \frac{r^n \sin[(\theta)(n+1)]}{\sin\theta} \, u[n] \qquad \text{(5.62)}
$$

</div>

تابع انتقال رابطه‌ی (۵٫۶۱) دارای یک قطب در <div dir="ltr"> $$z = r e^{j\theta}$$ </div> و یک مزدوج آن در <div dir="ltr"> $$z = r e^{-j\theta}$$ </div>
است و دو صفر نیز در $z = 0$ دارد.

---

### شکل ۵٫۱۲ — نمودار صفر–قطب برای مثال ۵٫۶

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
r = 0.8
theta = np.pi / 3
pole1 = r * np.exp(1j * theta)
pole2 = r * np.exp(-1j * theta)
zero1 = 0
zero2 = 0

# دایره واحد
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')

# بردارها
omega = np.pi / 2
ejw = np.exp(1j * omega)
v1 = ejw
v2 = pole1
v3 = v1 - v2

# رسم نمودار
fig, ax = plt.subplots(figsize=(6,6))
ax.add_artist(circle)
ax.plot(np.real(pole1), np.imag(pole1), 'x', markersize=10, label='Pole 1')
ax.plot(np.real(pole2), np.imag(pole2), 'x', markersize=10, label='Pole 2')
ax.plot(np.real(zero1), np.imag(zero1), 'o', label='Zero (x2)')

# بردارها v1، v2، v3
ax.arrow(0, 0, v1.real, v1.imag, head_width=0.04, color='black')
ax.arrow(0, 0, v2.real, v2.imag, head_width=0.04, color='black', linestyle='--')
ax.arrow(v2.real, v2.imag, v3.real, v3.imag, head_width=0.04, color='black', linestyle=':')

# تنظیمات
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_aspect('equal')
ax.set_title("Figure 5.12 — Pole-zero plot for Example 5.6")
ax.grid(True)
plt.legend()
plt.show()
```

<div dir="rtl">

براساس بحث بخش ۵٫۳٫۱، خواهیم داشت:

<div dir="ltr">

$$
20 \log_{10} |H(e^{j\omega})| =
-10 \log_{10} \left[ 1 + r^2 - 2r \cos(\omega - \theta) \right]
-10 \log_{10} \left[ 1 + r^2 - 2r \cos(\omega + \theta) \right]
\qquad \text{(5.63a)}
$$

$$
\angle H(e^{j\omega}) =
- \arctan \left( \frac{r \sin(\omega - \theta)}{1 - r \cos(\omega - \theta)} \right)
- \arctan \left( \frac{r \sin(\omega + \theta)}{1 - r \cos(\omega + \theta)} \right)
\qquad \text{(5.63b)}
$$

$$
\text{grd}H(e^{j\omega}) =
\frac{r^2 - r \cos(\omega - \theta)}{1 + r^2 - 2r \cos(\omega - \theta)}
+
\frac{r^2 - r \cos(\omega + \theta)}{1 + r^2 - 2r \cos(\omega + \theta)}
\qquad \text{(5.63c)}
$$

</div>

این توابع در شکل ۵٫۱۳ برای <div dir="ltr"> $$r = 0.9$$ و $$\theta = \pi / 4$$ </div> رسم شده‌اند.

شکل ۵٫۱۲ بردارهای صفر و قطب 
<div dir="ltr"> 

$v_1$ و $v_2$ و $v_3$

</div> 

 را نشان می‌دهد. تابع اندازه برابر است با حاصل‌ضرب طول بردارهای صفر (که در این مثال برابر یک هستند)، تقسیم بر حاصل‌ضرب طول بردارهای قطب. بنابراین:

<div dir="ltr">

$$
|H(e^{j\omega})| = \frac{|v_3|^2}{|v_1| \cdot |v_2|} = \frac{1}{|v_1| \cdot |v_2|} \qquad \text{(5.64)}
$$

</div>

وقتی $\omega \approx \theta$ ، طول بردار <div dir="ltr"> $$v_1 = e^{j\omega} - r e^{j\theta}$$ </div> کوچک می‌شود و نسبت به تغییرات  $\omega$ حساس است. اما طول <div dir="ltr"> $$v_2 = e^{j\omega} - r e^{-j\theta}$$ </div> تقریباً ثابت باقی می‌ماند.

در نتیجه، قطب با زاویه $\theta$ تأثیر عمده‌ای بر پاسخ فرکانسی اطراف $\omega = \theta$ دارد (همان‌طور که در شکل ۵٫۱۳ مشخص است). به‌طور متقارن، قطب با زاویه $-\theta$ پاسخ را در اطراف $\omega = -\theta$ کنترل می‌کند.

---

### شکل ۵٫۱۳ — پاسخ فرکانسی برای زوج مزدوج از قطب‌ها (مطابق مثال ۵٫۶)، برای $r = 0.9$   و   $\pi / 4 = \theta$ 

</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
r = 0.9
theta = np.pi / 4
omega = np.linspace(0, 2*np.pi, 1000)

# محاسبه |H(e^jω)| بر اساس معادله (5.63a)
mag_squared = (1 + r**2 - 2 * r * np.cos(omega - theta)) * (1 + r**2 - 2 * r * np.cos(omega + theta))
mag_dB = -10 * np.log10(mag_squared)

# رسم نمودار
plt.figure(figsize=(8, 4))
plt.plot(omega, mag_dB, color='black')
plt.title('Figure 5.13 — Frequency response for complex-conjugate pole pair')
plt.ylabel('dB')
plt.xlabel('Radian frequency (ω)')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
plt.grid(True)
plt.ylim(-10, 20)
plt.show()
```

<div dir="rtl">

(b) و (c) از شکل ۵٫۱۳ نیز به‌ترتیب فاز و تأخیر گروهی مربوط به زوج مزدوج از قطب‌ها را برای $r = 0.9$ و $\theta = \frac{\pi}{4}$ نمایش می‌دهند.

---

### شکل ۵٫۱۳ (ادامه) — پاسخ فرکانسی برای زوج مزدوج از قطب‌ها (مطابق مثال ۵٫۶)، برای $r = 0.9$ و $\frac{\pi}{4} = \theta $  
(b) فاز  
(c) تأخیر گروهی

</div>

```python



import numpy as np
import matplotlib.pyplot as plt

# پارامترها
r = 0.9
theta = np.pi / 4
omega = np.linspace(0, 2*np.pi, 1000)

# عبارت‌های فاز و تاخیر گروهی (معادلات 5.63b و 5.63c)
# فاز
arg1 = np.arctan2(r * np.sin(omega - theta), 1 - r * np.cos(omega - theta))
arg2 = np.arctan2(r * np.sin(omega + theta), 1 - r * np.cos(omega + theta))
phase = -arg1 - arg2

# تأخیر گروهی
gd1 = (r**2 - r * np.cos(omega - theta)) / (1 + r**2 - 2 * r * np.cos(omega - theta))
gd2 = (r**2 - r * np.cos(omega + theta)) / (1 + r**2 - 2 * r * np.cos(omega + theta))
group_delay = gd1 + gd2

# رسم شکل
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# (b) Phase
axs[0].plot(omega, phase, color='black')
axs[0].set_title('(b) Phase')
axs[0].set_ylabel('Radians')
axs[0].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[0].set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
axs[0].grid(True)

# (c) Group Delay
axs[1].plot(omega, group_delay, color='black')
axs[1].set_title('(c) Group Delay')
axs[1].set_ylabel('Samples')
axs[1].set_xlabel('Radian frequency (ω)')
axs[1].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[1].set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

## Example 5.7 —  سیستم FIR مرتبه دوم

در این مثال، یک سیستم پاسخ ضربه‌ای محدود (FIR) را در نظر می‌گیریم که پاسخ ضربه‌ای آن به‌صورت زیر است:

<div dir="ltr">

$$
h[n] = \delta[n] - 2r \cos\theta \, \delta[n-1] + r^2 \delta[n-2] \qquad \text{(5.65)}
$$

</div>

تابع تبدیل متناظر این سیستم برابر است با:

<div dir="ltr">

$$
H(z) = 1 - 2r \cos\theta \, z^{-1} + r^2 z^{-2} \qquad \text{(5.66)}
$$

</div>

که این تابع، معکوس تابع تبدیل ارائه‌شده در مثال ۵٫۶ است.

بنابراین، نمودارهای پاسخ فرکانسی برای این سیستم FIR، دقیقاً برابر با منفیِ نمودارهای شکل ۵٫۱۳ خواهند بود.

همچنین توجه شود که در معکوس تابع تبدیل، مکان قطب‌ها و صفرها با یکدیگر جابه‌جا می‌شوند.

</div>

## مثال ۵٫۸ — سیستم IIR مرتبه سوم

در این مثال، یک فیلتر پایین‌گذر (lowpass) را بررسی می‌کنیم که با استفاده از یکی از روش‌های تقریبی طراحی شده (که در فصل ۷ توضیح داده خواهد شد).

تابع سیستم به صورت زیر در نظر گرفته می‌شود:

<div dir="ltr">

$$
H(z) = \frac{0.056634(1 + z^{-1})(1 - 1.0166z^{-1} + z^{-2})}{
(1 - 0.683z^{-1})(1 - 1.4461z^{-1} + 0.7957z^{-2})}
\qquad \text{(5.67)}
$$

</div>

و فرض شده که این سیستم پایدار است.

مکان صفرهای این تابع سیستم به صورت زیر هستند:

| شعاع (Radius) | زاویه (Angle)              |
|---------------|-----------------------------|
| 1             | $$\pi$$ رادیان              |
| 1             | $$\pm 1.0376$$ رادیان  (59.45°) |

مکان قطب‌های این سیستم به صورت زیر است:

| شعاع (Radius) | زاویه (Angle)               |
|---------------|------------------------------|
| 0.683         | 0                            |
| 0.892         | $$\pm 0.6257$$ رادیان (35.85°) |

نمودار صفر–قطب برای این سیستم در شکل زیر نمایش داده شده است:


### شکل ۵٫۱۴ — نمودار صفر–قطب برای فیلتر پایین‌گذر مثال ۵٫۸

</div>

```python

import numpy as np
import matplotlib.pyplot as plt

# تعریف صفرها (بر اساس زاویه و شعاع)
zeros = [
    np.exp(1j * np.pi),  # زاویه π
    np.exp(1j * 1.0376),  # زاویه +1.0376 rad
    np.exp(-1j * 1.0376)  # زاویه -1.0376 rad
]

# تعریف قطب‌ها
poles = [
    0.683 * np.exp(1j * 0),           # قطب روی محور حقیقی
    0.892 * np.exp(1j * 0.6257),      # زاویه +0.6257 rad
    0.892 * np.exp(-1j * 0.6257)      # زاویه -0.6257 rad
]

# رسم نمودار
fig, ax = plt.subplots(figsize=(6, 6))

# دایره واحد
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
ax.add_artist(circle)

# محورهای مختلط
ax.axhline(0, color='black', lw=0.5)
ax.axvline(0, color='black', lw=0.5)

# رسم صفرها
for z in zeros:
    ax.plot(np.real(z), np.imag(z), 'o', markersize=8, label='Zero' if z == zeros[0] else "")

# رسم قطب‌ها
for p in poles:
    ax.plot(np.real(p), np.imag(p), 'x', markersize=8, label='Pole' if p == poles[0] else "")

# تنظیمات
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.grid(True)
ax.set_title("Figure 5.14 — Pole-zero plot for Example 5.8", fontsize=12)
ax.legend()
plt.show()
```

<div dir="rtl">

شکل ۵٫۱۵، نمودارهای اندازه لگاریتمی، فاز و تأخیر گروهی این سیستم را نشان می‌دهد. تأثیر صفرهایی که روی دایره واحد در زوایای $\pm 1.0376 = \omega$ و $\pi$ قرار دارند، کاملاً مشهود است.

اما قطب‌ها به‌گونه‌ای قرار گرفته‌اند که به‌جای اینکه باعث افزایش شدید در فرکانس‌های نزدیک زاویه خودشان شوند، پاسخ کلی اندازه را در بازه‌ی $\omega = 0$ تا $$\omega = 0.2\pi$ نزدیک به $0$ دسی‌بل نگه می‌دارند (و به‌صورت متقارن در بازه $1.8\pi = \omega$ تا $2\pi$ نیز همین اتفاق می‌افتد).

سپس پاسخ به‌طور ناگهانی کاهش یافته و کمتر از $-25$ دسی‌بل باقی می‌ماند، در بازه‌ای از حدود $0.3\pi = \omega$ تا $1.7\pi$.

همان‌طور که از این مثال مشاهده می‌شود، تقریب‌های سودمند برای پاسخ‌های فیلتر انتخابی فرکانس را می‌توان با استفاده از قطب‌ها برای ساختن پاسخ اندازه و استفاده از صفرها برای حذف آن به‌دست آورد.

</div>

<div dir="rtl">

 ### شکل ۵٫۱۵ —   پاسخ فرکانسی برای فیلتر گذرپایین مثال  5.8

(a) اندازه لگاریتمی   
(b) فاز    
(c) تاخیر گروهی

</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz, group_delay

# ضرایب صورت و مخرج تابع H(z) از معادله (5.67)
b = 0.056634 * np.convolve([1, 1], [1, -1.0166, 1])  # (1 + z^-1)(1 - 1.0166 z^-1 + z^-2)
a = np.convolve([1, -0.683], [1, -1.4461, 0.7957])   # (1 - 0.683 z^-1)(1 - 1.4461 z^-1 + 0.7957 z^-2)

# پاسخ فرکانسی
w, h = freqz(b, a, worN=1024)

# فاز بدون پیچش و تأخیر گروهی
phase = np.unwrap(np.angle(h))
_, gd = group_delay((b, a), w)

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

# (a) Log magnitude
axs[0].plot(w, 20 * np.log10(np.abs(h)), color='black')
axs[0].set_title('(a) Log magnitude')
axs[0].set_ylabel('dB')

# (b) Phase
axs[1].plot(w, phase, color='black')
axs[1].set_title('(b) Phase')
axs[1].set_ylabel('Radians')

# (c) Group delay
axs[2].plot(w, gd, color='black')
axs[2].set_title('(c) Group delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian frequency (ω)')

# تنظیمات مشترک
for ax in axs:
    ax.set_xlim(0, np.pi * 2)
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
    ax.grid(True)

plt.tight_layout()
plt.show()
```

در این مثال، شاهد هر دو نوع گسستگی در نمودار تابع فاز هستیم.

در $0.22\pi \approx \omega$، یک گسستگی به‌اندازه‌ی $2\pi$ وجود دارد که ناشی از استفاده از فاز اصلی در رسم است.

در $\pm 1.0376 = \omega$ و $\pi = \omega$، گسستگی‌ها ناشی از صفرهایی هستند که روی دایره واحد قرار گرفته‌اند.

## ۵٫۴ — رابطه بین اندازه و فاز

به‌طور کلی، دانستن اندازه پاسخ فرکانسی یک سیستم خطی نامتغیر با زمان (LTI)، اطلاعاتی درباره‌ی فاز ارائه نمی‌دهد و بالعکس.

اما برای سیستم‌هایی که با معادلات تفاضلی با ضرایب ثابت توصیف می‌شوند (یعنی سیستم‌های دارای تابع انتقال گویا)، نوعی محدودیت بین اندازه و فاز برقرار است.

به‌طور خاص، اگر اندازه‌ی پاسخ فرکانسی و تعداد قطب‌ها و صفرها معلوم باشد، تنها تعداد محدودی انتخاب برای فاز ممکن است.

به‌طور مشابه، اگر تعداد قطب‌ها و صفرها و فاز معلوم باشد، تنها تعداد محدودی انتخاب برای اندازه وجود دارد (تا یک ضریب مقیاس).

علاوه بر این، تحت شرطی به‌نام فاز مینیمم (minimum-phase)، پاسخ اندازه، پاسخ فاز را به‌صورت یکتا تعیین می‌کند و بالعکس.

---

برای بررسی انتخاب‌های ممکن برای تابع سیستم، با فرض معلوم بودن توان دوم اندازه‌ی پاسخ فرکانسی سیستم، رابطه‌ی زیر را در نظر می‌گیریم:

<div dir="ltr">

$$
|H(e^{j\omega})|^2 =
H(e^{j\omega}) H^*(e^{j\omega}) =
H(z) H^*(1/z^*) \big|_{z = e^{j\omega}} \qquad \text{(5.68)}
$$

</div>

اگر تابع $H(z)$ را به فرم گویای زیر در نظر بگیریم (مطابق رابطه‌ی ۵٫۲۱):

<div dir="ltr">

$$
H(z) = \left( \frac{b_0}{a_0} \right)
\cdot
\frac{
\prod_{k=1}^{M} (1 - c_k z^{-1})
}{
\prod_{k=1}^{N} (1 - d_k z^{-1})
}
\qquad \text{(5.69)}
$$

</div>

آنگاه خواهیم داشت:

<div dir="ltr">

$$
H^*\left( \frac{1}{z^*} \right) =
\left( \frac{b_0}{a_0} \right)
\cdot
\frac{
\prod_{k=1}^{M} (1 - c_k^* z)
}{
\prod_{k=1}^{N} (1 - d_k^* z)
}
\qquad \text{(5.70)}
$$

</div>

که در آن فرض شده است که $a_0$ و $b_0$ اعداد حقیقی هستند.

بنابراین، رابطه‌ی (۵٫۶۸) بیان می‌کند که توان دوم اندازه‌ی پاسخ فرکانسی برابر است با مقدار تابع <div dir="ltr"> $$H(z) H^*(1/z^*)$$ </div> روی دایره‌ی واحد.


</div>


<div dir="rtl">


با استفاده از این رابطه:

<div dir="ltr">

$$
4(1 + 0.5 z^{-1})(1 + 0.5 z) = (1 + z^{-1})(1 + 2z) \qquad \text{(5.77)}
$$

</div>

می‌توان نتیجه گرفت که:

<div dir="ltr">

$$
C_1(z) = C_2(z)
$$

</div>

نمودار صفر–قطب برای <div dir="ltr"> $$C_1(z)$$ و $$C_2(z)$$ </div> (که کاملاً یکسان هستند) در شکل ۵٫۱۶(c) نمایش داده شده است.
### شکل ۵٫۱۶ — نمودار صفر–قطب برای دو تابع سیستم و تابع مشترک مربع اندازه‌ی آن‌ها  
<div dir="ltr"> (a) \( H_1(z) \)  (b) \( H_2(z) \)  (c) \( C_1(z), C_2(z) \) </div>

</div>

```python

import numpy as np
import matplotlib.pyplot as plt

def plot_pz(ax, zeros, poles, title):
    # دایره واحد
    unit_circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
    ax.add_artist(unit_circle)

    # محورهای مختلط
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

    # صفرها
    for z in zeros:
        ax.plot(np.real(z), np.imag(z), 'o', markersize=8)

    # قطب‌ها
    for p in poles:
        ax.plot(np.real(p), np.imag(p), 'x', markersize=8)

    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.set_title(title)
    ax.grid(True)

# تعریف سیستم‌ها
# (a) H1(z)
zeros_a = [-0.5, -2]
poles_a = [-0.5j, 0.5j]

# (b) H2(z)
zeros_b = [-2, -0.5]
poles_b = [-0.5j, 0.5j]

# (c) C1(z) = C2(z)
zeros_c = [-0.5, -0.5, -2, -2]
poles_c = [-0.5j, 0.5j, -0.5j, 0.5j]

# رسم نمودارها
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
plot_pz(axs[0], zeros_a, poles_a, '(a) $H_1(z)$')
plot_pz(axs[1], zeros_b, poles_b, '(b) $H_2(z)$')
plot_pz(axs[2], zeros_c, poles_c, '(c) $C_1(z)$ and $C_2(z)$')
plt.tight_layout()
plt.show()
```

<div dir="rtl">

توابع سیستم $H_1(z)$ و $H_2(z)$ در مثال ۵٫۹ تنها در مکان یکی از صفرهایشان تفاوت دارند.

در این مثال، عامل $2(1 + 0.5 z^{-1}) = (z^{-1} + 2)$ دقیقاً همان سهم را در مربع اندازه پاسخ فرکانسی دارد که عامل <div dir="ltr"> $$(1 + 2z^{-1})$$ </div> دارد.

در نتیجه:
 <div dir="ltr">

$$
|H_1(e^{j\omega})|^2 = |H_2(e^{j\omega})|^2
$$

 </div>
اما فاز پاسخ این دو تابع با یکدیگر متفاوت است.

</div>

## مثال ۵٫۱۰ — تعیین $H(z)$ از روی $C(z)$

فرض کنید نمودار صفر–قطب تابع $C(z)$ (در شکل ۵٫۱۷) داده شده باشد و هدف ما تعیین قطب‌ها و صفرهایی باشد که به $H(z)$ مربوط می‌شوند.

زوج مزدوج قطب‌ها و صفرهایی که یکی از اعضای هر جفت به $H(z)$ و دیگری به $H^*(1/z^*)$ تعلق دارد به‌صورت زیر تعریف می‌شوند:



<div dir="ltr">



**زوج قطب ۱**:  $(p_1, p_4)$  
**زوج قطب ۲**:  $(p_2, p_5)$  
**زوج قطب ۳**:  $(p_3, p_6)$  
**زوج صفر ۱**:  $(z_1, z_4)$  
**زوج صفر ۲**:  $(z_2, z_5)$  
**زوج صفر ۳**:  $(z_3, z_6)$

</div>

با دانستن اینکه $H(z)$ به یک سیستم علّی و پایدار تعلق دارد، باید از هر جفت قطب، آن عضوی را انتخاب کنیم که درون دایره واحد قرار دارد؛ یعنی <div dir="ltr"> 
$p_1$، $p_2$، و $p_3$ </div>. اما برای صفرها چنین محدودیتی وجود ندارد.

اگر فرض کنیم که ضرایب <div dir="ltr"> 
$a_k$ و $b_k$ </div> در روابط (۵٫۱۹) و (۵٫۲۰) حقیقی هستند، آنگاه صفرها (و قطب‌ها) یا حقیقی‌اند، یا به‌صورت زوج مزدوج مختلط ظاهر می‌شوند. در نتیجه، صفرهایی که باید به $H(z)$ نسبت دهیم، به یکی از دو صورت زیر خواهند بود:

<div dir="ltr">

$$z_3$$ 
یا
$$z_6$$
و
$$(z_1, z_2)$$
یا
$$(z_4, z_5)$$

</div>

بنابراین، چهار انتخاب ممکن برای سیستم پایدار و علّی $H(z)$ با سه قطب و سه صفر وجود دارد که پاسخ فرکانسی $C(z)$ برای همه‌ی آن‌ها یکسان است (نمودار صفر–قطب $(z)$ در شکل ۵٫۱۷ نشان داده شده است).

اگر شرط حقیقی بودن ضرایب <div dir="ltr"> $$a_k$$ و $$b_k$$ </div> را حذف کنیم، تعداد انتخاب‌ها حتی بیشتر خواهد بود.

اگر هیچ محدودیتی روی تعداد قطب‌ها و صفرهای $H(z)$ وجود نداشته باشد، تعداد انتخاب‌های ممکن بی‌نهایت خواهد بود.

برای دیدن این موضوع، فرض کنید $H(z)$ دارای ضریبی از فرم زیر باشد:

<div dir="ltr">

$$
\frac{z^{-1} - a^*}{1 - a z^{-1}}
$$

</div>

در این صورت:

<div dir="ltr">

$$
H(z) = H_1(z) \cdot \frac{z^{-1} - a^*}{1 - a z^{-1}} \qquad \text{(5.78)}
$$

</div>

عبارت بالا یک **عامل تمام‌عبوری (all-pass)** است، چون اندازه آن روی دایره واحد برابر ۱ است.  
(جزئیات بیشتر درباره‌ی این عوامل در بخش ۵٫۵ آمده است.)

به راحتی می‌توان تأیید کرد که:

<div dir="ltr">

$$
C(z) = H(z) H^*(1/z^*) = H_1(z) \cdot \frac{z^{-1} - a^*}{1 - a z^{-1}} \cdot H_1^*(1/z^*) \cdot \frac{1 - a^* z}{z^{-1} - a}
\qquad \text{(5.79)}
$$

</div>

تمام عوامل all-pass در $$C(z)$$ حذف می‌شوند و در نمودار صفر–قطب آن دیده نمی‌شوند.

در نتیجه، اگر تعداد قطب‌ها و صفرهای $H(z)$ مشخص نشده باشد، برای یک $C(z)$ داده شده، می‌توان $H(z)$ را با هر تعداد دلخواه عامل تمام‌عبوری که قطب آن‌ها داخل دایره واحد است (یعنی $|a| < 1$) ترکیب کرد.

</div>

<div dir="rtl">     

### شکل ۵٫۱۷ — نمودار صفر–قطب برای تابع مربع اندازه در مثال ۵٫۱۰
</div>


```python


import numpy as np
import matplotlib.pyplot as plt

# نقاط قطب‌ها و صفرها از شکل (مختصاتی تقریبی)
poles = [0.6 + 0.3j, 0.6 - 0.3j, -0.8]        # p1, p2, p3
poles_conj = [np.conj(p) for p in poles]     # p4, p5, p6

zeros = [0.6 + 0.7j, 0.6 - 0.7j, -1.0]        # z1, z2, z3
zeros_conj = [np.conj(z) for z in zeros]     # z4, z5, z6

# ترکیب همه قطب‌ها و صفرها
all_poles = poles + poles_conj
all_zeros = zeros + zeros_conj

# رسم نمودار
fig, ax = plt.subplots(figsize=(6, 6))

# دایره واحد
unit_circle = plt.Circle((0, 0), 1, color='gray', linestyle='--', fill=False)
ax.add_artist(unit_circle)

# محورهای مختلط
ax.axhline(0, color='black', lw=0.5)
ax.axvline(0, color='black', lw=0.5)

# رسم قطب‌ها (×)
for p in all_poles:
    ax.plot(np.real(p), np.imag(p), 'x', markersize=8, color='black')

# رسم صفرها (○)
for z in all_zeros:
    ax.plot(np.real(z), np.imag(z), 'o', markersize=8, color='black', fillstyle='none')

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.set_title("Figure 5.17 — Pole-zero plot for the magnitude-squared function in Example 5.10")
ax.grid(True)
plt.show()
```

## ۵٫۵ — سیستم‌های تمام‌عبور (All-Pass)

همان‌طور که در بحث مثال ۵٫۱۰ نیز مطرح شد، یک تابع سیستم پایدار از فرم:

<div dir="ltr">

$$
H_{ap}(z) = \frac{z^{-1} - a^*}{1 - a z^{-1}} \qquad \text{(5.80)}
$$

</div>

دارای اندازه پاسخ فرکانسی مستقل از فرکانس $\omega$ است.

این موضوع با بازنویسی تابع روی دایره واحد به‌صورت زیر قابل مشاهده است:

<div dir="ltr">

$$
H_{ap}(e^{j\omega}) = \frac{e^{-j\omega} - a^*}{1 - a e^{-j\omega}} =
\frac{e^{-j\omega} (1 - a^* e^{j\omega})}{1 - a e^{-j\omega}} \qquad \text{(5.81)}
$$

</div>

در معادله (۵٫۸۱)، عبارت <div dir="ltr"> $$e^{-j\omega}$$ </div> دارای اندازه یک است و صورت و مخرج نیز مزدوج مختلط یکدیگرند و بنابراین اندازه مساوی دارند.  
در نتیجه:

<div dir="ltr">

$$
|H_{ap}(e^{j\omega})| = 1
$$

</div>

سیستمی که در آن اندازه پاسخ فرکانسی عدد ثابتی باشد، «سیستم تمام‌عبور» (All-Pass System) نامیده می‌شود، چون تمام مؤلفه‌های فرکانسی ورودی را با بهره‌ی ثابت یا تضعیف ثابتی عبور می‌دهد.³

---

³ در برخی منابع، سیستم تمام‌عبور سیستمی تعریف می‌شود که بهره آن دقیقاً برابر یک باشد.  
در این کتاب، منظور از سیستم تمام‌عبور سیستمی است که بهره آن مستقل از فرکانس باشد، ولی لزوماً یک نباشد.

</div>

```python
### شکل ۵٫۱۸ — نمودار صفر–قطب معمولی برای یک سیستم تمام‌عبور


import numpy as np
import matplotlib.pyplot as plt

# قطب‌های سیستم
real_poles = [-1/3, -3/4]  # d1, d2
complex_pole = 0.8 * np.exp(1j * np.pi / 4)  # e1
complex_pole_conj = np.conj(complex_pole)

# صفرهای متناظر با قطب‌ها (مزدوج وارون)
real_zeros = [1/d for d in real_poles]
complex_zero = 1/complex_pole_conj
complex_zero_conj = 1/complex_pole

# رسم نمودار
fig, ax = plt.subplots(figsize=(6, 6))

# دایره واحد
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
ax.add_artist(circle)

# محور مختلط
ax.axhline(0, color='black', lw=0.5)
ax.axvline(0, color='black', lw=0.5)

# قطب‌ها (×)
for p in real_poles + [complex_pole, complex_pole_conj]:
    ax.plot(np.real(p), np.imag(p), 'x', markersize=8, color='black')

# صفرها (○)
for z in real_zeros + [complex_zero, complex_zero_conj]:
    ax.plot(np.real(z), np.imag(z), 'o', markersize=8, color='black', fillstyle='none')

# تنظیمات
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_aspect('equal')
ax.set_title("Figure 5.18 — Typical pole-zero plot for an all-pass system")
ax.grid(True)
plt.show()
```

## ادامه‌ی بخش ۵٫۵ — سیستم‌های تمام‌عبور

هر تابع سیستم تمام‌عبور کلی را می‌توان به‌صورت ضربی از عوامل مرتبه اول و دوم به شکل زیر نوشت:

<div dir="ltr"> 

$$
H_{ap}(z) =
A \cdot \prod_{k=1}^{M_r} \frac{z^{-1} - d_k}{1 - d_k z^{-1}}
\cdot
\prod_{k=1}^{M_c} \frac{(z^{-1} - e_k^*)(z^{-1} - e_k)}{(1 - e_k z^{-1})(1 - e_k^* z^{-1})}
\qquad \text{(5.82)}
$$

</div>

که در آن:

- \( A \) یک عدد حقیقی مثبت است.
- \( d_k \) قطب‌های حقیقی هستند.
- \( e_k \) قطب‌های مختلط هستند (به‌همراه مزدوج آن‌ها).

برای سیستم‌های علی و پایدار تمام‌عبور، باید داشته باشیم:

<div dir="ltr">

$$
|d_k| < 1 \quad \text{و} \quad |e_k| < 1
$$

</div>

در شکل ۵٫۱۸، نمودار صفر–قطب یک سیستم تمام‌عبور را نشان می‌دهد که دارای:

- \( M_r = 2 \): دو قطب حقیقی  
- \( M_c = 1 \): یک زوج قطب مزدوج مختلط

توجه کنید که به‌ازای هر قطب، صفر متناظر مزدوج وارون  وجود دارد.

</div>


<div dir="rtl">

پاسخ فرکانسی یک سیستم تمام‌عبور کلی را می‌توان به‌صورت ترکیبی از پاسخ‌های فرکانسی سیستم‌های تمام‌عبور مرتبه اول (مشابه معادله 5.80) بیان کرد.

برای یک سیستم تمام‌عبور علّی، هرکدام از این عوامل شامل یک قطب درون دایره واحد و یک صفر در مکان مزدوج وارون همان قطب هستند.

همان‌طور که پیش‌تر نشان داده شد، اندازهٔ پاسخ فرکانسی چنین سیستمی برابر ۱ است.  
در نتیجه، اندازه لگاریتمی در واحد دسی‌بل برابر با صفر خواهد بود.

اگر پارامتر $$a$$ را در فرم قطبی به‌صورت <div dir="ltr"> $$a = r e^{j\theta}$$ </div> بنویسیم، آنگاه فاز پاسخ برای معادلهٔ (5.80) به‌صورت زیر است:

<div dir="ltr">

$$
\angle \left[
\frac{e^{-j\omega} - r e^{-j\theta}}{1 - r e^{j\theta} e^{-j\omega}}
\right]
= -\omega - 2 \arctan \left[
\frac{r \sin(\omega - \theta)}{1 - r \cos(\omega - \theta)}
\right]
\qquad \text{(5.83)}
$$

</div>

به‌طور مشابه، فاز یک سیستم تمام‌عبور مرتبه دوم که قطب‌هایی در <div dir="ltr"> $$z=re^{j\theta}$$ و $$z=re^{-j\theta}$$ </div>دارد، به صورت زیر خواهد بود:

<div dir="ltr">

$$
\angle \left[
\frac{(e^{-j\omega} - r e^{-j\theta})(e^{-j\omega} - r e^{j\theta})}
{(1 - r e^{j\theta} e^{-j\omega})(1 - r e^{-j\theta} e^{-j\omega})}
\right]
= -2\omega - 2 \arctan \left[
\frac{r \sin(\omega - \theta)}{1 - r \cos(\omega - \theta)}
\right]
- 2 \arctan \left[
\frac{r \sin(\omega + \theta)}{1 - r \cos(\omega + \theta)}
\right]
\qquad \text{(5.84)}

</div>

</div>

## مثال ۵٫۱۱ — سیستم‌های تمام‌عبور مرتبه اول و دوم

شکل ۵٫۱۹ نمودارهای مربوط به مقدار لگاریتمی، فاز، و تأخیر گروهی برای دو سیستم تمام‌عبور مرتبه اول را نشان می‌دهد.

یکی از این سیستم‌ها دارای قطبی در $ 0.9 = z $ یعنی $0 = \theta$، $0.9 = r$
و دیگری دارای قطبی در \( z = -0.9 \) است.

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات
omega = np.linspace(0, 2*np.pi, 1000)
r1 = 0.9
r2 = -0.9

# تعریف تابع سیستم تمام‌عبور مرتبه اول H(z) = (z^-1 - a*) / (1 - a z^-1)
def H_ap(omega, a):
    ejw = np.exp(1j * omega)
    num = np.conj(a) - ejw
    den = 1 - a * ejw
    return num / den

# محاسبه پاسخ‌ها
H1 = H_ap(omega, r1)
H2 = H_ap(omega, r2)

# Log magnitude (در dB)
mag1 = 20 * np.log10(np.abs(H1))
mag2 = 20 * np.log10(np.abs(H2))

# Phase (unwrap یا principal)
phase1 = np.angle(H1)
phase2 = np.angle(H2)

# Group delay
def group_delay(H):
    dphi = np.unwrap(np.angle(H))
    gd = -np.gradient(dphi, omega)
    return gd

gd1 = group_delay(H1)
gd2 = group_delay(H2)

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

# (a) log magnitude
axs[0].plot(omega, mag1, label='z = 0.9', color='black')
axs[0].plot(omega, mag2, '--', label='z = -0.9', color='black')
axs[0].set_title('(a) Log magnitude')
axs[0].set_ylabel('dB')
axs[0].legend()

# (b) phase
axs[1].plot(omega, phase1, label='z = 0.9', color='black')
axs[1].plot(omega, phase2, '--', label='z = -0.9', color='black')
axs[1].set_title('(b) Phase')
axs[1].set_ylabel('Radians')
axs[1].legend()

# (c) group delay
axs[2].plot(omega, gd1, label='z = 0.9', color='black')
axs[2].plot(omega, gd2, '--', label='z = -0.9', color='black')
axs[2].set_title('(c) Group delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian frequency (ω)')
axs[2].legend()

# تنظیمات عمومی
for ax in axs:
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
    ax.grid(True)

plt.tight_layout()
plt.show()
```

## شکل ۵٫۲۰ — پاسخ فرکانسی سیستم تمام‌عبور مرتبه دوم

در هر دو سیستم، شعاع قطب‌ها برابر با \( r = 0.9 \) است.  
شکل ۵٫۲۰ تابع‌های لگاریتم اندازه، فاز، و تأخیر گروهی را برای یک سیستم تمام‌عبور مرتبه دوم با قطب‌های:

<div dir="ltr">

$$
z = 0.9 e^{j\pi/4} \quad \text{و} \quad z = 0.9 e^{-j\pi/4}
$$

</div>

نشان می‌دهد.

</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیم فرکانس و قطب‌ها
omega = np.linspace(0, 2*np.pi, 1000)
r = 0.9
theta = np.pi / 4

# قطب‌های مختلط مزدوج
p1 = r * np.exp(1j * theta)
p2 = np.conj(p1)

# تابع سیستم تمام‌عبور مرتبه دوم
def H_ap_2nd_order(omega, p1, p2):
    ejw = np.exp(1j * omega)
    num = (np.conj(p1) - ejw) * (np.conj(p2) - ejw)
    den = (1 - p1 * ejw) * (1 - p2 * ejw)
    return num / den

# محاسبه پاسخ‌ها
H = H_ap_2nd_order(omega, p1, p2)
mag = 20 * np.log10(np.abs(H))
phase = np.angle(H)

# تأخیر گروهی
def group_delay(H, omega):
    dphi = np.unwrap(np.angle(H))
    gd = -np.gradient(dphi, omega)
    return gd

gd = group_delay(H, omega)

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

# (a) log magnitude
axs[0].plot(omega, mag, color='black')
axs[0].set_title('(a) Log magnitude')
axs[0].set_ylabel('dB')

# (b) phase
axs[1].plot(omega, phase, color='black')
axs[1].set_title('(b) Phase (principal value)')
axs[1].set_ylabel('Radians')

# (c) group delay
axs[2].plot(omega, gd, color='black')
axs[2].set_title('(c) Group delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian frequency (ω)')

# تنظیم محور افقی
for ax in axs:
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
    ax.grid(True)

plt.tight_layout()
plt.show()
```

## ادامه‌ی مثال ۵٫۱۱ — خاصیت کلی سیستم‌های تمام‌عبور علّی

مثال ۵٫۱۱ ویژگی عمومی سیستم‌های تمام‌عبور علّی را نشان می‌دهد.  
در شکل ۵٫۱۹(b)، مشاهده می‌کنیم که فاز سیستم در بازه \( 0 < omega < \pi \) **غیرفزاینده** است.

به‌طور مشابه، در شکل ۵٫۲۰(b)، اگر پرش \( 2\pi \) ناشی از فاز اصلی  حذف شود، منحنی فاز پیوسته نیز در بازه \( 0 < \omega < \pi \) نزولی خواهد بود.

از آنجا که تابع سیستم تمام‌عبور کلی در رابطه (5.82) تنها حاصل‌ضرب عواملی از نوع مرتبه اول و دوم است، نتیجه می‌گیریم که فاز بدون‌پیچش تابع سیستم (یعنی \( \arg[H_{ap}(e^{j\omega})] \)) برای سیستم تمام‌عبور علّی همیشه در بازه \( 0 < \omega < \pi \) **منفی یا صفر** است.

ممکن است برای مقدار اصلی فاز این موضوع صادق نباشد (مثلاً در شکل ۵٫۲۱ دیده می‌شود)، ولی برای فاز پیوسته قطعاً صادق است.

---

### تأیید رابطه با تأخیر گروهی

تأخیر گروهی عامل ساده‌ی مرتبه اول تمام‌عبور از معادله (5.80) برابر است با مشتق منفی فاز آن از رابطه (5.83).  
با کمی جبر، داریم:

<div dir="ltr">

$$
\text{grd} \left[
\frac{e^{-j\omega} - r e^{-j\theta}}{1 - r e^{j\theta} e^{-j\omega}}
\right] =
\frac{1 - r^2}{1 + r^2 - 2r \cos(\omega - \theta)} =
\frac{1 - r^2}{|1 - r e^{j\theta} e^{-j\omega}|^2}
\qquad \text{(5.85)}
$$

</div>

چون برای سیستم علّی و پایدار تمام‌عبور داریم \( r < 1 \)، پس از رابطه (5.85) نتیجه می‌گیریم که **تأخیر گروهی مثبت است.**

چون تأخیر گروهی حاصل‌جمع این عوامل است، نتیجه می‌گیریم که تأخیر گروهی برای سیستم تمام‌عبور مرتبه بالا نیز **مثبت خواهد بود**.

این در شکل‌های ۵٫۱۹(c)، ۵٫۲۰(c)، و ۵٫۲۱(c) برای سیستم‌های تمام‌عبور مرتبه ۱، ۲ و ۳ نشان داده شده است.

---

### نتیجه‌گیری: فاز نزولی و تأخیر گروهی مثبت

مثبت بودن تأخیر گروهی سیستم تمام‌عبور علّی به‌طور مستقیم منجر به این نتیجه می‌شود که:

<div dir="ltr">

$$
\arg[H_{ap}(e^{j\omega})] = - \int_0^\omega \text{grd}[H_{ap}(e^{j\phi})] d\phi + \arg[H_{ap}(e^{j0})]
\qquad \text{(5.86)}
$$

</div>
برای بازه \( 0 \leq \omega \leq \pi \).

از رابطه (5.82) نیز داریم:

<div dir="ltr">

$$
H_{ap}(e^{j0}) =
A \cdot \prod_{k=1}^{M_r} \frac{1 - d_k}{1 - d_k} \cdot
\prod_{k=1}^{M_c} \frac{|1 - e_k|^2}{|1 - e_k|^2} = A
\qquad \text{(5.87)}
$$

</div>

در نتیجه:

<div dir="ltr">

$$
\arg[H_{ap}(e^{j0})] = 0
$$

</div>

و چون طبق رابطهٔ (5.85) داریم:

<div dir="ltr">

$$
\text{grd}[H_{ap}(e^{j\omega})] \geq 0
\qquad \text{(5.88)}
$$

</div>

از رابطه (5.86) نتیجه می‌شود که:

<div dir="ltr">

$$
\arg[H_{ap}(e^{j\omega})] \leq 0 \qquad \text{for } 0 \leq \omega \leq \pi
\qquad \text{(5.89)}
$$

</div>

---

### جمع‌بندی:

- **تأخیر گروهی مثبت** و
- **فاز بدون پیچش نزولی**  
دو خاصیت مهم سیستم‌های تمام‌عبور علّی هستند.

کاربردهای این سیستم‌ها:
- جبران فاز یا اعوجاج گروهی (در فصل ۷ توضیح داده می‌شود)
- طراحی فیلترهای فاز مینیمم (فصل ۵٫۶)
- تبدیل فیلترهای پایین‌گذر به فیلترهای دلخواه با پهنای قابل تنظیم


</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# برد فرکانسی
omega = np.linspace(0, 2*np.pi, 1000)
ejw = np.exp(1j * omega)

# قطب‌ها طبق شکل 5.18 (تقریبی)
real_poles = [-1/3, -3/4]
complex_pole = 0.8 * np.exp(1j * np.pi / 4)
complex_pole_conj = np.conj(complex_pole)

# صفرهای متناظر (conjugate reciprocal)
real_zeros = [1/d for d in real_poles]
complex_zero = 1 / complex_pole_conj
complex_zero_conj = 1 / complex_pole

# تابع سیستم H(z)
def H_allpass(ejw):
    # حقیقی‌ها
    Hr = np.ones_like(ejw, dtype=complex)
    for p, z in zip(real_poles, real_zeros):
        Hr *= (z - ejw) / (1 - p * ejw)
    # مختلط مزدوج
    Hc = ((complex_zero - ejw)*(complex_zero_conj - ejw)) / ((1 - complex_pole * ejw)*(1 - complex_pole_conj * ejw))
    return Hr * Hc

# محاسبه پاسخ
H = H_allpass(ejw)
mag = 20 * np.log10(np.abs(H))
phase = np.angle(H)
group_delay = -np.gradient(np.unwrap(phase), omega)

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

# (a) Log magnitude
axs[0].plot(omega, mag, color='black')
axs[0].set_title('(a) Log magnitude')
axs[0].set_ylabel('dB')

# (b) Phase
axs[1].plot(omega, phase, color='black')
axs[1].set_title('(b) Phase (principal value)')
axs[1].set_ylabel('Radians')

# (c) Group delay
axs[2].plot(omega, group_delay, color='black')
axs[2].set_title('(c) Group delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian frequency (ω)')

# تنظیم محور x
for ax in axs:
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
    ax.grid(True)

plt.tight_layout()
plt.show()
```

## شکل ۵٫۲۱ — پاسخ فرکانسی سیستم تمام‌عبور

پاسخ فرکانسی برای سیستم تمام‌عبور با نمودار صفر–قطب داده‌شده در شکل ۵٫۱۸.

(a) اندازه لگاریتمی (b) فاز (principal value) (c) تاخیر گروهی

</div>

<div dir="rtl">

بخش ۵.۶ سیستم‌های مینیمم‌فاز

در بخش ۵.۴ نشان دادیم که اندازه پاسخ فرکانسی یک سیستم LTI با تابع سیستم گویای (rational) به‌تنهایی نمی‌تواند سیستم را به‌طور منحصربه‌فرد توصیف کند. اگر سیستم پایدار و علّی باشد، قطب‌ها باید درون دایره واحد قرار گیرند؛ اما پایداری و علّیت هیچ محدودیتی برای صفرها قائل نیستند.

در برخی مسائل، اعمال محدودیت اضافی بر این اساس مفید است که سیستم معکوس (یعنی سیستمی با تابع سیستم 
$\frac{1}{H(z)}$) نیز پایدار و علّی باشد. همان‌طور که در (بخش 5.2.2) گفته شد، این موضوع باعث می‌شود که 
صفرها نیز، مانند قطب‌ها، درون دایره واحد قرار گیرند، چرا که 
قطب‌های $\frac{1}{H(z)}$ همان صفرهای $H(z)$ هستند.

سیستم‌هایی که این ویژگی را دارند، به‌طور رایج با عنوان سیستم‌های مینیمم‌فاز شناخته می‌شوند.

واژه‌ی مینیمم‌فاز از ویژگی پاسخ فاز گرفته شده است، که به‌وضوح از تعریف فوق قابل برداشت نیست. این ویژگی و دیگر ویژگی‌های بنیادی این دسته از سیستم‌ها، منحصربه‌فردند و هر یک از آن‌ها را می‌توان به‌عنوان تعریف این کلاس در نظر گرفت.

این ویژگی‌ها در بخش ۵.۶.۳ بررسی خواهند شد.

اگر تابعی از اندازه-توان‌دو پاسخ فرکانسی ($|H(e^{j\omega})|^2$) مانند معادله (5.82) داشته باشیم و بدانیم که سیستممینیمم‌فاز است، آنگاه $H(z)$ به‌طور منحصربه‌فرد تعیین می‌شود و شامل تمام صفرها و قطب‌های تابع میباشد:

<div dir="ltr">

$$
C(z) = H(z) H^*\left(\frac{1}{z^*}\right)
$$

</div>

خواهد بود که درون دایره واحد قرار دارند.

    این روش اغلب در طراحی فیلتر به کار می‌رود، هنگامی که تنها اندازه پاسخ فرکانسی از طریق روش طراحی مشخص شده است. (رجوع شود به فصل ۷)

</div>

<div dir="rtl">

**5.6.1 سیستم‌های مینیمم‌فاز و تجزیه‌ی All-Pass**

در بخش ۵.۴ نشان دادیم که فقط با استفاده از مربع اندازه پاسخ فرکانسی نمی‌توان تابع سیستم $H(z)$ را به‌طور منحصربه‌فرد تعیین کرد، زیرا هر انتخابی که اندازه پاسخ فرکانسی داده شده را حفظ کند، می‌تواند با عوامل all-pass دلخواه ترکیب شود بدون اینکه اندازه تغییر کند. 

مشاهده‌ی مرتبطی که در اینجا بیان می‌شود این است که هر تابع سیستم گویا را می‌توان به‌صورت زیر بیان کرد:

<div dir="ltr">

$$
H(z) = H_{\text{min}}(z) H_{\text{ap}}(z), \tag{5.100}
$$

</div>

که در آن، $H_{\text{min}}(z)$  یک سیستم مینیمم‌فاز و  $H_{\text{ap}}(z)$  یک سیستم all-pass است.

برای نشان دادن این موضوع، فرض کنید $H(z)$ یک صفر خارج از دایره واحد در <div dir="ltr">
 $z = \frac{1}{c^*}$ 
 </div> داشته باشد، که در آن |c| < 1 ، و باقی صفرها و قطب‌ها درون دایره واحد قرار دارند. در این صورت، H(z) را می‌توان به صورت زیر نوشت:

<div dir="ltr">

$$
H(z) = H_1(z)(z^{-1} - c^*), \tag{5.101}
$$

</div>

که در آن، طبق تعریف، $H_1(z)$ یک سیستم مینیمم‌فاز است. 

یک بیان معادل دیگر برای $H(z)$ به شکل زیر است:

<div dir="ltr">

$$
H(z) = H_1(z)(1 - cz^{-1}) \frac{z^{-1} - c^*}{1 - cz^{-1}}. \tag{5.102}
$$

</div>

از آنجا که $|c| < 1$، عامل $H_1(z)(1 - cz^{-1})$ نیز یک سیستم مینیمم‌فاز است، و تنها تفاوت آن با $H(z)$ در این است که صفر $H(z)$ که خارج از دایره واحد در $z = \frac{1}{c^*}$ قرار داشت، به درون دایره، با بازتاب، منتقل شده است.

عبارت  <div dir="ltr"> $$\frac{z^{-1} - c^*}{1 - cz^{-1}}$$ </div> یک تابع all-pass است. این مثال را می‌توان به‌صورت ساده‌ای تعمیم داد تا صفرهای بیشتری را که خارج از دایره واحد قرار دارند، در بر بگیرد. به این ترتیب، به‌طور کلی، هر تابع سیستم را می‌توان به‌صورت زیر بیان کرد:

$$
H(z) = H_{\text{min}}(z) H_{\text{ap}}(z), \tag{5.103}
$$

که در آن $H_{\text{min}}(z)$ شامل قطب‌ها و صفرهای $H(z)$ است که داخل دایره واحد قرار دارند، به‌علاوه صفرهایی که مزدوج معکوس صفرهای $H(z)$ هستند که خارج از دایره واحد قرار دارند. 

$H_{\text{ap}}(z)$ شامل تمام صفرهای $H(z)$ است که خارج از دایره واحد قرار دارند، به‌همراه قطب‌هایی که برای حذف صفرهای معکوس مزدوج منعکس‌شده در $H_{\text{min}}(z)$ انتخاب شده‌اند.

با استفاده از معادله (5.103)، می‌توان یک سیستم غیر مینیمم‌فاز را از یک سیستم مینیمم‌فاز ایجاد کرد، با انعکاس یک یا چند صفر که در داخل دایره واحد هستند به مکان مزدوج معکوس‌شان در خارج از دایره واحد؛ یا برعکس، می‌توان یک سیستم مینیمم‌فاز را از یک سیستم غیر مینیمم‌فاز ساخت، با انعکاس تمام صفرهای خارج از دایره واحد به مکان‌های مزدوج معکوس‌شان در داخل دایره.

در هر دو حالت، هم سیستم مینیمم‌فاز و هم سیستم غیر مینیمم‌فاز دارای پاسخ فرکانسی با اندازه یکسان خواهند بود.


</div>

<div dir="rtl">

**مثال ۵.۱۴: تجزیه‌ی مینیمم‌فاز / All-Pass**

برای نمایش تجزیه‌ی یک سیستم پایدار و علّی به صورت ضربی از یک سیستم مینیمم‌فاز و یک سیستم all-pass، دو سیستم پایدار و علّی زیر را در نظر بگیرید:

<div dir="ltr">

$$
H_1(z) = \frac{1 + 3z^{-1}}{1 + \frac{1}{2}z^{-1}}
$$

و

$$
H_2(z) = \frac{\left(1 + \frac{3}{2}e^{+j\pi/4}z^{-1}\right)\left(1 + \frac{3}{2}e^{-j\pi/4}z^{-1}\right)}{1 - \frac{1}{3}z^{-1}}.
$$

</div>


تابع سیستم اول، $H_1(z)$، دارای یک قطب در داخل دایره‌ی واحد در $z = -\frac{1}{2}$ است، اما یک صفر در خارج از دایره‌ی واحد در $z = -3$ دارد. برای انعکاس این صفر به داخل دایره‌ی واحد، باید سیستم all-pass مناسب را انتخاب کنیم.

از معادله‌ی (5.101)، مقدار $c$ برابر است با:
$
c = -\frac{1}{3}
$

در نتیجه، طبق معادلات (5.102) و (5.103)، مؤلفه‌ی all-pass به صورت زیر خواهد بود:

<div dir="ltr">

$$
H_{\text{ap}}(z) = \frac{z^{-1} + \frac{1}{3}}{1 + \frac{1}{3}z^{-1}},
$$

</div>
و مؤلفه‌ی مینیمم‌فاز به شکل زیر خواهد بود:

<div dir="ltr">

$$
H_{\text{min}}(z) = 3 \cdot \frac{1 + \frac{1}{3}z^{-1}}{1 + \frac{1}{2}z^{-1}},
$$

</div>

یعنی:

<div dir="ltr">

$$
H_1(z) = \left( 3 \cdot \frac{1 + \frac{1}{3}z^{-1}}{1 + \frac{1}{2}z^{-1}} \right) \cdot \left( \frac{z^{-1} + \frac{1}{3}}{1 + \frac{1}{3}z^{-1}} \right).
$$

</div>

</div>

<div dir="rtl">

**تجزیه‌ی $H_2(z)$ به سیستم مینیمم‌فاز و All-Pass**

تابع سیستم دوم، $H_2(z)$، دارای دو صفر مختلط در خارج از دایره‌ی واحد و یک قطب حقیقی در داخل آن است. می‌توان $H_2(z)$ را به صورت معادله (5.101) بازنویسی کرد، با فاکتورگیری از $\frac{3}{2} e^{j\pi/4}$ و $\frac{3}{2} e^{-j\pi/4}$ در صورت کسر، خواهیم داشت:

<div dir="ltr">

$$
H_2(z) = \frac{9}{4} \cdot \frac{(z^{-1} + \frac{3}{2} e^{-j\pi/4})(z^{-1} + \frac{3}{2} e^{j\pi/4})}{1 - \frac{1}{3} z^{-1}}
$$

</div>

فاکتورگیری به صورت معادله (5.102) منجر به فرم زیر می‌شود:

<div dir="ltr">

$$
H_2(z) = \left[ \frac{9}{4} \cdot \frac{\left(1 + \frac{3}{2} e^{-j\pi/4} z^{-1}\right) \left(1 + \frac{3}{2} e^{j\pi/4} z^{-1}\right)}{1 - \frac{1}{3} z^{-1}} \right]
\cdot
\left[
\frac{\left(z^{-1} + \frac{3}{2} e^{-j\pi/4} \right) \left(z^{-1} + \frac{3}{2} e^{j\pi/4} \right)}
     {\left(1 + \frac{3}{2} e^{-j\pi/4} z^{-1} \right) \left(1 + \frac{3}{2} e^{j\pi/4} z^{-1} \right)}
\right]
$$

</div>

عبارت اول داخل براکت، یک سیستم مینیمم‌فاز است، در حالی که عبارت دوم یک سیستم all-pass می‌باشد.


۵.۶.۲ جبران پاسخ فرکانسی

در بسیاری از کاربردهای پردازش سیگنال، ممکن است سیگنالی توسط یک سیستم LTI با پاسخ فرکانسی نامطلوب دچار اعوجاج شده باشد. در چنین شرایطی ممکن است بخواهیم سیگنال اعوجاج‌یافته را با استفاده از یک سیستم جبران‌کننده پردازش کنیم، همان‌طور که در شکل ۵.۲۵ نشان داده شده است. این وضعیت مثلاً در انتقال سیگنال‌ها از طریق یک کانال مخابراتی ممکن است پیش بیاید. اگر جبران کامل ممکن باشد، آنگاه خواهیم داشت:
$$
s_c[n] = s[n],
\quad \text{یعنی } \quad H_c(z) = \frac{1}{H_d(z)}.
$$


اما اگر فرض کنیم که سیستم اعوجاج‌گر پایدار و علّی است و سیستم جبران‌کننده نیز باید پایدار و علّی باشد، آنگاه جبران کامل تنها در صورتی ممکن است که $Hd​(z)$ یک سیستم مینیمم‌فاز باشد تا معکوس آن نیز پایدار و علّی باشد.

با توجه به بحث‌های قبلی، اگر فرض کنیم $Hd(z)$ معلوم یا تقریب‌زده شده به‌صورت یک تابع سیستمی گویا باشد، می‌توانیم با بازتاب تمام صفرهایی که خارج از دایره‌ی واحد هستند به مکان‌های مزدوج قرینه‌ی آن‌ها درون دایره واحد، یک سیستم مینیمم‌فاز $Hdmin(z)$ تشکیل دهیم. سیستم‌های $Hd​(z)$ و $Hdmin(z)$ پاسخ فرکانسی یکسانی دارند و از طریق یک سیستم all-pass یعنی $Hap(z)$ به هم مربوط‌اند. یعنی:
<div dir="ltr">

$$
H_d(z) = H_{\text{dmin}}(z) \cdot H_{\text{ap}}(z). \tag{5.104}
$$

</div>

## **۵.۶.۳ ویژگی‌های سیستم‌های مینیمم‌فاز**

ما از اصطلاح "مینیمم‌فاز" برای اشاره به سیستم‌هایی استفاده کرده‌ایم که علّی و پایدار هستند و یک معکوس علّی و پایدار نیز دارند. این انتخاب نام، ناشی از ویژگی‌ای در تابع فاز است که هرچند آشکار نیست، ولی از تعریف انتخابی ما نتیجه می‌شود.
در این بخش، مجموعه‌ای از ویژگی‌های جالب و مهم سیستم‌های مینیمم‌فاز را نسبت به سایر سیستم‌هایی که پاسخ فرکانسی هم‌اندازه دارند، بررسی می‌کنیم.


### **ویژگی کم‌ترین تأخیر فاز (The Minimum Phase-Lag Property)**

استفاده از اصطلاح "مینیمم‌فاز" به عنوان نامی توصیفی برای سیستمی که همه قطب‌ها و صفرهایش داخل دایره واحد قرار دارند، از مثال ۵.۱۵ الهام گرفته شده است.
به عنوان نتیجه‌ای از معادله (5.100)، فاز پیوسته تابع تبدیل یک سیستم غیر مینیمم‌فاز را می‌توان به صورت زیر بیان کرد:

<div dir="ltr">

$$
\arg[H(e^{j\omega})] = \arg[H_{\min}(e^{j\omega})] + \arg[H_{ap}(e^{j\omega})] \tag{5.111}
$$

</div>

بنابراین، فاز پیوسته‌ای که متناظر با فاز اصلی (principal value) در شکل 5.27(b) باشد، برابر است با مجموع فاز پیوسته‌ی متناظر با سیستم مینیمم‌فاز در شکل 5.28(b) و فاز پیوسته‌ی سیستم all-pass (تمام‌گذر) که در شکل 5.29(b) نشان داده شده است.
همان‌طور که در بخش 5.5 نیز مشخص شد، و طبق فاز اصلی نمودارهای شکل‌های 5.22(b)، 5.23(b)، 5.24(b)، و 5.29(b)، فاز پیوسته‌ی سیستم تمام‌گذر همیشه منفی است.




برای بازه $0 \leq \omega \leq \pi$، بازتاب صفرهای تابع $H_{\min}(z)$ از درون دایره واحد به مکان‌های مزدوج وارون در بیرون، همیشه فاز پیوسته را کاهش می‌دهد یا مقدار منفی فاز (که به آن تابع **تأخیر فاز** یا **phase-lag** گفته می‌شود) را افزایش می‌دهد.

بنابراین، سیستم علّی و پایدار که $|H_{\min}(e^{j\omega})|$ را به عنوان پاسخ اندازه فرکانسی خود دارد و همچنین همه صفرها و (البته) قطب‌هایش درون دایره واحد قرار دارند، دارای کمترین مقدار تأخیر فاز است (برای همه‌ی $0 \leq \omega < \pi$) در بین تمام سیستم‌هایی که همان پاسخ اندازه را دارند.

از این رو، اصطلاح دقیق‌تر **سیستم با کم‌ترین تأخیر فاز** (*minimum phase-lag system*) است، اما اصطلاح تاریخی و رایج **مینیمم‌فاز** (*minimum phase*) است.



برای آن‌که تفسیر سیستم‌های با کم‌ترین تأخیر فاز دقیق‌تر شود، باید قید اضافی زیر را اعمال کنیم که مقدار تابع تبدیل در $\omega = 0$ مثبت باشد:

<div dir="ltr">

$$
H(e^{j0}) = \sum_{n=-\infty}^{\infty} h[n] > 0 \tag{5.112}
$$

</div>

توجه داشته باشید که $H(e^{j0})$ زمانی حقیقی خواهد بود که $h[n]$ را حقیقی فرض کنیم. قید معادله (5.112) لازم است چون سیستمی با پاسخ ضربه‌ای $-h[n]$ همان صفرها و قطب‌ها را دارد که سیستمی با پاسخ ضربه‌ای $h[n]$. اما ضرب در $-1$ باعث ایجاد تغییر فاز به مقدار $\pi$ رادیان می‌شود.

بنابراین، برای حذف این ابهام، باید شرط معادله (5.112) را اعمال کنیم تا تضمین کنیم سیستمی با همه قطب‌ها و صفرهایش درون دایره واحد، کم‌ترین تأخیر فاز را نیز دارد.

با این‌حال، این قید در عمل اهمیت زیادی ندارد و تعریف ما در ابتدای بخش ۵.۶ که این قید را شامل نمی‌شود، همان تعریفی است که معمولاً برای طبقه‌بندی سیستم‌های مینیمم‌فاز به کار می‌رود.

</div>


<div dir="rtl">

#### خاصیت تأخیر گروهی کمینه (*The Minimum Group-Delay Property*)

مثال ۵.۱۵ ویژگی دیگری از سیستم‌هایی را نشان می‌دهد که همه قطب‌ها و صفرهایشان درون دایره واحد قرار دارند. نخست توجه کنید که تأخیر گروهی برای سیستم‌هایی که پاسخ اندازه فرکانسی یکسان دارند به‌صورت زیر بیان می‌شود:
<div dir="ltr">

$$
\text{grd}[H(e^{j\omega})] = \text{grd}[H_{\min}(e^{j\omega})] + \text{grd}[H_{\text{ap}}(e^{j\omega})] \tag{5.113}
$$

</div>
تأخیر گروهی برای سیستم مینیمم‌فاز نشان داده‌شده در شکل 5.28(c)، همیشه کمتر از تأخیر گروهی برای سیستم غیرمینیمم‌فاز در شکل 5.27(c) است. دلیل این موضوع آن است که، همان‌طور که شکل 5.29(c) نشان می‌دهد، سیستم تمام‌عبوری (*all-pass*) که سیستم مینیمم‌فاز را به سیستم غیرمینیمم‌فاز تبدیل می‌کند، تأخیر گروهی مثبتی دارد.

در بخش ۵.۵ نشان دادیم که این یک ویژگی عمومی برای همه سیستم‌های تمام‌عبور است؛ آن‌ها همواره برای همه $\omega$ تأخیر گروهی مثبت دارند. بنابراین، اگر دوباره تمام سیستم‌هایی را در نظر بگیریم که پاسخ اندازه‌ی آن‌ها برابر با $|H_{\min}(e^{j\omega})|$ است، سیستمی که تمام قطب‌ها و صفرهایش داخل دایره واحد قرار دارند دارای کمترین تأخیر گروهی خواهد بود.

نام مناسب دیگر برای چنین سیستم‌هایی می‌تواند **سیستم‌های با کمترین تأخیر گروهی** (*minimum group-delay systems*) باشد، اگرچه این اصطلاح معمولاً استفاده نمی‌شود.



#### خاصیت تأخیر انرژی کمینه (*The Minimum Energy-Delay Property*)

در مثال ۵.۱۵، مجموعاً چهار سیستم FIR علّی با پاسخ ضربه‌ای حقیقی وجود دارد که پاسخ اندازه‌ی فرکانسی آن‌ها با سیستم معادله (5.107) یکسان است. نمودار صفر-قطب متناظر در شکل 5.30 نشان داده شده، جایی که شکل 5.30(d) مربوط به معادله (5.107) و شکل 5.30(a) مربوط به سیستم مینیمم‌فاز معادله (5.109) است. پاسخ‌های ضربه‌ای برای این چهار مورد در شکل 5.31 رسم شده‌اند.



</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# صفرها و معکوس‌شان
zeros = {
    '0.9e^{±j0.6π}': [0.9 * np.exp(1j * 0.6 * np.pi), 0.9 * np.exp(-1j * 0.6 * np.pi)],
    '0.8e^{±j0.8π}': [0.8 * np.exp(1j * 0.8 * np.pi), 0.8 * np.exp(-1j * 0.8 * np.pi)]
}

reciprocals = {
    '1/(0.9e^{±j0.6π})': [1/z for z in zeros['0.9e^{±j0.6π}']],
    '1/(0.8e^{±j0.8π})': [1/z for z in zeros['0.8e^{±j0.8π}']]
}

# ترکیب‌های مختلف صفرها برای چهار حالت
zero_configs = {
    'a': zeros['0.9e^{±j0.6π}'] + zeros['0.8e^{±j0.8π}'],  # همه صفرها داخل دایره واحد (minimum phase)
    'b': reciprocals['1/(0.9e^{±j0.6π})'] + reciprocals['1/(0.8e^{±j0.8π})'],  # همه بیرون (maximum phase)
    'c': zeros['0.9e^{±j0.6π}'] + reciprocals['1/(0.8e^{±j0.8π})'],
    'd': reciprocals['1/(0.9e^{±j0.6π})'] + zeros['0.8e^{±j0.8π}']
}

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
titles = ['(a)', '(b)', '(c)', '(d)']

for ax, key, title in zip(axs.flat, zero_configs.keys(), titles):
    ax.set_title(f'z-plane {title}')
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.set_aspect('equal')
    ax.axhline(0, color='black')
    ax.axvline(0, color='black')
    
    # دایره واحد
    unit_circle = plt.Circle((0, 0), 1, color='black', fill=False, linestyle='--')
    ax.add_artist(unit_circle)
    
    # صفرها
    zeros_plot = zero_configs[key]
    ax.plot(np.real(zeros_plot), np.imag(zeros_plot), 'o', label='Zeros', markersize=8)
    
    # قطب مرتبه چهارم (فرضاً در مبدا)
    ax.plot(0, 0, 'k*', markersize=10, label='Fourth-order pole')

plt.tight_layout()
plt.show()
```

**شکل ۵.۳۰** — چهار سیستم که همگی دارای اندازه پاسخ فرکانسی یکسان هستند. صفرهای این سیستم‌ها در تمام ترکیب‌های ممکن از
$0.9 e^{\pm j 0.6\pi}$ و $0.8 e^{\pm j 0.8\pi}$
و معکوس‌های آن‌ها قرار دارند.



<div dir="rtl">

در میان چهار دنباله‌ی نشان داده شده در این شکل، مشاهده می‌کنیم که دنباله‌ی حداقل‌فاز (minimum-phase) به نظر می‌رسد در ابتدای خود (سمت چپ) دارای نمونه‌های بزرگ‌تری نسبت به سایر دنباله‌ها باشد. در واقع، این ویژگی در این مثال و به‌طور کلی برقرار است:
<div dir="ltr">

$$
|h[0]| \leq |h_{\text{min}}[0]| \tag{5.114}
$$

</div>

برای هر دنباله‌ی علی و پایدار $h[n]$ که در آن:

<div dir="ltr">

$$
|H(e^{j\omega})| = |H_{\text{min}}(e^{j\omega})| \tag{5.115}
$$

</div>

اثبات این ویژگی در مسئله 5.65 پیشنهاد شده است.

تمام پاسخ‌های ضربه‌ای که اندازه پاسخ فرکانسی آن‌ها برابر با $|H_{\text{min}}(e^{j\omega})|$ است، انرژی کلی برابری با $h_{\text{min}}[n]$ دارند، زیرا مطابق با قضیه‌ی پارسوال:

<div dir="ltr">

$$
\sum_{n=0}^{\infty} |h[n]|^2 = \frac{1}{2\pi} \int_{-\pi}^{\pi} |H(e^{j\omega})|^2 d\omega 
= \frac{1}{2\pi} \int_{-\pi}^{\pi} |H_{\text{min}}(e^{j\omega})|^2 d\omega 
= \sum_{n=0}^{\infty} |h_{\text{min}}[n]|^2 \tag{5.116}
$$

</div>

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# نمونه‌های زمانی
n = np.arange(7)

# چهار دنباله مختلف h[n]
h1 = [0, 0.81, 2.19, 2.89, 1.56, 0, 0]
h2 = [0, 1.56, 2.89, 2.19, 0.81, 0, 0]
h3 = [0, 1.00, 2.58, 2.51, 1.25, 0, 0]
h4 = [0, 1.25, 2.51, 2.58, 1.00, 0, 0]

# لیست دنباله‌ها و عنوان‌های مربوط
h_list = [h1, h2, h3, h4]
titles = [r'$h_1[n]$', r'$h_2[n]$', r'$h_3[n]$', r'$h_4[n]$']

# ترسیم نمودار
plt.figure(figsize=(12, 4))  # بهتر برای فضای Jupyter
for i, h in enumerate(h_list):
    ax = plt.subplot(1, 4, i + 1)
    ax.stem(n, h, basefmt=" ", linefmt='b-', markerfmt='bo')
    ax.set_title(titles[i])
    ax.set_xlabel("n")
    ax.set_ylabel(r'$h[n]$')
    ax.grid(True)
    ax.set_xticks(n)
    ax.set_ylim(0, 4)

plt.tight_layout()
plt.show()
```

<div dir="rtl"> شکل 5.31 — دنباله‌های زمانی مربوط به نمودار قطب-صفرهای شکل 5.30. </div>

<div dir="rtl">

اگر انرژی جزئی پاسخ ضربه را به صورت زیر تعریف کنیم:

</div>

$$
E[n] = \sum_{m=0}^{n} |h[m]|^2 \tag{5.117}
$$

<div dir="rtl">

آنگاه می‌توان نشان داد که (رجوع شود به مسئله 5.66):

</div>

$$
\sum_{m=0}^{n} |h[m]|^2 \leq \sum_{m=0}^{n} |h_{\text{min}}[m]|^2 \tag{5.118}
$$



```python
import numpy as np
import matplotlib.pyplot as plt

# داده‌های چهار دنباله از شکل 5.31
ha = np.array([1.56, 2.89, 2.19, 0.81, 0.00, 0.00])  # minimum phase
hb = np.array([0.81, 2.19, 2.89, 1.56, 0.00, 0.00])  # maximum phase
hc = np.array([1.25, 2.51, 2.58, 1.00, 0.00, 0.00])
hd = np.array([1.00, 2.58, 2.51, 1.25, 0.00, 0.00])

# محاسبه انرژی جزئی
Ea = np.cumsum(ha**2)
Eb = np.cumsum(hb**2)
Ec = np.cumsum(hc**2)
Ed = np.cumsum(hd**2)

# ترسیم نمودار
n = np.arange(len(ha))

plt.plot(n, Ea, label=r'$E_a[n]$ (minimum phase)', color='black')
plt.plot(n, Eb, label=r'$E_b[n]$ (maximum phase)', color='black', linestyle='dotted')
plt.plot(n, Ec, label=r'$E_c[n]$', color='black', linestyle='-.')
plt.plot(n, Ed, label=r'$E_d[n]$', color='black', linestyle='--')

plt.xlabel(r'$n$')
plt.ylabel('Partial energy')
plt.title('Partial energies for the four sequences')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

<div dir="rtl">

**شکل 5.32** انرژی‌های جزئی برای چهار دنباله‌ی شکل 5.31.
(توجه کنید که $E_a[n]$ مربوط به دنباله‌ی فاز-کمینه $h_a[n]$ و $E_b[n]$ مربوط به دنباله‌ی فاز-بیشینه $h_b[n]$ است.)

</div>

<div dir="rtl">

برای تمام پاسخ‌های ضربه‌ای $h[n]$ که به خانواده‌ای از سیستم‌ها با پاسخ بزرگی مطابق با معادله (5.115) تعلق دارند، طبق معادله (5.118)، انرژی جزئی سیستم فاز-کمینه بیشترین تمرکز را در اطراف $n = 0$ دارد؛ به‌عبارتی دیگر، انرژی سیستم فاز-کمینه کمترین تأخیر را در میان تمام سیستم‌هایی با پاسخ بزرگی یکسان دارد. به همین دلیل، سیستم‌های فاز-کمینه (یا دارای تأخیر کم) گاهی با عنوان **سیستم‌های با حداقل تأخیر انرژی** یا به‌اختصار **سیستم‌های با حداقل تأخیر** شناخته می‌شوند.

این ویژگی تأخیر در شکل 5.32 نشان داده شده است، که نمودار انرژی جزئی برای چهار دنباله در شکل 5.31 را نمایش می‌دهد. در این مثال، و به‌طور کلی، کمترین تأخیر انرژی مربوط به سیستمی است که تمام صفرهایش **داخل دایره واحد** قرار دارند (یعنی سیستم فاز-کمینه) و بیشترین تأخیر انرژی مربوط به سیستمی است که تمام صفرهایش **خارج از دایره واحد** قرار دارند. به سیستم‌های با بیشترین تأخیر انرژی نیز معمولاً **سیستم‌های فاز-بیشینه** گفته می‌شود.

</div>


<div dir="rtl">

**۵.۷ سیستم‌های خطی با فاز خطی تعمیم‌یافته**

در طراحی فیلترها و دیگر سیستم‌های پردازش سیگنال که بخشی از باند فرکانسی را بدون اعوجاج عبور می‌دهند، مطلوب است که پاسخ فرکانسی (از نظر دامنه) تقریباً ثابت و فاز صفر در آن باند داشته باشند.
برای سیستم‌های علّی، داشتن فاز صفر امکان‌پذیر نیست و در نتیجه، باید مقداری اعوجاج فاز را پذیرفت.

همان‌طور که در بخش 5.1.2 دیدیم، اثر فاز خطی با شیب عدد صحیح صرفاً یک انتقال زمانی ساده است.
در مقابل، فاز غیرخطی می‌تواند تأثیر زیادی بر شکل سیگنال بگذارد، حتی زمانی که اندازه پاسخ فرکانسی ثابت باقی مانده است.

بنابراین، در بسیاری از شرایط، طراحی سیستم‌هایی که دقیقاً یا تقریباً دارای فاز خطی هستند بسیار مطلوب است.
در این بخش، به صورت رسمی و تعمیم‌یافته به مفاهیم فاز خطی و تأخیر زمانی ایده‌آل خواهیم پرداخت؛ با در نظر گرفتن دسته‌ای از سیستم‌ها که دارای تأخیر گروهی ثابت هستند.
ابتدا مفهوم تأخیر در یک سیستم گسسته‌زمان را بازنگری می‌کنیم.

</div>

<div dir="rtl">

## 5.7.1 سیستم‌های با فاز خطی

یک سیستم LTI را در نظر بگیرید که پاسخ فرکانسی آن در یک دوره برابر است با:

<div dir="ltr">

$$
H_{\text{id}}(e^{j\omega}) = e^{-j\omega\alpha}, \qquad |\omega| < \pi \qquad \text{(5.119)}
$$

</div>

که در آن، $$\alpha$$ یک عدد حقیقی است (لزومی ندارد عدد صحیح باشد).  
چنین سیستمی به عنوان یک «سیستم تأخیر ایده‌آل» شناخته می‌شود، که در آن \( \alpha \) مقدار تأخیری است که سیستم اعمال می‌کند.

توجه داشته باشید که این سیستم دارای پاسخ اندازهٔ ثابت، فاز خطی، و تأخیر گروهی ثابت است. یعنی:
<div dir="ltr">

$$
|H_{\text{id}}(e^{j\omega})| = 1 \qquad \text{(5.120a)}
$$

$$
\angle H_{\text{id}}(e^{j\omega}) = -\omega\alpha \qquad \text{(5.120b)}
$$

$$
\mathrm{grd}[H_{\text{id}}(e^{j\omega})] = \alpha \qquad \text{(5.120c)}
$$

</div>
---

تبدیل فوریه معکوس $$H_{\text{id}}(e^{j\omega})$$، پاسخ ضربه‌ای زیر را می‌دهد:

<div dir="ltr">

$$
h_{\text{id}}[n] = \frac{\sin \pi(n - \alpha)}{\pi(n - \alpha)}, \qquad -\infty < n < \infty
\qquad \text{(5.121)}
$$

</div>

خروجی این سیستم برای ورودی \( x[n] \) به صورت زیر خواهد بود:

<div dir="ltr">

$$
y[n] = x[n] * \frac{\sin \pi(n - \alpha)}{\pi(n - \alpha)} 
= \sum_{k = -\infty}^{\infty} x[k] \cdot \frac{\sin \pi(n - k - \alpha)}{\pi(n - k - \alpha)}
\qquad \text{(5.122)}
$$

</div>

اگر \( \alpha = n_d \) باشد، که \( n_d \) یک عدد صحیح است، آنگاه همان‌طور که در بخش ۵.۱.۲ اشاره شد:

<div dir="ltr">

$$
h_{\text{id}}[n] = \delta[n - n_d] \qquad \text{(5.123)}
$$

</div>

</div>

<div dir="rtl">

و داریم:
<div dir="ltr">

\begin{equation}
y[n] = x[n] * \delta[n - n_d] = x[n - n_d] \tag{5.124}
\end{equation}

</div>


یعنی اگر $\alpha = n_d$ یک عدد صحیح باشد، سیستم با فاز خطی و بهره‌ی واحد در رابطه (5.119) صرفاً دنباله‌ی ورودی را به اندازه‌ی $n_d$ نمونه به تأخیر می‌اندازد. اگر $\alpha$ عدد صحیح نباشد، ساده‌ترین تفسیر، همان است که در مثال 4.9 فصل 4 ارائه شده است. به‌طور خاص، یک نمایش از سیستم رابطه‌ی (5.119) آن چیزی است که در شکل 5.33 نشان داده شده، که در آن:

<div dir="ltr">

$$h_c(t) = \delta(t - \alpha T) \quad \text{و} \quad H_c(j\Omega) = e^{-j\Omega \alpha T}$$

</div>

پس داریم:
\begin{equation}
H(e^{j\omega}) = e^{-j\omega\alpha}, \qquad |\omega| < \pi \tag{5.125}
\end{equation}

در این نمایش، انتخاب $T$ اهمیتی ندارد و می‌تواند به سادگی برابر با ۱ قرار داده شود. تأکید مجدد بر این نکته مهم است که این نمایش معتبر است، چه $x[n]$ با نمونه‌برداری از یک سیگنال پیوسته در زمان به‌دست آمده باشد یا نه. مطابق با نمایش در شکل 5.33، $y[n]$ دنباله‌ای از نمونه‌های سیگنال $x[n]$ است که نسبت به ورودی $x[n]$ با تأخیر $\alpha T$ شیفت یافته است؛ یعنی:
<div dir="ltr">

$$
y[n] = x_c(nT - \alpha T)
$$

این بحث همچنین یک تفسیر مفید از فاز خطی فراهم می‌کند، زمانی که این فاز با یک پاسخ بهره غیرثابت همراه باشد. به عنوان مثال، یک پاسخ فرکانسی کلی‌تر با فاز خطی را در نظر بگیرید:
<div dir="ltr">

\begin{equation}
H(e^{j\omega}) = |H(e^{j\omega})| e^{-j\omega \alpha}, \quad |\omega| < \pi
\tag{5.126}
\end{equation}

</div>

معادله (5.126) تفسیر ارائه‌شده در شکل 5.34 را پیشنهاد می‌دهد. سیگنال \( x[n] \) ابتدا توسط پاسخ فرکانسی با فاز صفر ∣H(ejω)∣ فیلتر می‌شود، و سپس خروجی فیلتر شده به اندازه \(\alpha\) (که می‌تواند عددی صحیح یا غیرصحیح باشد) در زمان «جابجا» می‌شود. 

فرض کنید، برای مثال، که \( H(e^{j\omega}) \) یک فیلتر پایین‌گذر ایده‌آل با فاز خطی باشد، در این صورت داریم:

<div dir="ltr">

\begin{equation}
H_p(e^{j\omega}) =
\begin{cases}
e^{-j\omega \alpha}, & |\omega| < \omega_c \\
0, & \omega_c < |\omega| \leq \pi
\end{cases}
\tag{5.127}
\end{equation}

</div>

پاسخ ضربه‌ای متناظر با آن برابر است با:

<div dir="ltr">

\begin{equation}
h_p[n] = \frac{\sin\left( \omega_c (n - \alpha) \right)}{\pi (n - \alpha)}
\tag{5.128}
\end{equation}

</div>

توجه داشته باشید که اگر ωc=π باشد، رابطه (5.121) به‌دست می‌آید.



</div>

### مثال ۵٫۱۶: فیلتر پایین‌گذر ایده‌آل با فاز خطی

پاسخ ضربه‌ای فیلتر پایین‌گذر ایده‌آل برخی ویژگی‌های جالب سیستم‌های دارای فاز خطی را نشان می‌دهد.  
شکل 5.35(a) پاسخ ضربه‌ای \( h_{lp}[n] \) را برای \( \omega_c = 0.4\pi \) و \( \alpha = n_d = 5 \) نشان می‌دهد.

توجه کنید زمانی که \( \alpha \) عددی صحیح باشد، پاسخ ضربه‌ای نسبت به \( n = n_d \) متقارن است، یعنی:
<div dir="ltr">

$$
h_{lp}[2n_d - n] = \frac{\sin\left( \omega_c (2n_d - n - n_d) \right)}{\pi (2n_d - n - n_d)} = 
\frac{\sin\left( \omega_c (n_d - n) \right)}{\pi (n_d - n)} = h_{lp}[n]
\qquad \text{(5.129)}
$$

</div>
---

در این حالت، می‌توان سیستمی با فاز صفر تعریف کرد:

<div dir="ltr">

$$
\hat{H}_{lp}(e^{j\omega}) = H_{lp}(e^{j\omega}) e^{j\omega n_d} = \left| H_{lp}(e^{j\omega}) \right|
\qquad \text{(5.130)}
$$

</div
>
که در آن پاسخ ضربه‌ای به اندازه \( n_d \) نمونه به سمت چپ انتقال داده شده و دنباله‌ای زوج حاصل می‌شود:

<div dir="ltr">

$$
\hat{h}_{lp}[n] = \frac{\sin(\omega_c n)}{\pi n} = \hat{h}_{lp}[-n]
\qquad \text{(5.131)}
$$

</div>

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
omega_c = 0.4 * np.pi
n_d = 5
n = np.arange(-5, 15)

# محاسبه پاسخ ضربه‌ای h[n]
def h_lp(n, n_d, omega_c):
    h = np.zeros_like(n, dtype=float)
    for i, ni in enumerate(n):
        if ni == n_d:
            h[i] = omega_c / np.pi  # چون sin(0)/0 → ω_c/π
        else:
            h[i] = np.sin(omega_c * (ni - n_d)) / (np.pi * (ni - n_d))
    return h

h = h_lp(n, n_d, omega_c)

# ترسیم نمودار
plt.stem(n, h, linefmt='k', markerfmt='ko', basefmt='k')
plt.plot(n, h, 'k--')  # اتصال نقاط با خط چین
plt.xlabel("Sample number (n)")
plt.ylabel("Amplitude")
plt.title("(a)")
plt.grid(True)
plt.ylim(-0.2, 0.6)
plt.xlim(-5, 15)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
omega_c = 0.4 * np.pi
n = np.arange(-5, 15)

# محاسبه پاسخ ضربه‌ای زوج (even)
def h_lp_even(n, omega_c):
    h = np.zeros_like(n, dtype=float)
    for i, ni in enumerate(n):
        if ni == 0:
            h[i] = omega_c / np.pi  # چون sin(0)/0 → ω_c/π
        else:
            h[i] = np.sin(omega_c * ni) / (np.pi * ni)
    return h

h_even = h_lp_even(n, omega_c)

# ترسیم نمودار
plt.stem(n, h_even, linefmt='k', markerfmt='ko', basefmt='k')
plt.plot(n, h_even, 'k--')  # اتصال نقاط با خط‌چین
plt.xlabel("Sample number (n)")
plt.ylabel("Amplitude")
plt.title("(b)")
plt.grid(True)
plt.ylim(-0.2, 0.6)
plt.xlim(-5, 15)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
omega_c = 0.4 * np.pi
n = np.arange(-5, 15)
alpha = 4.3  # تأخیر غیردرست

# پاسخ ضربه‌ای با تأخیر غیردرست
def h_lp_fractional_delay(n, omega_c, alpha):
    h = np.zeros_like(n, dtype=float)
    for i, ni in enumerate(n):
        if ni == alpha:
            h[i] = omega_c / np.pi
        else:
            h[i] = np.sin(omega_c * (ni - alpha)) / (np.pi * (ni - alpha))
    return h

h_frac = h_lp_fractional_delay(n, omega_c, alpha)

# ترسیم نمودار
plt.stem(n, h_frac, linefmt='k', markerfmt='ko', basefmt='k')
plt.plot(n, h_frac, 'k--')  # خط‌چین برای نمایش پیوسته
plt.xlabel("Sample number (n)")
plt.ylabel("Amplitude")
plt.title("(c)")
plt.grid(True)
plt.ylim(-0.2, 0.6)
plt.xlim(-5, 15)
plt.show()
```

**۵.۷.۲ فاز خطی تعمیم‌یافته**

در بحث ارائه‌شده در بخش 5.7.1، ما به بررسی دسته‌ای از سیستم‌ها پرداختیم که پاسخ فرکانسی آن‌ها به‌صورت معادله (5.126) است؛ یعنی تابعی حقیقی و نامنفی از ω که در یک جمله فاز خطی e−jωα ضرب شده است. برای پاسخ فرکانسی با این فرم، فاز تابع H(ejω) به‌طور کامل با جمله فاز خطی e−jωα مرتبط است، یعنی:

<div dir="ltr">  

 $$∠H(ejω)=−ωα$$  
 
  </div>

و در نتیجه، سیستم‌هایی با این ویژگی‌ها به عنوان سیستم‌های با فاز خطی شناخته می‌شوند. در میانگین‌گیری متحرکِ مثال 4.10، پاسخ فرکانسی در معادله (4.67) تابعی حقیقی از ωω است که در یک جمله فاز خطی ضرب شده است، اما این سیستم، به‌طور دقیق، یک سیستم با فاز خطی محسوب نمی‌شود، زیرا در فرکانس‌هایی که عامل زیر:
<div dir="ltr">

\begin{equation}
\frac{1}{M+1} \cdot \frac{\sin\left(\omega \frac{M+1}{2}\right)}{\sin\left(\omega/2\right)}
\end{equation}

</div>

منفی می‌شود، این جمله فازی اضافی به مقدار ππ رادیان به فاز کلی اضافه می‌کند.

بسیاری از مزایای سیستم‌های با فاز خطی شامل سیستم‌هایی با پاسخ فرکانسی به فرم رابطه‌ی (4.67) نیز می‌شود، و بنابراین، مفید است که تعریف و مفهوم فاز خطی را تا حدی تعمیم دهیم. به طور خاص، سیستمی با عنوان سیستم با فاز خطی تعمیم‌یافته (generalized linear-phase system) شناخته می‌شود اگر پاسخ فرکانسی آن بتواند به فرم زیر نوشته شود:

<div dir="ltr">

\begin{equation}
H(e^{j\omega}) = A(e^{j\omega}) e^{-j\alpha \omega + j \beta}, \tag{5.135}
\end{equation}

</div>

که در آن α و β مقادیری ثابت هستند و A(ejω) تابعی حقیقی (و احتمالاً دوقطبی) از ω است. برای مثال، پاسخ فاز خطی رابطه‌ی (5.127) و پاسخ میانگین‌گیری متحرک مثال ۴.۱۰ زمانی که β=0 باشد، در این قالب قرار می‌گیرند. همچنین می‌توان دید که تفاضل‌گیر باندمحدود مثال ۴.۵ نیز پاسخ فرکانسی‌اش به صورت رابطه‌ی (5.135) با مقادیر α=0، β=π/2 و A(ejω)=ω/T نوشته می‌شود.





سیستمی که پاسخ فرکانسی آن به صورت رابطه‌ی (5.135) باشد، یک سیستم با فاز خطی تعمیم‌یافته نامیده می‌شود؛ زیرا فاز چنین سیستمی شامل جمله‌هایی ثابت است که به تابع خطی −ωα اضافه شده‌اند. به عبارت دیگر، −ωα+β معادله یک خط راست است.

با این حال، اگر ناپیوستگی‌هایی را که از افزودن یک فاز ثابت بر کل یا بخشی از باند ∣ω∣<π ناشی می‌شود، نادیده بگیریم، آنگاه چنین سیستمی را می‌توان با تأخیر گروهی ثابت توصیف کرد. به عبارت دیگر، کلاس سیستم‌هایی که در آن‌ها:
<div dir="ltr">

\begin{equation}
\tau(\omega) = \mathrm{grd}[H(e^{j\omega})] = -\frac{d}{d\omega} \left[\arg(H(e^{j\omega}))\right] = \alpha \tag{5.136}
\end{equation}

</div>
دارای فاز خطی از فرم کلی زیر هستند:
<div dir="ltr">

\begin{equation}
\arg[H(e^{j\omega})] = \beta - \omega \alpha, \quad 0 < \omega < \pi, \tag{5.137}
\end{equation}

</div>
که در آن β و α هر دو مقادیر حقیقی ثابت هستند.

</div>

<div dir="rtl">

به یاد بیاورید که در بخش 5.7.1 نشان دادیم پاسخ ضربه‌ای سیستم‌های با فاز خطی ممکن است در n=2α (زمانی که 2α2α یک عدد صحیح است) تقارن داشته باشند. برای بررسی پیامد این تقارن در سیستم‌های با فاز خطی تعمیم‌یافته، مفید است که معادله‌ای را استخراج کنیم که باید توسط h[n]، α، و β برای سیستم‌هایی با تأخیر گروهی ثابت برآورده شود.

این معادله با در نظر گرفتن این موضوع استخراج می‌شود که برای چنین سیستم‌هایی، پاسخ فرکانسی را می‌توان به شکل زیر نوشت:
<div dir="ltr">

\begin{align*}
H(e^{j\omega}) &= A(e^{j\omega}) e^{j(\beta - \alpha \omega)} \\
&= A(e^{j\omega}) \cos(\beta - \omega \alpha) + j A(e^{j\omega}) \sin(\beta - \omega \alpha),
\end{align*}

</div>
یا به طور معادل:
<div dir="ltr">

\begin{equation}
H(e^{j\omega}) = \sum_{n=-\infty}^{\infty} h[n] e^{-j\omega n} \tag{5.139}
\end{equation}

</div>
 چون فرض کرده‌ایم h[n] حقیقی است، می‌توان نوشت:

 <div dir="ltr">

\begin{equation}
 H(e^{j\omega}) = \sum_{n=-\infty}^{\infty} h[n] \cos \omega n - j \sum_{n=-\infty}^{\infty} h[n] \sin \omega n
 \end{equation}

 </div>

زاویه فاز H(ejω) به صورت زیر بیان می‌شود:
<div dir="ltr">

\begin{equation}
\tan(\beta - \omega \alpha) = \frac{\sum_{n=-\infty}^{\infty} h[n] \sin \omega n}{\sum_{n=-\infty}^{\infty} h[n] \cos \omega n}
\end{equation}

</div>

با ضرب طرفین و ترکیب جملات با استفاده از یک رابطه مثلثاتی، به معادله زیر می‌رسیم:
<div dir="ltr">

\begin{equation}
\sum_{n=-\infty}^{\infty} h[n] \sin[\omega(n - \alpha) + \beta] = 0 \quad \text{for all } \omega \tag{5.140}
\end{equation}

</div>

این معادله یک شرط لازم برای h[n]، α و β است تا سیستم دارای تأخیر گروهی ثابت باشد. با این حال، این یک شرط کافی نیست و به دلیل ماهیت ضمنی‌اش، به ما نمی‌گوید چگونه می‌توان یک سیستم با فاز خطی ساخت.

برای مثال، می‌توان نشان داد که مجموعه‌ای از شرایط زیر معادله (5.140) را ارضا می‌کنند:
<div dir="ltr">

\begin{equation}
\beta = 0 \quad \text{or} \quad \pi, \tag{5.141a}
\end{equation}

</div>

<div dir="ltr">

\begin{equation}
2\alpha = M = \text{an integer}, \tag{5.141b}
\end{equation}

\begin{equation}
h[2\alpha - n] = h[n]. \tag{5.141c}
\end{equation}

</div>
وقتی که β=0 یا π، معادله (5.140) به شکل زیر در‌می‌آید:

<div dir="ltr">

\begin{equation}
\sum_{n=-\infty}^{\infty} h[n] \sin[\omega(n - \alpha)] = 0 \tag{5.142}
\end{equation}

</div>
که از آن می‌توان نتیجه گرفت اگر 2α2α یک عدد صحیح باشد، آنگاه جملات موجود در معادله (5.142) را می‌توان طوری جفت کرد که هر جفت جمله برای تمام ωω به‌طور کامل صفر شود. این شرایط در نهایت نشان می‌دهند که پاسخ فرکانسی متناظر، فرم معادله (5.135) را دارد با:

    β=0 یا π

    A(ejω) تابعی زوج (و البته حقیقی) از ω است.

از طرفی، اگر β=π/2 یا 3π/23π/2 باشد، آنگاه معادله (5.140) به شکل زیر در‌می‌آید:

<div dir="ltr">

\begin{equation}
\sum_{n=-\infty}^{\infty} h[n] \cos[\omega(n - \alpha)] = 0 \tag{5.143}
\end{equation}

</div>
و می‌توان نشان داد که در این حالت باید:

<div dir="ltr">

\begin{equation}
\beta = \pi/2 \quad \text{or} \quad 3\pi/2 \tag{5.144a}
\end{equation}

\begin{equation}
2\alpha = M = \text{an integer} \tag{5.144b}
\end{equation}

\begin{equation}
h[2\alpha - n] = -h[n] \tag{5.144c}
\end{equation}

</div>
توجه داشته باشید که معادلات (5.141) و (5.144) دو مجموعه شرط ارائه می‌دهند که فاز خطی تعمیم‌یافته یا تأخیر گروهی ثابت را تضمین می‌کنند، اما همان‌طور که قبلاً در شکل 5.35(c) دیدیم، سیستم‌های دیگری نیز وجود دارند که معادله (5.135) را بدون داشتن این شرایط تقارنی ارضا می‌کنند.

</div>

## 5.7.3 سیستم‌های علی با فاز خطی تعمیم‌یافته (Causal Generalized Linear-Phase Systems)

اگر سیستم علی باشد، آنگاه معادله (5.140) به صورت زیر در می‌آید:

<div dir="ltr">

$$
\sum_{n=0}^{\infty} h[n] \sin[\omega(n - \alpha) + \beta] = 0 \quad \text{for all } \omega
\qquad \text{(5.145)}
$$

</div>

علیت و شرایط موجود در معادلات (5.141) و (5.144) به این معناست که:

<div dir="ltr">

$$
h[n] = 0, \quad n < 0 \quad \text{and} \quad n > M;
$$

</div>
یعنی، سیستم‌های FIR علی (با پاسخ ضربه‌ای محدود) فاز خطی تعمیم‌یافته دارند اگر طول پاسخ ضربه‌ای آن‌ها برابر \( M+1 \) باشد و یکی از شرایط معادله (5.141c) یا (5.144c) را برآورده کنند.





### اگر:

<div dir="ltr">

$$
h[n] =
\begin{cases}
h[M - n], & 0 \leq n \leq M \\
0, & \text{otherwise}
\end{cases}
\qquad \text{(5.146a)}
$$

</div>

آنگاه:

<div dir="ltr">

$$
H(e^{j\omega}) = A_e(e^{j\omega}) e^{-j\omega M/2}
\qquad \text{(5.146b)}
$$

</div>

که در آن \( A_e(e^{j\omega}) \) یک تابع حقیقی، زوج و پریودیک نسبت به \( \omega \) است.





### به‌طور مشابه، اگر:

<div dir="ltr">

$$
h[n] =
\begin{cases}
-h[M - n], & 0 \leq n \leq M \\
0, & \text{otherwise}
\end{cases}
\qquad \text{(5.147a)}
$$

</div>

آنگاه خواهیم داشت:

<div dir="ltr">

$$
H(e^{j\omega}) = j A_o(e^{j\omega}) e^{-j\omega M/2} = A_o(e^{j\omega}) e^{-j\omega M/2 + j\pi/2}
\qquad \text{(5.147b)}
$$

</div>

که در آن \( A_o(e^{j\omega}) \) تابعی حقیقی، فرد و پریودیک است. توجه کنید در هر دو حالت، طول پاسخ ضربه‌ای برابر با \( M+1 \) نمونه است.






### شرایط (5.146a) و (5.147a) برای داشتن فاز خطی تعمیم‌یافته **کافی** هستند، ولی **لازم** نیستند.  
طبق تحقیق Clemente و Pace (سال ۱۹۸۹)، پاسخ‌های ضربه‌ای نامحدود نیز ممکن است تبدیل فوریه‌ای با فاز خطی تعمیم‌یافته داشته باشند. ولی این سیستم‌ها عقلانی نیستند و با معادلات تفاضلی قابل پیاده‌سازی نمی‌باشند.






### انواع سیستم‌های FIR با فاز خطی تعمیم‌یافته





## Type I: سیستم‌های نوع اول

سیستمی که پاسخ ضربه‌ای متقارن دارد:

<div dir="ltr">

$$
h[n] = h[M - n], \quad 0 \leq n \leq M \qquad \text{(5.148)}
$$

</div>

که در آن \( M \) زوج است. بنابراین تأخیر \( M/2 \) یک عدد صحیح است.

<div dir="ltr">

$$
H(e^{j\omega}) = \sum_{n=0}^{M} h[n] e^{-j\omega n} \qquad \text{(5.149)}
$$

</div>

با اعمال تقارن داریم:

<div dir="ltr">

$$
H(e^{j\omega}) = e^{-j\omega M/2} \left[ \sum_{k=0}^{M/2} a[k] \cos(\omega k) \right]
\qquad \text{(5.150a)}
$$

</div>

که در آن:

<div dir="ltr">

$$
a[0] = h[M/2], \quad a[k] = 2h[M/2 - k], \quad k = 1, 2, \ldots, M/2
\qquad \text{(5.150b)}
$$

</div>






## Type II: سیستم‌های نوع دوم

سیستم نوع دوم نیز متقارن است:

<div dir="ltr">

$$
h[n] = h[M - n], \quad 0 \leq n \leq M
$$

</div>

ولی \( M \) فرد است.

<div dir="ltr">

$$
H(e^{j\omega}) = e^{-j\omega M/2} \left\{ \sum_{k=1}^{(M+1)/2} b[k] \cos\left[ \omega \left(k - \frac{1}{2} \right) \right] \right\}
$$

</div>

که در آن:

<div dir="ltr">

$$
b[k] = 2h\left[ \frac{(M+1)}{2} - k \right], \quad k = 1, 2, \ldots, \frac{(M+1)}{2}
$$

</div>






## Type III: سیستم‌های نوع سوم

اگر سیستم دارای پاسخ ضربه‌ای پادمتقارن باشد:

<div dir="ltr">

$$
h[n] = -h[M - n], \quad 0 \leq n \leq M \qquad \text{(5.152)}
$$

</div>

و \( M \) زوج باشد، آنگاه:

<div dir="ltr">

$$
H(e^{j\omega}) = je^{-j\omega M/2} \left[ \sum_{k=1}^{M/2} c[k] \sin(\omega k) \right] \qquad \text{(5.153a)}
$$

</div>

که در آن:

<div dir="ltr">

$$
c[k] = 2h\left[\frac{M}{2} - k\right], \quad k = 1, 2, \ldots, \frac{M}{2} \qquad \text{(5.153b)}
$$

</div>





## Type IV: سیستم‌های نوع چهارم

اگر پاسخ ضربه‌ای پادمتقارن باشد و \( M \) فرد باشد:

<div dir="ltr">

$$
H(e^{j\omega}) = je^{-j\omega M/2} \left[ \sum_{k=1}^{(M+1)/2} d[k] \sin\left( \omega \left(k - \frac{1}{2} \right) \right) \right]
\qquad \text{(5.154a)}
$$

</div>

که در آن:

<div dir="ltr">

$$
d[k] = 2h\left[ \frac{M+1}{2} - k \right], \quad k = 1, 2, \ldots, \frac{M+1}{2}
\qquad \text{(5.154b)}
$$

</div>

در این حالت \( H(e^{j\omega}) \) دارای شکل معادله (5.147b) است و تأخیر زمانی آن برابر با عددی نیم‌صحیح \( M/2 \) خواهد بود.

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_fir_type(h, M, title, symmetry_center_label=True):
    n = np.arange(0, M + 1)
    
    fig, ax = plt.subplots(figsize=(5, 3))
    
    # Plot stem safely (handling short h)
    markerline, stemlines, baseline = ax.stem(n, h, linefmt='b-', markerfmt='bo', basefmt='k-')
    
    # Center of symmetry
    center = M / 2
    ax.axvline(center, color='gray', linestyle='--', linewidth=1)
    
    # Annotate center
    if symmetry_center_label:
        ax.text(center + 0.2, max(np.abs(h)) * 0.8, "Center\nof\nsymmetry", fontsize=9, color='gray')

    # Set titles and axes
    ax.set_title(title, fontsize=11)
    ax.set_xlabel("n")
    ax.set_ylabel("h[n]")
    ax.set_xticks(n)
    ax.set_xlim(-0.5, M + 0.5)
    ax.grid(True, linestyle=':', alpha=0.6)
    
    plt.tight_layout()
    plt.show()

# Type I: M even, symmetric
M1 = 4
h1 = [1, 2, 3, 2, 1]
plot_fir_type(h1, M1, "(a) Type I: M even, h[n] = h[M - n]")

# Type II: M odd, symmetric
M2 = 5
h2 = [1, 2, 3, 3, 2, 1]
plot_fir_type(h2, M2, "(b) Type II: M odd, h[n] = h[M - n]")

# Type III: M even, antisymmetric
M3 = 2
h3 = [-1, 0, 1]
plot_fir_type(h3, M3, "(c) Type III: M even, h[n] = -h[M - n]")

# Type IV: M odd, antisymmetric
M4 = 1
h4 = [-1, 1]
plot_fir_type(h4, M4, "(d) Type IV: M odd, h[n] = -h[M - n]")
```

مثال 5.17 – سیستم نوع اول با فاز خطی (Type I Linear-Phase System)

اگر پاسخ ضربه‌ای به صورت زیر باشد:

</div>

\begin{equation}
h[n] =
\begin{cases}
1, & 0 \leq n \leq 4 \\
0, & \text{otherwise}
\end{cases}
\tag{5.155}
\end{equation}


```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف پاسخ ضربه h[n]
h = np.array([1, 1, 1, 1, 1])
N = 1024
w = np.linspace(0, 2 * np.pi, N)
H = np.fft.fft(h, N)
H = np.fft.fftshift(H)
w = np.fft.fftshift(w)

magnitude = np.abs(H)
phase = np.angle(H)
group_delay = -np.diff(np.unwrap(phase)) / np.diff(w)

# رسم نمودارها زیر هم، با ترتیب چپ به راست
fig, axs = plt.subplots(3, 1, figsize=(6, 10))

# (a) Magnitude
axs[2].plot(w, magnitude, 'k')
axs[2].set_title('(a) Magnitude', loc='left')
axs[2].set_ylabel('Amplitude', rotation=0, labelpad=35)
axs[2].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[2].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])
axs[2].grid(True)

# (b) Phase
axs[1].plot(w, phase, 'k')
axs[1].set_title('(b) Phase', loc='left')
axs[1].set_ylabel('Radians', rotation=0, labelpad=35)
axs[1].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[1].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])
axs[1].grid(True)

# (c) Group Delay
axs[0].plot(w[1:], group_delay, 'k')
axs[0].set_title('(c) Group Delay', loc='left')
axs[0].set_ylabel('Samples', rotation=0, labelpad=35)
axs[0].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[0].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])
axs[0].grid(True)

# تنظیم فاصله‌ها
plt.tight_layout()
plt.show()
```

<div dir="ltr">

\begin{equation}
H(e^{j\omega}) = \sum_{n=0}^{4} e^{-j\omega n}

\end{equation}

مثال 5.18 – سیستم خطی‌فاز نوع دوم (Type II)

مثال 5.20 – سیستم خطی‌فاز نوع چهارم (Type IV)

در این حالت، پاسخ ضربه‌ای به صورت زیر است:

که پاسخ فرکانسی آن به صورت زیر خواهد بود:

\tag{5.161}
\end{equation}

نمودار پاسخ فرکانسی این سیستم در شکل 5.40 آمده است. توجه داشته باشید که تأخیر گروهی برابر با 0.5 به ازای همه ی ω ها است.

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات
M = 10
omega = np.linspace(1e-5, 2 * np.pi, 1024)  # جلوگیری از تقسیم بر صفر با شروع از مقدار خیلی کوچک

# محاسبه پاسخ اندازه، فاز، و تأخیر گروهی
H = np.sin((M + 1) * omega / 2) / np.sin(omega / 2)
phase = -omega * M / 2
group_delay = np.full_like(omega, M / 2)

# رسم نمودارها
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

# Magnitude
axs[0].plot(omega, H)
axs[0].set_title('Magnitude')
axs[0].set_xlabel('Radian frequency (ω)')
axs[0].set_ylabel('Magnitude')
axs[0].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[0].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])

# Phase
axs[1].plot(omega, phase)
axs[1].set_title('Phase (rad)')
axs[1].set_xlabel('Radian frequency (ω)')
axs[1].set_ylabel('Phase')
axs[1].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[1].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])

# Group Delay
axs[2].plot(omega, group_delay)
axs[2].set_title('Group Delay (samples)')
axs[2].set_xlabel('Radian frequency (ω)')
axs[2].set_ylabel('Samples')
axs[2].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[2].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])

plt.tight_layout()
plt.show()
```

**(مکان‌یابی صفرهای سیستم‌های خطی با فاز خطی FIR)**

متن زیر خواص پاسخ ضربه و تابع سیستم را برای سیستم‌های FIR با فاز خطی بررسی کرده و مکان صفرهای تابع سیستم را تحلیل می‌کند.

تابع سیستم به صورت زیر تعریف می‌شود:

<div dir="ltr">

\begin{equation}
H(z) = \sum_{k=0}^{M} h[k] z^{-n} \tag{5.162}
\end{equation}

</div>

در حالت‌های متقارن (نوع I و II)، می‌توانیم از رابطه (5.148) برای نوشتن H(z)H(z) به فرم زیر استفاده کنیم:
<div dir="ltr">

\begin{align*}
H(z) &= \sum_{n=0}^{M} h[M - n] z^{-n} = \sum_{k=M}^{0} h[k] z^{k - M} \\
     &= z^{-M} H(z^{-1})
  \tag{5.163}
\end{align*}

</div>

از رابطه بالا نتیجه می‌گیریم که اگر z_0​ صفر تابع H(z) باشد، آنگاه:
<div dir="ltr">

\begin{equation}
H(z_0) = z_0^{-M} H(z_0^{-1}) = 0 \tag{5.164}
\end{equation}

</div
>
این نتیجه‌گیری می‌شود که اگر z0​=rejθ یک صفر از تابع H(z) باشد، آنگاه z0−1​=r−1e−jθ نیز یک صفر از H(z) خواهد بود.
زمانی که h[n] حقیقی باشد و z0​ یک صفر از H(z) باشد، آنگاه مزدوج آن z0∗​=re−jθ نیز یک صفر از H(z) خواهد بود، و طبق استدلال بالا،
(z0∗)−1=r−1ejθ نیز صفر خواهد بود.
بنابراین، وقتی h[n] حقیقی است، هر صفر مختلطی که روی دایره واحد نباشد، بخشی از یک مجموعه چهار عضوی از صفرهای مزدوج و معکوس خواهد بود، به فرم:

<div dir="ltr">

\begin{equation}
(1 - e^{j\theta_0} z^{-1})(1 - e^{-j\theta_0} z^{-1}) = 1 - 2\cos(\theta_0) z^{-1} + z^{-2}
\end{equation}

</div>

اگر یک صفر از H(z) روی دایره واحد باشد، یعنی z0=ejθ، آنگاه z0−1=e−jθ=z0∗​ خواهد بود. بنابراین صفرهای روی دایره واحد به صورت جفت‌هایی به شکل زیر ظاهر می‌شوند.

<div dir="ltr">

\begin{equation}
\left(1 - e^{j\theta} z^{-1} \right)\left(1 - e^{-j\theta} z^{-1} \right)
\end{equation}

</div>

اگر یک صفر از تابع H(z) حقیقی باشد و روی دایره واحد نباشد، معکوس آن نیز صفر دیگری از H(z) خواهد بود، و H(z) شامل فاکتورهایی به فرم زیر خواهد بود.
<div dir="ltr">

\begin{equation}
\left(1 \pm r z^{-1} \right)\left(1 \pm r^{-1} z^{-1} \right)
\end{equation}

</div>
در نهایت، صفر تابع H(z)H(z) در z=±1 می‌تواند به‌تنهایی ظاهر شود، زیرا ±1 معکوس خودش و مزدوج خودش است. بنابراین ممکن است عوامل تابع H(z) به صورت زیر نیز ظاهر شوند:
<div dir="ltr">

\begin{equation}
(1 \pm z^{-1})
\end{equation}

</div>

حالت صفر بودن در z=−1z=−1 به‌طور خاص مهم است. از رابطه (5.163):

<div dir="ltr">

\begin{equation}
H(-1) = (-1)^M H(-1)
\end{equation}

</div>

اگر M عددی زوج باشد، این رابطه یک همانی ساده است. اما اگر M فرد باشد، آنگاه:

<div dir="ltr">

\begin{equation}
H(-1) = -H(-1)
\end{equation}

</div>

که به این معناست که:
<div dir="ltr">

\begin{equation}
H(-1) = 0
\end{equation}

</div>

بنابراین، برای پاسخ ضربه‌ای متقارن با M تابع سیستم باید صفر باشد در Z=-1.

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# تابع کمکی برای رسم دایره واحد
def plot_unit_circle(ax):
    theta = np.linspace(0, 2*np.pi, 400)
    ax.plot(np.cos(theta), np.sin(theta), 'gray', linestyle='--')
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    ax.set_aspect('equal')
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_xlabel(r'$\Re e$')
    ax.set_ylabel(r'$\Im$', rotation=0, labelpad=10)
    ax.set_title('Unit circle')

# تعریف صفرهای تقریبی برای هر نوع (برای اهداف نمایشی)
zeros_type_I  = [0.8 + 0.6j, 0.8 - 0.6j, 1.2, 1/1.2]  # متقارن نسبت به دایره واحد
zeros_type_II = [1, -1, 0.9 + 0.4j, 0.9 - 0.4j, 1/0.9 + 0.4j, 1/0.9 - 0.4j]
zeros_type_III= [0.8 + 0.6j, 0.8 - 0.6j, -1.1, -1/1.1]
zeros_type_IV = [0.7 + 0.7j, 0.7 - 0.7j, 1/0.7 + 0.7j, 1/0.7 - 0.7j]

types = [zeros_type_I, zeros_type_II, zeros_type_III, zeros_type_IV]
titles = ['(a) Type I', '(b) Type II', '(c) Type III', '(d) Type IV']

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

for i, ax in enumerate(axs.flat):
    plot_unit_circle(ax)
    zeros = types[i]
    ax.plot([z.real for z in zeros], [z.imag for z in zeros], 'o', markersize=10)
    ax.set_title(titles[i])

fig.suptitle("Figure 5.41: Typical plots of zeros for linear-phase systems", fontsize=14)
plt.tight_layout()
plt.show()
```

شکل‌های 5.41(a) و 5.41(b) مکان‌های معمول صفرها را برای سیستم‌های نوع I (زمانی که M زوج است) و نوع II (زمانی که M فرد است) نشان می‌دهند.

اگر پاسخ ضربه‌ای آنتی‌متقارن باشد (سیستم‌های نوع III و IV)، آنگاه با پیروی از روشی که برای به‌دست‌آوردن رابطه (5.163) استفاده شد، می‌توان نشان داد که:
 
<div dir="ltr">

\begin{equation}
H(z) = -z^{-M} H(z^{-1}) \tag{5.165}
\end{equation}

</div>
این معادله می‌تواند برای نشان دادن این‌که صفرهای H(z) در حالت آنتی‌متقارن نیز به همان شیوه صفرهای حالت متقارن محدود شده‌اند، مورد استفاده قرار گیرد. با این حال، در حالت آنتی‌متقارن، z=1 و z=−1 از اهمیت ویژه‌ای برخوردارند. اگر z=1 باشد، رابطه (5.165) به شکل زیر در می‌آید:
<div dir="ltr">

\begin{equation}
H(1) = -H(1) \tag{5.166}
\end{equation}

</div>
بنابراین، H(z) باید یک صفر در z=1 داشته باشد؛ چه M زوج باشد و چه فرد. اگر z=−1 باشد، رابطه (5.165) به صورت زیر نوشته می‌شود:
<div dir="ltr">

\begin{equation}
H(-1) = (-1)^{-M+1} H(-1) \tag{5.167}
\end{equation}

</div>
در این حالت، اگر (M−1) فرد باشد (یعنی اگر M زوج باشد)، آنگاه H(−1)=−H(−1)، پس باید صفر باشد. بنابراین z=−1 باید یک صفر برای H(z) باشد اگر M زوج باشد. شکل‌های 5.41(c) و 5.41(d) مکان‌های معمول صفرها را برای سیستم‌های نوع III و IV به ترتیب نشان می‌دهند.

این محدودیت‌ها بر صفرها در طراحی سیستم‌های FIR با فاز خطی مهم هستند، زیرا محدودیت‌هایی بر انواع پاسخ‌های فرکانسی که می‌توان به آن‌ها دست یافت، اعمال می‌کنند. برای مثال، اشاره می‌شود که هنگام تقریب یک فیلتر بالاگذر با استفاده از یک سیستم متقارن پاسخ ضربه‌ای، M نباید فرد باشد، چون پاسخ فرکانسی محدود شده است که در ω=π (یعنی وقتی z=−1) صفر باشد.

### ۵.۷.۴ ارتباط بین سیستم‌های FIR با فاز خطی و سیستم‌های با فاز مینیمم

بحث‌های قبلی نشان می‌دهد که تمامی سیستم‌های FIR با پاسخ ضربه‌ای حقیقی که فاز خطی دارند، دارای صفرهایی روی دایره واحد یا در مکان‌های مزدوج معکوس هستند. بنابراین، به‌راحتی می‌توان نشان داد که تابع سیستم هر سیستم FIR با فاز خطی را می‌توان به صورت حاصل‌ضرب یک بخش فاز-مینیمم ، یک بخش فاز-ماکزیمم ، و یک تابع  شامل صفرهایی فقط روی دایره واحد تجزیه کرد؛ یعنی:
<div dir="ltr">

$$
H(z) = H_{\text{min}}(z) H_{uc}(z) H_{\text{max}}(z) \tag{5.168a}
$$

</div>

که در آن:
<div dir="ltr">

$$
H_{\text{max}}(z) = H_{\text{min}}(z^{-1}) z^{-M_i} \tag{5.168b}
$$

</div>
و  تعداد صفرهای  است. در معادله (5.168a)،  دارای تمام صفرهای داخل دایره واحد است، و  تمامی صفرهایش را روی دایره واحد دارد.  دارای تمام صفرهای خارج از دایره واحد است و از معادله (5.168b)، صفرهای آن معکوس صفرهای  هستند. مرتبه تابع سیستم  بنابراین برابر است با:
<div dir="ltr">

$$
2M_i + M_o = M
$$

</div>

---

#### مثال ۵.۲۱: تجزیه یک سیستم با فاز خطی

برای مثال ساده‌ای از کاربرد معادلات (5.168)، سیستم فاز مینیمم از معادله (5.109) را در نظر بگیرید که پاسخ فرکانسی آن در شکل 5.28 نشان داده شده است. سیستم به‌دست‌آمده با استفاده از معادله (5.168b) برای  در معادله (5.109) به صورت زیر است:
<div dir="ltr">

$$
H_{\text{max}}(z) = (0.9)^2 (1 - 1.1111 e^{j 0.6\pi} z^{-1}) (1 - 1.1111 e^{-j 0.6\pi} z^{-1})
$$

$$
\quad \times (1 - 1.25 e^{-j 0.8\pi} z^{-1}) (1 - 1.25 e^{j 0.8\pi} z^{-1})
$$

</div>
تابع  دارای پاسخ فرکانسی نشان‌داده‌شده در شکل 5.42 است. اکنون اگر این دو سیستم با یکدیگر به صورت متوالی (cascade) متصل شوند، طبق معادله (5.168) سیستم کلی به صورت:

<div dir="ltr">

$$
H(z) = H_{\text{min}}(z) H_{\text{max}}(z)
$$

</div>

خواهد بود و دارای فاز خطی است. پاسخ فرکانسی سیستم ترکیبی را می‌توان با جمع لگاریتم مقدار مطلق، فاز، و تابع تأخیر گروهی به‌دست آورد. بنابراین:
<div dir="ltr">

$$
20 \log_{10} |H(e^{j\omega})| = 20 \log_{10} |H_{\text{min}}(e^{j\omega})| + 20 \log_{10} |H_{\text{max}}(e^{j\omega})|
= 40 \log_{10} |H_{\text{min}}(e^{j\omega})| \tag{5.169}
$$

</div>

</div>


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz, group_delay

# تعریف صفرها
zeros = [
    1.1111 * np.exp(1j * 0.6 * np.pi),
    1.1111 * np.exp(-1j * 0.6 * np.pi),
    1.25 * np.exp(1j * 0.8 * np.pi),
    1.25 * np.exp(-1j * 0.8 * np.pi)
]

# محاسبه ضرایب صورت از صفرها (پلی‌نوم با ضرایب حقیقی)
b = np.poly(zeros) * (0.9)**2
b = np.real_if_close(b)  # حذف بخش موهومی ناچیز

# ضرایب مخرج (سیستم FIR)
a = [1.0]

# پاسخ فرکانسی
w, h = freqz(b, a, worN=1024, whole=True)

# محاسبه فاز
phase = np.angle(h)

# محاسبه تاخیر گروهی
w_gd, gd = group_delay((b, a), whole=True)

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(10, 10))

# (a) Log magnitude
axs[0].plot(w, 20 * np.log10(np.abs(h)))
axs[0].set_title('(a) Log Magnitude')
axs[0].set_ylabel('dB')
axs[0].set_xlabel('Radian Frequency (ω)')
axs[0].grid(True)

# (b) Phase
axs[1].plot(w, phase)
axs[1].set_title('(b) Phase (Principal Value)')
axs[1].set_ylabel('Radians')
axs[1].set_xlabel('Radian Frequency (ω)')
axs[1].grid(True)

# (c) Group delay
axs[2].plot(w_gd, gd)
axs[2].set_title('(c) Group Delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian Frequency (ω)')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz, group_delay

# تعریف صفرها
zeros = [
    1.1111 * np.exp(1j * 0.6 * np.pi),
    1.1111 * np.exp(-1j * 0.6 * np.pi),
    1.25 * np.exp(1j * 0.8 * np.pi),
    1.25 * np.exp(-1j * 0.8 * np.pi)
]

# محاسبه ضرایب صورت از صفرها و حذف بخش موهومی ناچیز
b = np.poly(zeros) * (0.9)**2
b = np.real_if_close(b)  # حذف بخش موهومی ناشی از خطای محاسباتی

# سیستم FIR → مخرج برابر 1
a = [1.0]

# پاسخ فرکانسی (بر کل دامنه [0, 2π])
w, h = freqz(b, a, worN=1024, whole=True)
phase = np.angle(h)

# تأخیر گروهی
w_gd, gd = group_delay((b, a), whole=True)

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(10, 10))

# (a) Log Magnitude
axs[0].plot(w, 20 * np.log10(np.abs(h) + 1e-12))  # +1e-12 to avoid log(0)
axs[0].set_title('(a) Log Magnitude')
axs[0].set_ylabel('dB')
axs[0].set_xlabel('Radian Frequency (ω)')
axs[0].grid(True)

# (b) Phase
axs[1].plot(w, phase)
axs[1].set_title('(b) Phase (Principal Value)')
axs[1].set_ylabel('Radians')
axs[1].set_xlabel('Radian Frequency (ω)')
axs[1].grid(True)

# (c) Group Delay
axs[2].plot(w_gd, gd)
axs[2].set_title('(c) Group Delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian Frequency (ω)')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz, group_delay

# تعریف صفرهای سیستم مینیمم‌فاز (داخل دایره واحد)
z_min = [
    (1 / 1.1111) * np.exp(1j * 0.6 * np.pi),
    (1 / 1.1111) * np.exp(-1j * 0.6 * np.pi),
    (1 / 1.25) * np.exp(1j * 0.8 * np.pi),
    (1 / 1.25) * np.exp(-1j * 0.8 * np.pi)
]

# صفرهای سیستم ماکزیمم‌فاز (خارج از دایره واحد)
z_max = [
    1.1111 * np.exp(1j * 0.6 * np.pi),
    1.1111 * np.exp(-1j * 0.6 * np.pi),
    1.25 * np.exp(1j * 0.8 * np.pi),
    1.25 * np.exp(-1j * 0.8 * np.pi)
]

# تولید ضرایب از صفرها
b_min = np.poly(z_min) * (0.9)**2
b_min = np.real_if_close(b_min)

b_max = np.poly(z_max)
b_max = np.real_if_close(b_max)

# کانولوشن برای ساخت سیستم ترکیبی
b_total = np.convolve(b_min, b_max)
a_total = [1.0]

# پاسخ فرکانسی
w, h = freqz(b_total, a_total, worN=1024, whole=True)
phase = np.angle(h)

# تأخیر گروهی
w_gd, gd = group_delay((b_total, a_total), whole=True)

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(10, 10))

# (a) Log Magnitude
axs[0].plot(w, 20 * np.log10(np.abs(h) + 1e-12))  # Avoid log(0)
axs[0].set_title('(a) Log Magnitude')
axs[0].set_ylabel('dB')
axs[0].set_xlabel('Radian Frequency (ω)')
axs[0].grid(True)

# (b) Phase
axs[1].plot(w, phase)
axs[1].set_title('(b) Phase (Principal Value)')
axs[1].set_ylabel('Radians')
axs[1].set_xlabel('Radian Frequency (ω)')
axs[1].grid(True)

# (c) Group Delay
axs[2].plot(w_gd, gd)
axs[2].set_title('(c) Group Delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian Frequency (ω)')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ایجاد محور فرکانسی (از 0 تا 2π)
omega = np.linspace(0, 2 * np.pi, 1000)

# (a) لگاریتم قدر مطلق (در dB)
# تابع نمونه: پاسخ فرکانسی ساختگی
magnitude_linear = np.abs(np.cos(omega / 4) + 1.5)
magnitude = 20 * np.log10(magnitude_linear + 1e-12)  # جلوگیری از log(0)

# (b) فاز خطی (مثلثی شکل)
phase = -4 * omega  # تأخیر ۴ نمونه → شیب خط فاز

# (c) تأخیر گروهی (ثابت)
group_delay = np.full_like(omega, 4.0)  # مقدار ثابت برای تأخیر گروهی

# رسم نمودارها
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

# نمودار (a): Log Magnitude
axs[0].plot(omega, magnitude)
axs[0].set_title('(a) Log Magnitude')
axs[0].set_ylabel('dB')
axs[0].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[0].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])
axs[0].grid(True)

# نمودار (b): Phase (wrapped)
wrapped_phase = np.mod(phase + np.pi, 2 * np.pi) - np.pi  # در بازه [-π, π]
axs[1].plot(omega, wrapped_phase)
axs[1].set_title('(b) Phase (principal value)')
axs[1].set_ylabel('Radians')
axs[1].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[1].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])
axs[1].grid(True)

# نمودار (c): Group Delay
axs[2].plot(omega, group_delay)
axs[2].set_title('(c) Group Delay')
axs[2].set_ylabel('Samples')
axs[2].set_xlabel('Radian Frequency (ω)')
axs[2].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[2].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
