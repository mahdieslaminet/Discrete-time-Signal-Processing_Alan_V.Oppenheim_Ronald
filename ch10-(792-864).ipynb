تحلیل فوریه سیگنال با استفاده از تبدیل فوریه گسسته   ص 792

10.0 مقدمه
در فصل 8، تبدیل فوریه گسسته (DFT) را به عنوان نمایش فوریه سیگنال های با طول محدود توسعه دادیم. از آنجا که DFT می تواند به طور موثر محاسبه شود، نقش اصلی را در طیف گسترده ای از برنامه های پردازش سیگنال، از جمله فیلتر کردن و تجزیه و تحلیل طیف ایفا می کند. در این فصل، نگاهی مقدماتی به تحلیل فوریه سیگنال ها با استفاده از DFT می اندازیم.
در برنامه‌ها و الگوریتم‌های مبتنی بر ارزیابی صریح تبدیل فوریه، در حالت ایده‌آل تبدیل فوریه گسسته زمان (DTFT) مورد نظر است، اگرچه این DFT است که می‌تواند واقعاً محاسبه شود. برای سیگنال‌های با طول محدود، DFT نمونه‌های دامنه فرکانس DTFT را ارائه می‌کند و پیامدهای این نمونه‌برداری باید به وضوح درک و در نظر گرفته شود.
به عنوان مثال، همانطور که در بخش 8.7 در نظر گرفته شد، در فیلتر کردن خطی یا کانولوشن که با ضرب DFTها به جای DTFTها اجرا می شود، یک کانولوشن دایره ای اجرا می شود، و باید دقت ویژه ای برای اطمینان از اینکه نتایج معادل یک انحراف خطی هستند، انجام شود.
علاوه بر این، در بسیاری از کاربردهای فیلترینگ و تحلیل طیف، سیگنال ها ذاتا طول محدودی ندارند. همانطور که بحث خواهیم کرد، این ناسازگاری بین نیاز طول محدود DFT و واقعیت سیگنال‌های طولانی مدت نامحدود را می‌توان دقیقاً یا تقریباً از طریق مفاهیم پنجره‌سازی، پردازش بلوک و تبدیل فوریه وابسته به زمان در نظر گرفت.






10.1 تجزیه و تحلیل فوریه سیگنال ها با استفاده از DFT
یکی از کاربردهای عمده DFT در تجزیه و تحلیل محتوای فرکانس سیگنال های زمان پیوسته است. برای مثال، همانطور که در بخش 10.4.1 توضیح دادیم، در تجزیه و تحلیل و پردازش گفتار، تجزیه و تحلیل فرکانس به ویژه در شناسایی و مدل‌سازی رزونانس‌های حفره صوتی مفید است. مثال دیگری که در بخش 10.4.2 معرفی شد، رادار داپلر است که در آن سرعت یک هدف با تغییر فرکانس بین سیگنال های ارسالی و دریافتی نشان داده می شود. مراحل اساسی در اعمال DFT به سیگنال های زمان پیوسته در شکل 10.1 نشان داده شده است.
فیلتر antialiasing برای حذف یا به حداقل رساندن اثر aliasing در زمانی که سیگنال زمان پیوسته با نمونه برداری به یک دنباله تبدیل می شود، تعبیه شده است. نیاز به ضرب x[n] در w[n]، یعنی پنجره‌بندی، نتیجه نیاز طول محدود DFT است. در بسیاری از موارد مورد علاقه عملی، sc(t) و در نتیجه، x[n] سیگنال های بسیار طولانی یا حتی نامحدود طولانی هستند (مانند سخنرانی یا موسیقی). بنابراین، یک پنجره با مدت زمان محدود w[n] قبل از محاسبه DFT به x[n] اعمال می شود. شکل 10.2 تبدیل فوریه سیگنال ها را در شکل 10.1 نشان می دهد. شکل 10.2 (الف) یک طیف زمان پیوسته را نشان می‌دهد که در فرکانس‌های بالا کاهش می‌یابد اما باند محدود نیست. همچنین نشان دهنده وجود مقداری انرژی سیگنال باند باریک است که توسط قله های باریک نشان داده می شود. پاسخ فرکانسی یک فیلتر ضد آلیازینگ در شکل 10.2 (b) نشان داده شده است. همانطور که در شکل 10.2 (c) نشان داده شده است، تبدیل فوریه زمان پیوسته X، (fil) حاوی اطلاعات مفید کمی در مورد Se(j C2) برای فرکانس های بالاتر از فرکانس قطع فیلتر است. از آنجایی که Haa(jf2) نمی تواند ایده آل باشد، اجزای فوریه ورودی در باند عبور و باند انتقال نیز توسط پاسخ فرکانسی فیلتر اصلاح می شوند.

10.2 تجزیه و تحلیل DFT سیگنال های سینوسی
10.2 تجزیه و تحلیل DFT سیگنال های سینوسی DTFT یک سیگنال سینوسی $A cos(w_{0}n +\phi )$(موجود برای همه n) یک جفت تکانه در $+w_{0}$ و $-w_{0}$ است (تکرار دوره ای با دوره $2\pi$). در تجزیه و تحلیل سیگنال‌های سینوسی با استفاده از DFT، نمونه‌برداری پنجره‌ای و طیفی (حوزه فرکانس) تأثیرات مهمی دارد. همانطور که در بخش 10.2.1 خواهیم دید، پنجره سازی تکانه های نمایش فوریه را لکه دار یا گسترده می کند، بنابراین، فرکانس دقیق کمتر تعریف شده است. پنجره سازی همچنین توانایی تشخیص سیگنال های سینوسی که از نظر فرکانس نزدیک به هم هستند را کاهش می دهد. نمونه برداری طیفی ذاتی در DFT به طور بالقوه یک تصویر اشتباه یا نادرست از طیف واقعی سیگنال سینوسی ارائه می دهد. این اثر در بخش 10.2.3 مورد بحث قرار گرفته است.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import iv  # Bessel function

# Parameters
L = 10  # L value
alpha = 5.09  # alpha value
pi_alpha = 16  # pi * alpha value
x = np.linspace(-L/2, L/2, 1000)  # x values in range [-L/2, L/2]

# Define w0(x) function
w0_x = np.zeros_like(x)
for i, xi in enumerate(x):
    if abs(xi) <= L/2:
        w0_x[i] = (1/L) * (iv(0, pi_alpha * np.sqrt(1 - (2 * xi / L)**2))) / iv(0, pi_alpha)
    else:
        w0_x[i] = 0

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, w0_x, label=r'$w_0(x)$', color='b')
plt.title(r'Plot of $w_0(x)$ with $\pi \alpha = 16$ and $\alpha = 5.09$')
plt.xlabel(r'$x$')
plt.ylabel(r'$w_0(x)$')
plt.grid(True)
plt.legend()
plt.show()

# Part b) Fourier series related to part a
import numpy as np
import matplotlib.pyplot as plt

# Kaiser window parameters
N = 64  # window length
beta = 8  # shape parameter (beta)

# Create Kaiser window
window = np.kaiser(N, beta)

# Compute Fourier transform
fft_result = np.fft.fft(window, n=2048)  # n is number of points for FFT
fft_magnitude = np.abs(fft_result)  # Fourier transform magnitude
fft_magnitude_db = 20 * np.log10(fft_magnitude)  # convert to dB

# Frequencies for horizontal axis
frequencies = np.fft.fftfreq(len(fft_result))

# Plot
plt.plot(frequencies[:len(frequencies)//2], fft_magnitude_db[:len(fft_result)//2])
plt.title('Fourier Transform of Kaiser Window')
plt.xlabel('Frequency (Normalized)')
plt.ylabel('Magnitude (dB)')
plt.grid(True)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftshift

# پارامترها
a0 = 15 * np.pi * 1e-6  # ضریب a0
N_total = 1024           # طول کل سیگنال
N_window = 400           # طول پنجره همنگ
eta = 320                # موقعیت پنجره

# تولید سیگنال x[n] = cos(a0 * n^2)
n = np.arange(N_total)
x = np.cos(a0 * n**2)

# تعریف پنجره همنگ برای N_window نمونه
window = np.hamming(N_window)

# انتخاب بخشی از سیگنال حول eta و اعمال پنجره
start_idx = max(0, eta - N_window // 2)
end_idx = min(N_total, eta + N_window // 2)
x_segment = x[start_idx:end_idx]

# اگر طول بخش انتخاب شده کمتر از N_window بود، padding انجام دهید
if len(x_segment) < N_window:
    padding = N_window - len(x_segment)
    x_segment = np.pad(x_segment, (0, padding), mode='constant')

x_windowed = x_segment * window

# محاسبه DTFT (با استفاده از FFT با اندازه بزرگ برای دقت بیشتر)
N_fft = 2048
X = fftshift(fft(x_windowed, N_fft))
freq = np.linspace(-0.5, 0.5, N_fft)  # فرکانس نرمال‌شده

# رسم نتایج
plt.figure(figsize=(15, 6))

# سیگنال اصلی و پنجره
plt.subplot(1, 2, 1)
plt.plot(n, x, label='$x[n] = \cos(a_0 n^2)$')
plt.axvspan(start_idx, end_idx, color='red', alpha=0.3, label='پنجره همنگ')
plt.axvline(x=eta, color='black', linestyle='--', label='$\eta = 320$')
plt.xlabel('$n$')
plt.ylabel('$x[n]$')
plt.title('سیگنال اصلی و موقعیت پنجره')
plt.legend()
plt.grid(True)

# طیف DTFT
plt.subplot(1, 2, 2)
plt.plot(freq, np.abs(X), label='$|X(\lambda)|$', color='green')
plt.xlabel('فرکانس نرمال‌شده ($\lambda$)')
plt.ylabel('دامنه')
plt.title('DTFT در $\eta = 320$ (پنجره همنگ)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

تجزیه و تحلیل از نظر بانک های فیلتر. از آنجایی که علاقه اصلی ما به کاربردهای DFT است،
ما بیشتر بحث های خود را بر مبنای معادله قرار خواهیم داد. (10.18).
10.3.3 اثر پنجره
هدف اصلی پنجره در تبدیل فوریه وابسته به زمان محدود کردن است
وسعت دنباله ای که باید تبدیل شود، به طوری که ویژگی های طیفی هستند
تقریباً در طول مدت پنجره ثابت است. هر چه سیگنال سریعتر باشد
ویژگی ها تغییر می کند، پنجره باید کوتاه تر باشد. ما در بخش 10.2 دیدیم که به عنوان
پنجره کوتاه تر می شود، وضوح فرکانس کاهش می یابد. همین اثر درست است،
البته، برای $ X[n,\lambda )$ از سوی دیگر، با کاهش طول پنجره، توانایی
برای حل تغییرات با افزایش زمان. در نتیجه، انتخاب طول پنجره
تبدیل به یک مبادله بین وضوح فرکانس و وضوح زمان می شود. این داد و ستد
در مثال 10.10 نشان داده شد.
تأثیر پنجره بر ویژگی‌های فرعی فوریه وابسته به زمان
شکل را می توان با فرض اینکه سیگنال x[n] دارای یک DTFT X معمولی $X(e^{jw})$ است، مشاهده کرد.
ابتدا فرض کنیم که پنجره برای همه m واحد است. یعنی فرض کنید پنجره ای وجود ندارد

$ X[n,\lambda )= X(e^{jw})e^(j\lambda n)$
\

البته، یک پنجره معمولی برای تجزیه و تحلیل طیف وابسته به زمان به صفر می رسد، بنابراین
به عنوان انتخاب تنها بخشی از سیگنال برای تجزیه و تحلیل. از سوی دیگر همانطور که بحث شد
در بخش 10.2، طول و شکل پنجره به گونه ای انتخاب شده است که فوریه
تبدیل پنجره در A در مقایسه با تغییرات در A فوریه باریک است
تبدیل سیگنال بنابراین، نیاز به وضوح زمان خوب و فرکانس خوب است
حل و فصل اغلب نیاز به مصالحه دارد. تبدیل فوریه یک پنجره معمولی است

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftshift

# پارامترهای سیگنال
N = 64  # طول سیگنال
n = np.arange(N)
omega = np.linspace(-np.pi, np.pi, 1024)  # محدوده فرکانس

# تعریف سیگنال نمونه (مثلا یک سیگنال سینوسی)
x = np.sin(2 * np.pi * 0.1 * n)  # سیگنال با فرکانس 0.1 (نرمالیزه شده)

# محاسبه DTFT (تقریب با FFT بزرگ)
X = np.zeros(len(omega), dtype=complex)
for i, w in enumerate(omega):
    X[i] = np.sum(x * np.exp(-1j * w * n))

# تعریف lambda (مثلا 0.2)
lambda_val = 0.2

# محاسبه X[n, lambda) = X(e^{jw}) * e^{j lambda n}
X_n_lambda = np.outer(np.exp(1j * lambda_val * n), X)

# رسم بزرگی (Magnitude) تابع
plt.figure(figsize=(12, 6))

# رسم |X(e^{jw})|
plt.subplot(1, 2, 1)
plt.plot(omega, np.abs(X), label='$|X(e^{j\omega})|$')
plt.xlabel('فرکانس ($\omega$)', fontname='DejaVu Sans')
plt.ylabel('بزرگی', fontname='DejaVu Sans')
plt.title('طیف DTFT اصلی', fontname='DejaVu Sans')
plt.grid(True)

# رسم |X[n, lambda)| برای n=0
plt.subplot(1, 2, 2)
plt.plot(omega, np.abs(X_n_lambda[0, :]), label='$|X[0, \lambda)|$')
plt.xlabel('فرکانس ($\lambda$)', fontname='DejaVu Sans')
plt.ylabel('بزرگی', fontname='DejaVu Sans')
plt.title('$X[n, \lambda)$ برای $n=0$', fontname='DejaVu Sans')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# تعریف سیگنال اولیه x[n]
x = np.sin(2 * np.pi * 0.05 * np.arange(100))  # یک سینوس ساده

# تعریف پنجره w[m]
L = 20                      # طول پنجره
w = np.hanning(L)           # پنجره‌ی هنینگ

R = 10                      # فاصله‌ی بین پنجره‌ها (hop size)
r = 3                       # شماره‌ی پنجره مورد نظر

# محاسبه x_r[m]
m = np.arange(L)
xr = x[r * R + m] * w

# ترسیم نمودار x_r[m]
plt.stem(m, xr, basefmt=" ")
plt.title(f"$x_r[m] = x[rR + m] \cdot w[m]$ برای r = {r}")
plt.xlabel("m")
plt.ylabel("$x_r[m]$")
plt.grid(True)
plt.show()

import numpy as np
import matplotlib.pyplot as plt

# تعریف سیگنال اولیه x[n]
x = np.sin(2 * np.pi * 0.05 * np.arange(100))  # سینوس ساده

# پارامترها
L = 32                     # طول پنجره
R = 16                     # hop size
N = 64                     # تعداد نقاط DFT (باید ≥ L باشد)
w = np.hanning(L)          # پنجره هنینگ

r = 2                      # شماره پنجره
m = np.arange(L)
xr = x[r * R + m] * w      # سیگنال پنجره‌ای‌شده

# محاسبه DFT (X[rR, k])
k = np.arange(N)
X_r = np.zeros(N, dtype=complex)
for i in k:
    X_r[i] = np.sum(xr * np.exp(-1j * 2 * np.pi * i * m / N))

# ترسیم قدر مطلق طیف
plt.stem(k, np.abs(X_r), basefmt=" ")
plt.title(f"$|X[rR, k]|$ برای r = {r}")
plt.xlabel("k")
plt.ylabel("$|X[rR, k]|$")
plt.grid(True)
plt.show()

```

تقارن DFT
به عنوان یک مثال خاص، سیگنال را می توان دقیقاً از نمونه برداری شده بازسازی کرد
تبدیل فوریه وابسته به زمان برای حالت خاص R = L = N. در این مورد، N
نمونه های یک سیگنال واقعی با N عدد واقعی نشان داده می شوند و این حداقل است
که می‌توانیم انتظار داشته باشیم برای یک سیگنال انتخابی خودسرانه به آن دست پیدا کنیم. برای R = L = N می توانیم
بازیابی x[m] = x[rR + m]w[m] برای $0\le m \le N-1$ با محاسبه DFT معکوس
از X، [k]. بنابراین، می توانیم x[n] را برای rR ≤ n ≤ [(r+1)R - 1] بر حسب
بخش های پنجره ای x، [m] به عنوان
\
\
10.42
\
$x[n]=\frac{x_{r}[n-rR]}{w[n-rR]}, rR\le n \le [(r+1)R-1]$

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, freqz, windows

# پارامترهای طراحی فیلتر
M = 50          # طول پنجره (به تعداد نقاط)
fs = 10000      # فرکانس نمونه‌برداری (10 کیلوهرتز)
fc_low = 100    # فرکانس قطع پایین (100 هرتز)
fc_high = 1000  # فرکانس قطع بالا (1000 هرتز)

# ایجاد پنجره‌های بارتلت و هان
bartlett = windows.bartlett(M+1)  # پنجره بارتلت
hann = windows.hann(M+1)          # پنجره هان

# طراحی فیلتر میان‌گذر با پنجره‌های مختلف
taps_bartlett = firwin(M, [fc_low, fc_high], fs=fs,
                      pass_zero=False, window='bartlett')
taps_hann = firwin(M, [fc_low, fc_high], fs=fs,
                  pass_zero=False, window='hann')

# محاسبه پاسخ فرکانسی فیلترها
f, h_bartlett = freqz(taps_bartlett, fs=fs)  # پاسخ فیلتر با پنجره بارتلت
f, h_hann = freqz(taps_hann, fs=fs)          # پاسخ فیلتر با پنجره هان

# رسم نمودارها
plt.figure(figsize=(12, 8))

# نمودار پنجره‌ها
plt.subplot(2, 1, 1)
plt.plot(bartlett, label='پنجره بارتلت')
plt.plot(hann, label='پنجره هان')
plt.title('نمودار پنجره‌های زمانی')
plt.xlabel('نمونه')
plt.ylabel('دامنه')
plt.legend()
plt.grid()

# نمودار پاسخ فرکانسی
plt.subplot(2, 1, 2)
plt.plot(f, 20*np.log10(np.abs(h_bartlett)), label='فیلتر با پنجره بارتلت')
plt.plot(f, 20*np.log10(np.abs(h_hann)), label='فیلتر با پنجره هان')
plt.title('پاسخ فرکانسی فیلتر میان‌گذر (100-1000 هرتز)')
plt.xlabel('فرکانس (هرتز)')
plt.ylabel('بزرگی (dB)')
plt.xlim(0, 2000)  # محدوده فرکانسی
plt.ylim(-80, 5)   # محدوده بزرگی
plt.axvline(fc_low, color='r', linestyle='--', label='فرکانس قطع پایین')
plt.axvline(fc_high, color='r', linestyle='--', label='فرکانس قطع بالا')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

یکپارچگی آن به عنوان تبدیل فوریه وابسته به زمان، سپس یک سیگنال پردازش شده y[n] می تواند باشد

سپس از سنتز فوریه برای بازسازی سیگنال y[n] برای گوش دادن استفاده می شود. این اساس است
برای مثال برای کدگذاری صوتی MP3. یکی دیگر از برنامه های کاربردی سرکوب نویز صوتی است،
که در آن طیف نویز آکوستیک تخمین زده می شود و سپس یا از آن کم می شود
طیف فوریه وابسته به زمان سیگنال ورودی یا به عنوان پایه ای برای وینر استفاده می شود
فیلتر اعمال شده به X,[k]. (به Quatieri، 2002 مراجعه کنید.) اینها و بسیاری از کاربردهای دیگر
توسط الگوریتم‌های FFT که برای محاسبات کارآمد در دسترس هستند، بسیار تسهیل می‌شوند
تبدیل فوریه وابسته به زمان گسسته.

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, freqz
from scipy.signal.windows import kaiser  # اصلاح شده

# پارامترهای طراحی
fs = 10000  # فرکانس نمونه‌برداری (هرتز)
num_filters = 8  # تعداد فیلترها در فیلتربانک
M = 100  # طول فیلتر
beta = 8  # پارامتر پنجره کایزر

# فرکانس‌های مرکز فیلترها (به صورت لگاریتمی)
center_freqs = np.logspace(np.log10(200), np.log10(4000), num_filters)

# طراحی فیلتربانک
plt.figure(figsize=(14, 6))

# پنجره مستطیلی
plt.subplot(1, 2, 1)
for fc in center_freqs:
    taps = firwin(M, [fc-200, fc+200], fs=fs, pass_zero=False, window='boxcar')
    w, h = freqz(taps, worN=2000, fs=fs)
    plt.semilogx(w, 20*np.log10(np.abs(h)))
plt.title('(a) پاسخ فرکانسی فیلتربانک با پنجره مستطیلی')
plt.xlabel('فرکانس (هرتز)')
plt.ylabel('بزرگی (dB)')
plt.grid(True)
plt.xlim(100, fs/2)
plt.ylim(-60, 5)

# پنجره کایزر
plt.subplot(1, 2, 2)
for fc in center_freqs:
    # روش صحیح استفاده از پنجره کایزر
    taps = firwin(M, [fc-200, fc+200], fs=fs, pass_zero=False, window=('kaiser', beta))
    w, h = freqz(taps, worN=2000, fs=fs)
    plt.semilogx(w, 20*np.log10(np.abs(h)))
plt.title('(b) پاسخ فرکانسی فیلتربانک با پنجره کایزر')
plt.xlabel('فرکانس (هرتز)')
plt.ylabel('بزرگی (dB)')
plt.grid(True)
plt.xlim(100, fs/2)
plt.ylim(-60, 5)

plt.tight_layout()
plt.show()
```

```python

# وارد کردن کتابخانه‌های مورد نیاز
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram, windows

# تنظیم پارامترهای اصلی
fs = 1000  # فرکانس نمونه‌برداری (هرتز)
duration = 5  # مدت زمان سیگنال (ثانیه)
t = np.linspace(0, duration, int(fs*duration), endpoint=False)  # محور زمان

# تولید سیگنال رادار داپلر با فرکانس متغیر
f0 = 50  # فرکانس پایه (هرتز)
doppler_shift = 20 * np.sin(2*np.pi*0.5*t)  # تغییر فرکانس داپلر به صورت سینوسی
signal = np.cos(2*np.pi*(f0 + doppler_shift)*t)  # سیگنال نهایی

# پارامترهای تحلیل زمان-فرکانس
nperseg = 256  # طول پنجره
noverlap = nperseg // 2  # میزان همپوشانی پنجره‌ها (50%)

# محاسبه طیف‌نگاره (STFT)
f, t_spec, Sxx = spectrogram(signal, fs=fs, window=windows.hann(nperseg),
                           nperseg=nperseg, noverlap=noverlap)

# یافتن فرکانس غالب در هر لحظه
peak_freqs = f[np.argmax(Sxx, axis=0)]

# رسم نمودارها
plt.figure(figsize=(12, 8))

# (الف) نمایش طیف‌نگاره
plt.subplot(2, 1, 1)
plt.pcolormesh(t_spec, f, 10*np.log10(Sxx), shading='gouraud')
plt.colorbar(label='چگالی طیف توان (dB)')
plt.title('(الف) تبدیل فوریه زمان-وابسته سیگنال رادار داپلر')
plt.xlabel('زمان (ثانیه)')
plt.ylabel('فرکانس (هرتز)')
plt.ylim(0, 150)  # محدوده فرکانسی

# (ب) نمایش تغییرات فرکانس داپلر
plt.subplot(2, 1, 2)
plt.plot(t_spec, peak_freqs - f0, 'r-', linewidth=2)
plt.title('(ب) فرکانس داپلر تخمین زده شده')
plt.xlabel('زمان (ثانیه)')
plt.ylabel('جابجایی داپلر (هرتز)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import windows, welch

# تابع محاسبه انرژی نرمال‌شده پنجره
def calculate_U(window):
    return np.sum(window**2) / len(window)

# تنظیمات پارامترهای مختلف
configs = [
    {'L': 64, 'N': 1024, 'color': 'blue', 'linestyle': '-', 'title': 'پیکربندی (b): طول پنجره 64، طول DFT 1024'},
    {'L': 256, 'N': 1024, 'color': 'green', 'linestyle': '--', 'title': 'پیکربندی (c): طول پنجره 256، طول DFT 1024'},
    {'L': 1024, 'N': 1024, 'color': 'red', 'linestyle': ':', 'title': 'پیکربندی (d): طول پنجره 1024، طول DFT 1024'}
]

# انواع پنجره‌ها برای تحلیل
window_types = ['hann', 'hamming', 'boxcar']

# تولید و نمایش نمودارها برای هر پیکربندی
for i, cfg in enumerate(configs):
    plt.figure(figsize=(12, 6))

    for w_type in window_types:
        # تولید پنجره
        window = windows.get_window(w_type, cfg['L'])

        # محاسبه انرژی نرمال‌شده
        U = calculate_U(window)

        # تولید نویز سفید
        np.random.seed(0)
        white_noise = np.random.randn(cfg['L'])

        # محاسبه پریودوگرام
        f, Pxx = welch(white_noise, fs=1, window=w_type,
                      nperseg=cfg['L'], nfft=cfg['N'])

        # رسم نمودار
        plt.plot(f[:cfg['N']//2], 10*np.log10(Pxx[:cfg['N']//2]),
                color=cfg['color'], linestyle=cfg['linestyle'],
                label=f'{w_type} (U={U:.3f})')

    plt.title(cfg['title'])
    plt.xlabel('فرکانس نرمال‌شده (×π رادیان/نمونه)')
    plt.ylabel('چگالی طیف توان (dB)')
    plt.grid(True)
    plt.legend()
    plt.xlim(0, 0.5)
    plt.ylim(-100, 20)
    plt.tight_layout()
    plt.show()

# نمایش جدول نتایج
print("\nمقایسه انرژی نرمال‌شده پنجره‌ها (U):")
print("| {:<10} | {:<10} | {:<10} | {:<10} |".format(
    "پیکربندی", "هان", "همینگ", "مستطیلی"))
for i, cfg in enumerate(configs):
    U_values = []
    for w_type in window_types:
        window = windows.get_window(w_type, cfg['L'])
        U_values.append(calculate_U(window))

    print("| {:<10} | {:<10.4f} | {:<10.4f} | {:<10.4f} |".format(
        f"L={cfg['L']}", U_values[0], U_values[1], U_values[2]))
```

```python


import numpy as np
import matplotlib.pyplot as plt

# تولید دنباله نمونه (الف)
L = 16  # طول دنباله
n = np.arange(L)
x = np.sin(2*np.pi*0.1*n) + 0.5*np.random.randn(L)  # سیگنال ترکیبی سینوسی و نویز
w = np.hanning(L)  # پنجره هان

# محاسبه سیگنال پنجره‌گذاری شده
v = x * w

# (الف) نمایش دنباله اصلی
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
markerline, stemlines, baseline = plt.stem(n, v, basefmt=' ')
plt.setp(stemlines, 'linewidth', 1)  # تنظیم ضخامت خطوط
plt.setp(markerline, 'markersize', 5)  # تنظیم اندازه مارکرها
plt.title('(الف) دنباله با طول محدود $v[n] = x[n]w[n]$')
plt.xlabel('نمونه (n)')
plt.ylabel('دامنه')
plt.grid(True)
plt.xlim(-1, L+5)

# (ب) نمایش دنباله شیفت داده شده برای m=3
m = 3
plt.subplot(3, 1, 2)
marker1, stem1, _ = plt.stem(n, v, basefmt='b', linefmt='b-', markerfmt='bo', label='$v[n]$')
marker2, stem2, _ = plt.stem(n+m, v, basefmt='r', linefmt='r--', markerfmt='ro', label='$v[n+m]$ (m=3)')
plt.setp(stem1, 'linewidth', 1)
plt.setp(stem2, 'linewidth', 1)
plt.title('(ب) دنباله اصلی و شیفت داده شده برای m>0')
plt.xlabel('نمونه (n)')
plt.ylabel('دامنه')
plt.grid(True)
plt.legend()
plt.xlim(-1, L+5)

# محاسبه تابع خودهمبستگی
max_shift = L-1
C_vv = np.zeros(2*max_shift+1)
shifts = np.arange(-max_shift, max_shift+1)

for i, m in enumerate(shifts):
    if m >= 0:
        C_vv[i] = np.sum(v[:L-m] * v[m:])
    else:
        C_vv[i] = np.sum(v[-m:] * v[:L+m])

# (ج) نمایش تابع خودهمبستگی
plt.subplot(3, 1, 3)
marker3, stem3, _ = plt.stem(shifts, C_vv, basefmt=' ')
plt.setp(stem3, 'linewidth', 1)
plt.title('(ج) تابع خودهمبستگی $C_{vv}[m]$')
plt.xlabel('تاخیر (m)')
plt.ylabel('مقدار همبستگی')
plt.grid(True)
plt.tight_layout()
plt.show()
```


10.5.3 میانگین پریودوگرام
میانگین پریودوگرام ها در تخمین طیف ابتدا به طور گسترده توسط
بارتلت (1953); بعدها، پس از توسعه الگوریتم های سریع برای محاسبه DFT،
ولش (1970) این الگوریتم های محاسباتی را با استفاده از یک پنجره داده ترکیب کرد
w[n] برای توسعه روش میانگین گیری پریودوگرام های اصلاح شده. در پریودوگرام میانگین
پیری، یک دنباله داده x[n]، 0 ≤ n ≤ Q-1، به بخش هایی از نمونه های طول-L تقسیم می شود،
با پنجره ای به طول L که برای هر کدام اعمال می شود. یعنی ما بخش ها را تشکیل می دهیم
\
\
10.82
\
\
$x_r[n]=x[rR+n]w[n],0\len n \le L-1$
\
\
اگر R < L قطعات با هم همپوشانی داشته باشند، و برای R = L قطعات به هم پیوسته هستند. توجه داشته باشید
که Q نشان دهنده طول داده های موجود است. تعداد کل بخش ها بستگی دارد
در مقادیر و رابطه بین R، L و Q. به طور خاص، K کامل وجود خواهد داشت
قطعات طول، که در آن K بزرگترین عدد صحیح است که برای آن (K -1)R+(L-1) ≤ Q-1 است.
پریودوگرام قطعه r ام است
\
\
10.83
\
\
$I_r(w)=\frac{1}{LU}|X_r(e^{jw})|^2$
\
\
که در آن X,(ejw) DTFT x,[n] است. هر I,(w) دارای خواص پریودوگرام است
قبلا شرح داده شد میانگین پریودوگرام شامل میانگین گیری پریودوگرام K است
برآورد I(w); به عنوان مثال، ما پریودوگرام میانگین زمانی را تشکیل می دهیم که به صورت تعریف شده است
\
\
10.84
\
\
$\bar{I}(w)=\frac{1}{L}\sum^{K-1}_{r=0}I_r(w)$
\
\
برای بررسی بایاس و واریانس I(0)، اجازه دهید L = R را در نظر بگیریم، به طوری که بخش ها انجام دهند
همپوشانی نداشته باشند، و فرض کنید که øxx[m] برای m > L کوچک است. یعنی نمونه های سیگنال بیش از
L از هم تقریباً همبسته نیستند. اگر فرض کنیم که پریودوگرام های I, (w) هستند
متغیرهای تصادفی مستقل به طور یکسان توزیع می شوند، سپس مقدار مورد انتظار I (c) است
\
\
10.85
\
\
$\varepsilon {\bar{I}(w)} =\frac{1}{K-1},\varepsilon {I_r(w)}$
\
\
یا، از آنجایی که فرض می کنیم پریودوگرام ها مستقل و به طور یکسان توزیع شده اند،
\
\
10.86
\
\
$\ varepsilon {\bar{I(w)}}=\ varepsilon {I_r(w)},  for any r$
\
\
از معادله (10.76) چنین می شود که
\
\
10.87
\

$\ varepsilon {\bar{I}(w)}=\ varepsilon {I_r(w)}=\frac{1}{2\pi LU}\int^{\pi }_{-\pi }P_{xx}(\theta)C_{ww}(e^{j(w-\theta )})d\theta $
\
\
که در آن L طول پنجره است. هنگامی که پنجره w[n] یک پنجره مستطیل شکل است،
روش میانگین گیری پریودوگرام را روش بارتلت می نامند و در این صورت می تواند
نشان داده شود که

با استفاده از  DFT
همانند پریودوگرام، پریودوگرام متوسط را فقط می توان به صراحت ارزیابی کرد
در مجموعه ای گسسته از فرکانس ها به دلیل در دسترس بودن الگوریتم های FFT برای
محاسبه DFT، یک انتخاب بسیار راحت و پرکاربرد مجموعه ای است
فرکانس $w_k = 2\pi k/N$ برای انتخاب مناسب N. از معادله. (10.84)، می بینیم که

نمونه هایی را در فرکانس های DFT $w_k = 2\pi k/N$  برای k = 0,1، ...، N -1 بدست آورید.
به طور خاص، با Xr[k] که نشان دهنده DFT x، [n] است،
\
\
10.92a
$I_r[k]=I_r(w_k)=\frac{1}{LU}|X_r[k]|^2$
\
\
10.92b
\
\
$\bar{I}[k]=\bar{I}(w_k)=\frac{1}{K}\sum^{k-1}_{r=0}I_r[k]$
\
\
شایان ذکر است که رابطه بین میانگین گیری پریودوگرام و
تبدیل فوریه وابسته به زمان همانطور که در بخش 10.3 به تفصیل مورد بحث قرار گرفت. معادله
(10.92a) نشان می دهد که به جز معرفی ثابت نرمال کننده 1/(LU)،
هر پریودوگرام منفرد صرفاً مجذور بزرگی وابسته به زمان است
تبدیل فوریه در زمان r R و فرکانس 2xk/N. بنابراین، برای هر شاخص فرکانس k،
میانگین تخمین طیف توان در فرکانس مربوط به k میانگین زمان است.

```python

import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1
omega0 = np.pi / 8
theta = 0
sigma = 0.2
N = 100
Q = 60
M = 30

# Generate signal x[n]
n = np.arange(N)
x = A * np.cos(omega0 * n + theta) + sigma * np.random.randn(N)

# Compute autocorrelation
phi_hat = np.zeros(M)
for m in range(M):
    sum_val = 0
    for n in range(Q - m):
        sum_val += x[n] * x[n + m]
    phi_hat[m] = sum_val / Q

# Symmetrize for negative lags
phi_hat_full = np.concatenate([phi_hat[:0:-1], phi_hat])
lags = np.arange(-(M-1), M)

# Plot
plt.figure(figsize=(10, 5))
plt.stem(lags, phi_hat_full, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.xlabel('Lag (m)')
plt.ylabel(r'$\hat{\phi}_{xx}[m]$')
plt.title('Estimated Autocorrelation of $x[n]$')
plt.grid(True)
plt.show()

```

```python

import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_samples = 1000  # Number of samples per line
bits = 10           # Quantization resolution (10-bit)
magnification = 2**9  # Error magnification factor (512x)

# Generate simulated speech signal (replace with real data if available)
np.random.seed(0)
speech_signal = np.cumsum(np.random.randn(2 * num_samples))  # Fixed missing parenthesis
speech_signal = speech_signal / np.max(np.abs(speech_signal))  # Normalize to [-1, 1]

# 10-bit quantization
quantized_signal = np.round(speech_signal * (2**(bits-1))) / (2**(bits-1))
quantization_error = (speech_signal - quantized_signal) * magnification

# Split data into 4 segments (2 speech, 2 error)
speech_lines = [speech_signal[:num_samples], speech_signal[num_samples:]]
error_lines = [quantization_error[:num_samples], quantization_error[num_samples:]]

# Plot
plt.figure(figsize=(12, 8))

for i in range(2):
    # Speech signal (odd rows: 1st and 3rd)
    plt.subplot(4, 1, 2*i + 1)
    plt.plot(speech_lines[i], 'b-', linewidth=1)
    plt.ylabel('Speech Signal')
    plt.ylim(-1.1, 1.1)
    plt.grid(True, linestyle='--', alpha=0.6)

    # Quantization error (even rows: 2nd and 4th)
    plt.subplot(4, 1, 2*i + 2)
    plt.plot(error_lines[i], 'r-', linewidth=1)
    plt.ylabel(f'Error (×{magnification})')
    plt.ylim(-0.6, 0.6)
    plt.grid(True, linestyle='--', alpha=0.6)

plt.xlabel('Sample Index')
plt.tight_layout()
plt.show()

```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import windows

# Parameters
Q = 3000  # Record length
bits = 10  # Quantization bits
M1 = 100   # Window length for first spectrum
M2 = 50    # Window length for second spectrum

# Generate quantization noise (uniform distribution)
np.random.seed(0)
quant_noise = np.random.uniform(-2**(-bits), 2**(-bits), Q)

# (a) Autocorrelation estimate
max_lag = 100
autocorr = np.zeros(2*max_lag + 1)
for m in range(-max_lag, max_lag+1):
    n_start = max(0, -m)
    n_end = min(Q, Q - m)
    autocorr[m + max_lag] = np.sum(quant_noise[n_start:n_end] * quant_noise[n_start + m:n_end + m]) / Q

# (b) Power spectrum estimation (Blackman-Tukey method)
def blackman_tukey_psd(autocorr, M):
    window = windows.bartlett(2*M + 1)
    autocorr_windowed = autocorr[len(autocorr)//2 - M : len(autocorr)//2 + M + 1] * window
    psd = np.fft.fftshift(np.abs(np.fft.fft(autocorr_windowed, 1024)))
    return 10 * np.log10(psd)

psd_M100 = blackman_tukey_psd(autocorr, M1)
psd_M50 = blackman_tukey_psd(autocorr, M2)
freq = np.linspace(-0.5, 0.5, len(psd_M100))
noise_floor = 10 * np.log10(2**(-2*bits)/12)  # Theoretical noise floor

# Plotting
plt.figure(figsize=(12, 8))

# Part (a): Autocorrelation
plt.subplot(2, 1, 1)
lags = np.arange(-max_lag, max_lag + 1)
markerline, stemlines, baseline = plt.stem(lags, autocorr, basefmt=" ")
plt.setp(stemlines, 'linewidth', 1)
plt.setp(markerline, 'markersize', 3)
plt.xlabel('Lag (η)')
plt.ylabel('Autocorrelation')
plt.title('(a) Autocorrelation Estimate of 10-bit Quantization Noise (Q=3000)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlim(-max_lag, max_lag)

# Part (b): Power spectrum
plt.subplot(2, 1, 2)
plt.plot(freq, psd_M100, label=f'M=100')
plt.plot(freq, psd_M50, label=f'M=50')
plt.axhline(noise_floor, color='r', linestyle='--', label='Theoretical Noise Floor')
plt.xlabel('Normalized Frequency (×π rad/sample)')
plt.ylabel('Power/frequency (dB)')
plt.title('(b) Power Spectrum Estimates (Blackman-Tukey Method)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlim(-0.5, 0.5)
plt.ylim(noise_floor - 10, max(psd_M100) + 5)

plt.tight_layout()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
import librosa  # For loading audio files (install with: pip install librosa)

# Parameters
num_samples = 1000  # Samples per line
bits = 4            # 4-bit quantization
magnification = 2**3  # 8x error magnification
num_lines = 4       # Total lines to display

# Generate or load speech signal
def generate_speech_like_signal(length):
    """Create a synthetic speech-like signal if real data isn't available"""
    t = np.linspace(0, 10, length)
    signal = (np.sin(2*np.pi*100*t) * (1 + 0.5*np.sin(2*np.pi*2*t)) +
             0.3*np.sin(2*np.pi*300*t) +
             0.2*np.random.randn(length))
    return signal / np.max(np.abs(signal))

try:
    # Try to load real speech (replace with your file)
    speech_signal, _ = librosa.load('speech.wav', sr=None, mono=True)
    speech_signal = speech_signal[:num_samples*num_lines//2]  # Use first part
except:
    print("Using synthetic speech signal (install librosa for real audio)")
    speech_signal = generate_speech_like_signal(num_samples*num_lines//2)

# Normalize and quantize
speech_signal = speech_signal / np.max(np.abs(speech_signal))
quantized = np.round(speech_signal * (2**(bits-1))) / (2**(bits-1))
error = (speech_signal - quantized) * magnification

# Prepare plot data
plot_signals = []
for i in range(num_lines):
    if i % 2 == 0:  # Speech lines (1st and 3rd)
        plot_signals.append(speech_signal[i//2*num_samples:(i//2+1)*num_samples])
    else:            # Error lines (2nd and 4th)
        plot_signals.append(error[i//2*num_samples:(i//2+1)*num_samples])

# Create figure
plt.figure(figsize=(12, 8))

for i, signal in enumerate(plot_signals):
    plt.subplot(num_lines, 1, i+1)

    if i % 2 == 0:
        plt.plot(signal, 'b-', linewidth=0.8)
        plt.ylabel('Speech Signal')
        plt.ylim(-1.1, 1.1)
    else:
        plt.plot(signal, 'r-', linewidth=0.8)
        plt.ylabel(f'Error (×{magnification})')
        plt.ylim(-0.6, 0.6)

    plt.grid(True, linestyle='--', alpha=0.5)
    if i < num_lines-1:
        plt.xticks([])

plt.xlabel('Sample Index')
plt.tight_layout()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import windows, spectrogram, welch

# Parameters
N = 1024  # Total signal length
M = 50    # Window length

# Generate test signal (sinusoid + noise)
np.random.seed(0)
t = np.arange(N)
signal = 0.8 * np.sin(2*np.pi*0.1*t) + 0.2*np.random.randn(N)

# Blackman-Tukey Method
def blackman_tukey(x, M, window_type='bartlett'):
    autocorr = np.correlate(x, x, mode='full')[len(x)-1-M:len(x)+M] / len(x)
    if window_type == 'bartlett':
        window = windows.bartlett(2*M + 1)
    else:
        window = windows.hamming(2*M + 1)
    autocorr_windowed = autocorr * window
    psd = np.fft.fftshift(np.abs(np.fft.fft(autocorr_windowed, 1024)))
    return 10 * np.log10(psd + 1e-12)

# Bartlett window estimate
psd_bartlett = blackman_tukey(signal, M, 'bartlett')

# Hamming window estimate
psd_hamming = blackman_tukey(signal, M, 'hamming')

# Welch's method (averaged periodograms)
f_welch, psd_welch = welch(signal, window='hamming', nperseg=M*2,
                           noverlap=M, nfft=1024, scaling='spectrum')
psd_welch = 10 * np.log10(psd_welch)

# Frequency axis
freq = np.linspace(-0.5, 0.5, len(psd_bartlett))
freq_welch = f_welch/np.pi  # Normalize to π rad/sample

# Plot
plt.figure(figsize=(12, 6))
plt.plot(freq[len(freq)//2:], psd_bartlett[len(freq)//2:], 'b-', linewidth=2.5,
        label='Blackman-Tukey (Bartlett)')
plt.plot(freq[len(freq)//2:], psd_hamming[len(freq)//2:], 'g-', linewidth=1.5,
        label='Blackman-Tukey (Hamming)')
plt.plot(freq_welch, psd_welch, 'r--', linewidth=2,
        label='Welch (Averaged Periodograms)')

plt.xlabel('Normalized Frequency (×π rad/sample)')
plt.ylabel('Power Spectrum (dB)')
plt.title('Power Spectrum Estimates (M=50)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlim(0, 0.5)  # Show only positive frequencies
plt.ylim(-60, 10)
plt.tight_layout()
plt.show()
```
