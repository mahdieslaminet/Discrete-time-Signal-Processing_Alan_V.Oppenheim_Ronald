## فصل ۷: تکنیک‌های طراحی فیلتر (Filter Design Techniques)




## ۷.۰ مقدمه

فیلترها یک دسته‌ی مهم از سیستم‌های LTI به‌شمار می‌روند. به‌طور دقیق‌تر، اصطلاح **فیلتر انتخابی فرکانس** به سیستمی اطلاق می‌شود که مؤلفه‌های فرکانسی خاصی از سیگنال ورودی را عبور می‌دهد و سایر مؤلفه‌ها را کاملاً حذف می‌کند. البته در یک مفهوم گسترده‌تر، هر سیستمی که برخی فرکانس‌ها را نسبت به دیگر فرکانس‌ها تغییر دهد نیز فیلتر نامیده می‌شود.

در حالی‌که تأکید اصلی این فصل بر طراحی فیلترهای انتخابی فرکانس است، بسیاری از تکنیک‌ها کاربردی وسیع‌تر دارند. ما تمرکز خود را روی طراحی فیلترهای علّی می‌گذاریم، با این‌حال در بسیاری از کاربردها، فیلترها الزاماً نباید علّی باشند. فیلترهای غیرعلّی نیز طراحی و پیاده‌سازی می‌شوند، به‌ویژه از طریق تغییر طراحی فیلترهای آنالوگ.

طراحی فیلترهای گسسته‌زمانی معمولاً شامل تعیین پارامترهای تابع تبدیل یا معادله‌ی تفاضلی است که پاسخ فرکانسی موردنظر را تقریب می‌زند. همان‌طور که در فصل ۲ بیان شد، سیستم‌های گسسته‌زمانی را می‌توان به دو دسته‌ی کلی تقسیم کرد: سیستم‌هایی با پاسخ ضربه‌ای نامتناهی (IIR) و سیستم‌هایی با پاسخ ضربه‌ای متناهی (FIR).

طراحی فیلترهای IIR معمولاً به تعیین تابع تبدیل گویای $H(z)$ می‌انجامد، در حالی‌که فیلترهای FIR عمدتاً به روش‌های تقریب چندجمله‌ای وابسته هستند. تکنیک‌های متداول طراحی برای این دو دسته متفاوت‌اند. به‌ویژه در طراحی فیلترهای IIR، روش‌های قدیمی‌تر شامل نگاشت طراحی‌های آنالوگ به گسسته‌زمانی از طریق روش‌هایی مانند "وارون‌سازی ضربه" (impulse invariance) و "تبدیل bilinear" بوده‌اند.

در مقابل، برای فیلترهای FIR، چون طراحی آنالوگی معادلی وجود نداشته، روش‌های طراحی مستقیم‌تری مطرح شده‌اند، مانند استفاده از پنجره‌ها (windowing) و الگوریتم‌های تکراری مانند Parks-McClellan که در فصل ۷ بررسی خواهند شد.

ما در این فصل، روش‌های متنوعی برای طراحی هر دو دسته فیلترهای IIR و FIR را بررسی خواهیم کرد. هدف این فصل، معرفی تکنیک‌های پرکاربرد طراحی و اشاره به برخی ملاحظات و مبادله‌های موجود (trade-off) میان آن‌هاست.

## ۷.۱ مشخصات فیلتر (Filter Specifications)

بسیاری از فیلترهایی که در عمل استفاده می‌شوند، بر اساس طرح تلورانس مشابه شکل زیر (شکل 7.1) مشخص می‌شوند؛ بدون هیچگونه قید خاصی روی فاز پاسخ، جز آنچه پایداری و علیت تحمیل می‌کنند.

به‌عنوان مثال، برای یک سیستم IIR علّی و پایدار، قطب‌های تابع سیستم باید داخل دایره واحد قرار گیرند. در طراحی فیلترهای FIR نیز اغلب قید «فاز خطی» اعمال می‌شود. این قید باعث حذف فاز از ملاحظات طراحی می‌شود.

---

### مثال 7.1 – تعیین مشخصات برای یک فیلتر گسسته‌زمان (Discrete-Time)

یک فیلتر پایین‌گذر گسسته‌زمانی را در نظر بگیرید که برای فیلتر کردن یک سیگنال پیوسته‌زمانی استفاده می‌شود، با استفاده از ساختار پایه‌ای نشان‌داده‌شده در شکل 7.2.

همان‌طور که در بخش 4.4 نیز دیده شد، اگر یک سیستم گسسته‌زمانی برای فیلتر کردن سیگنال پیوسته‌زمانی به‌کار رود و نرخ نمونه‌برداری به اندازه کافی زیاد باشد، آنگاه سیستم کلی مانند یک سیستم LTI پیوسته‌زمانی با پاسخ فرکانسی زیر رفتار می‌کند:

<div dir="ltr">

$$
H_{\text{eff}}(j\Omega) = 
\begin{cases}
H(e^{j\omega}), & |\Omega| < \pi / T \\
0, & |\Omega| \geq \pi / T
\end{cases}
\tag{7.1a}
$$

</div>

در چنین حالتی، به راحتی می‌توان از مشخصات فیلتر پیوسته‌زمانی به مشخصات فیلتر گسسته‌زمانی رسید، با استفاده از رابطه‌ی $$\omega = \Omega T$$. بنابراین:

<div dir="ltr">

$$
H(e^{j\omega}) = H_{\text{eff}}(j\Omega) \quad \text{for} \quad \omega = \Omega T
\tag{7.1b}
$$

</div>

---

### شکل‌ها

- **شکل 7.1:** طرح تلورانس فیلتر پایین‌گذر، شامل ناحیه عبوری (Passband)، گذار (Transition)، و توقف (Stopband).
- **شکل 7.2:** ساختار پایه برای فیلتر کردن سیگنال‌های پیوسته‌زمانی به‌کمک سیستم گسسته‌زمانی.

</div>


<div dir="rtl">

### ادامه‌ی مثال ۷.۱:

برای این مثال، سیستم کلی در شکل 7.2 باید دارای مشخصات زیر باشد، زمانی که نرخ نمونه‌برداری برابر است با \( 10^4 \) نمونه بر ثانیه (یعنی \( T = 10^{-4} \) ثانیه):

۱. بهره \( |H_{\text{eff}}(j\Omega)| \) باید در بازه‌ی فرکانسی \( 0 \leq \Omega \leq 2\pi(2000) \)، با دقت ±0.01 از مقدار واحد باشد.

۲. بهره نباید بیشتر از 0.001 باشد در باند فرکانسی \( \Omega \geq 2\pi(3000) \).

از آنجا که معادله (7.1a) تنها نگاشتی بین فرکانس‌های پیوسته‌زمان و گسسته‌زمان است، تنها فرکانس‌های مرزی را تحت تأثیر قرار می‌دهد نه تلورانس‌ها. برای این مثال خاص، مقادیر زیر به‌دست می‌آید:

<div dir="ltr">

$$
\delta_{p1} = \delta_{p2} = 0.01 \\
\delta_s = 0.001 \\
\omega_p = 0.4\pi \quad \text{radians} \\
\omega_s = 0.6\pi \quad \text{radians}
$$

</div>

بنابراین، در این حالت، بهره‌ی عبوری ایده‌آل برابر با ۱ است و مجاز است تا بین \(1+\delta_{p1}\) و \(1-\delta_{p2}\) نوسان کند، و بهره‌ی توقف مجاز است تا بین ۰ و \( \delta_s \) باشد.

بر حسب دسی‌بل:

- بهره‌ی عبوری ایده‌آل: \( 20\log_{10}(1) = 0 \ \text{dB} \)
- بیشینه بهره عبوری: \( 20\log_{10}(1.01) = 0.0864 \ \text{dB} \)
- کمینه بهره عبوری: \( 20\log_{10}(0.99) = -0.0873 \ \text{dB} \)
- بیشینه بهره توقف: \( 20\log_{10}(0.001) = -60 \ \text{dB} \)

---

این مثال در چارچوب استفاده از یک فیلتر گسسته‌زمان برای پردازش سیگنال پیوسته‌زمان بعد از نمونه‌برداری دوره‌ای بیان شده بود. در بسیاری از کاربردها، ممکن است سیگنال گسسته‌زمان از سیگنال پیوسته مشتق نشده باشد و فقط نیاز به فیلتر شدن داشته باشد. در این موارد، ممکن است مرحله‌ی نمونه‌برداری اصلاً وجود نداشته باشد یا با شکل خاصی انجام شود.

در چنین حالاتی، طراحی فیلتر باید از مشخصات سیستم گسسته‌زمان شروع شود و نه از مشخصات فیلتر پیوسته‌زمان.

---

## ۷.۲ طراحی فیلتر IIR گسسته‌زمان از روی فیلترهای پیوسته‌زمان

در گذشته، زمانی که پردازش سیگنال دیجیتال در حال توسعه بود، طراحی فیلترهای IIR گسسته‌زمان عمدتاً با تبدیل فیلترهای پیوسته‌زمان انجام می‌شد تا فیلتر گسسته‌زمانی حاصل شود که شرایط طراحی را برآورده کند.

این رویکرد هنوز هم متداول و منطقی است، به دلایل زیر:

- طراحی فیلتر IIR پیوسته‌زمان از لحاظ تئوری بسیار توسعه‌یافته است و نتایج مفیدی از آن قابل دستیابی است.
- بنابراین، استفاده از روش‌های طراحی موجود برای فیلترهای پیوسته‌زمان برای طراحی فیلترهای گسسته‌زمان مزیت دارد.

</div>

## ۷.۲ طراحی فیلترهای IIR گسسته‌زمان از روی فیلترهای پیوسته‌زمان (ادامه)

- بسیاری از روش‌های طراحی فیلتر IIR پیوسته‌زمان دارای فرمول‌های بسته‌ی ساده برای طراحی هستند. بنابراین، روش‌های طراحی فیلتر IIR گسسته‌زمان که بر اساس همین فرمول‌های استاندارد پیوسته‌زمان بنا شده‌اند، به‌سادگی قابل پیاده‌سازی هستند.
- روش‌های تقریب استانداردی که برای فیلترهای IIR پیوسته‌زمان مؤثر هستند، در حالت کلی منجر به فرمول‌های بسته‌ی ساده برای فیلترهای گسسته‌زمان نمی‌شوند، زیرا پاسخ فرکانسی فیلتر گسسته‌زمان تناوبی است در حالی که پاسخ فرکانسی فیلتر پیوسته‌زمان این‌گونه نیست.

واقعیت اینکه طراحی فیلترهای پیوسته‌زمان قابل نگاشت به طراحی فیلترهای گسسته‌زمان هستند، کاملاً مستقل از اینکه فیلتر گسسته‌زمان مطابق با پیکربندی شکل ۷.۲ استفاده شود یا نه. با این حال، در اینجا تأکید می‌شود که در تمامی روش‌هایی که بررسی می‌کنیم، فرض می‌کنیم که مشخصات فیلتر گسسته‌زمان داده شده‌اند و روش طراحی بر اساس آن مشخصات انجام می‌گیرد. به عبارت دیگر، استفاده از روش‌های طراحی پیوسته‌زمان صرفاً برای راحتی و سادگی در برآورده کردن مشخصات گسسته‌زمانی است.

---

### ۷.۲.۱ طراحی فیلتر با استفاده از **ناوری ضربه‌ای (Impulse Invariance)**

در بخش ۴.۴.۲، مفهوم **ناوری ضربه‌ای** را بررسی کردیم، که در آن یک سیستم گسسته‌زمان با نمونه‌برداری از پاسخ ضربه‌ای یک سیستم پیوسته‌زمان تعریف می‌شود.

ما نشان دادیم که ناوری ضربه‌ای یک روش مستقیم برای محاسبه نمونه‌هایی از خروجی یک سیستم پیوسته‌زمان محدود باند است که ورودی آن نیز محدود باند باشد. در برخی کاربردها، این روش کاملاً مناسب و راحت است. به‌طور خاص، اگر هدف کلی شبیه‌سازی یک سیستم پیوسته‌زمان در محیط گسسته‌زمان باشد، می‌توان طراحی را به‌گونه‌ای انجام داد که پاسخ ضربه‌ای سیستم گسسته‌زمان، نمونه‌هایی از پاسخ ضربه‌ای سیستم پیوسته‌زمان باشد.

</div>


<div dir="rtl">

در سایر کاربردها، ممکن است هدف این باشد که در محیط گسسته‌زمان، ویژگی‌های زمانی خاصی از فیلتر پیوسته‌زمان حفظ شوند، مانند **overshoot** (بیش‌جهش)، **تراکم انرژی**، **نوسان محدود زمانی** و غیره.

در این حالت نیز می‌توان از روش ناوری ضربه‌ای استفاده کرد تا یک فیلتر گسسته‌زمان با پاسخ فرکانسی مشتق‌شده از فیلتر پیوسته‌زمان تولید شود.

در **روش ناوری ضربه‌ای** برای تبدیل فیلترهای پیوسته‌زمان به گسسته‌زمان، پاسخ ضربه‌ای فیلتر گسسته‌زمان به صورت نمونه‌برداری شده از پاسخ ضربه‌ای فیلتر پیوسته‌زمان تعریف می‌شود، یعنی:

<div dir="ltr">

$$
h[n] = T_d h_c(nT_d) \tag{7.2}
$$

</div>

که در آن \( T_d \) دوره‌ی نمونه‌برداری است. (همان‌طور که در ادامه خواهیم دید، مقدار \( T_d \) هیچ تأثیری در طراحی ندارد و صرفاً به دلایل فرمالی در معادله آورده می‌شود.)

اگر از پیکربندی شکل 7.2 استفاده شود، دوره‌ی نمونه‌برداری \( T_d \) الزاماً برابر با دوره‌ی نمونه‌برداری \( T \) در تبدیل C/D یا D/C نیست.

---

در طراحی فیلتر با استفاده از ناوری ضربه‌ای، تمرکز اصلی ما روی **رابطه بین پاسخ‌های فرکانسی** فیلتر گسسته‌زمان و پیوسته‌زمان است. همان‌طور که در فصل ۴ دیدیم، این دو پاسخ با رابطه‌ی زیر مرتبط‌اند:

<div dir="ltr">

$$
H(e^{j\omega}) = \sum_{k=-\infty}^{\infty} H_c\left( \frac{\omega}{T_d} + \frac{2\pi k}{T_d} \right) \tag{7.3}
$$

</div>

اگر فیلتر پیوسته‌زمان محدودباند باشد، به‌طوری‌که:

<div dir="ltr">

$$
H_c(j\Omega) = 0 \quad \text{برای} \quad |\Omega| \geq \frac{\pi}{T_d} \tag{7.4}
$$

</div>

آنگاه:

<div dir="ltr">

$$
H(e^{j\omega}) = H_c\left( \frac{\omega}{T_d} \right), \quad |\omega| \leq \pi \tag{7.5}
$$

</div>

به عبارت دیگر، پاسخ فرکانسی گسسته‌زمان دقیقاً نگاشتی از پاسخ پیوسته‌زمان است، اگر شرایط بالا برقرار باشد.

---

در غیر این صورت، **پدیده تداخل طیفی (aliasing)** بین ضرایب \( k \) مختلف در جمع (7.3) رخ خواهد داد. با این حال، اگر پاسخ فرکانسی فیلتر پیوسته‌زمان در فرکانس‌های بالا بسیار کوچک باشد، این تداخل ممکن است ناچیز باشد.

---

در نتیجه، اگر روش ناوری ضربه‌ای برای طراحی یک فیلتر گسسته‌زمان استفاده شود، ابتدا مشخصات پاسخ فرکانسی مورد نظر به مشخصات یک فیلتر پیوسته‌زمان تبدیل می‌شوند (با استفاده از معادله 7.5)، و سپس با نمونه‌برداری از پاسخ ضربه‌ای فیلتر پیوسته‌زمان، پاسخ ضربه‌ای فیلتر گسسته‌زمان تولید می‌شود.

</div>


<div dir="rtl">

<div align="center">  
شکل 7.3 – نمایش تداخل طیفی (Aliasing) در روش طراحی با ناوری ضربه‌ای  
</div>

---

در صورتی که تداخل طیفی ناشی از تبدیل \( H_c(j\Omega) \) به \( H(e^{j\omega}) \) ناچیز باشد، با استفاده از رابطه:

<div dir="ltr">

$$
\Omega = \frac{\omega}{T_d} \tag{7.6}
$$

</div>

می‌توان مشخصات مورد نظر روی \( H(e^{j\omega}) \) را به مشخصات معادل روی \( H_c(j\Omega) \) تبدیل کرد.

---

پس از به‌دست‌آوردن فیلتر پیوسته‌زمانی که این مشخصات را برآورده می‌کند، آن را به فیلتر گسسته‌زمان با تابع تبدیل \( H(z) \) تبدیل می‌کنیم. جزئیات جبری این تبدیل در ادامه ارائه خواهد شد.

لازم به ذکر است که در تبدیل به حوزه‌ی گسسته‌زمان، رابطه (7.3) باعث می‌شود که فرکانس‌های درون \( H(e^{j\omega}) \) با فرکانس‌های درون \( H_c(j\Omega) \) از طریق جمع تناوبی در امتداد محور فرکانس در ارتباط باشند. این یعنی علی‌رغم استفاده از رابطه (7.6)، هنوز خطر aliasing وجود دارد.

---

چون پارامتر \( T_d \) را نمی‌توان برای کنترل aliasing استفاده کرد (چون مشخصات اولیه روی دامنه فرکانسی گسسته‌زمان داده شده‌اند)، تنها راه کاهش aliasing افزایش نرخ نمونه‌برداری است، یعنی کاهش \( T_d \). اگر فرکانس قطع فیلتر پیوسته‌زمانی زیاد باشد، باید نرخ نمونه‌برداری متناسب افزایش یابد.

در عمل، فیلتر پیوسته‌زمان باید چنان طراحی شود که انرژی طیفی آن در نواحی بالا (که منجر به aliasing می‌شوند) ناچیز باشد، و بدین ترتیب، تداخل بین باندها کمینه گردد.

---

در طراحی فیلتر با روش ناوری ضربه‌ای، اگر خواسته باشیم آن را به‌صورت **زمانی** انجام دهیم (یعنی از روی تابع تبدیل \( H_c(s) \))، این کار ساده‌تر از حالت فرکانسی خواهد بود. برای این منظور، فرض می‌کنیم \( H_c(s) \) دارای گسترش به صورت زیر باشد (کسر جزئی):

<div dir="ltr">

$$
H_c(s) = \sum_{k=1}^{N} \frac{A_k}{s - s_k} \tag{7.7}
$$

</div>

در این صورت، پاسخ ضربه‌ای متناظر برابر است با:

<div dir="ltr">

$$
h_c(t) =
\begin{cases}
\sum\limits_{k=1}^{N} A_k e^{s_k t}, & t \geq 0 \\
0, & t < 0
\end{cases}
\tag{7.8}
$$

</div>

<div style="font-size: smaller">
¹ برای ساده‌سازی، فرض شده که تمام قطب‌های \( H(s) \) از مرتبه اول هستند. در مسئله 7.41، حالت قطب‌های با مرتبه بالاتر بررسی خواهد شد.
</div>

</div>


<div dir="rtl">

تابع پاسخ ضربه‌ای فیلتر گسسته‌زمان علی که با نمونه‌برداری از \( T_d h_c(t) \) به دست می‌آید برابر است با:

<div dir="ltr">

$$
h[n] = T_d h_c(n T_d) = \sum_{k=1}^{N} T_d A_k e^{s_k n T_d} u[n] \tag{7.9}
$$

</div>

تابع تبدیل سیستم گسسته‌زمان علی متناظر برابر است با:

<div dir="ltr">

$$
H(z) = \sum_{k=1}^{N} \frac{T_d A_k}{1 - e^{s_k T_d} z^{-1}} \tag{7.10}
$$

</div>

---

با مقایسه‌ی معادلات (7.7) و (7.10) مشاهده می‌کنیم که هر قطب \( s_k \) در صفحه‌ی s به قطب \( z_k = e^{s_k T_d} \) در صفحه‌ی z نگاشته می‌شود. ضرایب در بسط کسر جزئی \( H_c(s) \) و \( H(z) \) یکسان هستند، بجز عامل مقیاس \( T_d \).

اگر فیلتر پیوسته‌زمانی پایدار باشد، یعنی بخش حقیقی \( s_k \) منفی باشد، آنگاه \( |e^{s_k T_d}| < 1 \) و بنابراین قطب متناظر درون دایره‌ی واحد قرار دارد. لذا، فیلتر گسسته‌زمان نیز پایدار خواهد بود.

نکته‌ی مهم: نگاشت قطب‌ها از s به z به این معنا نیست که صفرهای سیستم هم به‌همان صورت نگاشت می‌شوند. ضرایب \( T_d A_k \) ممکن است منجر به تغییر مکان صفرها شوند. بنابراین، رفتار صفرها ممکن است با نگاشت \( z_k = e^{s_k T_d} \) مطابقت نداشته باشد.

### مثال 7.2: ناوری ضربه‌ای برای فیلتر Butterworth

در این مثال، طراحی یک فیلتر پایین‌گذر گسسته‌زمان را با روش ناوری ضربه‌ای بررسی می‌کنیم. فیلتر پیوسته‌زمان مرجع ما یک فیلتر Butterworth خواهد بود (در ضمیمه B به‌تفصیل توضیح داده شده‌اند).

مشخصات موردنیاز برای فیلتر به‌صورت زیر داده شده‌اند:

<div dir="ltr">

$$
0.89125 \leq |H(e^{j\omega})| \leq 1, \quad 0 \leq \omega \leq 0.3\pi \tag{7.11a}
$$

$$
|H(e^{j\omega})| \leq 0.17783, \quad 0.35\pi \leq \omega \leq \pi \tag{7.11b}
$$

</div>

---

در طراحی فیلتر با ناوری ضربه‌ای از روی یک فیلتر Butterworth، ابتدا باید مشخصات فرکانسی گسسته‌زمان را به مشخصات معادل پیوسته‌زمان تبدیل کنیم. برای این مثال، فرض می‌کنیم aliasing ناشی از رابطه‌ی (7.3) قابل صرف‌نظر باشد.

پس از طراحی، می‌توان پاسخ فرکانسی حاصل را با مشخصات (7.11a) و (7.11b) مقایسه و ارزیابی کرد.

<div style="font-size: smaller">
² فیلترهای Butterworth و Chebyshev پیوسته‌زمان در ضمیمه‌ی B توضیح داده شده‌اند.
</div>

</div>


<div dir="rtl">

---

با توجه به ملاحظات قبل، می‌خواهیم فیلتر Butterworth پیوسته‌زمان با تابع اندازه زیر طراحی کنیم:

<div dir="ltr">

$$
0.89125 \leq |H_c(j\Omega)| \leq 1, \quad 0 \leq |\Omega| \leq 0.2\pi \tag{7.12a}
$$

$$
|H_c(j\Omega)| \leq 0.17783, \quad 0.3\pi \leq |\Omega| \leq \pi \tag{7.12b}
$$

</div>

---

از آنجا که تابع اندازه‌ی Butterworth تک‌مدوله است، روابط (7.12a) و (7.12b) با موارد زیر ارضا می‌شوند:

<div dir="ltr">

$$
|H_c(j0.2\pi)| \geq 0.89125 \tag{7.13a}
$$

$$
|H_c(j0.3\pi)| \leq 0.17783 \tag{7.13b}
$$

</div>

تابع اندازه-توان دوم برای فیلتر Butterworth برابر است با:

<div dir="ltr">

$$
|H_c(j\Omega)|^2 = \frac{1}{1 + (\Omega / \Omega_c)^{2N}} \tag{7.14}
$$

</div>

هدف طراحی، تعیین پارامترهای \( \Omega_c \) و \( N \) است که این مشخصات را ارضا کنند.

با جای‌گذاری در معادله (7.14) و اعمال برابری از معادلات (7.13) داریم:

<div dir="ltr">

$$
1 + \left( \frac{0.2\pi}{\Omega_c} \right)^{2N} = \frac{1}{(0.89125)^2} \tag{7.15a}
$$

$$
1 + \left( \frac{0.3\pi}{\Omega_c} \right)^{2N} = \frac{1}{(0.17783)^2} \tag{7.15b}
$$

</div>

---

حل همزمان معادلات بالا منجر به مقادیر تقریبی زیر می‌شود:

- \( N = 5.888 \)
- \( \Omega_c = 0.70474 \)

از آن‌جا که \( N \) باید عدد صحیح باشد، مقدار آن به بالا گرد می‌شود:  
👉 بنابراین، \( N = 6 \) و مقدار \( \Omega_c \) طوری تنظیم می‌شود که یکی از نواحی گذر یا توقف به‌درستی برآورده شود.

اگر معادله (7.15a) را با \( N = 6 \) حل کنیم، به مقدار جدید \( \Omega_c = 0.7032 \) می‌رسیم. این مقدار باعث می‌شود که شرایط گذر (passband) با دقت بیشتری برآورده شوند، حتی اگر شرایط توقف کمی نقض شود (حداکثر نشت: \( |H_c(j0.3\pi)| \approx 0.182 \)).

---

قطب‌های تابع مربع اندازه Butterworth \( H_c(s)H_c(-s) \) با \( N = 6 \) روی دایره‌ای به شعاع \( 0.7032 \) در صفحه‌ی s به‌طور یکنواخت پخش شده‌اند. بنابراین، ۳ جفت قطب مختلط مزدوج خواهیم داشت:

**Pole pair 1:**
<div dir="ltr">\( -0.182 \pm j(0.679) \)</div>

**Pole pair 2:**
<div dir="ltr">\( -0.497 \pm j(0.497) \)</div>

**Pole pair 3:**
<div dir="ltr">\( -0.679 \pm j(0.182) \)</div>

---

</div>

### ادامه مثال ۷.۲ – طراحی فیلتر با ناوری ضربه‌ای (Impulse Invariance)

قطب‌های تابع \( H_c(s)H_c(-s) \) مربوط به فیلتر مرتبه ۶ Butterworth در صفحه s مطابق شکل زیر توزیع شده‌اند:

<div dir="ltr">

**Figure 7.4** – مکان قطب‌ها در صفحه s برای فیلتر Butterworth مرتبه ۶  
\( \Omega_c = 0.7032 \)

</div>

با استفاده از مکان قطب‌ها، تابع تبدیل پیوسته‌زمان فیلتر Butterworth به صورت زیر نوشته می‌شود:

<div dir="ltr">

$$
H_c(s) = \frac{0.12093}{(s^2 + 0.3640s + 0.4945)(s^2 + 0.9945s + 0.4945)(s^2 + 1.3585s + 0.4945)} \tag{7.16}
$$

</div>

اکنون اگر \( H_c(s) \) را به بسط کسر جزئی تبدیل کنیم، و تبدیل ناوری ضربه‌ای را با ترکیب قطب‌های مزدوج مختلط انجام دهیم، تابع تبدیل گسسته‌زمان نهایی برابر خواهد بود با:

<div dir="ltr">

$$
H(z) =
\frac{0.2871 - 0.4466z^{-1}}{1 - 1.2971z^{-1} + 0.6949z^{-2}} +
\frac{-2.1428 + 1.1455z^{-1}}{1 - 1.0691z^{-1} + 0.3699z^{-2}} +
\frac{1.8557 - 0.6303z^{-1}}{1 - 0.9972z^{-1} + 0.2570z^{-2}}
\tag{7.17}
$$

</div>

در اینجا از رابطه (7.10) و بسط کسر جزئی (7.7) برای انتقال از سیستم پیوسته به گسسته استفاده شده است.

### نکات تکمیلی:

- همان‌طور که از رابطه (7.17) پیداست، می‌توان فیلتر نهایی را به صورت **موازی از سه فیلتر مرتبه‌دوم** پیاده‌سازی کرد (شکل مستقیم II یا حالت cascade قابل استفاده است).
- پاسخ فرکانسی این فیلتر گسسته‌زمان (که در شکل 7.5 نمایش داده شده) نشان می‌دهد که مشخصات طراحی به‌خوبی رعایت شده‌اند.
- اثر ناوری (aliasing) در این مثال ناچیز بوده و تأثیر قابل توجهی نداشته است.
- اگر اثر aliasing بیشتر بود و طراحی شکست می‌خورد، باید از فیلتر مرتبه بالاتر یا پارامترهای جدیدی برای طراحی استفاده می‌کردیم.
