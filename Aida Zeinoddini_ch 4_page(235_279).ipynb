### تحلیل و مستر در ده  
#### ۶-۷-۴. بانک‌های فیلتر چندآهنگی  

ساختارهای چندفازی برگزینش و درون‌یابی کاربرد گسترده‌ای در بانک‌های فیلتر مورد استفاده در تحقیقات سیگنال‌های شنیداری و گفتاری، به‌ویژه در یک فیلتر دوکاناله، پیدا کرده‌اند. شکل ۲-۲۴، نمودار بلوکی یک بانک فیلتر تحلیل و سنتز متداول را در کاربردهای کدگذاری گفتار نشان می‌دهد.

هدف بخش تحلیل سیستم، تجزیه‌ی طیف فرکانسی ورودی به یک باند پایین‌گذر (نمایش یافته با سیگنال کم‌نمونه‌برداری شده \( a[n] \)) و یک باند بالاگذر (نمایش یافته با سیگنال \( n[n] \)) است. در کاربردهای کدگذاری گفتار، سیگنال‌های این دو کانال برای انتقال و ذخیره‌سازی کوانتیزه می‌شوند.

چون باند اصلی در بخش مساوی دارای پهنای \(3.2\) رادیان تقسیم می‌شود، نرخ نمونه‌برداری خروجی فیلترها می‌تواند نصف نرخ ورودی باشد، طوری که تعداد کل نمونه‌ها در کل ثابت بماند. توجه کنید که کم‌نمونه‌برداری از خروجی فیلتر پایین‌گذر باعث می‌شود باند فرکانس پایین تمام گستره را بپوشاند، و به‌طور مشابه، کم‌نمونه‌برداری از خروجی فیلتر بالاگذر نیز باند فرکانس بالا را به سمت پایین سوق داده و کل گستره را پر می‌کند.

این تجزیه مستلزم آن است که \( a[n] \) و \( h[n] \) به ترتیب پاسخ ضربه‌ی فیلترهای پایین‌گذر و بالاگذر باشند. یک رهیافت متداول گرفتن فیلتر بالاگذر از فیلتر پایین‌گذر به صورت \( h[n] = (-1)^n a[n] \) است. این وضعیت باعث می‌شود داشته باشیم:

- اگر \( A(e^{j\omega}) \) یک فیلتر پایین‌گذر با باند عبور نامی باشد، آنگاه \( H(e^{j\omega}) \) فیلتر بالاگذر با باند نامی مکمل آن خواهد بود.

هدف بخش سمت راست (سنتز) در شکل ۲-۲۴، بازسازی تقریبی سیگنال ورودی بر اساس دو سیگنال \( a[n] \) و \( h[n] \) است. به این منظور، هر دو سیگنال بیش‌نمونه‌برداری شده، به ترتیب از فیلترهای پایین‌گذر \( L_a \) و \( L_h \) عبور داده می‌شوند. سپس سیگنال‌های درونیابی‌شده‌ی حاصل با هم جمع شده و سیگنال خروجی \( x[n] \) را تشکیل می‌دهند که همان نرخ نمونه‌برداری ورودی را دارد.

اعمال نتایج حوزه‌ی فرکانس برای کم‌نمونه‌برداری و بیش‌نمونه‌برداری به سیستم شکل ۲-۲۴ منجر به روابط زیر می‌شود:

#### معادله (۲-۱۱۱):
\[
Y(e^{j\omega}) = \frac{1}{2} \left[ H_0(e^{j\omega})H_0(e^{j\omega}) + H_1(e^{j\omega})H_1(e^{j\omega}) \right] X(e^{j\omega})
\]

این سیستم‌ها که تعداد نمونه در واحد زمان را ثابت نگه می‌دارند، بانک‌های فیلتر با کاهش یکنواخت (uniformly decimated filter banks) نامیده می‌شوند.

در صورتی که فیلترهای تحلیل و سنتز ایده‌آل باشند، طوری که دقیقاً باندها را به بخش‌های مساوی و بدون هم‌پوشانی تقسیم کنند، بانک فیلتر قادر به بازسازی کامل سیگنال ورودی خواهد بود. اما در حالت فیلترهای غیرایده‌آل، هم‌پوشانی رخ می‌دهد. با این حال، با طراحی مناسب، همچنان بازسازی تقریباً کامل ممکن است.

#### شرط هم‌پوشانی معکوس (معادله ۲-۱۱۲):
\[
H_0(e^{j\omega})G_0(e^{j\omega}) + H_1(e^{j\omega})G_1(e^{j\omega}) = 2
\]

فیلترهای \( H_0 \) و \( H_1 \) را فیلترهای آینه‌ای می‌نامند، زیرا:

#### معادله (۲-۱۱۳):
\[
G_0(e^{j\omega}) = H_1(e^{j(\pi - \omega)})
\quad , \quad
G_1(e^{j\omega}) = -H_0(e^{j(\pi - \omega)})
\]

گذاشتن این روابط در معادله (۲-۱۱۱) رابطه زیر را به دست می‌دهد:

#### معادله (۲-۱۱۴):
\[
Y(e^{j\omega}) = X(e^{j\omega}) e^{-j\omega M}
\]

که نشان می‌دهد بازسازی کامل با تأخیر \( M \) نمونه، ممکن است اگر شرط زیر برقرار باشد:

#### معادله (۲-۱۱۵):
\[
H_0(e^{j\omega})H_0(e^{j\omega}) + H_1(e^{j\omega})H_1(e^{j\omega}) = 2
\]

وی‌دیناتان (Vaidyanathan, 1993) نشان داده است که فیلترهایی با پاسخ ضربه از نوع:

\[
h[n] = \delta[n - 28] - \delta[n - 8]
\]

قابل ساخت هستند که می‌توانند بازسازی کامل را انجام دهند، اما ممکن است برای کاربردهایی مثل کدگذاری گفتار، ویژگی‌های فرکانسی مناسب را نداشته باشند.

برای نمونه، در نظر بگیرید فیلتر پایین‌گذر میانگین متحرک زیر را:

#### معادله (۲-۱۶۲):
\[
h[n] = \frac{1}{2} (\delta[n] + \delta[n - 1])
\]

پاسخ فرکانسی این فیلتر از معادله (۲-۱۱۴) و جایگذاری در (۲-۱۱۲) به‌دست می‌آید.

### استفاده از فیلترهای FIR و IIR در سیستم تحلیل و سنتز

در سیستم تحلیل–سنتز شکل ۲-۲۲ می‌توان از فیلترهای **FIR** یا **IIR** استفاده کرد، مشروط بر این‌که فیلتر پایین‌گذر، تقریب قابل قبولی از فیلتر ایده‌آل ارائه دهد. برای بازسازی تقریباً کامل، باید روابط بین فیلترها طبق معادلات زیر برقرار باشد:

$$
\text{معادلات (۲-۱۱۳-الف تا ۲-۱۱۳-ج)}
$$

مجموعه‌ای از این فیلترها و الگوریتم طراحی آن‌ها توسط **جات (ع)** معرفی شده‌اند. همچنین **اسمیت و بارتول (۱۹۸۲)** و **مینتزر (۱۹۸۵)** نشان داده‌اند که بازسازی کامل بانک فیلتر دوکاناله شکل ۲-۲۲ تنها در صورتی ممکن است که فیلترها رابطه خاصی داشته باشند که در معادلات (۲-۱۱۳) تعریف نشده و در نتیجه به دسته‌ای از فیلترها به نام **فیلترهای QMF اصلاح‌شده (Modified QMF)** منتهی می‌شود.

### ساختار چندفازی

برای کاهش محاسبات، می‌توان ساختار چندفازی را به سیستم تحلیل–سنتز اعمال کرد. این کار با به‌کارگیری کم‌نمونه‌برداری چندفازی (Polyphase) انجام می‌شود. شکل ۲۵-۲ (الف) ساختار تحلیل چندفازی را نشان می‌دهد.

$$
\text{معادله (۲-۱۱۷-الف)}
$$

شکل ۲۵-۴ نمایش چندفازی سیستم تحلیل دوکاناله را نشان می‌دهد. فیلترهای چندفازی در این ساختار، به‌گونه‌ای انتخاب می‌شوند که فقط یک مجموعه از فیلترها نیاز باشد. به این ترتیب، فقط نیمی از محاسبات لازم خواهد بود.

### توسعه به ساختارهای چندکاناله

این سیستم تحلیل–سنتز دوکاناله را می‌توان به ۸ کانال با پهنای باند برابر توسعه داد که در **کدگذاری صوتی** کاربرد دارد. این توسعه، هماهنگ با ویژگی‌های سیستم شنوایی انسان بوده و به **فشرده‌سازی اطلاعات دیجیتال** کمک می‌کند.

استفاده از فیلترهای **CQF** باعث بازسازی دقیق سیگنال می‌شود و سیستم حاصل در واقع همان **تبدیل موجک گسسته (DWT)** است.

### ۸-۴ پردازش دیجیتال سیگنال‌های آنالوگ

تا این مرحله، تمرکز روی نمایش سیگنال‌های پیوسته در زمان به‌صورت گسسته بود. این نمایش مبتنی بر سیستم‌های تبدیل به گسسته در زمان (CD) و تبدیل به پیوسته (DIC) بود. برای مثال، در بخش ۲-۴، از این مدل‌ها برای پیاده‌سازی سیستم‌های LTI پیوسته استفاده شد، مشروط بر این‌که ورودی باند-محدود باشد و نرخ نمونه‌برداری از **نرخ نایکویست** بیشتر باشد.

در عمل، سیگنال‌ها دقیقاً باند-محدود نیستند و فیلتر ایده‌آل هم قابل پیاده‌سازی نیست. بنابراین از مبدل‌های A/D و D/A استفاده می‌شود. شکل ۲۷-۹ (ب) نمودار واقع‌گرایانه‌ای از پردازش دیجیتال سیگنال‌های آنالوگ است.

---

### ۱-۸-۴ پیش‌فیلتر برای جلوگیری از درهم‌روی

در پردازش سیگنال‌های آنالوگ، پایین نگه‌داشتن نرخ نمونه‌برداری مطلوب است؛ چراکه حجم محاسبات کاهش می‌یابد. اما اگر ورودی باند-محدود نباشد، ممکن است **پیش‌فیلتر (Anti-Aliasing Filter)** لازم باشد.

برای مثال، در پردازش گفتار، اطلاعات مهم در باند فرکانسی تا حدود ۴ کیلوهرتز قرار دارند. اگرچه سیگنال گفتار ممکن است محتوای فرکانسی بالاتری نیز داشته باشد، ولی باید با فیلتر پایین‌گذر قبل از مبدل A/D، محتوای بالا را حذف کرد.

فیلتر ضد درهم‌روی ایده‌آل باید دارای پاسخ فرکانسی زیر باشد:

$$
\text{معادله (۲-۱۱۸)}
$$

شکل ۲۸-۴ این کاربرد را به تصویر می‌کشد. در نهایت، ترکیب پاسخ فرکانسی پیش‌فیلتر و مبدل‌ها باعث می‌شود که کل سیستم مانند یک سیستم **LTI** رفتار کند، مشروط بر این‌که ورودی باند-محدود باشد و فرکانس‌های بالاتر از $\pi$ حذف شده باشند.

### ادامه‌ی بحث فیلتر ضد درهم‌روی و پردازش دیجیتال سیگنال‌های آنالوگ

پس به ازای **فیلتر ضد درهم‌روی ایده‌آل پایین‌گذر**، سیستم شکل ۲-۲۸ به‌صورت یک سیستم **LTI** با پاسخ فرکانسی داده‌شده توسط معادله (۲-۱۱۹) رفتار می‌کند، حتی اگر ورودی‌ها باند-محدود نباشند.

در عمل، پاسخ فرکانسی \( H_0(e^{j\omega}) \) نمی‌تواند باند-محدود ایده‌آل باشد، ولی می‌توان آن را (در رابطه‌ی ۲٫۷) به اندازه‌ای کوچک کرد که درهم‌روی مینیمم شود. در این صورت پاسخ فرکانسی کل سیستم شکل ۲۰۰۲ را می‌توان به صورت زیر تقریب زد:

$$
H(e^{j\omega}) = H_H(e^{j\omega}) \cdot H_0(e^{j\omega}) \quad \text{(معادله ۲-۱۲۰)}
$$

برای این که پاسخ فرکانسی در فرکانس‌های بالاتر از \(\pi\) ناچیز باشد، باید \(H_0(e^{j\omega})\) از فرکانسی پایین‌تر از \(\pi\) شروع به افت کند.

معادله (۲-۱۲۰) بیان می‌کند که این افت در فیلتر ضد درهم‌روی و دیگر اعوجاج‌های **LTI** (که بعداً بررسی خواهند شد) هنگام طراحی سیستم گسسته در زمان تا حدی قابل جبران هستند. این نکته در مسئله ۲-۶۲ بررسی خواهد شد.

در این بحث فرض کردیم فیلتر ضد درهم‌روی افت تیزی دارد. فیلتر آنالوگ با افت تیز را می‌توان با شبکه‌های فعال و به‌صورت مدار مجتمع پیاده‌سازی کرد. ولی در کاربردهایی که از پردازشگرهای دیجیتال استفاده می‌شود (که توانمند و ارزان هستند)، این فیلترهای پیوسته در زمان ممکن است بخش عمده‌ای از هزینه را تشکیل دهند. همچنین، پیاده‌سازی فیلترهایی با شیب قطع تند مشکل و پرهزینه است و اگر بخواهیم با نرخ نمونه‌برداری متغیر کار کنیم، نیاز به فیلترهای قابل تنظیم خواهیم داشت.

فیلترهای با شیب تند نیز معمولاً پاسخ فازی شدیداً غیرخطی دارند، به‌ویژه در ناحیه‌ی باند عبور. پس ساده‌سازی فیلترهای پیوسته در زمان به دلایل مختلف مطلوب است.

### رهیافت ساده‌سازی فیلتر ضد درهم‌روی

یک رهیافت در شکل ۲۹-۴ نشان داده شده است. اگر بالاترین مؤلفه فرکانسی سیگنال (پس از عبور از فیلتر ضد درهم‌روی) محدود باشد، می‌توان از یک فیلتر بسیار ساده با افت تدریجی استفاده کرد. سپس مبدل \( C/D \) را با نرخ نمونه‌برداری بسیار بزرگ‌تر از \(\omega_N\) پیاده‌سازی کرده و نرخ نمونه‌برداری را کاهش داد.

برای این کار از یک فیلتر ضد درهم‌روی **گسسته در زمان** استفاده می‌شود. حال می‌توان پردازش بعدی را با نرخ نمونه‌برداری کمتر انجام داد تا محاسبات کاهش یابد.

### نمونه‌سازی و تغییر نرخ نمونه‌برداری

کاربرد **پیش‌نمونه‌برداری** و **تغییر نرخ نمونه‌برداری** در شکل ۲-۵۰ نمایش داده شده است.

- **شکل ۲-۵۰ (الف):** تبدیل فوریه‌ی سیگنال در باند فرکانسی ۰ تا ۲۰ کیلوهرتز، به‌همراه مؤلفه‌های ناخواسته در فرکانس‌های بالا.
- **شکل ۲-۵۰ (ب):** تبدیل فوریه‌ی خروجی فیلتر ضد درهم‌روی ساده با افت تدریجی.
- **شکل ۲-۵۰ (ج):** سیگنال پس از نمونه‌برداری با نرخ بالا و تأثیر درهم‌روی.
- **شکل ۲-۵۰ (د):** خروجی فیلتر گسسته در زمان با ضریب کاهش یافته‌ی نرخ نمونه‌برداری.

به این ترتیب، تنها یک فیلتر پیوسته در زمان ساده کافی است. چون فیلترهای FIR در زمان گسسته می‌توانند پاسخ فازی کاملاً خطی داشته باشند، این رهیافت برای کاربردهایی که **حفظ شکل موج** اهمیت دارد، مزایای بسیاری خواهد داشت.

### ۲-۸-۴ تبدیل آنالوگ به دیجیتال

مبدل \( C/D \) سیگنال پیوسته در زمان را به سیگنال گسسته در زمان تبدیل می‌کند که هر نمونه آن دقت محدودی دارد (کوانتیزه‌شده).

**شکل ۲-۵۱:** آرایش فیزیکی مبدل A/D شامل:

- فیلتر ضد درهم‌روی
- نمونه‌بردار
- مبدل A/D

در این سیستم، سیگنال آنالوگ به سیگنال دیجیتال (رشته‌ای از نمونه‌های دودویی) تبدیل می‌شود. مبدل A/D توسط سیگنال ساعت کنترل می‌شود و عملکرد آن با نمونه‌بردار-نگهدارنده ترکیب می‌شود.

پاسخ ضربه‌ی سیستم نگهدارنده درجه صفر به صورت زیر است:

$$
h(t) =
\begin{cases}
1, & 0 \leq t < T \\\\
0, & \text{در غیر این صورت}
\end{cases}
$$

که می‌توان آن را به شکل معادل زیر نیز نوشت:

$$
\text{(معادله ۲-۱۲۳)}
$$

نمونه‌بردار نگهدار ایدئال، معادل یک قطار ضربه مدوله‌شده است که مطابق شکل ۱۲۰۲ العنابی، دنبال آن یک فیلتر عالی به‌صورت نگهدار رتبه صفر قرار دارد. رابطه بین تبدیل فوریه آن و تابع زمانی را می‌توان مشابه تحلیل بخش ۴-۲ به‌دست آورد. اما چنین تحلیلی در اینجا ضروری نیست، زیرا رفتار سیستم را از حوزه زمان نیز می‌توان به‌خوبی درک کرد.

به‌طور مشخص، خروجی نگهدار مرتبه صفر یا همان "رنه مدرید"، شکل موج پلکانی دارد که مقدار نمونه‌ها را در دوره نمونه‌برداری Δ ثابت نگه می‌دارد.

### نمونه‌بردار - نگهدار

در شکل ۲-۵۲ (الف)، نمایش نمونه‌بردار - نگهدار ایدئال آمده است. شکل ۲-۵۲ (ب) خروجی این سیستم را نشان می‌دهد.

مدارهای واقعی نمونه‌بردار - نگهدار طوری طراحی می‌شوند که تا حد ممکن به‌صورت لحظه‌ای نمونه‌برداری انجام شود و مقدار نمونه تا زمان نمونه‌برداری بعدی ثابت بماند. هدف از این کار فراهم‌کردن ولتاژ یا جریان ثابت برای مبدل A/D است.

بررسی جزئیات طراحی این مدارها و ساختارهای آن‌ها خارج از حیطه این کتاب است، ولی مطالعه‌هایی توسط Hanatek (1988) و Selmiid (1976) انجام شده‌اند.

---

### کوانتش و مبدل A/D

در این بخش، اثرات کوانتش در تبدیل آنالوگ به دیجیتال بررسی می‌شود.

کوانتایزر یک سیستم غیرخطی است که نمونه‌های ورودی \( x_{\text{in}} \) را به نزدیک‌ترین سطح از یک مجموعه محدود از سطوح کوانتش تبدیل می‌کند. این عملیات منجر به نمونه‌های کوانتیده می‌شود.

کوانتایزر می‌تواند سطوح یکنواخت یا نایکنواخت داشته باشد، اما برای انجام عملیات محاسباتی، معمولاً سطوح یکنواخت انتخاب می‌شوند. شکل ۲-۵۲ مشخصه یک کوانتایزر یکنواخت را نشان می‌دهد که در آن، نمونه‌ها به نزدیک‌ترین سطح گرد می‌شوند.

---

### کدگذاری دودویی

در شکل ۲-۵۲، کدگذاری سطوح کوانتش نیز نشان داده شده است. با داشتن ۸ سطح کوانتش، کدگذاری با ۳ بیت ممکن است.

در حالت کلی، با n بیت می‌توان \( 2^n \) سطح کوانتش را نمایش داد. یکی از روش‌های مرسوم، استفاده از کد متمم ۲ (Two's Complement) است که در آن بیت سمت چپ، بیت علامت است و باقی بیت‌ها مقدار عددی را تعیین می‌کنند.

در این حالت، عدد دودویی کسری ۳ بیتی در متمم ۲ به شکل زیر تعریف می‌شود:

\[
\text{عدد دودویی} = -b_0 + b_1 \cdot 2^{-1} + b_2 \cdot 2^{-2}
\]

رابطه عددی بین کد و مقدار کوانتیده به پارامتر مقیاس کامل مبدل A/D بستگی دارد.

---

### اثرات کوانتش و خطا

کوانتایزر معمولاً باعث ایجاد خطاهایی به نام "خطای کوانتش" می‌شود. اگر نمونه‌ای خارج از گستره کوانتش باشد، اصطلاحاً بریده می‌شود.

مدل خطای کوانتش به‌صورت نویز سفید جمع‌شونده در نظر گرفته می‌شود و تحلیل آن بر مبنای فرض‌های آماری زیر انجام می‌شود:

- خطای کوانتش مستقل و تصادفی است.
- توزیع یکنواخت دارد.
- با سیگنال ورودی همبستگی ندارد.

در بسیاری از سیگنال‌های واقعی (مثل گفتار و موسیقی)، این فرض‌ها قابل قبول‌اند.

---

### مدل‌سازی ریاضی خطای کوانتش

در این مدل، خطای کوانتش معمولاً به صورت زیر تعریف می‌شود:

\[
e[n] = x[n] - x_q[n]
\]

که در آن \( x[n] \) مقدار اصلی و \( x_q[n] \) مقدار کوانتیده‌شده است.

تحلیل آماری این خطاها، به تخمین نسبت سیگنال به نویز (SNR) کمک می‌کند.

---

### نتیجه‌گیری

در تبدیل آنالوگ به دیجیتال، ترکیب نمونه‌بردار - نگهدار و کوانتایزر باعث ایجاد شکل پلکانی در سیگنال خروجی و همچنین ایجاد خطای کوانتش می‌شود. مدل‌سازی آماری این خطاها برای تحلیل دقیق‌تر سیستم‌های دیجیتال حیاتی است.

### شکل ۴-۵۷ (ادامه)

**(ب)** نمونه‌های کوانتیده‌شده شکل موج کسینوسی (الف) با کوانتایزر گردکننده.

**(ج)** رشته خطای کوانتش برای کوانتایزر ۳ بیتی.

**(د)** رشته خطای کوانتش برای کوانتایزر ۸ بیتی.

برای کوانتایزرهایی که مقدار نمونه را به نزدیک‌ترین سطح کوانتش گرد می‌کنند، مانند آنچه در شکل ۵۲-۴ نشان داده شده، دامنه نویز کوانتش در گستره‌ی زیر قرار دارد:

$$
-\frac{\Delta}{2} \leq e_q[n] \leq \frac{\Delta}{2}
$$

به ازای Δ کوچک، این فرض که $e_q[n]$ یک متغیر تصادفی با توزیع یکنواخت در بازه‌ی $[-\frac{\Delta}{2}, \frac{\Delta}{2}]$ است، فرضی معقول به نظر می‌رسد. چگالی احتمال مرتبه اول فرض شده برای نویز کوانتش در شکل ۵۲-۵ نشان داده شده است.

اگر در کوانتایزر به جای گرد کردن از بریدن استفاده شود، خطا همیشه منفی خواهد بود و باید چگالی احتمال را در بازه‌ای مانند $[0, \Delta]$ یکنواخت در نظر گرفت.

برای کامل کردن مدل آماری نویز کوانتش، فرض می‌کنیم نمونه‌های متوالی نویز با همدیگر همبستگی ندارند و همچنین $e_q[n]$ با $x[n]$ نیز همبستگی ندارد. بنابراین $e_q[n]$ یک نویز سفید با توزیع یکنواخت است. مقدار میانگین آن صفر است و واریانس آن برابر است با:

$$
\sigma^2 = \frac{\Delta^2}{12}
$$

برای کوانتایزر $B$ بیتی با دامنه کامل $X_m$، مقدار $\Delta$ به صورت زیر است:

$$
\Delta = \frac{2X_m}{2^B}
$$

---

برای بررسی اعتبار مدل نویز کوانتش، سیگنال $x[n] = A \cos(\omega_0 n)$ را در نظر بگیرید که با دقت ممیز شناور ۶۴ بیتی محاسبه شده است. سپس این سیگنال با $B$ بیت کوانتیده می‌شود.

برای تخمین توزیع احتمال نویز کوانتش، معمولاً از هیستوگرام استفاده می‌شود. شکل ۲-۵۹ هیستوگرام نویز کوانتش را برای کوانتایزرهای ۸ و ۱۶ بیتی نشان می‌دهد. چون تعداد کل نمونه‌ها ۱۰۱۰۰۰ و تعداد فواصل ۱۰۱ است، انتظار می‌رود در هر فاصله حدوداً ۱۰۰۰ نمونه باشد، تا توزیع یکنواخت قابل قبول باشد.

---

در فصل ۱۰ نشان خواهیم داد که چگونه می‌توان طیف چگالی توان نویز را تخمین زد. در اینجا، طیف نویز کوانتش برای تعداد نمونه‌های زیاد به شکل یکنواخت و تخت دیده می‌شود که با مدل نویز سفید با توزیع یکنواخت سازگار است.

نسبت سیگنال به نویز (SNR) برای کوانتایزر یکنواخت $B$ بیتی به‌صورت زیر بیان می‌شود:

$$
SNR_{dB} \approx 6.02B + 1.76
$$

افزایش یک بیت در طول کلمه کوانتایزر، تقریباً موجب افزایش ۶ dB در SNR می‌شود. اگر دامنه سیگنال از دامنه کامل کوانتایزر بیشتر شود، خطای برش (clipping) رخ خواهد داد و مدل دیگر معتبر نیست.

برای جلوگیری از این مشکل، فیلترها و تقویت‌کننده‌ها قبل از مبدل باید طوری طراحی شوند که سیگنال به‌طور کامل در دامنه کوانتایزر جای بگیرد.

---

### ۴-۸-۴ تبدیل دیجیتال به آنالوگ (DAC)

در بخش قبلی، بازسازی سیگنال باندمحدود با استفاده از فیلتر پایین‌گذر ایده‌آل بررسی شد. خروجی یک مبدل DAC با فیلتر پایین‌گذر ایده‌آل به صورت زیر بازسازی می‌شود:

$$
x(t) = \sum_{n=-\infty}^{\infty} x[n] \cdot \text{sinc}\left(\frac{t - nT}{T}\right)
$$

اگر سیگنال نمونه‌برداری‌شده $x[n]$ از یک سیگنال پیوسته $x(t)$ با نرخ $1/T$ باشد، بازسازی دقیق ممکن خواهد بود به شرطی که باند سیگنال اولیه از $\pi/T$ تجاوز نکند.

### فصل ۲ – نمونه‌برداری از سیگنال‌های پیوسته

فیلم درونیابی ایده‌آل

**شکل ۲-۱۳:** پاسخ فرکانسی نگهدار دسته صفر و مقایسه آن با فیلتر بازسازی ایده‌آل اعمال‌شده برای جبران‌سازی جهت کار با نگهدار دسته صفر.

با قرار دادن این فیلتر بازسازی جبران‌شده، رابطه زیر به دست می‌آید:

معادله (۱۵۰۰۲) در شکل ۱۳-الف نشان داده شده و با فیلتر درونیاب ایده‌آل معادل معادله (۱۳۹۳) مقایسه شده است. بهره هر دو فیلتر در ω = ۱۵۰۰ برابر ۱ است، ولی نگهدار دسته صفر ذاتاً یک فیلتر پایین‌گذر با قطعیت تیز ندارد.

شکل ۲۳۰۲ (ب) اندازه پاسخ فرکانسی بازسازی جبران‌شده‌ای را نشان می‌دهد که باید در سیستم بازسازی به دنبال نگهدار دسته صفر اعمال شود. این سیستم در حالت ایده‌آل با تقدم زمانی ۳٫۲ ثانیه هم‌راستا است ولی چون این تقدم را نمی‌توان در عمل ایجاد کرد، معمولاً از تأخیر استفاده می‌شود. در عمل، بهره نگهدار دسته صفر معمولاً حدود ۱٫۵ تا ۲۵٪ افت دارد.

**شکل ۲۲۰۴:** یک مدل مبدل DA را نشان می‌دهد که به دنبال آن یک فیلتر بازسازی جبران‌شده قرار گرفته است.

---

همان‌طور که در مباحث پیشین نشان داده شد، اگر به جای مدل DIA، بازسازی جبران‌شده ایده‌آل اعمال شود، سیگنال خروجی بازسازی‌شده به صورت زیر خواهد بود:

$$
y(t) = \sum x(nT) \cdot h(t - nT)
$$

که در آن، \( h(t) \) تابع پاسخ ضربه فیلتر بازسازی است و سیگنال \( x(nT) \) حاصل نمونه‌برداری سیگنال پیوسته ورودی است.

---

اگر فرض کنیم نویز ورودی یک نویز سفید باند-محدود باشد و پاسخ فیلترها خطی و زمان‌ناوابسته باشد، آن‌گاه پاسخ کل سیستم به صورت ضرب فرکانسی فیلترها است:

$$
H_{total}(f) = H_{anti-alias}(f) \cdot H_{ZOH}(f) \cdot H_{reconstruction}(f)
$$

در صورتی که بازسازی ایده‌آل نباشد، مدل ساده‌شده‌ای برای پاسخ مؤثر کل سیستم داریم که در معادله (۱۹۹۴) آمده است.

---

### ۲-۹: پیش‌نمونه‌برداری و شکل‌دهی نویز در تبدیل A/D و D/A

در این بخش نشان می‌دهیم که چگونه با استفاده از **پیش‌نمونه‌برداری** و **فیلتر دیجیتال** می‌توان اثر نویز کوانتش را کاهش داد. 

در بخش ۲-۹-۱ خواهیم دید که با پیش‌نمونه‌برداری و کاهش نرخ نمونه‌برداری می‌توان اندازه گام کوانتایزر را افزایش داد، یا نیاز به دقت بالا در تبدیل A/D را کاهش داد.

---

### ۲-۱۹: تبدیل A/D با پیش‌نمونه‌برداری و کوانتش مستقیم

شکل ۴-۲۵ سیستم تبدیل A/D با پیش‌نمونه‌برداری را نشان می‌دهد. در این مدل، سیگنال ورودی یک فرآیند تصادفی با میانگین صفر و چگالی طیف محدود فرض شده است.

برای ساده‌سازی فرض می‌کنیم سیگنال ورودی باند-محدود بوده و با نسبت بیش‌نمونه‌برداری $$ M $$ نمونه‌برداری می‌شود.

در این حالت و با فرض نویز کوانتش سفید و مستقل از سیگنال، می‌توان توان خروجی سیگنال و نویز را جداگانه محاسبه کرد:

$$
P_{signal} = \int_{-B}^{B} S_x(f) df
$$

$$
P_{noise} = \frac{\Delta^2}{12} \cdot \frac{1}{M}
$$

که در آن $$ \Delta $$ اندازه گام کوانتایزر است و $$ M $$ نسبت بیش‌نمونه‌برداری.

---

**شکل ۲-۶۷:** نمایش تغییر مقیاس فرکانسی بر اساس آهنگ نمونه‌برداری.

### تحلیل توان سیگنال و نویز پس از پیش‌نمونه‌برداری و شکل‌دهی نویز

با توجه به اینکه بود و با جایگذاری مقدار $\lambda = 12$ در معادله (۲-۱۶۵)، به‌دست می‌آوریم:

> بنابراین، کل توان سیگنال نمونه‌برداری شده و کل توان سیگنال آنالوگ اصلی یکسان‌اند، همان‌طور که در معادله (۱۲) نشان داده شد.  
> چون فیلتر برگزیده، یک فیلتر پایین‌گذر ایده‌آل با فرکانس قطع $\omega_c = 2018$ است، خروجی نیز همین مقدار را حفظ می‌کند. طیف توان نیز همین موضوع را تأیید می‌کند. سیگنال ما بدون تغییر از این فیلتر عبور می‌کند، بنابراین توان کل مؤلفه سیگنال نمونه‌برداری شده محدود باقی می‌ماند.

با استفاده از معادله (۲-۱۶۶)، به‌دست می‌آوریم:

> که نشان می‌دهد توان مؤلفه سیگنال با عبور از کل سیستم، ثابت باقی می‌ماند.  
> دلیل آن این است که برای هر تغییر مقیاس فرکانسی ناشی از نمونه‌برداری، دامنه نیز تغییر می‌کند و این دو تغییر، هم‌دیگر را موازنه می‌کنند.  
> در نتیجه، مساحت زیر طیف توان هنگام گذار از حوزه $t$ به $f$ و سپس به حوزه نمونه‌برداری‌شده ($n$)، ثابت می‌ماند.

#### نویز کوانتش و مدل‌سازی آن

اکنون حالت نویز ناشی از کوانتش را در نظر می‌گیریم. بر اساس مدل بخش ۲-۸-۳، فرض می‌کنیم نویز $a[n]$ یک نویز سفید ایستا با میانگین صفر و واریانس مشخص $\sigma_q^2$ است. بنابراین:

$$
E[a[n]] = 0 \quad \text{و} \quad R_{aa}[m] = \sigma_q^2 \delta[m]
$$

> طیف چگالی توان نویز $a[n]$ نیز یک مقدار ثابت در تمام فرکانس‌ها خواهد بود، زیرا نویز سفید فرض شده است.

#### کاهش نویز در اثر پیش‌نمونه‌برداری

با افزایش ضریب بیش‌نمونه‌برداری، بخش کمتری از طیف نویز با طیف سیگنال هم‌پوشانی پیدا می‌کند. فیلتر پایین‌گذر ایده‌آل این امکان را می‌دهد تا سیگنال بدون تغییر عبور کرده و نویز کوانتش حذف شود.

توان نویز در خروجی فیلتر پایین‌گذر:

$$
P_{n, \text{out}} = \frac{\sigma_q^2}{M}
$$

که $M$ ضریب بیش‌نمونه‌برداری است.

اما توان سیگنال:

$$
P_s = \text{ثابت}
$$

#### کاهش بیت مورد نیاز

معادله (۲-۱۷۰) نشان می‌دهد:

> با دو برابر شدن نرخ نمونه‌برداری، برای دستیابی به نسبت سیگنال به نویز (SNR) یکسان، نیم بیت کمتر نیاز است.

#### تبدیل A/D با شکل‌دهی نویز

در سیستم‌های بیش‌نمونه‌برداری‌شده، اگر از **مدولاتور دلتا-سیگما** استفاده شود، می‌توان نویز کوانتش را به باند فرکانسی بالاتر هدایت کرد. سپس با اعمال فیلتر پایین‌گذر، نویز حذف شده و تنها سیگنال اصلی باقی می‌ماند.

در این حالت، مدل سیستم به‌صورت زیر بیان می‌شود:

$$
y[n] = x[n] + \hat{e}[n]
$$

که $\hat{e}[n]$ نویز فیلترشده‌ی کوانتش است.

#### طیف چگالی توان نویز پس از شکل‌دهی:

$$
\Phi_{\hat{e}}(e^{j\omega}) = \sigma_q^2 |H(e^{j\omega})|^2
$$

که $H(z)$ تابع انتقال شکل‌دهی نویز است، مثل:

$$
H(z) = 1 - z^{-1}
$$

در نتیجه:

$$
|H(e^{j\omega})|^2 = 4 \sin^2\left(\frac{\omega}{2}\right)
$$

> این تابع نشان می‌دهد که بیشتر توان نویز در فرکانس‌های بالا متمرکز شده و با فیلتر پایین‌گذر حذف می‌شود.

#### معادله نهایی توان نویز در خروجی:

$$
P_{n,\text{final}} = \frac{1}{2\pi} \int_{-\pi}^{\pi} \Phi_{\hat{e}}(e^{j\omega}) \, d\omega
$$

و می‌توان نتیجه گرفت:

> با شکل‌دهی نویز و بیش‌نمونه‌برداری، نسبت سیگنال به نویز به‌شکل چشم‌گیری بهبود یافته و تعداد بیت مورد نیاز برای دقت مشخص کاهش می‌یابد.

### کاهش نویز و بیت در شکل‌دهی نویز

معادله زیر:

$$
=-\log M + \log(7.9) - 10P + 108
$$

مقایسه معادله (۲-۱۷۷) با معادله (۲-۱۷۱) نشان می‌دهد که در کوانتش مستقیم، دو برابر کردن به کاهش نیم بیت منجر می‌شود، در حالی که شکل‌دهی نویز، کاهش ۱٫۵ بیت را به دنبال دارد.

**جدول ۱۰۴** میزان کاهش بیت کوانتش‌کننده را نسبت به حالت کوانتش مستقیم بدون پیش‌نمونه‌برداری برای:
- کوانتش مستقیم با بیش‌نمونه‌برداری (بخش ۲-۱۰۹)،
- و شکل‌دهی نویز با بیش‌نمونه‌برداری  
نشان می‌دهد.

راهبرد شکل‌دهی موج شکل ۷۰۰۲ را می‌توان با افزودن یک انباره دیگر مطابق شکل ۷۵۴ تعمیم داد. در این حالت نیز با مدل‌سازی کوانتش‌کننده به صورت یک منبع نویز جمع‌شونده، می‌توان تحلیل را ادامه داد.

**جدول ۲-۱** کاهش تعداد بیت نسبت به معادله (۲-۱۷۱):

| روش | تعداد بیت |
|-----|-----------|
| کوانتش مستقیم | ۷۵ |
| شکل‌دهی نویز | ۳۷، ۵۱، ۲۵ |

**شکل ۲-۷۵:** کوانتش‌کننده پیش‌نمونه‌بردار با شکل‌دهی نویز مرتبه دوم.  
در این حالت دو طبقه‌ی پردازش نویز کوانتش با تابع تبدیل:

$$
H_x(z) = (1 - z^{-1})^7
$$

و خروجی:

$$
y[n] = x[n] + \tilde{e}[n]
$$

نویز موجود در خروجی سیستم شکل‌دهی نویز، اگرچه از نظر کل توان بیشتر است، اما عمدتاً در خارج از باند سیگنال قرار دارد.

در حالت کلی، اگر در طبقه‌ی nام استفاده شود، توان نویز به‌صورت زیر است:

$$
(M - 1) \cdot \text{Tain}(0.1)
$$

**جدول ۲۰۲**: کاهش بیت به‌عنوان تابعی از مرتبه‌ی شکل‌دهی موج و نسبت بیش‌نمونه‌برداری.  
مثلاً برای \( M = 2 \) و نسبت \( 64 \)، دقت تقریباً ۱۳ بیت افزایش می‌یابد. در نتیجه یک کوانتش‌کننده‌ی ۱ بیتی می‌تواند دقتی معادل ۱۲ بیت ایجاد کند.

با این حال، افزایش طبقات فیدبک باعث افزایش احتمال ناپایداری می‌شود. در مسئله ۱۸۰۹ ساختار جایگزینی به نام **شکل‌دهی نویز چند طبقه (MASH)** پیشنهاد شده است.

---

### ۳-۹-۴ بیش‌نمونه‌برداری و شکل‌دهی نویز در تبدیل D/A

در بخش‌های ۲-۹-۲ و ۲-۱۰۹ درباره‌ی کاربرد بیش‌نمونه‌برداری در ساده‌سازی فرآیند تبدیل A/D صحبت شد. همین اصول را می‌توان در تبدیل D/A نیز به‌کار برد. شکل ۴-۷۶ سیستم پایه‌ی مربوطه را نشان می‌دهد. در این سیستم:

1. سیگنال بیش‌نمونه‌برداری می‌شود،
2. سپس کوانتش مجدد انجام می‌گیرد،
3. در نهایت خروجی به مبدل D/A داده می‌شود.

اگر نویز کوانتش در خارج از باند سیگنال قرار گیرد، می‌توان از یک فیلتر آنالوگ ساده برای حذف آن استفاده کرد.

در تحلیل سیستم‌های شکل ۴-۷۶ و ۴-۷۷، فرض می‌شود کوانتش‌کننده به شکل یک نویز سفید خطی مدل شده باشد. در این صورت، طیف چگالی توان نویز کوانتش در خروجی به صورت زیر خواهد بود:

$$
\Phi_e(e^{j\omega}) = \sigma^2 \cdot \left| \sin(\omega \tau) \right|^2
$$

### شکل ۲-۷۹: طیف‌های مختلف سیستم شکل‌دهی نویز

**(الف)** طیف توان ورودی  
**(ب)** طیف توان سیگنال بیش‌نمونه‌برداری‌شده  
**(ج)** طیف نویز کوانتش در خروجی  
**(د)** طیف ترکیبی سیگنال و نویز در خروجی D/A

در این ساختار، فرض شده از فیلتر بازسازی ایده‌آل با فرکانس قطع \( MT \) استفاده شده که نویز را حذف می‌کند. اما در عمل می‌توان از فیلترهای ساده‌تری استفاده کرد زیرا شکل‌دهی نویز باعث سوق دادن نویز به فرکانس‌های بالاتر می‌شود.

### ۱۰-۴ خلاصه

در این فصل رابطه‌ی بین سیگنال‌های پیوسته در زمان و رشته‌های گسسته در زمان بررسی شد. قضیه نایکوئیست-شنون بیان می‌کند که اگر نرخ نمونه‌برداری بیشتر از دو برابر بیشینه فرکانس سیگنال باشد، می‌توان سیگنال را بدون اعوجاج بازسازی کرد.

همچنین با روش‌هایی نظیر:
- کم‌نمونه‌برداری و بیش‌نمونه‌برداری،
- تغییر نرخ نمونه‌برداری با ضریب غیرصحیح،
- فیلترهای پیشگیری از درهم‌روی،

امکان پردازش بهتر سیگنال‌ها فراهم می‌شود. در پایان نیز به موضوعاتی نظیر خطای کوانتش، فیلترهای بازسازی، و ساده‌سازی تبدیل‌های A/D و D/A با استفاده از شکل‌دهی نویز پرداخته شد.

**۲-۲۴**  
برای نمونه، نمودار بلوکی یک بانک فیلتر تحلیل و سنتز متداول در کاربردهای کدگذاری گفتار نشان داده شده است. هدف بخش تحلیل سیستم، تجزیه طیف فرکانسی ورودی به یک باند پایین‌گذر (نمایش یافته با سیگنال کم‌نمونه‌برداری‌شده \[a\]) و یک باند بالاگذر (نمایش یافته با سیگنال \[n\]) است.

در کاربردهای کدگذاری گفتار، سیگنال کانال‌ها برای انتقال یا ذخیره‌سازی، کوانتیده می‌شوند. چون باند اصلی به دو بخش مساوی با پهنای \( \pi \) رادیان تقسیم می‌شود، نرخ نمونه‌برداری خروجی فیلترها می‌تواند نصف نرخ ورودی باشد، به‌طوری‌که تعداد کل نمونه‌ها ثابت بماند.

توجه کنید که کم‌نمونه‌برداری از خروجی فیلتر پایین‌گذر باعث می‌شود باند فرکانس پایین، تمام گستره را پر کند. از طرف دیگر، کم‌نمونه‌برداری از خروجی فیلتر بالاگذر نیز باند فرکانس بالا را به سمت پایین سوق داده و آن را در گستره \( [0, \pi] \) جای می‌دهد.

این تجزیه مستلزم آن است که \[a[n]\] و \[d[n]\] به‌ترتیب پاسخ ضربه‌ی فیلترهای پایین‌گذر و بالاگذر باشند. یک رویکرد رایج، گرفتن فیلتر بالاگذر از فیلتر پایین‌گذر با رابطه \( d[n] = (-1)^n a[M - 1 - n] \) است.

این وضعیت باعث می‌شود که اگر \( A(e^{j\omega}) \) یک فیلتر پایین‌گذر با باند عبور نامی باشد، آنگاه \( D(e^{j\omega}) \) یک فیلتر بالاگذر مکمل خواهد بود.

هدف بخش سنتز در سمت راست شکل ۲-۲۴، بازسازی تقریبی سیگنال اصلی بر اساس سیگنال‌های \[a[n]\] و \[d[n]\] است. به این منظور، هر دو سیگنال به‌ترتیب از فیلترهای پایین‌گذر \( H_0[n] \) و بالاگذر \( H_1[n] \) عبور داده شده و سپس جمع می‌شوند تا سیگنال خروجی \( \hat{x}[n] \) با همان نرخ نمونه‌برداری ورودی به دست آید.

نتایج حوزه فرکانس برای کم‌نمونه‌برداری و بیش‌نمونه‌برداری در این سیستم به روابطی منجر می‌شوند که در معادله‌های زیر خلاصه شده‌اند:

> شرط بازسازی کامل (بدون اعوجاج) برای این سیستم در معادله زیر آمده است:

\[
H_0(e^{j\omega})G_0(e^{j\omega}) + H_1(e^{j\omega})G_1(e^{j\omega}) = c \cdot e^{-j\omega n_0}
\]

که در آن \( G_0, G_1 \) فیلترهای سنتز و \( H_0, H_1 \) فیلترهای تحلیل هستند. اگر این شرط برقرار باشد، بازسازی کامل سیگنال با تأخیر \( n_0 \) امکان‌پذیر خواهد بود.

این فیلترها اگر به‌گونه‌ای طراحی شوند که رابطه‌ی متقارن (آینه‌ای) بینشان برقرار باشد (به‌ویژه نسبت به \( \pi/2 \))، به آن‌ها **فیلترهای آینه‌ای کوادریچر (QMF)** گفته می‌شود.

برای نمونه، اگر فیلتر پایین‌گذر ساده میانگین متحرک در نظر گرفته شود:

\[
h[n] = \frac{1}{2}(\delta[n] + \delta[n - 1])
\]

پاسخ فرکانسی این فیلتر:

\[
H(e^{j\omega}) = \frac{1}{2}(1 + e^{-j\omega})
\]

می‌تواند به‌عنوان مثال ساده‌ای از طراحی فیلتر در این سیستم باشد.

# چکیده‌ای از بانک فیلتر و کاربرد آن در کدگذاری گفتار

## ۱. معرفی کلی سیستم تحلیل و سنتز
- سیستم‌های تحلیل و سنتز سیگنال در کدگذاری گفتار برای **تجزیه طیف فرکانسی** و **بازسازی مجدد سیگنال** استفاده می‌شوند.
- این سیستم‌ها معمولاً شامل دو بخش هستند:
  - **تحلیل (Analysis)**: تقسیم سیگنال به باندهای فرکانسی (پایین‌گذر و بالاگذر)
  - **سنتز (Synthesis)**: ترکیب دوباره باندهای فیلترشده برای بازسازی سیگنال

## ۲. فرآیند تحلیل
- سیگنال ورودی به کمک دو فیلتر:
  - **پایین‌گذر (Low-pass)** با خروجی \[a[n]\]
  - **بالاگذر (High-pass)** با خروجی \[d[n]\]
- سپس هر خروجی با نرخ پایین‌تر (مثلاً نصف نرخ اصلی) نمونه‌برداری می‌شود.
- هدف این است که اطلاعات در فرکانس‌های پایین و بالا جدا شده و با حجم داده کمتر ذخیره یا منتقل شوند.

## ۳. فرآیند سنتز
- سیگنال‌های \[a[n]\] و \[d[n]\] مجدداً **پیش‌نمونه‌برداری** (Upsample) می‌شوند.
- سپس از فیلترهای سنتز (پایین‌گذر و بالاگذر) عبور می‌کنند:
  - \( G_0[n] \) و \( G_1[n] \)
- خروجی‌ها با هم جمع شده و سیگنال بازسازی‌شده \( \hat{x}[n] \) تولید می‌شود.

## ۴. شرط بازسازی کامل
- برای این‌که بازسازی سیگنال بدون اعوجاج باشد، باید شرط زیر برقرار باشد:

\[
H_0(e^{j\omega})G_0(e^{j\omega}) + H_1(e^{j\omega})G_1(e^{j\omega}) = c \cdot e^{-j\omega n_0}
\]

- در این شرط، \( c \) ثابت و \( n_0 \) تأخیر است.
- اگر برقرار باشد، **بازسازی کامل با تأخیر ثابت** امکان‌پذیر است.

## ۵. فیلترهای آینه‌ای (QMF)
- در طراحی بانک فیلتر، فیلتر بالاگذر می‌تواند از پایین‌گذر مشتق شود:

\[
d[n] = (-1)^n a[M - 1 - n]
\]

- به این فیلترها، فیلترهای آینه‌ای (Quadrature Mirror Filters - QMF) می‌گویند.
- شرط تقارن طیفی در آن‌ها موجب حذف اعوجاج در بازسازی می‌شود.

## ۶. طراحی فیلتر نمونه
- یک فیلتر پایین‌گذر ساده میانگین متحرک:

\[
h[n] = \frac{1}{2}(\delta[n] + \delta[n - 1])
\]

- پاسخ فرکانسی آن:

\[
H(e^{j\omega}) = \frac{1}{2}(1 + e^{-j\omega})
\]

- این فیلتر ساده می‌تواند برای تست مفاهیم تحلیل و سنتز مورد استفاده قرار گیرد.

## ۷. نتیجه‌گیری
- طراحی دقیق فیلترها و رعایت شرایط بازسازی، امکان ساخت بانک‌های فیلتری را فراهم می‌کند که با کاهش نرخ نمونه‌برداری، باز هم سیگنال اصلی را بدون اعوجاج بازسازی کنند.
- چنین سیستم‌هایی در کدگذاری گفتار، فشرده‌سازی صوت، و پردازش چندنرخی بسیار پرکاربرد هستند.
