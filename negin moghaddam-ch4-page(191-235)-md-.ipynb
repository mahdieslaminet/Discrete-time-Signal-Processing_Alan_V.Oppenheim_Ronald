x_s(t) = x(t)s(t) = x(t) \sum_{n=-\infty}^{\infty} \delta(t - nT) = \sum_{n=-\infty}^{\infty} x(nT) \delta(t - nT)

x_s(t) = \sum_{n=-\infty}^{\infty} x[n] \delta(t - nT)

۴-۳ بازسازی سیگنال محدود-باند از روی نمونه‌ها
طبق قضیه‌ی نمونه‌برداری، نمونه‌های یک سیگنال باند-محدود پیوسته در زمان، در صورتی که به اندازه‌ لازم گرفته شده باشند، به تنهایی برای نمایش کامل آن سیگنال کافی هستند. از این جهت، که سیگنال را بتوان از روی نمونه‌ها با استفاده از فیلتر مناسب بازسازی کرد. این فیلتر بازسازی که در ادامه معرفی می‌شود، امکان بازسازی سیگنال پیوسته از روی نمونه‌ها را فراهم می‌کند.
بازسازی سیگنال باند-محدود پیوسته در زمان از روی دنباله‌ای از نمونه‌های گسسته با استفاده از فیلتر ایده‌آل امکان‌پذیر است. اگر یک سیگنال پیوسته باند-محدود با باند فرکانسی باشد، و اگر نمونه‌برداری مطابق شکل زیر انجام گیرد:
x_s(t) = \sum_{n=-\infty}^{\infty} x[n] \delta(t - nT) \quad \text{(معادله ۴-۲۲)} 
که در آن ، و تناوب نمونه‌برداری است.
اگر این قطار ضربه به یک فیلتر پایین‌گذر ایده‌آل در زمان گذر داده شود با پاسخ فرکانسی ، پاسخ سیستم بازسازی ایده‌آل به دست می‌آید:
x_r(t) = \sum_{n=-\infty}^{\infty} x[n] h_r(t - nT) \quad \text{(معادله ۴-۲۳)} 
که در آن پاسخ ضربه فیلتر پایین‌گذر ایده‌آل به صورت زیر است:
h_r(t) = \frac{\sin(\pi t / T)}{\pi t / T} \quad \text{(معادله ۴-۲۴)} 
نمودارهای شکل ۴-۷
شکل ۴-۷ سه نمودار مختلف را نشان می‌دهد:
(الف) سیستم ایده‌آل بازسازی سیگنال محدود-باند
(ب) پاسخ فرکانسی فیلتر بازسازی ایده‌آل:
H_r(j\Omega) = \begin{cases} T, & |\Omega| \leq \frac{\pi}{T} \\ 0, & \text{otherwise} \end{cases} 


۴-۲ پردازش گسسته در زمان سیگنال‌های پیوسته در زمان
مدل کلی سیستم
برای پردازش دیجیتال سیگنال‌های پیوسته، ابتدا آن را به سیگنال گسسته تبدیل می‌کنیم، پردازش دیجیتال انجام می‌دهیم، و در پایان آن را دوباره به سیگنال پیوسته تبدیل می‌کنیم:
x(t) → C/D → x[n] → H[n] → y[n] → D/C → y(t) 
شکل ۴-۳: ساختار کلی یک سیستم پردازش دیجیتال برای سیگنال‌های پیوسته در زمان
بازسازی سیگنال
برای بازسازی سیگنال پیوسته از نمونه‌های آن، از رابطه زیر استفاده می‌شود:
x(t) = \sum_{n=-\infty}^{\infty} x[n] \cdot \frac{\sin\left(\pi \left(\frac{t}{T} - n\right)\right)}{\pi \left(\frac{t}{T} - n\right)} 
که به آن تابع سینک گفته می‌شود و نشان می‌دهد که اگر سیگنال باند محدود باشد، بازسازی دقیق آن از نمونه‌ها ممکن است.
در اینجا داریم:
ساختار سیستم کلی
پردازش در این نوع سیستم‌ها به صورت زیر انجام می‌شود:
x(t) → C/D → x[n] → H[n] → y[n] → D/C → y(t) 
که در آن سیستم گسسته‌زمان است. خروجی کلی در حوزه فرکانس به صورت زیر است:
X_r(\Omega) = \sum_{n=-\infty}^{\infty} H[n] X\left( j \frac{\Omega}{T} \right) e^{-j n \Omega}

(الف) سیگنال نمونه‌برداری‌شده 
(ب) بازسازی توسط فیلتر سینک
(ج) سیگنال بازسازی‌شده نهایی

معادله (۴-۳۲):
y_r(t) = \sum_{n=-\infty}^{\infty} y[n] \frac{\sin\left(\pi \frac{t - nT}{T}\right)}{\pi \frac{t - nT}{T}} 
تبدیل فوریه سیگنال پیوسته بازسازی‌شده:
Y_r(j\Omega) = \begin{cases} H_r(j\Omega) Y(e^{j\Omega T}) & |\Omega| < \pi/T \\ 0 & \text{غیر از این} \end{cases} 
۴-۴: پردازش گسسته در زمان LTI سیگنال‌های پیوسته در زمان
اگر سیستم گسسته در زمان شکل ۴-۱۰ خطی و مستقل از زمان باشد، داریم:
معادله (۴-۳۳):
Y(e^{j\omega}) = H(e^{j\omega}) X(e^{j\omega}) 
که $H(e^{j\omega})$ پاسخ فرکانسی سیستم است. تبدیل به حوزه پیوسته:
معادله (۴-۳۴):
Y_r(j\Omega) = H_r(j\Omega) H(e^{j\Omega T}) X(e^{j\Omega T}) 
با استفاده از معادله $ \omega = \Omega T $:
معادله (۴-۳۵):
Y_r(j\Omega) = H_r(j\Omega) \frac{1}{T} \sum_{k=-\infty}^{\infty} X_c\left(j\left(\Omega - \frac{2\pi k}{T}\right)\right) H\left(e^{j\left(\Omega - \frac{2\pi k}{T}\right) T}\right) 
اگر $X_c(j\Omega)$ باند-محدود باشد، آنگاه:
معادله (۴-۳۶):
Y_r(j\Omega) = \begin{cases} H(e^{j\Omega T}) X_c(j\Omega) & |\Omega| < \pi/T \\ 0 & |\Omega| \geq \pi/T \end{cases} 
که می‌توان به صورت یک پاسخ فرکانسی مؤثر نیز نوشت:
معادله (۴-۳۷):
Y_r(j\Omega) = H_{\text{eff}}(j\Omega) X_c(j\Omega) 
که در آن:
معادله (۴-۳۸):
H_{\text{eff}}(j\Omega) = \begin{cases} H(e^{j\Omega T}) & |\Omega| < \pi/T \\ 0 & |\Omega| \geq \pi/T \end{cases} 
مثال ۴-۳: فیلتر پایین‌گذر در زمان پیوسته با استفاده از فیلتر پایین‌گذر گسسته در زمان
فرض کنید:
H(e^{j\omega}) = \begin{cases} 1 & |\omega| < \omega_c \\ 0 & \text{غیر از این} \end{cases} 
اگر $\omega_c = \pi/2$ و $T = 1$, آنگاه پاسخ فرکانسی مؤثر:
H_{\text{eff}}(j\Omega) = \begin{cases} 1 & |\Omega| < \Omega_c = \omega_c/T = \pi/2 \\ 0 & \text{غیر از این} \end{cases} 
شکل ۴-۱۲:
الف) پاسخ فرکانسی $H(e^{j\omega})$
ب) پاسخ فرکانسی مؤثر $H_{\text{eff}}(j\Omega)$

۴-۲ پردازش گسسته در زمان

اگر پاسخ ضربه سیستم گسسته در زمان نگاشتی از پاسخ پیوسته باشد، داریم:

h[n] = h_c(nT) \tag{۴-۵۱}

و تابع تبدیل:

H(e^{j\omega}) = \frac{1}{T} \sum_{k=-\infty}^{\infty} H_c\left(j\left(\frac{\omega}{T} - \frac{2\pi k}{T}\right)\right) \tag{۴-۵۲}

در حالتی که شرط باند محدود برقرار باشد:

H(e^{j\omega}) = \frac{1}{T} H_c\left(j\frac{\omega}{T}\right), \quad |\omega| < \pi \tag{۴-۵۳}

که مجدداً برای  برقرار است.

مثال ۴: فیلتر پایین‌گذر گسسته در زمان به روش هم‌ضربگی

h_c(t) = \frac{\sin(\Omega_c t)}{\pi t}

پاسخ گسسته متناظر:

h[n] = T h_c(nT) = \frac{\sin(\Omega_c nT)}{\pi nT} = \frac{\sin(\omega_c n)}{\pi n} \tag{۴-۵۴}

مثال ۴-۶: اعمال هم‌ضربگی بر سیستم‌های پیوسته در زمان با تابع نمایی

فرض کنیم:

h_c(t) = A e^{s_1 t} u(t)

که تبدیل لاپلاس آن:

H_c(s) = \frac{A}{s - s_1}, \quad \text{Re}(s) > \text{Re}(s_1)

و با اعمال هم‌ضربگی:

h[n] = T h_c(nT) = A T e^{s_1 nT} u[n]

تابع تبدیل Z:

H(z) = \frac{A T}{1 - e^{s_1 T} z^{-1}}, \quad |z| > |e^{s_1 T}|

و پاسخ فرکانسی:

H(e^{j\omega}) = \frac{A T}{1 - e^{s_1 T} e^{-j\omega}} \tag{۴-۵۶}

۴-۵ پردازش پیوسته در زمان سیستم‌های گسسته در زمان

در این قسمت به بررسی سیستم‌هایی می‌پردازیم که سیگنال ورودی پیوسته در زمان است، اما پردازش آن به‌صورت گسسته در زمان انجام می‌شود، و سپس خروجی دوباره به سیگنال پیوسته در زمان تبدیل می‌شود.

ساختار کلی سیستم

x(t) --> C/D --> x[n] --> H(e^{jω}) --> y[n] --> D/C --> y(t)

C/D: مبدل پیوسته به گسسته (نمونه‌بردار)

H(e^{jω}): پاسخ فرکانسی سیستم گسسته در زمان

D/C: مبدل گسسته به پیوسته (بازساز)


فرمول کلی پاسخ سیستم:

y(t) = T \sum_{n=-\infty}^{\infty} y[n] \, \text{sinc}\left(\frac{t - nT}{T}\right)

با ترکیب تبدیل‌ها:

X(e^{jω}) = X_c(jΩ) \Big|_{\Omega = \frac{ω}{T}}, \quad |\omega| < π

Y_c(jΩ) = H(jΩ)X_c(jΩ), \quad H(jΩ) = H\left(e^{jΩT}\right)

y[n] = x[n - Δ], \quad Δ \in \mathbb{R}

خروجی پیوسته‌ در زمان:

y_c(t) = x_c(t - ΔT)

## ۴-۶ تغییر آهنگ نمونه‌برداری با پردازش گسسته در زمان

در بسیاری از کاربردها نیاز است که یک سیگنال گسسته در زمان با آهنگ جدیدی نمونه‌برداری شود. فرض کنید سیگنال اصلی به صورت زیر باشد:

$$
x[n] = x_c(nT)
$$

با تغییر آهنگ نمونه‌برداری به $T_a$، خواهیم داشت:

$$
x_d[n] = x(nM) = x_c(nMT)
$$

که در آن $T_a = MT$ است. این کار در شکل ۴-۱۹ نشان داده شده است:



### تحلیل طیفی کاهش آهنگ نمونه‌برداری

اگر $x[n] = x_c(nT)$، تبدیل فوریه آن برابر است با:

$$
X(e^{j\omega}) = \frac{T}{T_d} \sum_{k=-\infty}^{\infty} X_c\left( \frac{\omega}{T} - \frac{2\pi k}{T} \right)
$$

پس از کاهش نرخ نمونه‌برداری به $T_d = MT$:

$$
X_d(e^{j\omega}) = \frac{1}{MT} \sum_{k=-\infty}^{\infty} X_c\left( \frac{\omega}{MT} - \frac{2\pi k}{MT} \right)
$$

## خلاصه:

- فیلترهای میانگین‌گیر متحرک سیگنال را در یک بازه محدود هموار می‌کنند و برای کاهش نوفه یا تراز کردن سیگنال مناسب‌اند.
- کاهش نرخ نمونه‌برداری با فاکتور صحیح $M$ باعث فشرده‌سازی طیف و افزایش احتمال همپوشانی طیفی (aliasing) می‌شود.
- فرمول‌های تبدیل فوریه نشان‌دهنده تغییر ساختار طیف پس از کاهش نرخ نمونه‌برداری هستند.


۴-۶ تغییر آهنگ نمونه‌برداری

اگر

که در آن  و  اعداد صحیح هستند و ، واضح است که r نیز عدد صحیحی بین  و  است. بنابراین معادله (4-71) به صورت زیر بیان کرد:

X_d\left(e^{j\omega}\right) = \frac{1}{M} \sum_{r=0}^{M-1} \sum_{k=-\infty}^{\infty} x_c\left(\frac{kMT + rT}{T}\right) e^{-j\omega(kM + r)}

با تغییر متغیر  و با توجه به معادله (4-70)، معادله بالا به صورت زیر بیان می‌گردد:

۴-۶-۲ افزایش آهنگ نمونه‌برداری با ضریب صحیح

برای افزایش آهنگ نمونه‌برداری یک سیگنال گسسته در زمان با ضریب صحیح ، از روش درج صفرها و فیلتر پایین‌گذر استفاده می‌کنیم.

رابطه اصلی:

با فرض اینکه:

x_1[n] = x_c(nT_1), \quad T_1 = \frac{T}{L}

با استفاده از درج صفرها و فیلتر پایین‌گذر، داریم:

x[n] \xrightarrow{\text{درج صفر}} x_I[n] = 
\begin{cases}
x\left[\frac{n}{L}\right], & n = 0, \pm L, \pm 2L, \ldots \\
0, & \text{otherwise}
\end{cases}

x_1[n] = x_I[n] * h[n]

که در آن  پاسخ ضربه‌ی فیلتر پایین‌گذر است:

h[n] = \frac{\sin\left( \frac{\pi n}{L} \right)}{\pi n / L}


---

دیاگرام بلوکی سیستم افزایش آهنگ نمونه‌برداری:

x[n] → ↑L (درج صفر) → فیلتر پایین‌گذر با π/L → x1[n]


---

نمایش در حوزه فرکانس (DTFT):

X_{1}(e^{j\omega}) = 
\begin{cases}
X(e^{jL\omega}), & |\omega| \leq \frac{\pi}{L} \\
0, & \text{otherwise}
\end{cases}

۴-۶ فیلترهای بازسازی ساده و عملی

در بازسازی سیگنال از روی نمونه‌های آن، ایده‌آل‌ترین فیلتر بازسازی، فیلتر سینک (sinc) است. پاسخ ضربه این فیلتر به صورت زیر تعریف می‌شود:

h[n] = \frac{\sin(\pi n / L)}{\pi n / L}

که در آن  نرخ افزایش نمونه‌گیری است.

در عمل، استفاده از فیلتر ایده‌آل غیرممکن است چون طول پاسخ ضربه آن بی‌نهایت است. بنابراین، فیلترهای بازسازی با پاسخ ضربه محدود (FIR) یا فیلترهای ساده‌تری مانند فیلتر نگهدارنده مرتبه صفر یا مرتبه اول به کار می‌روند.

مثال ۴-۳:

در این مثال یک فیلتر مثلثی برای بازسازی استفاده شده است. پاسخ ضربه این فیلتر محدود و به صورت مثلثی است. اگر  پاسخ ضربه خطی باشد، آنگاه:

|h_{lin}[n]| \leq 1

و به صورت زیر تعریف می‌شود (برای مثال با ):

h_{lin}[n] =
\begin{cases}
1 - \frac{|n|}{L} & |n| \leq L \\
0 & \text{otherwise}
\end{cases}

۴-۷ فیلترهای بازسازی باند محدود

برای بازسازی دقیق سیگنال‌های پیوسته در زمان از روی نمونه‌های گسسته، می‌توان از فیلترهای بازسازی باند محدود استفاده کرد.

در این فیلترها، پاسخ فرکانسی  باند محدودی دارد. پاسخ ضربه متناظر در حوزه زمان به صورت زیر محاسبه می‌شود:

x_r[m] = \sum_{k=-\infty}^{\infty} x[k] h_{lin}[m - kL]

که در آن  از رابطه زیر به دست می‌آید:

h_{lin}[n] = \frac{\sin(\pi n / L)}{\pi n / L} \cdot w[n]

که  یک پنجره مناسب برای محدود کردن پاسخ ضربه است.

در شکل ۴-۳۷ (الف) پاسخ فرکانسی  نمایش داده شده است.

در شکل ۴-۳۷ (ب) نمونه‌برداری سیگنال با استفاده از فیلتر خطی نمایش داده شده است.

نمونه‌برداری مجدد به کمک فیلترهای بازسازی با استفاده از سری کسری

در سیستم‌های چندنرخی، بازسازی سیگنال می‌تواند به کمک فیلترهای بازسازی پیاده‌سازی شود. برای مثال، شکل 4-30 نمونه‌ای از سیگنال بازسازی‌شده را با استفاده از فیلتر ایده‌آل (خط ممتد) و فیلتر تربی (خط‌چین) نشان می‌دهد. مشاهده می‌شود که فیلتر تربی می‌تواند جایگزینی برای فیلتر ایده‌آل باشد.

شکل 4-31: پیاده‌سازی سیستم تغییر نرخ نمونه‌برداری کسری

در این شکل یک سیستم چندنرخی شامل افزایش نرخ نمونه‌برداری با فاکتور L و سپس کاهش با فاکتور M را مشاهده می‌کنیم. برای این منظور، ابتدا افزایش نرخ با درج صفر انجام می‌شود، سپس فیلتر پایین‌گذر اعمال شده و در نهایت زیرنمونه‌گیری صورت می‌گیرد.

معادله کلی تغییر نرخ:

x_{out}(n) = \sum_{k} h(k) x\left( \frac{Mn - k}{L} \right)

این شکل طیف سیگنال‌ها را در مراحل مختلف تغییر نرخ نشان می‌دهد:

(ب) بعد از درج صفرها (L = 2)

(ج) اعمال فیلتر پایین‌گذر

(د) بعد از کاهش نرخ با فاکتور M = 3

- x_r(t): سیگنال بازسازی‌شده
- x_p(t): سیگنال نمونه‌برداری شده
- نقاط: نمونه‌های گرفته شده

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(10, 4))

n = np.arange(0, 11, 1)

۴-۷ بازسازی چندنرخی

در اینجا به بررسی ساختارهایی می‌پردازیم که در آن‌ها سیگنال با نرخ پایین‌تر به نرخ بالاتر بازسازی می‌شود. این ساختارها معمولاً شامل فیلترهای بازسازی، فیلترهای پایین‌گذر ایده‌آل و افزایش نرخ نمونه‌برداری هستند.

فرمول بازسازی پایه‌ای برای تبدیل نرخ به شکل زیر است:

X_e(e^{j\omega}) = H(e^{j\omega}) X(e^{j\omega L})

و خروجی نهایی:

Y(e^{j\omega}) = \frac{1}{M} \sum_{k=0}^{M-1} X_e\left(e^{j(\omega - 2\pi k/M)}\right)

۴-۷-۲ ترکیب دو مرحله‌ای چندنرخی

در این بخش، ترکیب فیلترها در سیستم‌های چندنرخی را بررسی می‌کنیم.

فرمول نهایی برای این ترکیب:

Y(e^{j\omega}) = X(e^{j\omega}) H(e^{j\omega})

که در آن:

H(e^{j\omega}) = \frac{1}{M} \sum_{l=0}^{M-1} H_l(e^{j\omega/M}) e^{-j\omega l/M}
