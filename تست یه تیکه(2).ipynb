    <H2> مثال 9.23 </H2>
  <div dir="rtl">
    یک سیستم LTI با ورودی $x(t)$ و خروجی $y(t)$ را در نظر بگیرید که رابطه ورودی-خروجی آن از معادله
دیفرانسیل زیر تبعیت میکند
  <div dir="rtl">
$$\frac{dy(t)}{dt} + 3y(t) = x(t) \quad (9.126)$$
<div dir="rtl">
با گرفتن تبدیل لاپلاس از دو طرف معادله (9.126) و به کار بردن خواص خطی بودن و مشتق بیان شده در
بخشهای9.5.1 و 9.5.7 [معادله های (9.82) و (9.98)] معادله جبری زیر را به دست می آوریم.
<div dir="rtl">
$$sY(s) + 3Y(s) = X(s) \quad (9.127)$$
<div dir="rtl">
چون با توجه به معادله (9.112) تابع تبدیل به صورت زیرست

$$H(s) = \frac{Y(s)}{X(s)}$$
<div dir="rtl">
برای این مثال به دست می آوریم

$$H(s) = \frac{1}{s+3} \quad (9.128)$$
<div dir="rtl">
با این کار عبارت جبری تابع تبدیل به دست می آید، اما ناحیه همگرایی آن مشخص نمی شود. در واقع
معادله دیفرانسیل به تنهایی توصیف کاملی از سیستم LTI نیست و در حالت کلی پاسخ ضربه های متفاوتی
وجود دارند که تماماً با یک معادله دیفرانسیل سازگارند. اگر علاوه بر معادله دیفرانسیل بدانیم که سیستم علی
است، میتوانیم ROC را در سمت راست راست ترین قطب بگیریم، که در این مثال $\Re\{s\} > -3$ است. اگر
بدانیم سیستم ضدعلی است، ROC مربوط به $H(s)$ باید به صورت $\Re\{s\} < -3$ باشد. در حالت علی
پاسخ ضربه متناظر عبارت است از

$$h(t) = e^{-3t}u(t) \quad (9.129)$$
<div dir="rtl">
و در حالت ضد علی به صورت زیرست

$$h(t) = -e^{-3t}u(-t) \quad (9.130)$$
<div dir="rtl">
روش به کار رفته در یافتن $H(s)$ مثال بالا را میتوان در حالت کلی تر به کار برد. معادله دیفرانسیل خطی
دارای ضرایب ثابت زیر را در نظر بگیرید

$$\sum_{k=0}^{N} a_k \frac{d^k y(t)}{dt^k} = \sum_{k=0}^{M} b_k \frac{d^k x(t)}{dt^k} \quad (9.131)$$
<div dir="rtl">
با گرفتن تبدیل لاپلاس از دو طرف معادله و استفاده از خواص خطی بودن و مشتقگیری به دست می آوریم:

$$\left( \sum_{k=0}^{N} a_k s^k \right) Y(s) = \left( \sum_{k=0}^{M} b_k s^k \right) X(s) \quad (9.132)$$
<div dir="rtl">
یا

$$H(s) = \frac{\sum_{k=0}^{M} b_k s^k}{\sum_{k=0}^{N} a_k s^k} \quad (9.133)$$
<div dir="rtl">
بنابراین تابع تبدیل سیستم توصیف شده با معادله دیفرانسیل همیشه گویاست و صفرهای آن ریشه های
معادله زیرند

$$\sum_{k=0}^{M} b_k s^k = 0 \quad (9.134)$$
<div dir="rtl">
و قطبهای آن ریشه های معادله زیرست

$$\sum_{k=0}^{N} a_k s^k = 0 \quad (9.135)$$
<div dir="rtl">
همان طور که پیشتر گفتیم معادله (9.133) ناحیه همگرایی $H(s)$ را مشخص نمیکند، زیرا معادله
دیفرانسیل خطی دارای ضرایب ثابت قیدی بر ROC نمی گذارد. ولی با داشتن اطلاعاتی اضافی، مانند
پایداری یا علی بودن سیستم می توان ناحیه همگرایی را مشخص کرد.مثلا اگر شرط سکون ابتدایی را بگذاریم، سیستم علی می شود و می دانیم که ROC  باید سمت راست راست ترین قطب باشد.

    <H2> مثال 9.24 </H2>
 <div dir="rtl">
   یک مدار RLC که مقدار اولیه ولتاژ خازن و جریان القاگر آن صفرند، یک سیستم LTI توصیف شده با معادلات
دیفرانسیل خطی با ضرایب ثابت تشکیل میدهد. مدار RLC شکل 9.27 را در نظر بگیرید. ولتاژ منبع را
ورودی $x(t)$ و ولتاژ روی خازن را خروجی $y(t)$ فرض کنید. با برابر قرار دادن ولتاژ منبع و مجموع ولتاژهای
مقاومت، القاگر و خازن به دست می آوریم

$$RC\frac{dy(t)}{dt} + LC\frac{d^2y(t)}{dt^2} + y(t) = x(t) \quad (9.136)$$
<div dir="rtl">
به کمک معادله (9.133) به دست می آوریم

$$H(s) = \frac{1/LC}{s^2 + (R/L)s + (1/LC)} \quad (9.137)$$
<div dir="rtl">
در مسئله 9.64 نشان داده میشود که اگر $L$، $C$ و $R$ مثبت باشند، قطبهای این تابع تبدیل دارای بخش حقیقی
منفی هستند و سیستم پایدارست.

 <div dir="rtl">

در واقع تمام خواص تبدیل لاپلاس را میتوان به همین نحو به کار برد و رابطه ای بین رفتار سیستم و تابع

    <H2> مثال 9.25 </H2>
    <div dir="rtl">
   می‌دانیم که پاسخ یک سیستم LTI به ورودی

$$x(t) = e^{-3t}u(t)$$
<div dir="rtl">
به صورت زیرست

$$y(t) = \left[ e^{-t} - e^{-2t} \right] u(t)$$
<div dir="rtl">
اکنون نشان می‌دهیم که با توجه به این اطلاعات می‌توانیم تابع سیستم این سیستم را بیابیم و از آن بعضی خواص سیستم را نتیجه بگیریم. 
  با گرفتن تبدیل لاپلاس $x(t)$ و $y(t)$ را به دست می آوریم

$$X(s) = \frac{1}{s+3} \quad , \Re\{s\} > -3$$
<div dir="rtl"> 
و

$$Y(s) = \frac{1}{(s+1)(s+2)} \quad , \Re\{s\} > -1$$
<div dir="rtl"> 
با توجه به معادله (9.112) نتیجه می گیریم که

$$H(s) = \frac{Y(s)}{X(s)} = \frac{s+3}{(s+1)(s+2)} = \frac{s+3}{s^2 + 3s + 2}$$
<div dir="rtl"> 
می توانیم ROC این سیستم را نیز تعیین کنیم. از خاصیت کانولوشن بخش 9.5.6 می دانیم که ROC
تبدیل خروجی $Y(s)$ حداقل اشتراک نواحی همگرایی $X(s)$ و $H(s)$ را شامل می شود. بررسی سه گزینش
ممکن ROC برای $H(s)$ (یعنی سمت چپ قطب $s=-2$، بین قطبهای $s=-2$ و $s=-1$، و سمت راست
قطب $s=-1$) تنها ROC سازگار با نواحی همگرایی $X(s)$ و $Y(s)$، $\Re\{s\} > -1$ است. چون این ROC
سمت راست راست ترین قطب $H(s)$ است، نتیجه میگیریم که $H(s)$ علی است و چون هر دو قطب $H(s)$
بخشهای حقیقی منفی دارند سیستم پایدارست. به علاوه با توجه به رابطه معادلات (9.131) و (9.133)
می توانیم معادله دیفرانسیل زیر و شرط سکون ابتدایی را برای سیستم مشخص کنیم

$$\frac{d^2y(t)}{dt^2} + 3\frac{dy(t)}{dt} + 2y(t) = \frac{dx(t)}{dt} + 3x(t)$$

    <H2> مثال 9.26 </H2>
  <div dir="rtl">
 اطلاعات زیر در مورد یک سیستم LTI داده شده است:
<div dir="rtl">
۱. سیستم علی است.
<div dir="rtl">
۲. تابع سیستم گویاست و تنها دو قطب در $s=-2$ و $s=4$ دارد.
<div dir="rtl">
۳. به ازای $x(t) = 1 = e^{0t}$، $y(t) = 0$.
<div dir="rtl">
۴. پاسخ ضربه در $t=0^+$ برابر 4 است.
<div dir="rtl">
می خواهیم به کمک این اطلاعات تابع سیستم را بیابیم.
<div dir="rtl">
از دو شرط اول در می یابیم که سیستم ناپایدارست (چون سیستم علی است و یک قطب دارای بخش حقیقی
مثبت دارد) و تابع سیستم به صورت زیراست

$$H(s) = \frac{p(s)}{(s+2)(s-4)}$$
<div dir="rtl">
$p(s)$ یک چند جمله ای از $s$ است. چون پاسخ $y(t)$ به ورودی $x(t) = ۱ = e^{0t}$ باید $H(0) = 0$
باشد، از اطلاع ۳ نتیجه می گیریم که $p(0) = 0$، یعنی $p(s)$ باید یک ریشه در $s=0$ داشته باشد، و بنابراین به
صورت زیرست

$$p(s) = sq(s)$$
<div dir="rtl">
$q(s)$ نیز یک چند جمله ای از $s$ است.     
      
<div dir="rtl">      
سرانجام با توجه به اطلاع ۴ و قضیه مقدار اولیه بخش 9.5.10 داریم

$$\lim_{s \to \infty} sH(s) = \lim_{s \to \infty} \frac{s^2q(s)}{s^2 - 2s - 8} = 4 \quad (9.138)$$
<div dir="rtl"> 
در $s$ های بالاترین مرتبه های صورت و مخرج جملات غالب $sH(s)$ را تشکیل میدهند و در محاسبه معادله
(9.138) به کار میروند. همچنین اگر مرتبه صورت از مرتبه مخرج بزرگتر باشد، این حد واگراست. پس تنها
به شرطی حد مقداری محدود میشود که مرتبه چند جمله ایهای صورت و مخرج $sH(s)$ برابر باشند. چون
مرتبه مخرج ۲ است، برای صحت معادله (۹-۱۳۸) $q(s)$ باید ثابت باشد، یعنی $q(s) = K$. مقدار این ثابت را
می توان به صورت زیر تعیین کرد

$$\lim_{s \to \infty} \frac{Ks^2}{s^2 - 2s -8} = \lim_{s \to \infty} \frac{Ks^2}{s^2} = K \quad (9.139)$$
<div dir="rtl"> 
معادلات (9.138) و (9.139) نشان میدهند که $K = 4$ و در نتیجه

$$H(s) = \frac{4s}{(s+2)(s-4)}$$

    <H2> فیلترهای باتروث </H2>
در مثال 6.3 در مورد یک دسته سیستم LTI پر کاربرد، موسوم به فیلترهای باترورث به اختصار صحبت
کردیم. این فیلترها خصوصیاتی دارند، از جمله مشخصات اندازه پاسخ فرکانسی این فیلترها در باند گذر،
که آنها را از لحاظ عملی بسیار مطلوب میسازد. به عنوان یکی از کاربردهای تبدیل لاپلاس، در این بخش
آن را برای تعیین تابع تبدیل فیلترهای باترورث از روی اندازه پاسخ فرکانسی شان، به کار میگیریم.
پاسخ فرکانسی یک فیلتر باترورث مرتبه $N$، اندازه ای به صورت زیر دارد

$$|B(j\omega)|^2 = \frac{1}{1 + (j\omega/j\omega_c)^{2N}} \quad (9.140)$$
<div dir="rtl">
$N$ مرتبه فیلتر است. می خواهیم از معادله (9.140) یک تابع $B(s)$ بیابیم که $|B(j\omega)|^2$ آن به صورت داده
شده باشد. طبق تعریف داریم

$$|B(j\omega)|^2 = B(j\omega)B^*(j\omega) \quad (9.141)$$
 <div dir="rtl">   
  اگر پاسخ ضربه فیلتر باترورث را حقیقی فرض کنیم، بنا به خاصیت تقارن مزدوج تبدیل فوریه داریم

$$B^*(j\omega) = B(-j\omega) \quad (9.142)$$
 <div dir="rtl">
بنابراین

$$B(j\omega)B(-j\omega) = \frac{1}{1 + (j\omega/j\omega_c)^{2N}} \quad (9.143)$$
 <div dir="rtl">
می دانیم که $B(s)|_{s=j\omega} = B(j\omega)$ و در نتیجه معادله (9.143) به دست می دهد

$$B(s)B(-s) = \frac{1}{1 + (s/j\omega_c)^{2N}} \quad (9.144)$$
 <div dir="rtl">
ریشه های چند جمله ای مخرج که قطبهای $B(s)B(-s)$ هستند، عبارت اند از

$$B(s)B(-s) = (-1)^{1/2N}j\omega_c \quad (9.145)$$
 <div dir="rtl">
معادله (9.145) به ازای هر مقدار $s=s_p$ دارای مشخصات زیر برآورده میشود


$$|s_p| = \omega_c \quad (9.146)$$  
  
  
  <div dir="rtl"> 
    و
$$ \angle s_p = \frac{\pi (2k + 1)}{2N} + \frac{\pi}{2}  (9.147)$$ 
<div dir="rtl">
یعنی
    $$ s_p = \omega_c \exp \left( j \left[ \frac{\pi (2k + 1)}{2N} + \frac{\pi}{2} \right] \right)  (9.148) $$
 <div dir="rtl">
شکل 9.28  محل قطبهای $B(s)B(-s)$ را به ازای $N=1,2,3,6$ نشان میدهد. به طور کلی نکات
زیر مورد قطبهای $B(s)B(-s)$ صادق است:
 <div dir="rtl">
  <div dir="rtl">   
1. $2N$ قطب با زوایای مساوی بر روی دایره ای به شعاع $\omega_c$ در صفحه $s$ قرار دارند.
  <div dir="rtl">  
2. هرگز روی محور $j\omega$ قطب نداریم، به ازای $N$ فرد روی محور $\sigma$ قطب داریم ولی به ازای $N$ زوج
نه.
<div dir="rtl">
3. فاصله زاویه ای بین قطبهای $B(s)B(-s)$ برابر $\pi/N$ رادیان است.
 <div dir="rtl">
برای اینکه قطبهای $B(s)$ را از قطبهای $B(s)B(-s)$ به دست آوریم، توجه داریم که قطبهای $B(s)B(-s)$
زوج زوج هستند. بنابراین اگر قطبی در $s=s_p$ باشد، قطبی هم در $s=-s_p$ وجود دارد. در نتیجه برای
ساختن $B(s)$  از هر زوج قطب یکی را بر میگزینیم. اگر بخواهیم سیستم پایدار و علی باشد، قطبهایی که 
 نسبت میدهیم به $B(s)$ باید قطبهای روی دایره در نیم صفحه چپ باشند.   
<div dir="rtl">
این قطب ها B(S) را بدون ضریب
مقیاس مشخص میکنند. اما با توجه به معادله (9.144) میبینیم که $|B^2(s)|_{s=0}=1$، یعنی ضریب مقیاس
چنان انتخاب میشود که مجذور اندازه پاسخ فرکانسی در $\omega=0$ برابر یک باشد.
برای نشان دادن روند تعیین $B(s)$، حالتهای $N=1,2,3$ را در نظر میگیریم. در شکل 9.28 قطبهای
$B(s)B(-s)$ به دست آمده از معادله (9.148) نشان داده شده اند. شکل 9.29 قطبهای مربوط به $B(s)$ را
برای هر یک از این سه مقدار $N$ نشان میدهد. تابع تبدیل متناظر با هر کدام چنین است.
$$ N = 1: \quad B(s) = \frac{\omega_c}{s + \omega_c}  (9.149)$$



$$ N = ۲: \quad B(s) = \frac{\omega_c^2}{(s + \omega_c e^{j(\pi/4)})(s + \omega_c e^{-j(\pi/4)})}(9.150) $$

$$ = \frac{\omega_c^2}{s^2 + \sqrt{2}\omega_c s + \omega_c^2} $$


$$ N = 3: \quad B(s) = \frac{\omega_c^3}{(s + \omega_c)(s + \omega_c e^{j(\pi/3)})(s + \omega_c e^{-j(\pi/3)})} (9.151)$$

$$ = \frac{\omega_c^3}{(s + \omega_c)(s^2 + \omega_c s + \omega_c^2)} $$

$$ = \frac{\omega_c^3}{s^3 + 2\omega_c s^2 + 2\omega_c^2 s + \omega_c^3} $$
<div dir="rtl">
بر مبنای بحث بخش 9.7.3 میتوان معادله دیفرانسیل خطی دارای ضرایب ثابت مربوط به هر حالت را به
دست آورد. برای سه حالت ذکر شده فوق معادلات دیفرانسیل عبارت اند از:

$$ N = 1: \quad \frac{dy(t)}{dt} + \omega_c y(t) = \omega_c x(t) \quad (9.152) $$

$$ N = 2: \quad \frac{d^2y(t)}{dt^2} + \sqrt{2} \omega_c \frac{dy(t)}{dt} + \omega_c^2 y(t) = \omega_c^2 x(t) \quad (9.153) $$

$$ N = 3: \quad \frac{d^3y(t)}{dt^3} + 2 \omega_c \frac{d^2y(t)}{dt^2} + 2 \omega_c^2 \frac{dy(t)}{dt} + \omega_c^3 y(t) = \omega_c^3 x(t) \quad (9.154) $$

    <H2>8.9 جبر تابع سیستم و نمایش جعبه ای </H2>
 <div dir="rtl">
     کاربرد تبدیل لاپلاس این امکان را به ما میدهد که عملیات حوزه زمان، مانند مشتقگیری، کانولوشن،
جابجایی زمانی، و غیره را با عملیات جبری جایگزین کنیم. بسیاری از مزایای این کار در تحلیل سیستمهای
LTI را دیده ایم و در این بخش میخواهیم یک کاربرد مهم دیگر جبر تابع سیستم را بیان کنیم و آن تحلیل
اتصالهای مختلف سیستمهای LTI، و طراحی سیستم به کمک اتصال سیستمهای ابتدایی به یکدیگرست.
<div dir="rtl">
    <H2> 9.8.1 توابع تبدیل اتصال سیستم های LTI</H2>
        <div dir="rtl">
اتصال موازی دو سیستم ، به صورت نشان داده شده در شکل 9.30 (الف) را در نظر بگیرید. پاسخ ضربه
سیستم مرکب عبارت است از

$$h(t) = h_1(t) + h_2(t) \quad (9.155)$$
<div dir="rtl">
و با توجه به خطی بودن تبدیل لاپلاس داریم

$$ H(s) = H_1(s) + H_2(s) \quad (9.156) $$
<div dir="rtl">
به طور مشابه پاسخ ضربه یک اتصال سری، مطابق شکل 9.30 (ب) عبارت است از

$$ h(t) = h_1(t) * h_2(t) \quad (9.157) $$
<div dir="rtl">
و تابع تبدیل آن چنین است

$$ H(s) = H_1(s) H_2(s) \quad (9.158)$$
<div dir="rtl">
کاربرد تبدیل لاپلاس برای نمایش جبری ترکیب سیستمهای خطی را می توان به اتصال هایی بسیار پیچیده تر
   از این اتصالهای ساده سری و موازی تعمیم داد. برای نشان دادن چگونگی این امر اتصال فیدبک داری به
صورت شکل 9.31 را در نظر بگیرید. طراحی، کاربرد و تحلیل چنین اتصالهایی به تفصیل در فصل 11 بیان
شده است. گرچه تحلیل سیستم در حوزه زمان کار چندان ساده ای نیست، ولی یافتن تابع تبدیل کل سیستم،
از ورودی $x(t)$ به خروجی $y(t)$ یک عمل جبری ساده است. با توجه به شکل 9.31 داریم

$$Y(s) = H_1(s) E(s) \quad (9.159)$$

$$E(s) = X(s) - Z(s) \quad (9.160)$$
<div dir="rtl">
و

$$Z(s) = H_2(s) Y(s) \quad (9.161)$$
<div dir="rtl">
که رابطه زیر را به دست میدهد

$$Y(s) = H_1(s) [X(s) - H_2(s) Y(s)] \quad (9.162)$$
<div dir="rtl">
یا

$$\frac{Y(s)}{X(s)} = H(s) = \frac{H_1(s)}{1 + H_1(s) H_2(s)} \quad (9.163)$$

<H3>9.8.2 نمایش جعبه ای سیستم های LTI علی توصیف شونده با معادلات دیفرانسیل و توابع سیستم گویا</H3>

    <h2>مثال 9.30 </h2>
<div dir="rtl">
حال یک سیستم مرتبه دوم LTI با تابع تبدیل زیر در نظر می‌گیریم:

$$H(s) = \frac{1}{(s+1)(s+2)} = \frac{1}{s^2 + 3s + 2} \quad(9.165) $$
<div dir="rtl">
ورودی $x(t)$ و خروجی $y(t)$ این سیستم از معادله دیفرانسیل زیر تبعیت می‌کنند:

$$\frac{d^2y(t)}{dt^2} + 3\frac{dy(t)}{dt} + 2y(t) = x(t) \quad(9.166)$$
<div dir="rtl">
با استفاده از ایده‌هایی مشابه آنچه در مثال‌های قبل به کار گرفته شد، نمایش جعبه‌ای شکل 9.34 (الف) را به دست می‌آوریم. چون ورودی انتگرالگیر، مشتق خروجی انتگرالگیر است، سیگنال‌های این نمودار جعبه‌ای عبارت‌اند از:

$$f(t) = \frac{dy(t)}{dt}$$
$$e(t) = \frac{df(t)}{dt} = \frac{d^2y(t)}{dt^2}$$

$$\frac{d^2y(t)}{dt^2} = -3 \frac{dy(t)}{dt} - 2y(t) + x(t)$$
<div dir="rtl">
یا

$$e(t) = -3f(t) - 2y(t) + x(t)$$
<div dir="rtl">
که دقیقاً معادله نمایش یافته در شکل 9.34 (الف) است.
 <div dir="rtl">   
نمایش جعبه‌ای این شکل را گاهی نمایش مستقیم می‌نامند، زیرا ضرایب ظاهر شده در نمودار همان ضرایب تابع تبدیل یا معادله دیفرانسیل هستند. نمایش‌های جعبه‌ای دیگری نیز وجود دارند که اهمیت عملی فراوانی دارند و با کمی عملیات جبری بر تابع تبدیل به دست می‌آیند. معادله (9.165) را می‌توان به شکل زیر بازنویسی کرد:

$$H(s) = \left(\frac{1}{s+1}\right)\left(\frac{1}{s+2}\right)$$
 <div dir="rtl"> 
که نمایش این سیستم به صورت ترکیب سری دو سیستم مرتبه اول را تداعی می‌کند. نمایش سری $H(s)$ در شکل 9.34 (ب) نشان داده شده است.
<div dir="rtl">
به صورتی دیگر می‌توان با بسط $H(s)$ به کسرهای جزئی زیر به دست آورد:

$$H(s) = \frac{1}{s+1} - \frac{1}{s+2}$$
<div dir="rtl">
که نمایش موازی شکل 9.34 (ج) را به دست می‌دهد.

 <h2>9.9 تبدیل لاپلاس یک طرفه</h2>
 <div dir="rtl">   
 در بخش‌های قبل این فصل با آنچه تبدیل لاپلاس دو طرفه نام دارد کار کردیم. در این بخش تبدیل لاپلاس یکطرفه نسبتاً متفاوتی را معرفی و بررسی می‌کنیم، تبدیل لاپلاس یکطرفه که در تحلیل سیستم‌های علی، و مخصوصاً سیستمهایی که ابتدائاً ساکن نیستند (یعنی دارای شرایط اولیه ثابت)، نقش مهمی ایفا می‌کند.
<div dir="rtl">  
تبدیل لاپلاس یکطرفه سیگنال $x(t)$ به صورت زیر تعریف می‌شود:

$$X(s) = \mathcal{L}\{x(t)\} \triangleq \int_{0^-}^{\infty} x(t) e^{-st} dt \quad (9.170)$$
<div dir="rtl">  
حد پایین انتگرال، یعنی $0^-$ نشان می‌دهد که باید ضربه‌ها و توابع تکین مرتبه بالاتر موجود در $t=0$ را در فاصله انتگرالگیری منظور کرد. برای تبدیل لاپلاس یکطرفه نیز یک نماد اختصاری به صورت زیر وجود دارد:

$$x(t) \stackrel{\mathcal{L}}{\longleftrightarrow} X(s) = \mathcal{L}\{x(t)\} \quad (9.171)$$   
<div dir="rtl">     
مقایسه معادلات (9.170) و (9.3) نشان می‌دهد که تعریف تبدیل لاپلاس دو طرفه و یکطرفه تنها در حد پایین انتگرال تفاوت دارند. تبدیل دو طرفه به تمام سیگنال از $t = -\infty$ تا $t = +\infty$ بستگی دارد، ولی تبدیل لاپلاس یکطرفه تنها به مقدار سیگنال در $t = ۰^-$ تا $t = +\infty$ بستگی دارد. پس اگر دو سیگنال در $t < ۰$ یکسان باشند، ولی در $t \ge ۰$ با هم فرق بکنند، تبدیل لاپلاس دو طرفه متفاوتی دارند ولی تبدیل لاپلاس یکطرفه‌اش هر دو یکی است. همچنین اگر سیگنالی در $t < ۰$ متحد با صفر باشد، تبدیل لاپلاس یکطرفه و دو طرفه‌اش یکی است.
<div dir="rtl">  
چون تبدیل یکطرفه $\mathcal{L}\{x(t)\}$ مشابه تبدیل دو طرفه سیگنالی است که در $t < ۰$ با صفر برابر است و در $t \ge ۰$ با $x(t)$ برابر است، بسیاری از مفاهیم و نتایج به دست آمده برای تبدیل دو طرفه را می‌توان مستقیماً به حالت یکطرفه تعمیم داد. مثلاً با توجه به خاصیت 4 بیان شده در بخش 9.2 می‌بینیم که ROC معادله (9.170) همیشه یک    
راست-نیمصفحه است. عکس تبدیل لاپلاس یکطرفه نیز به همان صورت تبدیل لاپلاس دو طرفه حساب می‌شود، با این قید که ROC تبدیل یکطرفه همیشه یک راست-نیمصفحه است.
<div dir="rtl"> 
برای تشریح تبدیل لاپلاس یکطرفه چند مثال در نظر می‌گیریم.

$$x(t) = [e^{-t} - e^{-2t}]u(t), \quad t > 0^- \quad (9.180)$$
<div dir="rtl">
در این معادله بر یکطرفه بودن تبدیل لاپلاس، با نشان دادن این که اطلاعات به دست آمده در مورد سیگنال تنها در $t > 0^-$ معتبر است، تأکید کرده‌ایم.    

<div dir="rtl">
<h2>مثال 9.36</h2>
<div dir="rtl"> 
تبدیل لاپلاس یکطرفه زیر را در نظر بگیرید:

$$\mathcal{X}(s) = \frac{s^2 - 3}{s+2} \quad (9.181)$$
<div dir="rtl"> 
چون مرتبه صورت $\mathcal{X}(s)$ کوچکتر از مرتبه مخرج نیست، آن را به صورت زیر بسط می‌دهیم:

$$\mathcal{X}(s) = A + Bs + \frac{C}{s+2} \quad (9.182)$$
<div dir="rtl"> 
با برابر قرار دادن معادلات (9.181) و (9.182) و حذف مخرج به دست می‌آوریم:

$$s^2 - 3 = (A + Bs)(s+2) + C \quad (9.183)$$
<div dir="rtl"> 
با برابر قرار دادن ضریب جملات هم توان $s$ به دست می‌آوریم:

$$\mathcal{X}(s) = -2 + s + \frac{1}{s+2} \quad (9.184)$$
<div dir="rtl"> 
ناحیه همگرایی $\text{Re}\{s\} > -2$ است. با گرفتن عکس تبدیل لاپلاس جملات خواهیم داشت:

$$x(t) = -2\delta(t) + u_1(t) + e^{-2t}u(t), \quad t > ۰^- \quad (9.185)$$
<div dir="rtl">
<h2>خواص تبدیل لاپلاس یکطرفه</h2>
 <div dir="rtl">   
تبدیل لاپلاس یکطرفه نیز مانند تبدیل لاپلاس دو طرفه خواصی دارد که بسیاری از آنها مشابه خواص متناظر تبدیل لاپلاس دو طرفه است، ولی برخی از آنها تفاوتهایی قابل ملاحظه دارند. این خواص در جدول 9.3 خلاصه شده‌اند. ستونی برای مشخص کردن ROC تبدیل لاپلاس یکطرفه سیگنالها در نظر
نگرفته ایم،زیرا ROC هر تبدیل لاپلاس یک طرفه ای یک راست-نیم صفحه است. برای مثال ROC هر تبدیل لاپلاس یکطرفه گویایی سمت راست راست ترین قطب است.
مقایسه جدول ۹-۳ و جدول ۹-۱ تبدیل دو طرفه نشان می‌دهد که خواص خطی بودن، جابه‌جایی در حوزه $s$، تغییر مقیاس زمانی، مزدوج‌گیری و مشتق‌گیری در حوزه زمان با همتاهای دو طرفه شان مشابه اند. همچنین قضایای مقدار اولیه و مقدار نهایی بیان شده در بخش ۹-۲-۵ برای تبدیل لاپلاس یکطرفه نیز صادق‌اند. اشتقاق این خواص همانند همتاهای دو طرفه شان صورت می‌گیرد.
خاصیت کانولوشن تبدیل یکطرفه نیز کاملاً مشابۀ خاصیت کانولوشن تبدیل دو طرفه است. این خاصیت می‌گوید که اگر
 <div dir="rtl">  
     
$$ x_1(t) = x_2(t) = 0\quad t < 0$$
  <div dir="rtl">
     آنگاه:
      
$$x_1(t) * x_2(t) \stackrel{\mathcal{L}}{\longleftrightarrow} \mathcal{X}_1(s) \mathcal{X}_2(s) \quad (9.178)$$
<div dir="rtl">
معادله (9.187) بدون واسطه از خاصیت کانولوشن دو طرفه ناشی می‌شود. زیرا تحت شرایط معادله (9.186) تبدیلهای یکطرفه و دو طرفه $x_1(t)$ و $x_2(t)$ یکسان‌اند. پس ابزارهای تحلیلی و جبر تابع تبدیل که بر اساس این خاصیت تبدیل دو طرفه پی‌ریزی شدند، بدون تغییر برای تبدیلهای یکطرفه هم قابل اعمال‌اند، به شرطی که با سیستم‌های علی کار کنیم (که برایشان تبدیل یکطرفه و دو طرفه پاسخ ضربه یکسان است) و ورودیها در $t < ۰$ متحد با صفر باشند. خاصیت انتگرالگیری ذکر شده در جدول 9.3 نمونه‌ای از این مورد است: اگر در $t < ۰$، $x(t) = ۰$ آنگاه

$$\int_{0^-}^{t} x(\tau) d\tau = x(t) * u(t) \stackrel{\mathcal{L}}{\longleftrightarrow} \mathcal{X}(s) \{u(S)\} = \frac{1}{s} \mathcal{X}(s) \quad (9.188)$$
<div dir="rtl">
    به عنوان مورد دوم مثال زیر رو در نظر بگیرید.
<div dir="rtl">
<h2>مثال 9.37</h2>
<div dir="rtl"> 
یک سیستم LTI علی با معادله دیفرانسیل زیر و شرط سکون ابتدایی توصیف شده است.
$$\frac{d^2y(t)}{dt^2} + 3\frac{dy(t)}{dt} + 2y(t) = x(t) \quad (9.189)$$
<div dir="rtl"> 
با استفاده از معادله (9.189) تابع تبدیل این سیستم را به صورت زیر می‌یابیم:

$$\mathcal{H}(s) = \frac{1}{s^2 + 3s + 2} \quad (9.190)$$
<div dir="rtl">
ورودی این سیستم را $x(t) = au(t)$ فرض کنید. در این حالت تبدیل لاپلاس یکطرفه (و دو طرفه) خروجی $y(t)$ عبارت است از:

$$\begin{aligned}
\mathcal{Y}(s) &= \mathcal{H}(s) \mathcal{X}(s) = \frac{a}{s(s+1)(s+2)} \\
&= \frac{a/2}{s} - \frac{a}{s+1} + \frac{a/2}{s+2} \quad (9.191)
\end{aligned}$$
<div dir="rtl">
با استفاده از مثال 9.3 به دست می‌آوریم:

$$y(t) = a\left[\frac{1}{2} - e^{-t} + \frac{1}{2}e^{-2t}\right]u(t) \quad (9.192)$$    
<div dir="rtl">
توجه به این نکته مهم است که خاصیت کانولوشن تبدیل لاپلاس یکطرفه تنها هنگامی قابل اعمال است که سیگنالهای $x_1(t)$ و $x_2(t)$ معادله (9.187) در $t < ۰$ صفر باشند. تبدیل لاپلاس دو طرفه $x_1(t) * x_2(t)$ همیشه با حاصلضرب تبدیلهای دو طرفه $\mathcal{X}_1(t)$ و $\mathcal{X}_2(t)$ برابر است، ولی تبدیل لاپلاس یکطرفه $x_1(t) * x_2(t)$ با حاصلضرب تبدیلهای لاپلاس یکطرفه $\mathcal{X}_1(t)$ و $\mathcal{X}_2(t)$ برابر نیست، مگر این که $x_1(t)$ و $x_2(t)$ در $t < ۰$ صفر باشند. (مسئله 9.39 را ببینید.)
با یک تفاوت مهم خواص تبدیل لاپلاس یکطرفه و دو طرفه خاصیت مشتق‌گیری است. سیگنال $x(t)$ با
تبدیل لاپلاس یکطرفه $\mathcal{Y}(s)$ را در نظر بگیرید. با انتگرالگیری جزء به جزء، تبدیل لاپلاس یکطرفه $\frac{dx(t)}{dt}$ را به صورت زیر می‌یابیم:
$$\int_{0^-}^{\infty} \frac{dy(t)}{dt} e^{-st} dt = x(t) e^{-st} \Big|_{0^-}^{\infty} + s \int_{0^-}^{\infty} x(t) e^{-st} dt$$
$$= s \mathscr{X}(s) - x(0^-)$$


<div dir="rtl">
به نحوی مشابه، با اعمال دوباره این قضیه تبدیل لاپلاس یکطرفه $\frac{d^2x(t)}{dt^2}$ را به دست می‌آوریم:

$$s^2\mathcal{X}(s) - sx(0^-) - x'(0^-) \quad (9.194)$$
<div dir="rtl">
که در آن $x'(0^-)$ مشتق $x(t)$ در $t = 0^-$ را نشان می‌دهد. می‌توانیم این فرایند را ادامه دهیم و تبدیل فوریه مشتق‌های بالاتر را به دست آوریم.
<div dir="rtl">
 <h2> 9.9.3 حل معادلات دیفرانسیل به کمک تبدیل لاپلاس یک طرفه </h2>
<div dir="rtl">
<h2>مثال 9.38</h2>
<div dir="rtl">  
سیستم توصیف شده با معادله دیفرانسیل (9.189) و شرایط اولیه زیر را در نظر بگیرید:

$$y(0^-) = \beta, \quad y'(0^-) = \gamma \quad (9.195)$$
<div dir="rtl">
فرض کنید $x(t) = au(t)$. از دو طرف معادله (9.189) تبدیل لاپلاس یکطرفه می‌گیریم و به دست می‌آوریم:

$$s^2\mathcal{Y}(s) - \beta s - \gamma + 3[s\mathcal{Y}(s) - \beta] + 2\mathcal{Y}(s) = \frac{a}{s} \quad (9.196)$$
<div dir="rtl">
یا

$$\mathcal{Y}(s) = \frac{\beta(s+3)}{(s+1)(s+2)} + \frac{\gamma}{(s+1)(s+2)} + \frac{a}{s(s+1)(s+2)} \quad (9.197)$$ 
<div dir="rtl">
که در آن $\mathcal{Y}(s)$ تبدیل لاپلاس یکطرفه $y(t)$ است. با توجه به مثال 9.37 و معادلات (9.191) آن، می‌بینیم که جمله آخر طرف راست معادله (9.197) پاسخ سیستم به ازای شرایط اولیه صفر است ($y(0^-) = 0$ و $y'(0^-) = 0$). پس جمله آخر پاسخ سیستم LTI علی توصیف شده با معادله (9.189) و شرط سکون ابتدایی است. این پاسخ را غالباً پاسخ حالت صفر می‌نامند، که پاسخ به ازای شرایط اولیه صفر (توصیف شده در معادله (9.195)) است.
برای دو جمله اول سمت راست معادله (9.197) نیز تعبیر مشابهی وجود دارد. این جملات تبدیل یکطرفه پاسخ سیستم به ازای صفر بودن ورودی ($a = 0$) را نشان می‌دهند. این پاسخ را معمولاً پاسخ ورودی صفر می‌نامند. توجه کنید که پاسخ ورودی صفر، تابعی خطی از مقدار شرایط اولیه است (مثلاً با دو برابر شدن $\beta$ و $\gamma$، پاسخ ورودی صفر نیز دو برابر می‌شود). معادله (9.197) حقیقت مهم دیگری راجع به جواب معادلات دیفرانسیل خطی با ضرایب ثابت به ازای شرایط اولیه غیر صفر نشان می‌دهد، و آن این که پاسخ کل برهم‌نهش پاسخهای حالت صفر و ورودی صفر است. پاسخ حالت صفر پاسخی است که با صفر کردن شرایط اولیه به دست
می‌آید، یعنی پاسخ سیستم LTI توصیف شده با معادله دیفرانسیل و شرط سکون ابتدایی پاسخ ورودی صفر پاسخ به شرایط اولیه، هنگام صفر بودن ورودی است. مسائل 9.40، 9.45 و 9.46 مثال‌های دیگری از این وضعیت هستند.
سرانجام به ازای هر مقدار $\beta$ و $\gamma$ می‌توان $\mathcal{Y}(s)$ معادله (9.197) را به کسرهای جزئی بسط داد و با گرفتن عکس تبدیل لاپلاس $y(t)$ را یافت. برای مثال به ازای $a = 2$، $\beta = 3$ و $\gamma = -5$ بسط به کسرهای جزئی معادله (9.197) به دست می‌دهد:

$$\mathcal{Y}(s) = \frac{1}{s} - \frac{1}{s+2} + \frac{3}{s+2} \quad (9.198)$$
<div dir="rtl">
با استفاده از مثال 9.32 به دست می‌آوریم:

$$y(t) = [-1 - e^{-t} + 3e^{-2t}]u(t), \quad t > 0^- \quad (9.199)$$

<h2>خلاصه </h2>
در این فصل تبدیل لاپلاس را، که می‌توان آن را تعمیمی از تبدیل فوریه دانست، مورد مطالعه قرار دادیم. تبدیل لاپلاس مخصوصاً ابزار تحلیلی مفیدی برای تحلیل و مطالعه سیستم‌های LTI است. سیستم‌های LTI، از جمله سیستم‌های توصیف شده با معادلات دیفرانسیل خطی با ضرایب ثابت را می‌توان به خاطر خواص تبدیل لاپلاس در حوزه $s$ و به کمک عملیات جبری توصیف و تحلیل کرد. به علاوه عملیات جبری بر روی تابع تبدیل سیستم ابزار مناسبی برای تحلیل اتصال سیستم‌ها و یافتن نمایش جعبه‌ای سیستم‌های LTI توصیف شده با معادلات دیفرانسیل است.
برای سیگنال‌ها و سیستم‌هایی که تبدیل لاپلاس گویا دارند، غالباً تبدیل لاپلاس را با مشخص کردن محل قطب‌ها و صفرها و ناحیه همگرایی نشان می‌دهند. از روی نمودار قطب-صفر می‌توان تبدیل فوریه را به روش هندسی به دست آورد. با دانستن محل قطب‌ها و ناحیه همگرایی به راحتی می‌توان علی بودن، پایداری و دیگر مشخصه‌های سیستم را استنتاج کرد.
در این فصل عمدتاً تبدیل لاپلاس دو طرفه مورد توجه قرار داشت. تبدیل یکطرفه که نوع نسبتاً متفاوتی از تبدیل لاپلاس است نیز معرفی شد. در واقع تبدیل یکطرفه را می‌توان به عنوان تبدیل دو طرفه سیگنالی که مقادیر آن قبل از $t = 0^-$ صفر شده‌اند، در نظر گرفت. این نوع تبدیل لاپلاس در به دست آوردن پاسخ معادلات دیفرانسیل خطی دارای ضرایب ثابت با شرایط اولیه غیر صفر، بسیار به کار می‌رود.
