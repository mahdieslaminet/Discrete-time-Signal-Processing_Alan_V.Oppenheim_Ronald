## 5.1 LTI پاسخ فرکانسی سیستم‌های

پاسخ فرکانسی $ H(e^{j\omega}) $ یک سیستم LTI به عنوان بهره مختلط (مقدار ویژه) که سیستم به ورودی نمایی مختلط $ e^{j\omega t} $ اعمال می‌کند، تعریف شده است (بخش 2.6 را ببینید). 
علاوه بر این، همان‌طور که در بخش 2.9.6 بحث شد، از آنجا که تبدیل فوریه یک دنباله، تجزیه‌ای به صورت ترکیب خطی از نمایی‌های مختلط را نشان می‌دهد، تبدیل‌های فوریه ورودی و خروجی سیستم به‌صورت زیر به هم مرتبط هستند:

$$
Y(e^{j\omega}) = H(e^{j\omega}) X(e^{j\omega}) \quad (5.3)
$$

که در آن $ X(e^{j\omega}) $ و $ Y(e^{j\omega}) $ به ترتیب تبدیل فوریه ورودی و خروجی سیستم هستند.

### 5.1.1 فاز پاسخ فرکانسی و تأخیر گروهی

پاسخ فرکانسی به طور کلی در هر فرکانس یک عدد مختلط است. وقتی پاسخ فرکانسی به فرم قطبی بیان شود، دامنه و فاز تبدیل‌های فوریه ورودی و خروجی سیستم به صورت زیر به هم مرتبط هستند:

$$
|Y(e^{j\omega})| = |H(e^{j\omega})| \cdot |X(e^{j\omega})| \quad (5.4a)
$$

$$
\angle Y(e^{j\omega}) = \angle H(e^{j\omega}) + \angle X(e^{j\omega}) \quad (5.4b)
$$

که در آن $ |H(e^{j\omega})| $ نشان‌دهنده پاسخ دامنه یا بهره سیستم است و $ \angle H(e^{j\omega}) $ نشان‌دهنده پاسخ فاز یا شیفت فاز سیستم است.

اثرات دامنه و فاز بیان‌شده در معادلات بالا می‌توانند مطلوب باشند، اگر سیگنال ورودی به شکل مفیدی اصلاح شود، یا نامطلوب باشند، اگر سیگنال ورودی به صورت مخربی تغییر کند. در حالت دوم، معمولاً به اثرات یک سیستم LTI روی یک سیگنال به عنوان اعوجاج دامنه و اعوجاج فاز اشاره می‌کنیم.

فاز هر عدد مختلط به‌طور یکتا تعریف نشده است، زیرا هر مضرب صحیحی از $ 2\pi $ را می‌توان بدون تأثیرگذاری بر عدد اصلی به آن افزود. وقتی فاز به‌صورت عددی با استفاده از زیرروال تابع تانژانت معکوس محاسبه می‌شود، مقدار اصلی آن معمولاً به‌دست می‌آید. مقدار اصلی فاز $ H(e^{j\omega}) $ را به‌صورت $ \text{ARG}[H(e^{j\omega})] $ نشان می‌دهیم، که در بازه زیر قرار دارد:

$$
-\pi < \text{ARG}[H(e^{j\omega})] \leq \pi \quad (5.5)
$$


هر زاویه‌ی دیگری که مقدار مختلط صحیح تابع $ H(e^{j\omega}) $ را بدهد، می‌تواند به صورت مقدار اصلی زیر نمایش داده شود:

$$
\angle H(e^{j\omega}) = \text{ARG}[H(e^{j\omega})] + 2\pi r(\omega) \quad (5.6)
$$

که در آن $ r(\omega) $ یک عدد صحیح مثبت یا منفی است که می‌تواند در هر مقدار $ \omega $ متفاوت باشد. در بسیاری از موارد، مقدار اصلی دارای ناپیوستگی‌هایی به اندازه $ 2\pi $ رادیان در تابع $ \omega $ است. این موضوع در شکل زیر نشان داده شده است:



````python
import numpy as np
import matplotlib.pyplot as plt

omega = np.linspace(-np.pi, np.pi, 400)
phase_continuous = -2 * np.sin(omega) - 2 * omega
phase_principal = np.mod(phase_continuous + np.pi, 2*np.pi) - np.pi
r_values = np.round((phase_continuous - phase_principal) / (2 * np.pi))

fig, axs = plt.subplots(3, 1, figsize=(6, 8))

axs[0].plot(omega, phase_continuous, label='arg[H(e^{j\omega})]')
axs[0].set_ylabel('arg[H(e^{j\omega})]')
axs[0].set_xticks([-np.pi, 0, np.pi])
axs[0].set_xticklabels(['-π', '0', 'π'])
axs[0].grid()

axs[1].plot(omega, phase_principal, label='ARG[H(e^{j\omega})]', color='r')
axs[1].set_ylabel('ARG[H(e^{j\omega})]')
axs[1].set_xticks([-np.pi, 0, np.pi])
axs[1].set_xticklabels(['-π', '0', 'π'])
axs[1].grid()

axs[2].step(omega, r_values, label='r(\omega)', color='g', where='mid')
axs[2].set_ylabel('r(\omega)')
axs[2].set_xlabel('ω')
axs[2].set_xticks([-np.pi, 0, np.pi])
axs[2].set_xticklabels(['-π', '0', 'π'])
axs[2].grid()

plt.tight_layout()
plt.show()
````

هر زاویه‌ی دیگری که مقدار مختلط صحیح تابع $ H(e^{j\omega}) $ را بدهد، می‌تواند به صورت مقدار اصلی زیر نمایش داده شود:

$$
\angle H(e^{j\omega}) = \text{ARG}[H(e^{j\omega})] + 2\pi r(\omega) \quad (5.6)
$$

که در آن $ r(\omega) $ یک عدد صحیح مثبت یا منفی است که می‌تواند در هر مقدار $ \omega $ متفاوت باشد. 
ما به طور کلی در این متن از نماد زاویه در سمت چپ معادله‌ی بالا برای نمایش فاز نامعین استفاده خواهیم کرد، زیرا مقدار $ r(\omega) $ تا حدودی اختیاری است.

در بسیاری از موارد، مقدار اصلی دارای ناپیوستگی‌هایی به اندازه $ 2\pi $ رادیان در تابع $ \omega $ است. این موضوع در شکل 5.1 نشان داده شده است که یک تابع فاز پیوسته $ \arg[H(e^{j\omega})] $ و مقدار اصلی آن $ \text{ARG}[H(e^{j\omega})] $ را در دامنه‌ای مشخص نمایش می‌دهد.

یک نمایش مفید دیگر از فاز، از طریق تاخیر گروهی $ \tau(\omega) $ تعریف می‌شود به صورت:

$$
\tau(\omega) = \text{grd}[H(e^{j\omega})] = - \frac{d}{d\omega} \text{ARG}[H(e^{j\omega})]. \quad (5.7)
$$

شایان ذکر است که از آنجا که مشتق $ \arg[H(e^{j\omega})] $ و $ \text{ARG}[H(e^{j\omega})] $ فقط در نقاط ناپیوستگی اختلاف دارند، تاخیر گروهی را می‌توان با مشتق‌گیری از مقدار اصلی به دست آورد، به جز در ناپیوستگی‌ها.

به طور مشابه، می‌توان تاخیر گروهی را بر حسب فاز نامعین به شکل زیر بیان کرد:

$$
\text{grd}[H(e^{j\omega})] = - \frac{d}{d\omega} \angle H(e^{j\omega}). \quad (5.8)
$$

که در آن ضربه‌های ناشی از ناپیوستگی‌های به اندازه‌ی $ 2\pi $ در $ \angle H(e^{j\omega}) $ نادیده گرفته می‌شوند.

برای درک اثر فاز و به‌ویژه تأخیر گروهی یک سیستم خطی، ابتدا سیستم تأخیر ایده‌آل را در نظر می‌گیریم. پاسخ ضربه‌ای این سیستم به‌صورت زیر است:

$$
 h_{\text{id}}[n] = \delta[n - n_d]. \quad (5.9)
$$

و پاسخ فرکانسی آن برابر است با:

$$
H_{\text{id}}(e^{j\omega}) = e^{-j\omega n_d}. \quad (5.10)
$$

یا به صورت زیر بیان می‌شود:

$$
|H_{\text{id}}(e^{j\omega})| = 1, \quad (5.11a)
$$

$$
\angle H_{\text{id}}(e^{j\omega}) = -\omega n_d, \quad |\omega| < \pi. \quad (5.11b)
$$

که در آن، تناوبیت $ 2\pi $ در $ \omega $ مشاهده می‌شود. از معادله‌ی بالا نتیجه می‌گیریم که تاخیر زمانی (یا پیش‌افتادن اگر $ n_d < 0 $ باشد) با شیب فاز در ارتباط است.

w[n] =

0.54 - 0.46 \cos \left( \frac{2\pi n}{M} \right), & 0 \leq n \leq M \\
0, & \text{در غیر این صورت}
\end{cases}
\]

که در آن \( M = 60 \) است. این توالی به عنوان پنجره Hamming نیز شناخته می‌شود.

### ترکیب ورودی کامل که در شکل 5.5(a) نمایش داده شده:

\[
x[n] = x_3[n] + x_1[n - M - 1] + x_2[n - 2M - 2]
\]

یعنی، پالس با بیش‌ترین فرکانس اول می‌آید، سپس کم‌ترین، و در آخر پالس فرکانس میانی. فرکانس‌ها به صورت زیر هستند:

\[
\omega_1 = 0.2\pi, \quad \omega_2 = 0.4\pi, \quad \omega_3 = 0.8\pi
\]

مطابق با قضیه پنجره‌سازی تبدیل فوریه، پاسخ DTFT یک سینوس محدودشده حاصل کانولوشن DTFT آن سینوس با DTFT پنجره است.


````python
# نوت‌بوک Jupyter - فصل ۵ از کتاب Oppenheim (ویرایش دوم)
# صفحات 280 تا 290: پاسخ فاز و تأخیر گروهی

import numpy as np
import matplotlib.pyplot as plt

# تعریف محور فرکانس
omega = np.linspace(-np.pi, np.pi, 1024)

# تعریف تابع انتقال H(e^jω)
H = np.exp(1j * (4 * np.sin(omega))) * np.exp(-1j * omega**2)

# محاسبه فاز اصلی و فاز بازشده
phase_principal = np.angle(H)
phase_unwrapped = np.unwrap(phase_principal)

# رسم شکل 5.3
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

axs[0].plot(omega, phase_principal)
axs[0].set_title("(a) مقدار اصلی فاز")
axs[0].set_ylabel("ARG[H(e^{jω})]")
axs[0].grid(True)
axs[0].set_xlim([-np.pi, np.pi])

axs[1].plot(omega, phase_unwrapped)
axs[1].set_title("(b) فاز بازشده")
axs[1].set_ylabel("arg[H(e^{jω})]")
axs[1].set_xlabel("ω")
axs[1].grid(True)
axs[1].set_xlim([-np.pi, np.pi])

plt.tight_layout()
plt.show()
````

## صفحه 281

پاسخ فاز سیستم خطی و نامتغیر با زمان H(e^{jω}) اغلب توسط تابع فاز θ(ω) تعریف می‌شود:

$$
H(e^{jω}) = |H(e^{jω})|e^{jθ(ω)} \tag{5.18}
$$

تابع θ(ω) اغلب یک تابع غیرفرد است، حتی اگر |H(e^{jω})| زوج باشد. از آنجا که فاز به صورت زاویه در نظر گرفته می‌شود، تابع θ(ω) فقط تا مضربی از 2π مشخص است.

برای تحلیل سیستم‌ها، اغلب فاز "بازشده" استفاده می‌شود که به کمک عملیات ریاضی پیوستگی نسبی در θ(ω) ایجاد می‌کند.

در شکل 5.3، بخش (a) مقدار اصلی فاز را نشان می‌دهد و بخش (b) فاز بازشده را. همانطور که مشاهده می‌شود، فاز اصلی دارای پرش‌هایی در نقاط مختلف ω است که در فاز بازشده حذف شده‌اند.

اگر تابع فاز θ(ω) خطی باشد، به صورت:

$$
θ(ω) = -ωn_0
$$

در این صورت، پاسخ ضربی تبدیل فوریه برابر است با:

$$
H(e^{jω}) = e^{-jωn_0} \tag{5.19}
$$

که نشان‌دهنده یک تأخیر زمانی n₀ نمونه‌ای است. این نوع سیستم‌ها سیگنال ورودی را بدون اعوجاج تغییر شکل نمی‌دهند بلکه تنها آن را به تأخیر می‌اندازند.


````python

import numpy as np
import matplotlib.pyplot as plt

# Create signal x[n]
M = 60
def w(n, M):
    if 0 <= n <= M:
        return 0.54 - 0.46 * np.cos(2 * np.pi * n / M)
    return 0

# Frequencies
omega1 = 0.2 * np.pi
omega2 = 0.4 * np.pi
omega3 = 0.8 * np.pi
n = np.arange(0, 300)
x1 = np.array([np.cos(omega1 * (i - 150)) * w(i - 150 + 30, M) for i in n])
x2 = np.array([np.cos(omega2 * (i - 90)) * w(i - 90 + 30, M) for i in n])
x3 = np.array([np.cos(omega3 * (i - 30)) * w(i - 30 + 30, M) for i in n])
x = x3 + x1 + x2

# Plot waveform
plt.figure(figsize=(10, 4))
plt.plot(n, x)
plt.title("(a) Waveform of signal x[n]")
plt.xlabel("Sample number (n)")
plt.ylabel("x[n]")
plt.grid(True)
plt.tight_layout()
plt.show()

# DTFT using zero-padding
from scipy.fft import fft, fftshift
X = fftshift(fft(x, 2048))
X_mag = np.abs(X)
w = np.linspace(-np.pi, np.pi, len(X_mag))

# Plot DTFT magnitude
plt.figure(figsize=(10, 4))
plt.plot(w, X_mag)
plt.title("(b) Magnitude of DTFT of x[n]")
plt.xlabel("ω")
plt.ylabel("|X(e^{jω})|")
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftshift

# تنظیمات اولیه
M = 60
n = np.arange(0, 3*M)
w = 0.54 - 0.46 * np.cos(2*np.pi*np.arange(0, M)/M)

# فرکانس‌ها
w1 = 0.2 * np.pi
w2 = 0.4 * np.pi
w3 = 0.8 * np.pi

# تولید سیگنال‌های سینوسی ضرب در پنجره
x1 = np.zeros_like(n)
x2 = np.zeros_like(n)
x3 = np.zeros_like(n)

x1[0:M] = np.cos(w3*np.arange(0, M)) * w
x2[M:2*M] = np.cos(w2*np.arange(0, M)) * w
x3[2*M:3*M] = np.cos(w1*np.arange(0, M)) * w

# سیگنال ترکیبی
x = x3 + x1 + x2

# تعریف پاسخ فرکانسی سیستم H(e^{jω})
omega = np.linspace(-np.pi, np.pi, 1024)
H = np.exp(1j * (4 * np.sin(omega))) * np.exp(-1j * omega**2)

# اعمال سیستم به سیگنال ورودی
X = fftshift(fft(x, 1024))
Y = X * H
y = np.real(np.fft.ifft(np.fft.ifftshift(Y)))

# نمایش شکل موج y[n]
plt.figure(figsize=(10, 4))
plt.plot(n, y[:len(n)])
plt.title("Waveform of output signal y[n]")
plt.xlabel("Sample number (n)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()

# محاسبه و نمایش طیف DTFT خروجی
Y_mag = np.abs(Y)

plt.figure(figsize=(10, 4))
plt.plot(omega, Y_mag)
plt.title("Magnitude of DTFT of y[n]")
plt.xlabel("ω")
plt.ylabel("|Y(e^{jω})|")
plt.grid(True)
plt.tight_layout()
plt.show()
````

همان‌طور که در بخش 3.5 دیدیم، با اعمال تبدیل z به دو طرف معادله (5.19) و استفاده از خاصیت خطی بودن (بخش 3.4.1) و خاصیت انتقال زمانی (بخش 3.4.2)، نتیجه می‌گیریم که برای سیستمی که ورودی و خروجی آن یک معادله تفاضلی از فرم معادله (5.19) را ارضا می‌کند، تابع سیستم دارای فرم جبری زیر خواهد بود:

$$
H(z) = \frac{Y(z)}{X(z)} = \frac{\sum_{k=0}^{M} b_k z^{-k}}{\sum_{k=0}^{N} a_k z^{-k}} \tag{5.20}
$$

در معادله (5.20)، \( H(z) \) به فرم نسبت چندجمله‌ای‌هایی بر حسب \( z^{-1} \) است، زیرا معادله (5.19) از دو ترکیب خطی از ترم‌های تاخیر یافته تشکیل شده است.
اگرچه معادله (5.20) را می‌توان به‌گونه‌ای بازنویسی کرد که چندجمله‌ای‌ها بر حسب توان‌های \( z \) نوشته شوند به‌جای \( z^{-1} \)، اما معمولاً این کار انجام نمی‌شود.
همچنین، اغلب راحت‌تر است که معادله (5.20) به فرم فاکتورگیری شده زیر نوشته شود:

$$
H(z) = \left( \frac{b_0}{a_0} \right) \frac{\prod_{k=1}^{M} (1 - c_k z^{-1})}{\prod_{k=1}^{N} (1 - d_k z^{-1})} \tag{5.21}
$$

هر یک از فاکتورهای \( (1 - c_k z^{-1}) \) در صورت کسر یک صفر در \( z = c_k \) و یک صفر در \( z = 0 \) تولید می‌کند.
به‌طور مشابه، هر یک از فاکتورهای \( (1 - d_k z^{-1}) \) در مخرج کسر یک صفر در \( z = 0 \) و یک قطب در \( z = d_k \) تولید می‌کند.

یک رابطه مستقیم بین تابع سیستم و معادله تفاضلی متناظر آن وجود دارد.
به‌طور خاص، چندجمله‌ای صورت در معادله (5.20) دارای ضرایب و ساختار جبری یکسان با طرف راست معادله (5.19) است
(ترم‌هایی به فرم \( b_k x[n-k] \))، در حالی که چندجمله‌ای مخرج در معادله (5.20) دارای ضرایب و ساختار جبری یکسان با طرف چپ معادله (5.19) است
(ترم‌هایی به فرم \( a_k y[n-k] \)). بنابراین، با داشتن یا تابع سیستم به فرم معادله (5.20) یا معادله تفاضلی به فرم (5.19)، به راحتی می‌توان دیگری را به‌دست آورد.

این مطلب در مثال زیر نشان داده شده است:

### Example 5.1 - سیستم مرتبه دوم

H(z) = \frac{(1 + z^{-1})^2}{(1 - \frac{1}{2}z^{-1})(1 + \frac{3}{4}z^{-1})} \tag{5.22}
$$

برای یافتن معادله تفاضلی‌ای که توسط ورودی و خروجی این سیستم ارضا می‌شود، باید \( H(z) \) از معادله (5.20) به‌دست آید؛ با ضرب کردن صورت و مخرج در فاکتورهای مناسب برای تبدیل آن به نسبت چندجمله‌ای‌ها:

$$
H(z) = \frac{1 + 2z^{-1} + z^{-2}}{1 + \frac{1}{4}z^{-1} - \frac{3}{8}z^{-2}} = \frac{Y(z)}{X(z)}
$$
```



در نتیجه:

$$

$$

و معادله تفاضلی به شکل زیر خواهد بود:

$$
y[n] + \frac{1}{4}y[n-1] - \frac{3}{8}y[n-2] = x[n] + 2x[n-1] + x[n-2] \tag{5.24}
$$



## 5.2.1 پایداری و علیت


برای به‌دست آوردن معادله (5.20) از معادله (5.19)، فرض کردیم که سیستم خطی و تغییرناپذیر با زمان است تا بتوان از معادله (5.2) استفاده کرد،
اما هیچ فرضی درباره پایداری یا علیت نکردیم. به‌صورت متناظر، از معادله تفاضلی می‌توان فرم جبری تابع سیستم را به‌دست آورد، اما ناحیه همگرایی مشخص نمی‌شود.

به‌طور خاص، ناحیه همگرایی تابع \( H(z) \) از فرآیند استخراج معادله (5.20) به‌دست نمی‌آید، زیرا تنها چیزی که برای اعتبار معادله (5.20) نیاز است این است که \( X(z) \) و \( Y(z) \) دارای ناحیه‌های همگرایی مشترک باشند.
این موضوع با واقعیت مطرح‌شده در فصل 2 نیز سازگار است، که معادله تفاضلی پاسخ ضربه‌ای یک سیستم خطی و تغییرناپذیر با زمان را به‌طور یکتا مشخص نمی‌کند.

برای تابع سیستم به فرم (5.20) یا (5.21)، انتخاب‌های مختلفی برای ناحیه همگرایی وجود دارد. برای یک نسبت از چندجمله‌ای‌ها، هر انتخاب متفاوت برای ناحیه همگرایی به پاسخ ضربه‌ای متفاوتی منجر می‌شود، اما همگی به معادله تفاضلی یکسانی منجر خواهند شد.

اگر فرض کنیم سیستم علی باشد، نتیجه می‌گیریم که \( h[n] \) باید دنباله‌ای از سمت راست باشد (یعنی برای \( n < 0 \) صفر باشد)،
و بنابراین ناحیه همگرایی تابع \( H(z) \) باید بیرونی‌ترین ناحیه باشد.

از طرف دیگر، اگر فرض کنیم سیستم پایدار باشد، از بحث در بخش 2.4 می‌دانیم که پاسخ ضربه‌ای باید مجموع‌پذیر مطلق باشد، یعنی:

$$
\sum_{n=-\infty}^{\infty} |h[n]| < \infty \tag{5.25}
$$

از آنجا که رابطه (5.25) معادل شرط زیر است:

$$
\sum_{n=-\infty}^{\infty} |h[n]z^{-n}| < \infty \tag{5.26}
$$

برای \( |z| = 1 \)، شرط پایداری معادل این است که ناحیه همگرایی تابع \( H(z) \) شامل دایره واحد باشد.
تعیین ناحیه همگرایی متناظر با تابع سیستم که از معادله تفاضلی به‌دست آمده در مثال زیر نشان داده شده است:

## Example 5.2 - تعیین ناحیه همگرایی


در نظر بگیرید که سیستم خطی و تغییرناپذیر با زمان دارای ورودی و خروجی مرتبط با معادله تفاضلی زیر است:

$$
y[n] - \frac{5}{2}y[n - 1] + y[n - 2] = x[n] \tag{5.27}
$$

از بحث‌های پیشین، فرم جبری تابع سیستم \( H(z) \) به صورت زیر به‌دست می‌آید:

$$
H(z) = \frac{1}{1 - \frac{5}{2}z^{-1} + z^{-2}} = \frac{1}{(1 - \frac{1}{2}z^{-1})(1 - 2z^{-1})} \tag{5.28}
$$

نقاط صفر-قطب متناظر برای \( H(z) \) در شکل 5.7 نشان داده شده‌اند. سه انتخاب ممکن برای ناحیه همگرایی وجود دارد.
اگر سیستم علی باشد، ناحیه همگرایی باید بزرگ‌تر از بزرگ‌ترین مقدار قطب‌ها باشد.


اگر ناحیه همگرایی بیرون‌ترین قطب باشد، یعنی:

$$
|z| > 2
$$

در این حالت، سیستم پایدار نخواهد بود، چون ناحیه همگرایی شامل دایره واحد نمی‌شود. اگر فرض کنیم سیستم پایدار است، آنگاه ناحیه همگرایی خواهد بود:

$$
\frac{1}{2} < |z| < 2
$$

و تابع پاسخ ضربه‌ای \( h[n] \) یک دنباله دوطرفه خواهد بود.

برای انتخاب سوم ناحیه همگرایی یعنی:

$$
|z| < \frac{1}{2}
$$

سیستم نه پایدار است و نه علی.

## شکل 5.7 - نمودار قطب و صفر برای مثال 5.2


همان‌طور که مثال 5.2 نشان می‌دهد، علیت و پایداری الزاما با هم سازگار نیستند. برای یک سیستم خطی و تغییرناپذیر با زمان که ورودی و خروجی آن یک معادله تفاضلی از فرم معادله (5.19) را ارضا می‌کنند، برای اینکه هم علی و هم پایدار باشد، ناحیه همگرایی تابع سیستم باید بیرونی‌ترین قطب را شامل شود و دایره واحد را نیز در بر گیرد.

واضح است که این شرایط مستلزم آن است که تمام قطب‌های تابع سیستم درون دایره واحد قرار داشته باشند.


## 5.2.2 سیستم‌های معکوس


برای یک سیستم خطی و تغییرناپذیر با زمان با تابع سیستم \( H(z) \)، سیستم معکوس متناظر سیستمی است با تابع سیستم \( H_i(z) \) به‌طوری‌که اگر آن را با \( H(z) \) به‌صورت سری ترکیب کنیم، تابع سیستم کلی برابر با واحد خواهد شد. یعنی:

$$
G(z) = H(z) H_i(z) = 1 \\ \tag{5.29}
$$

از این رابطه نتیجه می‌گیریم:

$$
H_i(z) = \frac{1}{H(z)} \tag{5.30}
$$

معادل زمانی این رابطه به‌صورت:

$$
g[n] = h[n] * h_i[n] = \delta[n] \tag{5.31}
$$

از معادله (5.30)، پاسخ فرکانسی سیستم معکوس (اگر وجود داشته باشد) به شکل زیر خواهد بود:

$$
H_i(e^{j\omega}) = \frac{1}{H(e^{j\omega})} \tag{5.32}
$$


## مثال ۵.۳: سیستم معکوس برای سیستم مرتبه اول

یعنی \( H_i(e^{j\omega}) \) عکس \( H(e^{j\omega}) \) است. به‌طور معادل، اندازه لگاریتمی، فاز، و تأخیر گروهی سیستم معکوس منفی مقادیر متناظر برای سیستم اصلی هستند. همهٔ سیستم‌ها معکوس ندارند. برای مثال، فیلتر پایین‌گذر ایده‌آل معکوس ندارد. هیچ راهی برای بازیابی مؤلفه‌های فرکانسی که توسط چنین فیلتری صفر شده‌اند، وجود ندارد.

بسیاری از سیستم‌ها دارای معکوس هستند و کلاس سیستم‌هایی با تابع تبدیل گویا، یک مثال بسیار مفید و جالب را فراهم می‌کنند. به طور خاص، در نظر بگیرید:

$$
H(z) = \left( \frac{b_0}{a_0} \right) \frac{ \prod_{k=1}^M \left(1 - c_k z^{-1}\right) }{ \prod_{k=1}^N \left(1 - d_k z^{-1} \right) }\tag{5.33}
$$

که دارای صفرهایی در \( z = c_k \) و قطب‌هایی در \( z = d_k \) است، به‌علاوه صفرها یا قطب‌های ممکن در \( z = 0 \) و \( z = \infty \). در این‌صورت:

$$
H_i(z) = \left( \frac{a_0}{b_0} \right) \frac{ \prod_{k=1}^N \left(1 - d_k z^{-1}\right) }{ \prod_{k=1}^M \left(1 - c_k z^{-1}\right) }\tag{5.34}
$$

یعنی، قطب‌های \( H_i(z) \) همان صفرهای \( H(z) \) هستند و بالعکس. اکنون این پرسش مطرح می‌شود که چه ناحیه‌ای از همگرایی (ROC) باید برای \( H_i(z) \) در نظر گرفت. پاسخ این پرسش توسط قضیهٔ هم‌پیچی (convolution theorem) که در رابطهٔ (5.31) بیان شده، ارائه می‌شود. برای برقرار بودن رابطهٔ (5.31)، ناحیهٔ همگرایی \( H(z) \) و \( H_i(z) \) باید هم‌پوشانی داشته باشند. اگر \( H(z) \) علی باشد، ناحیهٔ همگرایی آن به صورت زیر است:

$$
|z| > \max_k |d_k|\tag{5.35}
$$

بنابراین، هر ناحیهٔ همگرایی برای \( H_i(z) \) که با ناحیهٔ فوق هم‌پوشانی داشته باشد، یک ناحیهٔ همگرایی معتبر برای \( H_i(z) \) است. مثال‌های ۵.۳ و ۵.۴ برخی از این حالات را نشان می‌دهند.

---

### مثال ۵.۳: سیستم معکوس برای سیستم مرتبه اول

فرض کنید:

$$
H(z) = \frac{1 - 0.5z^{-1}}{1 - 0.9z^{-1}}
$$

با ناحیهٔ همگرایی:

$$
|z| > 0.9
$$

در این صورت:

$$
H_i(z) = \frac{1 - 0.9z^{-1}}{1 - 0.5z^{-1}}
$$

از آن‌جا که \( H_i(z) \) فقط یک قطب دارد، تنها دو حالت برای ناحیهٔ همگرایی آن وجود دارد و تنها گزینه‌ای که با \( |z| > 0.9 \) هم‌پوشانی دارد، \( |z| > 0.5 \) است. بنابراین، پاسخ ضربه‌ای سیستم معکوس به صورت زیر است:

$$
h_i[n] = (0.5)^n u[n] - 0.9 (0.5)^{n-1} u[n - 1]
$$

در این حالت، سیستم معکوس هم علی است و هم پایدار.
