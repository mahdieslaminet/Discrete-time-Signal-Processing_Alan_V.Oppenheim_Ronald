# **مثال ۷.۱: طراحی یک فیلتر دیجیتال کم‌گذر با استفاده از روش نمونه‌برداری فرکانسی**

یک فیلتر کم‌گذر گسسته‌زمانی را در نظر بگیرید که قرار است برای فیلتر کردن یک سیگنال پیوسته‌زمانی با استفاده از پیکربندی پایه‌ای شکل 7.2 به کار رود. همان‌طور که در بخش 4.4 نشان داده شد، اگر یک سیستم گسسته‌زمانی تغییرناپذیر با زمان (LTI) مطابق شکل 7.2 استفاده شود و اگر ورودی باند محدود باشد و فرکانس نمونه‌برداری به اندازه کافی بالا باشد تا از پدیده تداخل طیفی (aliasing) جلوگیری کند، در این صورت، سیستم کلی مانند یک سیستم پیوسته‌زمانی LTI با پاسخ فرکانسی زیر رفتار خواهد کرد

$$
H_d(e^{j\omega}) = 
\begin{cases} 
1, & |\omega| \leq \omega_c \\
0, & \omega_c < |\omega| \leq \pi
\end{cases}
$$

در چنین مواردی، تبدیل مشخصات فیلتر پیوسته‌زمانی مؤثر به مشخصات فیلتر گسسته‌زمانی از طریق رابطه $$\omega = T\Omega$$ به‌سادگی انجام می‌شود. به این ترتیب، $$H(e^{j\omega})$$ در یک دوره با معادله زیر مشخص می‌شود


برای این مثال، سیستم کلی در شکل 7.2 باید دارای ویژگی‌های زیر باشد، زمانی که نرخ نمونه‌برداری 10^4 نمونه بر ثانیه است (T = 10^{-4} s):
	•	بهره |H_{eff}(j\Omega)| باید در بازه فرکانسی 0 \leq \Omega \leq 2\pi(2000) در محدوده 1 \pm 0.01 باشد.
	•	بهره نباید در بازه فرکانسی 2\pi(3000) \leq \Omega بیشتر از 0.001 باشد.

از آنجا که معادله (7.1a) یک نگاشت بین فرکانس‌های پیوسته‌زمانی و گسسته‌زمانی است، تنها فرکانس‌های مرزی باند عبور و باند توقف را تحت تأثیر قرار می‌دهد و تأثیری بر حدود تحمل قدر مطلق پاسخ فرکانسی ندارد. برای این مثال خاص، پارامترها به صورت زیر خواهند بود:

$$\delta_{p1} = \delta_{p2} = 0.01$$
$$\delta_s = 0.001$$
$$\omega_p = 0.4\pi \text{ radians}$$
$$\omega_s = 0.6\pi \text{ radians}$$

بنابراین، در این حالت:
	•	مقدار ایده‌آل بهره در باند عبور برابر 1 است و اجازه دارد بین (1 + \delta_{p1}) و (1 - \delta_{p2}) تغییر کند.
	•	مقدار بهره در باند توقف باید بین 0 و \delta_s باشد.

بیان شده بر حسب دسی‌بل:
	•	بهره ایده‌آل باند عبور در دسی‌بل:
$$20 \log_{10}(1) = 0 \text{ dB}$$
	•	حداکثر بهره در باند عبور:
$$20 \log_{10}(1.01) = 0.0864 \text{ dB}$$
	•	حداقل بهره در لبه باند عبور:
$$20 \log_{10}(0.99) = -0.873 \text{ dB}$$
	•	حداکثر بهره در باند توقف:
$$20 \log_{10}(0.001) = -60 \text{ dB}$$


مثال 7.1 در چارچوب پردازش یک سیگنال پیوسته‌زمانی توسط یک فیلتر گسسته‌زمانی پس از نمونه‌برداری متناوب مطرح شد. اما کاربردهای زیادی وجود دارند که در آن‌ها سیگنال گسسته‌زمانی مورد پردازش، مستقیماً از یک سیگنال پیوسته‌زمانی مشتق نشده است. همچنین، روش‌های مختلفی به‌جز نمونه‌برداری متناوب برای نمایش سیگنال‌های پیوسته‌زمانی به‌صورت دنباله‌های گسسته وجود دارد.

علاوه بر این، در بیشتر تکنیک‌های طراحی که مورد بحث قرار می‌گیرند، دوره نمونه‌برداری (T) هیچ نقشی در رویه تقریب ندارد. به همین دلیل، مسئله طراحی فیلتر را مستقیماً برحسب مشخصات موردنظر در متغیر فرکانسی گسسته‌زمانی \omega در نظر می‌گیریم. بسته به کاربرد یا زمینه خاص، این مشخصات ممکن است از فیلتر کردن در چارچوب شکل 7.2 مشتق شده باشند یا نباشند.





- از معادله دیفرانسیل برای به‌دست‌آوردن یک عبارت برای \( y_c(nT) \) استفاده کنید و این عبارت را در معادله به‌دست‌آمده در قسمت (a) جای‌گذاری کنید.  
- متغیرهای \( x[n] = x_c(nT) \) و \( y[n] = y_c(nT) \) را تعریف کنید. با استفاده از این نمادگذاری و نتیجه قسمت (b)، معادله تفاضلی مرتبط با \( x[n] \) و \( y[n] \) را به‌دست آورید و سپس تابع سیستم را تعیین کنید:  

$$
H(z) = \frac{Y(z)}{X(z)}
$$

برای سیستم گسسته‌ی حاصل.  
- نشان دهید که برای این مثال، داریم:  

$$
H(z) = H_c(s) \Big|_{s = \frac{2}{T} \frac{1 - z^{-1}}{1 + z^{-1}}}
$$



یعنی، نشان دهید که \( H(z) \) را می‌توان مستقیماً از \( H_c(s) \) با استفاده از تبدیل بی‌خطی (bilinear transformation) به‌دست آورد.  

(برای معادلات دیفرانسیل مرتبه بالاتر، اگر انتگرال‌گیری ذوزنقه‌ای تکرارشونده روی مشتق مرتبه بالای خروجی اعمال شود، همین نتیجه برای یک سیستم پیوسته با تابع سیستم گویا نیز حاصل خواهد شد.)  

### 7.50  

در این مسئله، یک روش طراحی فیلتر را بررسی می‌کنیم که می‌توان آن را **تغییرناپذیری همبستگی خودکار (autocorrelation invariance)** نامید.  

یک سیستم پایدار زمان‌پیوسته با پاسخ ضربه‌ای \( h_c(t) \) و تابع سیستم \( H_c(s) \) را در نظر بگیرید.  

تابع همبستگی خودکار پاسخ ضربه‌ای سیستم به‌صورت زیر تعریف می‌شود:  

$$
\phi_c(\tau) = \int_{-\infty}^{\infty} h_c(t) h_c(t + \tau) dt
$$


و برای یک پاسخ ضربه‌ای حقیقی، به‌راحتی نشان داده می‌شود که تبدیل لاپلاس \( \phi_c(\tau) \) به‌صورت زیر است:  

$$
\Phi_c(s) = H_c(s) H_c(-s)
$$  

به‌طور مشابه، یک سیستم زمان‌گسسته با پاسخ ضربه‌ای \( h[n] \) و تابع سیستم \( H(z) \) را در نظر بگیرید.  

تابع همبستگی خودکار پاسخ ضربه‌ای یک سیستم زمان‌گسسته به‌صورت زیر تعریف می‌شود:  

$$
\phi[m] = \sum_{n=-\infty}^{\infty} h[n] h[n + m]
$$



و برای یک پاسخ ضربه‌ای حقیقی، داریم:  

$$
\Phi(z) = H(z) H(z^{-1})
$$  

اصل **پایداری همبستگی خودکار** بیان می‌کند که یک فیلتر زمان‌گسسته با برابر قرار دادن تابع همبستگی خودکار سیستم زمان‌گسسته با نسخه نمونه‌برداری‌شده تابع همبستگی خودکار یک سیستم زمان‌پیوسته تعریف می‌شود؛ یعنی:  

$$
\phi[m] = T_d \phi_c(m T_d), \quad -\infty < m < \infty
$$



روش طراحی زیر برای **پایداری همبستگی خودکار** پیشنهاد می‌شود، زمانی که \( H_c(s) \) یک تابع گویا با **\( N \)** قطب مرتبه اول در **\( s_k \)** برای **\( k = 1, 2, \dots, N \)** و **\( M < N \)** صفر باشد:  

1. تجزیه‌ی کسرهای جزئی **\( \Phi_c(s) \)** را به‌دست آورید به‌طوری‌که به فرم زیر باشد:  

$$
\Phi_c(s) = \sum_{k=1}^{N} \frac{A_k}{s - s_k} + \sum_{k=1}^{N} \frac{A_k}{s + s_k}
$$  


2. تبدیل **Z** را به‌صورت زیر تشکیل دهید:  

$$
\Phi(z) = \sum_{k=1}^{N} \frac{A_k}{1 - e^{s_k T_d} z^{-1}} + \sum_{k=1}^{N} \frac{A_k}{1 - e^{-s_k T_d} z^{-1}}
$$  




3. قطب‌ها و صفرهای \( \Phi(z) \) را پیدا کنید، و یک تابع سیستم مینیمم-فاز \( H(z) \) از قطب‌ها و صفرهای \( \Phi(z) \) که داخل دایره واحد هستند، بسازید.

(a) هر مرحله از فرایند طراحی پیشنهادی را توجیه کنید؛ یعنی نشان دهید که تابع خودهمبستگی سیستم گسسته‌زمانی به‌دست‌آمده نسخه نمونه‌برداری‌شده از تابع خودهمبستگی سیستم پیوسته‌زمانی است. برای تأیید این فرایند، ممکن است مفید باشد که آن را بر روی سیستم مرتبه اول با پاسخ ضربه امتحان کنید.

$$
H(z) = H_c(s) \Big|_{s=\frac{2}{T} \frac{1 - z^{-1}}{1 + z^{-1}}}
$$

$$
\phi_c(\tau) = \int_{-\infty}^{\infty} h_c(t) h_c(t + \tau) dt
$$

$$
\Phi_c(s) = H_c(s) H_c(-s)
$$

$$
\phi[m] = \sum_{n=-\infty}^{\infty} h[n] h[n + m]
$$

$$
\Phi(z) = H(z) H(z^{-1})
$$

$$
\phi[m] = T_d \phi_c(mT_d), \quad -\infty < m < \infty
$$

$$
\Phi_c(s) = \sum_{k=1}^{N} \left( \frac{A_k}{s - s_k} + \frac{B_k}{s + s_k} \right)
$$

$$
\Phi(z) = \sum_{k=1}^{N} \left( \frac{T_d A_k}{1 - e^{s_k T_d} z^{-1}} + \frac{T_d B_k}{1 - e^{-s_k T_d} z^{-1}} \right)
$$

## 7.50 طراحی فیلتر IIR گسسته‌زمانی از فیلترهای پیوسته‌زمانی  

در این مسئله، روشی برای طراحی فیلتر در نظر گرفته می‌شود که **تغییرناپذیری خودهمبستگی (Autocorrelation Invariance)** نام دارد. یک سیستم پایدار پیوسته‌زمانی را با پاسخ ضربه‌ای \( h_c(t) \) و تابع سیستم \( H_c(s) \) در نظر بگیرید. تابع خودهمبستگی پاسخ ضربه‌ای این سیستم به صورت زیر تعریف می‌شود:  

$$ \phi_c(\tau) = \int_{-\infty}^{\infty} h_c(t) h_c(t + \tau) dt, $$

برای یک پاسخ ضربه‌ای حقیقی، به راحتی نشان داده می‌شود که تبدیل لاپلاس تابع خودهمبستگی برابر است با:  

$$ \Phi_c(s) = H_c(s) H_c(-s). $$

به‌طور مشابه، یک سیستم گسسته‌زمانی با پاسخ ضربه‌ای \( h[n] \) و تابع سیستم \( H(z) \) را در نظر بگیرید. تابع خودهمبستگی پاسخ ضربه‌ای این سیستم به‌صورت زیر تعریف می‌شود:  

$$ \phi[m] = \sum_{n=-\infty}^{\infty} h[n] h[n+m], $$

و برای یک پاسخ ضربه‌ای حقیقی، داریم:  

$$ \Phi(z) = H(z) H(z^{-1}). $$

**تغییرناپذیری خودهمبستگی** به این معناست که یک فیلتر گسسته‌زمانی با برابر گرفتن تابع خودهمبستگی سیستم گسسته‌زمانی و مقدار نمونه‌برداری‌شده تابع خودهمبستگی سیستم پیوسته‌زمانی تعریف می‌شود، یعنی:  

$$ \phi[m] = T_d \phi_c(mT_d), \quad -\infty < m < \infty. $$

### روش طراحی پیشنهادی  

روش طراحی زیر برای **تغییرناپذیری خودهمبستگی** پیشنهاد می‌شود، در حالتی که \( H_c(s) \) یک تابع گویا (rational function) است که دارای \( N \) قطب مرتبه اول در \( s_k, k = 1, 2, \dots, N \) و \( M < N \) صفر می‌باشد:  

1. بسط کسر جزئی (Partial Fraction Expansion) تابع \( \Phi_c(s) \) را به‌دست آورید:  

   $$ \Phi_c(s) = \sum_{k=1}^{N} \left( \frac{A_k}{s - s_k} + \frac{B_k}{s + s_k} \right). $$

2. تبدیل \( z \) را تشکیل دهید:  

   $$ \Phi(z) = \sum_{k=1}^{N} \left( \frac{T_d A_k}{1 - e^{s_k T_d} z^{-1}} + \frac{T_d B_k}{1 - e^{-s_k T_d} z^{-1}} \right). $$

3. قطب‌ها و صفرهای \( \Phi(z) \) را پیدا کنید و یک تابع سیستم **حداقل‌فاز** \( H(z) \) را از قطب‌ها و صفرهای \( \Phi(z) \) که **داخل دایره واحد** هستند، تشکیل دهید.  

(a) هر مرحله در روش طراحی پیشنهادی را توجیه کنید. به عبارت دیگر، نشان دهید که تابع خودهمبستگی سیستم گسسته‌زمانی حاصل، نسخه نمونه‌برداری‌شده تابع خودهمبستگی سیستم پیوسته‌زمانی است. برای بررسی روش، ممکن است مفید باشد که آن را روی یک سیستم مرتبه اول با پاسخ ضربه‌ای زیر آزمایش کنید:  

   $$ h_c(t) = e^{-\alpha t} u(t). $$
