# مقدمه‌ای بر تجزيه و تحليل سيستم‌ها

## مقدمه

در فصول قبلی توانایی فوق‌العاده تبدیل فوریه در تجزیه و تحلیل سیستم‌های خطی را مشاهده کردیم. این توانایی از آنجا ناشی می‌شود که بسیاری از سیگنال‌ها را می‌توان به صورت ترکیب خطی از توابع نمایی موهومی نوشت و چون توابع وی‍ژه سیستم‌ها نیز به صورت نمایی موهومی بودند، به‌سادگی توانستیم پاسخ به هر ورودی را به دست آوریم.

اما برخی سیگنال‌ها را نمی‌توان بر حسب توابع نمایی موهومی نوشت. در این فصل با معرفی **تبدیل لاپلاس**، که حالت کلی‌تر تبدیل فوریه است، می‌خواهیم بسط جدید سیگنال دلخواه را بر حسب توابع ویژه سیستم‌های LTI (توابع نمایی مختلط) ارائه کنیم.

## 6-1 تبدیل لاپلاس

برای یک سیستم LTI با پاسخ ضربه $h(t)$، دیدیم که پاسخ $y(t)$ سیستم به ورودی نمایی مختلط $e^{st}$ چنین می‌شود:

$$
y(t) = \int_{-\infty}^{+\infty} e^{s(t - \tau)} h(\tau)\, d\tau = e^{st} \int_{-\infty}^{+\infty} e^{-s\tau} h(\tau)\, d\tau = H(s) e^{st}
$$

که در آن:

$$
H(s) = \int_{-\infty}^{+\infty} e^{-s\tau} h(\tau)\, d\tau \tag{6-2}
$$

اگر $s = j\omega$ موهومی محض باشد، انتگرال معادله (6-2) همان **تبدیل فوریه** $h(t)$ بود. اما در حالت عمومی که $s$ عدد مختلط است، انتگرال (6-2) همان **تبدیل لاپلاس** $h(t)$ نامیده می‌شود که مانند تبدیل فوریه نقش مهمی در تجزیه و تحلیل سیگنال‌ها و سیستم‌های LTI دارد.

در حالت کلی، تبدیل لاپلاس سیگنال دلخواه $x(t)$ چنین تعریف می‌شود:

$$
X(s) \triangleq \int_{-\infty}^{+\infty} x(t) e^{-st}\, dt \tag{6-3}
$$

که در آن $s = \sigma + j\omega$ به‌طور کلی یک عدد مختلط است. انتگرال (6-3) همان‌طور که مشاهده می‌شود تابعی از $s$ است. گاهی اپراتور $\mathcal{L}$ را در مورد تبدیل لاپلاس به‌کار می‌برند و این‌گونه نمایش می‌دهند:

$$
x(t) \xleftrightarrow{\mathcal{L}} X(s)
$$

هنگامی که $s = j\omega$ باشد، معادله (6-3) به رابطه زیر تبدیل می‌شود:

$$
X(j\omega) = \int_{-\infty}^{+\infty} x(t) e^{-j\omega t}\, dt \tag{6-4}
$$

این همان **تبدیل فوریه** $x(t)$ است که به صورت $X(\omega)$ قبلاً نشان داده شد. بنابراین تبدیل لاپلاس وقتی $s = j\omega$ باشد، مساوی تبدیل فوریه $x(t)$ خواهد بود:

$$
|X(s)|_{s = j\omega} = \mathcal{F}\{x(t)\} = X(\omega) \tag{6-5}
$$

ارتباط جالب دیگری میان تبدیل لاپلاس و تبدیل فوریه وجود دارد که اگر معادله (6-3) را به صورت معادله (6-6) بنویسیم، می‌توان این ارتباط را پیدا کرد:

$$
X(\sigma + j\omega) = \int_{-\infty}^{+\infty} x(t) e^{-(\sigma + j\omega)t}\, dt \tag{6-6}
$$


# فصل ششم: تبدیل لاپلاس

یا  
$$
X(\sigma + j\omega) = \int_{-\infty}^{+\infty} x(t) e^{-(\sigma + j\omega)t} \, dt \tag{6-7}
$$

عبارت فوق، $X(s)$ را برحسب نقطه‌ای از صفحه مختلط تعریف می‌کند که $s = \sigma + j\omega$ است.  
از آن‌جا که تبدیل لاپلاس همان تبدیل فوریه برای $x(t)e^{-\sigma t}$ است، بسیاری از نتایج سیگنال‌های با مدت زمان محدود را می‌توان از تحلیل فوریه برای $x(t)e^{-\sigma t}$ به‌دست آورد.

---

## مثال

فرض کنید:

$$
x(t) = e^{-at} u(t), \quad a > 0 \tag{6-8}
$$

در این‌صورت تبدیل لاپلاس آن از رابطه‌ی (6-7) چنین محاسبه می‌شود:

$$
X(\omega) = \int_{-\infty}^{+\infty} e^{-at} e^{-j\omega t} \, dt = \frac{1}{j\omega + a}, \quad a > 0 \tag{6-9}
$$

و از معادله (6-7) داریم:

$$
X(s) = \int_{0}^{\infty} e^{-at} e^{-st} \, dt = \int_{0}^{\infty} e^{-(s+a)t} \, dt \tag{6-10}
$$

با فرض $s = \sigma + j\omega$ داریم:

$$
X(\sigma + j\omega) = \int_{0}^{\infty} e^{-(\sigma + a)t} e^{-j\omega t} \, dt \tag{6-11}
$$

می‌بینیم که تابع $x(t)$ همان است و از معادله (6-9) می‌توان نتیجه گرفت که:

$$
X(s) = \frac{1}{\sigma + a + j\omega}, \quad (\sigma + a) > 0 \tag{6-12}
$$

یا بطور معادل:

$$
X(s) = \frac{1}{s + a}, \quad \text{Re}(s) > -a \tag{6-13}
$$

---

## بررسی تابع دیگر:

اگر:

$$
x(t) = e^{at} u(t), \quad a > 0
$$

تبدیل فوریه آن وجود ندارد چون $x(t)$ محدود نیست.  
اما تبدیل لاپلاس آن برای $\text{Re}(s) > a$ برابر است با:

$$
X(s) = \frac{1}{s - a} \tag{6-14}
$$

---

در این حالت، $X(s)$ فقط برای بخش حقیقی بزرگ‌تر از $a$ تعریف شده است.

اگر $x(t) = e^{at}$ باشد و فقط برای $t > 0$ تعریف شود، باز هم تبدیل لاپلاس وجود ندارد زیرا ناحیه همگرایی شامل $\text{Re}(s) = 0$ نخواهد شد.


### مثال (6-3): مقایسه با مثال قبلی انجام دهید

اگر:
$$
x(t) = e^{-at} u(-t) \tag{6-15}
$$

بنویسید:
$$
X(s) = \int_{-\infty}^{0} e^{-at} e^{-st} \, dt = \int_{-\infty}^{0} e^{-(s + a)t} \, dt \tag{6-16}
$$

و با:
$$
X(s) = \frac{1}{s + a}, \quad \text{Re}\{s\} < -a \tag{6-17}
$$

توجه کنید که برای همگرا شدن تبدیل لاپلاس باید داشته باشیم:
$$
\text{Re}\{s\} < -a \tag{6-18}
$$

بنابراین:
$$
\mathcal{L}\{e^{-at} u(-t)\} = \frac{1}{s + a}, \quad \text{Re}\{s\} < -a \tag{6-19}
$$

عبارت تبدیل لاپلاس در هر دو مثال (6-3) و (6-1) یکی است اما ناحیه‌ی همگرایی (ROC) آن‌ها متفاوت است. این موضوع نشان می‌دهد که عبارت تبدیل لاپلاس به تنهایی برای بازسازی تابع زمانی کافی نیست، بلکه ناحیه همگرایی نیز باید مشخص شود.

---

در حالت کلی، بسیاری از سیگنال‌های پرکاربرد، دارای تبدیل لاپلاس یکسان اما نواحی همگرایی متفاوتی هستند. بنابراین تحلیل ناحیه همگرایی ROC بسیار اهمیت دارد.

اگر سیگنال $x(t)$ شامل هر دو قسمت $u(t)$ و $u(-t)$ باشد (مثلاً تابع زوج یا متناوب)، آنگاه تبدیل لاپلاس آن تنها زمانی موجود است که نواحی همگرایی دو قسمت با هم اشتراک داشته باشند.

مثلاً سیگنال ترکیبی از دو تابع مثال‌های (6-1) و (6-3) یعنی:
$$
x(t) = e^{-at}, \quad \text{برای تمام t}
$$

تبدیل لاپلاس آن وجود ندارد چون ROC مربوط به قسمت راست $\text{Re}\{s\} > -a$ و برای قسمت چپ $\text{Re}\{s\} < -a$ است، و هیچ اشتراکی ندارند.

---

**شکل (6-3)** روش نمایش ناحیه همگرایی (ROC) را برای دو حالت نشان می‌دهد:

- a) مربوط به مثال (6-1)
- b) مربوط به مثال (6-3)

> **نکته:** ROC = Region of Convergence


````python
import numpy as np
import matplotlib.pyplot as plt

# مقدار a
a = 2

# تنظیم شکل
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# ================================
# حالت a) Re{s} > -a (مثال 6-1)
axs[0].axvline(-a, color='black', linestyle='--')
axs[0].fill_betweenx([-10, 10], -a, 5, color='gray', alpha=0.4)
axs[0].set_xlim(-5, 5)
axs[0].set_ylim(-10, 10)
axs[0].set_title("a) Re{s} > -a")
axs[0].set_xlabel(r"$\sigma$")
axs[0].set_ylabel(r"$j\omega$")
axs[0].grid(True)
axs[0].text(-a + 0.1, 9, r"$\sigma = -a$", rotation=90)

# ================================
# حالت b) Re{s} < -a (مثال 6-3)
axs[1].axvline(-a, color='black', linestyle='--')
axs[1].fill_betweenx([-10, 10], -5, -a, color='gray', alpha=0.4)
axs[1].set_xlim(-5, 5)
axs[1].set_ylim(-10, 10)
axs[1].set_title("b) Re{s} < -a")
axs[1].set_xlabel(r"$\sigma$")
axs[1].set_ylabel(r"$j\omega$")
axs[1].grid(True)
axs[1].text(-a + 0.1, 9, r"$\sigma = -a$", rotation=90)

plt.tight_layout()
plt.show()
````

````python
## فصل ششم: تبدیل لاپلاس

مثال 1: یک سیگنال متناوب مانند در شکل (1-6)، ممکن است متشکل باشد از چندین u(t) با شیفت داده شده.  
فرض کنید:

$$
x(t) = e^{-t}u(t) + e^{-2t}u(t)
$$

عبارت تبدیل لاپلاس چنین است:

$$
X(s) = \int_{-\infty}^{\infty} \left[e^{-t}u(t) + e^{-2t}u(t)\right] e^{-st} dt
$$

$$
= \int_{0}^{\infty} e^{-(s+1)t} dt + \int_{0}^{\infty} e^{-(s+2)t} dt \tag{6-19}
$$

اگر در هر دو عبارت \( \text{Re}(s) > -1 \) و \( \text{Re}(s) > -2 \) در نظر گرفته شود، نتیجه نهایی برابر می‌شود با:

$$
X(s) = \frac{1}{s+1} + \frac{1}{s+2} \tag{6-20}
$$

برای تعیین ROC (مکان ناحیه همگرایی)، باید مجموع دو تابع خطی را در نظر بگیریم. در حالی که هر کدام از تبدیل‌های لاپلاس دارای یک ناحیه خاصی از s هستند، مجموع نواحی اشتراکی باید بررسی شود.  
برای این تابع داریم:

$$
\mathcal{L}\{x(t)\} = \frac{1}{s+1} + \frac{1}{s+2}, \quad \text{ROC: } \text{Re}(s) > -1 \tag{6-21}
$$

مثال دیگر:

$$
x(t) = e^{2t}u(-t)
$$

عبارت تبدیل لاپلاس آن:

$$
X(s) = \int_{-\infty}^{0} e^{2t} e^{-st} dt = \int_{-\infty}^{0} e^{-(s - 2)t} dt
$$

در این صورت:

$$
X(s) = \frac{1}{s - 2}, \quad \text{Re}(s) < 2 \tag{6-22}
$$

نکته:  
یک تابع مختلط کسری را می‌توان به صورت:

$$
X(s) = \frac{2s + 3}{s^2 + 2s + 1} = \frac{N(s)}{D(s)} \tag{6-23}
$$

کسی است که در آن \( N(s) \) و \( D(s) \) توابع چندجمله‌ای حقیقی از \( s \) هستند. در فصل بعدی نشان خواهیم داد که اگر \( X(s) \) ترکیب توابع کسری با صورت و مخرج چندجمله‌ای باشد، می‌توان آن را به عنوان پاسخ ضربه‌ای یک سیستم LTI که با معادلات دیفرانسیل خطی تعریف شده، در نظر گرفت.

---

<sup>1</sup> S-Plane  
<sup>2</sup> Rational
````

````python
## مقدمه‌ای دیگر بر تحلیل سیستم‌ها

تبدیل \( H(s) \) مربوط به کسری شامل صورت‌های \( N(s) \) (صورت‌ها) و مخرج‌های \( D(s) \) (مخرج‌ها) تبدیل هستند و محدوده‌ای در ROC آن ناحیه مشخصی می‌تواند برای مثال در شکل (6-6) رسم شود.

### ROC مربوط به شکل (6-6):

**شکل 6-6**: نمودار صفر و قطب، و ROC برای یک تابع \( X(s) \)

علامت × جهت نمایش قطب و علامت ○ جهت نمایش صفر بکار گرفته شده‌اند.

مثال 2: فرض کنید:

$$
x(t) = -\frac{4}{3}e^{-t}u(t) + \frac{1}{3}e^{-2t}u(t) \tag{6-7}
$$

تبدیل لاپلاس \( x(t) \) برای سیگنال بالا طبق تعریف تبدیل لاپلاس به صورت زیر خواهد بود:

$$
X(s) = \frac{4}{s+1} + \frac{1}{s+2}, \quad \text{Re}(s) > 2 \tag{6-8}
$$

ناحیه همگرایی از نظر عملی تک تک توان‌هایی سمت راست محور مختلط بوده که همراهی تبدیل برای زمانی‌های مثبت در نظر گرفته شده است. توجه کنید که همگرایی (معادله 6-8) وابسته به تابع \( x(t) \) می‌باشد.

$$
X(s) = \frac{(s - 1)^2}{(s+1)(s-2)} \quad \text{Re}(s) > 2 \tag{6-9}
$$

### رسم صفر و قطب، و ROC برای \( X(s) \) بصورت شکل (6-6) می‌باشد.

مطابق آنچه که در بالا نشان داده شد، هر تبدیل شامل صفرها و قطب‌هایی است. همگرایی در ROC شامل تمام نقاط s در صفحه مختلط است که موجب همگرایی انتگرال تبدیل لاپلاس می‌گردد.  
وجود ناحیه‌ای که در آن تبدیل لاپلاس وجود داشته باشد مستلزم این است که تابع نمایی زمان پیوسته از نوع نمایی نزولی باشد.

---

<sup>1</sup> Zeros  
<sup>2</sup> Poles
````

````python
import matplotlib.pyplot as plt
import numpy as np

# مقدار قطب‌ها و صفرها
poles = [-1, -2]
zeros = []

# ناحیه ROC (Re(s) > -2)
roc_left = -2
roc_right = -1

# رسم نمودار
fig, ax = plt.subplots(figsize=(6, 6))

# رسم قطب‌ها (×)
for p in poles:
    plt.plot(np.real(p), np.imag(p), 'rx', markersize=12, label='Pole' if p == poles[0] else "")

# رسم صفرها (○)
for z in zeros:
    plt.plot(np.real(z), np.imag(z), 'go', markersize=10, label='Zero' if z == zeros[0] else "")

# رسم ناحیه ROC با رنگ خاکستری
plt.axvspan(roc_left, roc_right, color='gray', alpha=0.3, label='ROC')

# اضافه کردن محورهای x و y
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# برچسب‌ها و عنوان
plt.xlabel('Re(s)')
plt.ylabel('Im(s)')
plt.title('Pole-Zero Plot with ROC')

# محدودیت‌های نمودار
plt.xlim(-4, 2)
plt.ylim(-3, 3)

# گرید و legend
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
````

### فصل ششم: تبدیل لاپلاس

در این مثال دو قطب \( s = -1 \) و \( s = -2 \) در صفحه s قرار دارند و ناحیه همگرایی (ROC) سمت راست بزرگترین قطب قرار گرفته است.

گفته شد برای بسیاری از تبدیل‌های لاپلاس معمولی (مثل پاسخ سیستم‌ها)، ناحیه ROC وجود دارد و شکل آن می‌تواند هموار باشد تا بتوان به راحتی مشخصه قطب‌ها، ناحیه ROC، و ساختار تابع تبدیل را تحلیل کرد (مانند شکل زیر).

### ۶-۴ خواص ناحیه همگرایی برای تبدیل لاپلاس

تبدیل لاپلاس \( X(s) \) فقط وقتی وجود دارد که تبدیل انتگرالی در تعریف آن دارای حد متناهی باشد. این اتفاق زمانی می‌افتد که تابع \( x(t) \) با سرعت کافی به سمت صفر میل کند و همزمان \( s \) به گونه‌ای انتخاب شود که انتگرال:

\[
X(s) = \int_{-\infty}^{\infty} x(t) e^{-st} dt
\]

همگرا باشد. شکل زیر چند حالت مختلف از ناحیه ROC برای سیگنال‌هایی با خواص متفاوت را نشان می‌دهد.


````python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# تنظیمات مشترک
poles = [[-1, -2], [-2, -3], [-1, -3]]
rocs = [[-1, 3], [-3, -2], [-3, -1]]  # [left, right] for each plot
titles = ["ROC سمت راست", "ROC سمت چپ", "ROC بین دو قطب"]

for i in range(3):
    ax = axs[i]
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    ax.set_xlim(-4, 4)
    ax.set_ylim(-2, 2)
    ax.set_aspect('equal', adjustable='box')
    ax.set_title(titles[i])
    ax.set_xlabel("Re")
    ax.set_ylabel("Im")

    # قطب‌ها
    for p in poles[i]:
        ax.plot(np.real(p), 0, 'rx', markersize=12)

    # ROC
    ax.axvspan(rocs[i][0], rocs[i][1], color='gray', alpha=0.4)

plt.tight_layout()
plt.show()
````

### مقدمه‌ای بر تجزیه و تحلیل سیستم‌ها

خاصیت **(2-3)**: اگر \( x(t) \) در حوزه زمان محدود (محدوده‌ی \([0,T]\)) و اگر حافظه‌ی تابع پاسخ از ورودی \( h(t) \) محدود باشد، در این‌صورت \( y(t) = \int_{-\infty}^{\infty} h(t-\tau)x(\tau)d\tau \) فقط در محدوده‌ای شامل \( [0,T] \) غیر صفر خواهد بود (شکل 2-5) به همین دلیل گفته می‌شود که:

> یک سیگنال دارای **حافظه محدود** است اگر فقط در ناحیه‌ای غیر صفر و داخل آن غیر صفر باشد (شکل 2-5) توجه کنید.

در صورتی که \( x(t) \) بطور مطلق قابل انتگرال‌گیری از \( (-\infty,\infty) \) باشد، به انتگرال‌پذیر بودن معروف است.

بنابراین

\[
\int_{-\infty}^{\infty} |x(t)| dt < \infty \tag{2-31}
\]

شکل (2-5): سیگنال با دوره محدود به محدوده‌ی \( [0,N] \) و دیگر نواحی صفر است. در حالی که سیگنال در محدوده غیر صفر دارای تغییرات سریع است.

در این‌جا اگر \( x(t) = 0 \) در ناحیه زمانی خارج از بازه‌ای دلخواه باشد، آن‌گاه \( \Re[x(t)] = 0 \) هم در ناحیه همگامی است.

بنابراین می‌توان نوشت:

\[
\int_{-\infty}^{\infty} |x(t)| e^{-a t} dt = \int_{0}^{T} |x(t)| e^{-a t} dt \tag{2-32}
\]

همانطور که گفته شد \( x(t) \) در محدوده‌ی \( [0,T] \) غیر صفر است و بنابراین تابعی قابل اندازه‌گیری خواهد بود. در حالتی که \( a > 0 \) و با توجه به محدود بودن تابع نمایی \( e^{-a t} \) نسبت به تابع \( |x(t)| \)، می‌توان نوشت:

\[
\int_{-\infty}^{\infty} |x(t)| e^{-a t} dt < \int_{0}^{T} |x(t)| dt < \infty \tag{2-33}
\]

با توجه به اینکه انتقال سمت راست محدود است، نتیجه می‌گیریم که حتی اگر \( \sigma_1 > 0 \)، ولی بازه‌ی همگرایی همیشه از راست امتداد می‌یابد اگر \( \sigma_2 = 0 \) به پایان آید.

````python
### فصل ششم: تبدیل لاپلاس

و دوباره می‌نویسیم که اگر \( \Re(s) = \sigma > 0 \)، مقدار مطلق انتگرال سمت راست برابر است با:

\[
\int_{T}^{\infty} |h(t)| e^{-\sigma t} dt < e^{-\sigma T} \int_{T}^{\infty} |h(t)| dt \tag{6-25}
\]

همگرایی تابع همانند تابع \( h(t) \) در شکل (6-6) قابل نمایش است. همان‌طور که می‌بینیم، تابع \( h(t) \) در ناحیه \( t > T \) دارای دامنه‌ی کوچک و دنباله‌ای در حال کاهش است، در نتیجه می‌توانیم آن را انتگرال‌پذیر در نظر بگیریم.

شکل (6-6): خاصیت همگرایی سیگنال برای تابعی که ناحیه‌ای غیر صفر دارد.

همان‌طور که شکل بالا نشان می‌دهد، برای این‌که انتگرال تبدیل لاپلاس وجود داشته باشد، نیاز است که تابع به اندازه‌ی کافی کاهش یافته باشد. اگر چنین باشد، آنگاه می‌توان نتیجه گرفت که:

> توجه داشته باشید که احتمالاً با محدود بودن رشد توابع تابع \( h(t) \) ارتباط داشته باشد.

خاصیت **(6-26)**: اگر \( h(t) = 0 \) برای \( t < 0 \)، یعنی تابع سمت راست یا causal باشد، و در ناحیه‌ی \( [0, \infty) \) دارای همگرایی نمایی باشد، آنگاه تبدیل لاپلاس آن نیز وجود دارد.

توجه کنید همان‌طور که در خاصیت قبل گفته شد، اگر \( \exists M > 0 \) و \( a > 0 \) به طوری که:

\[
|h(t)| < M e^{a t} \Rightarrow |h(t)| e^{-\sigma t} < M e^{-(\sigma - a) t} \tag{6-26}
\]

اگر \( \sigma > a \) باشد، آنگاه سمت راست دارای همگرایی خواهد بود و در نتیجه می‌توان نوشت که تبدیل لاپلاس آن وجود دارد. بنابراین اگر:

\[
\int_{0}^{\infty} |h(t)| e^{-\sigma t} dt < \infty \tag{6-27}
\]

ناحیه همگرایی چون \( \Re(s) > \sigma \) است.

بنابراین اگر \( \sigma_1 < \sigma_2 \) باشد، بازه همگرایی گسترده‌تر است.

> بنابراین اگر \( \Re(s) = \sigma > 0 \) باشد، مقدار مطلق انتگرال‌پذیر است، چون با افزایش \( t \)، \( e^{-\sigma t} \) به سرعت کاهش می‌یابد. که در شکل (6-7) نمایش داده شده است.

![شکل 6-7](example)

شکل (6-7): سیگنال راست‌طرفه

یک مثال از این نوع تابع \( h(t) = e^{a t} \) است، اما وقتی که تبدیل لاپلاس تنها برای مقادیر مشخصی از \( s \) همگرا باشد که ما با \( \Re(s) > a \) آن را نمایش می‌دهیم. چنین شرایطی:

\[
\int_{0}^{\infty} |h(t)| e^{-\sigma t} dt < \infty \tag{6-28}
\]

ناحیه همگرا چون \( \Re(s) > a \) است.
````

````python
### مقدمه‌ای بر تجزیه و تحلیل سیستم‌ها

شکل (6-6) اگر \( x(t) \) بطور مطلق انتگرال‌پذیر باشد، در صورت \( x(t) \approx 0 \) برای \( t > T_0 \)، با \( T_0 < \infty \)، به آن راست‌برد گفته می‌شود.

همچنین می‌توان نوشت:

\[
\int_{-\infty}^{\infty} |x(t)|e^{-\sigma t} dt = \int_{-\infty}^{T_0} |x(t)|e^{-\sigma t} dt
\]

و چون \( T_0 \) محدود است، معادله (6-7) مجموعه‌ای محدود که قسمت سمت راست معادله (6-6) محدود است، بدان معنی است. بنابراین مقدار مطلق انتگرال نیز محدود می‌شود. توجه کنید که چون:

\[
e^{-\sigma t} > e^{-\sigma T_0}, \quad \text{برای } t < T_0
\]

چون \( x(t) = 0 \) برای همه‌ی سمت راست \( T_0 \) بزرگتر از \( T_0 \) نیست و بنابراین:

\[
\int_{-\infty}^{T_0} |x(t)|e^{-\sigma t} dt \leq \int_{-\infty}^{T_0} |x(t)|e^{-\sigma T_0} dt = e^{-\sigma T_0} \int_{-\infty}^{T_0} |x(t)| dt
\]

اگر تابع \( x(t) \) (شکل پایین) مطلقاً انتگرال‌پذیر باشد، آنگاه قسمت سمت راست معادله فوق محدود خواهد بود، بنابراین \( x(t) \) در ناحیه همگرایی، برای \( \Re(s) > \sigma \)، تبدیل لاپلاس دارد.

شکل (6-9) توجه کنید: این تابع ممکن است اطلاعات عملی داشته باشد.

---

#### شکل 6-9: سیگنال چپ‌برد

خاصیت 6-7: اگر \( x(t) \) در جهت \( -\infty \) محدود باشد، به آن سیگنال چپ‌برد گفته می‌شود. شکل بالا گویای این موضوع است. سیگنال چپ‌برد سیگنالی است که فقط در بازه‌ی منفی زمان غیرصفر باشد.

یک سیگنال چپ‌برد نیز می‌تواند مطلقاً انتگرال‌پذیر باشد و بنابراین می‌تواند تبدیل لاپلاس داشته باشد، همانند شکل بالا.

بعضی از سیگنال‌ها نه راست‌برد و نه چپ‌برد هستند و به آنها سیگنال‌های دوطرفه گفته می‌شود. مانند \( x_1(t) \) و \( x_2(t) \) در شکل پایین. در ادامه مشاهده می‌کنید که در آن ناحیه هر دو سیگنال \( x_1(t) \) و \( x_2(t) \) همزمان موجود باشند.
````

````python
## فصل ششم: تبدیل لاپلاس

![شکل 6-1](attachment:image.png)
شکل (6-1): سیگنال دو جمله‌ای در حوزه‌های زمان، فرکانس فوریه و فرکانس لاپلاس نمایش داده شده است.

از شکل 6-1 می‌توان دید که تابع همگرا \( x(t) \) شامل قطعه‌ای از سیگنال است که در آن‌ها پاسخ فوریه وجود ندارد، اما پاسخ لاپلاس وجود دارد. فرض شامل قسمت‌هایی باشد که شامل سیگنال \( x(t) \) است و \( X(s) \) پاسخ لاپلاس آن. تابع همگرا \( x_1(t) \) شامل ناحیه‌ای از همپوشانی با \( x(t) \) است که در آن‌ها پاسخ فوریه \( X_1(j\omega) \) موجود است. ناحیه‌ی همپوشانی دو تابع و تبدیل آن‌ها، ناحیه‌ی اشتراک در تصویر (شکل 6-2) را تشکیل می‌دهد. تبدیل لاپلاس عمومی‌تر از فوریه است.

تابع \( x_2(t) \) که شامل قسمت‌هایی از \( x(t) \) است، که فوریه همگرا نیست، اما لاپلاس همگراست. شکل (6-2) برای مثال نشان‌دهنده‌ی یک خط مشترک ممکن بین سیگنال‌هاست.

![شکل 6-2](attachment:image.png)
شکل (6-2): نواحی همگرای فوریه \( x_1(t) \) و لاپلاس \( x_2(t) \) که همپوشانی دارند، می‌توانند به‌صورت مشترک تبدیل شوند.

اما ممکن است تابع \( x(t) \) دارای هر دو ناحیه‌ی همپوشانی مشترک باشد. وجود چنین تابعی می‌تواند تبدیل لاپلاس را برتر از فوریه و همگراتر نماید.

مثال زیر فرض کنید:

\[
x(t) =
\begin{cases}
e^{-2t} & 0 < t < 1 \\
0 & \text{else}
\end{cases}
\]

تبدیل لاپلاس آن به صورت زیر است:

\[
X(s) = \int_0^1 e^{-2t}e^{-st} dt = \frac{1 - e^{-(s+2)}}{s + 2}
\]

چون طول تابع \( x(t) \) محدود است، از خاصیت (6-2) متوجه می‌شویم که ناحیه همگرایی شامل تمام صفحه‌ی s است.
````

````python
### مقدمه‌ای بر تجزیه و تحلیل سیستم‌ها

۱) \( x(t) \) دارای تعداد نامتناهی صفر است که توان‌های \( s = \frac{j2\pi k}{T} \)، \( k = 0, \pm1, \pm2, \ldots \) می‌باشند.

بنابراین:

\[
e^{st} = e^{j2\pi kt/T}
\tag{1-6-9}
\]

\[
s = a + \frac{j2\pi k}{T}, \quad k = 0, \pm1, \pm2, \ldots
\tag{1-6-10}
\]

تمرین: در اصل همپوشانی، شرط را به وضعیت به‌خصوصی محدود کنید مثل \( x(t) = e^{st} \). شکل (1-6-11) را داشته باشیم:

\[
x(t) = e^{st}
\tag{1-6-11}
\]

(در این قسمت نمودار باید نمایش داده شود)

در شکل (1-6-11)، سیگنال \( x(t) = e^{bt} \)، بازه \( b < t < 0 \) رسم شده است.

در این شکل، دو طرف b در جهت مثبت t بازه‌ای را دو سیگنال راست‌رو و چپ‌رو تشکیل می‌دهند که می‌توان نوشت:

\[
x(t) = e^{bt} u(t) + e^{bt} u(-t)
\tag{1-6-12}
\]

اما داریم:

\[
\int_{-\infty}^{\infty} e^{bt} dt =
\begin{cases}
\frac{1}{s}, & \mathbb{R}\{s\} > 0 \\
\frac{-1}{s}, & \mathbb{R}\{s\} < 0
\end{cases}
= b
\tag{1-6-13}
\]

اگرچه سیگنال بالایی در دو قسمت سیگنال وجود دارد (یعنی خاصیت خطی دارد)، ولی اگر \( b < 0 \) باشد و ناحیه همپوشانی داشته باشیم، شرط مشترک داریم. بنابراین تبدیل لاپلاس خواهیم داشت. اگر \( b > 0 \) باشد، تبدیل لاپلاس نداریم.

تمرین: رابطهٔ زیر را اثبات کنید:

\[
X(s) = \frac{-2b}{(s - b)^2}, \quad \mathbb{R}\{s\} < -b
\tag{1-6-14}
\]
````

````python
### فصل ششم: تبدیل لاپلاس

- تمام صفحه \( s \)  
  (برای سیگنال‌های زمان نامحدود)  
- سمت چپ صفحه \( s \)  
  (برای سیگنال‌های یک‌طرفه)  
- سمت راست صفحه \( s \)  
  (برای سیگنال‌های چپ‌طرفه)  
- نوار عمودی صفحه \( s \)  
  (برای سیگنال‌های دوطرفه)

در تمامی حالات، ناحیه همگرایی از یک طرف توسط یک نقطه کاملاً بسته (نشان داده شده با دایره‌ی توپر بر روی خط مرز) محدود شده و از طرف دیگر با یک خط صاف باز (بدون دایره) که کاملاً خط تقاطع نمی‌باشد، ختم می‌شود.

آخرین نقطه در سمت چپ، شروع شده و به بینهایت مثبت (\(+\infty\)) ختم می‌شود.

مثل (1-6-7)، سیگنال زیر را به‌صورت حاصل‌ضرب کسر خطی داریم. رابطه (1-6-9) از نظر تجزیه:

\[
X(s) = \frac{(s + 1)(s + 2)}{(s - 1)(s - 2)}
\tag{1-6-9}
\]

همان‌طور که در شکل (1-6-13) دیده شده است (چهار حالت قطب‌ها و صفرها)، ناحیه همگرایی برای عبارت فوق در شکل‌های مختلف اثر گذار خواهد بود.

- برای سیگنال سمت راست: ناحیه همگرایی به سمت چپ خواهد بود.  
- برای سیگنال سمت چپ: ناحیه همگرایی به سمت راست خواهد بود.  
- برای سیگنال دو طرفه: نوار محدودی بین دو قطب (بین صفرها) خواهد بود.

---

#### شکل (1-6-13): تبدیل مختلط و ناحیه همگرایی سیگنال مثال (1-6-9)

(چهار نمودار قطب و صفر در زیر با استفاده از matplotlib رسم می‌شوند)
````

````python
## مقدمه‌ای بر تجزیه و تحلیل سیستم‌ها

اگر سیگنال شکلی چون موجود باشد (شکل (1-6-14))، ناحیه‌ی همگرایی فقط خواهد بود. همچنین اگر سیگنال \( h(t) \) به‌صورت زیر باشد که در آن دو تابع ضربه‌ی دیراک در زمان‌های خاصی قرار دارند:

\[
h(t) = a\delta(t - T) + 2\delta(t - 3T)
\tag{1-6-6}
\]

نمودار آن به‌صورت زیر خواهد بود:

### شکل (1-6-14): سیگنال \( h(t) \) مربوط به مثال (1-6-6)

(کد پایتون برای رسم نمودار در سلول کد جداگانه 👇)

---

### تبدیل لاپلاس عبارت بالا برابر است با:

\[
\mathcal{L} \{ a\delta(t - T) + 2\delta(t - 3T) \} = ae^{-Ts} + 2e^{-3Ts}
\tag{1-6-7}
\]

تبدیل لاپلاس دوم نیز به همین صورت است، این بار داریم:

\[
H(s) = ae^{-Ts} + a^2e^{-2Ts}
\tag{1-6-8}
\]

اگر:

\[
H(s) = a^2\left[1 + ae^{-Ts}\right]
\tag{1-6-9}
\]

---

### اگر شکل مربوط به صفرها و قطب‌های تابع را رسم کنیم، محور مختلط s در نظر گرفته می‌شود:

\[
a^2e^{-2Ts} = -1
\Rightarrow
e^{-2Ts} = \frac{-1}{a^2}
\Rightarrow
k = \frac{-1}{a^2}
\tag{1-6-10}
\]

با گرفتن لگاریتم طبیعی از دو طرف داریم:

\[
-2Ts = \ln\left(\frac{-1}{a^2}\right)
\tag{1-6-11}
\]

اگر بنویسیم \( s = \sigma + j\omega \) آنگاه:

\[
\sigma = -\frac{1}{2T} \ln\left(\frac{1}{a^2}\right)
\tag{1-6-12}
\]
````

````python
## فصل ششم: تبدیل لاپلاس

معادله:

$$
-2T\omega = -k\pi \\
\Rightarrow \omega = \frac{k\pi}{2T} \quad \text{k = ±1, ±3, ...} \tag{6-91}
$$

توجه کنید که همان‌طور که در رسم نمودار صفر-قطب مشاهده می‌شود:

![نمودار صفر-قطب](6-92.png)  
*شکل (6-13): نمودار صفر و قطب مثال (6-90)*

---

### 6-2 معکوس تبدیل لاپلاس

قبلاً دیدیم که تبدیل لاپلاس تابع \( x(t) \) داده شده به‌صورت:

$$
X(s) = \mathcal{L}\{x(t)\} = \int_{0^-}^{\infty} x(t)e^{-st} dt \tag{6-93}
$$

اگر \( s = \sigma + j\omega \)، پس:

$$
X(\sigma + j\omega) = F(j\omega) = \int_{0^-}^{\infty} x(t)e^{-(\sigma + j\omega)t} dt \tag{6-94}
$$

که تبدیل فوریه یک سیگنال ضرب شده در \( e^{-\sigma t} \) است. بنابراین می‌توانیم تبدیل لاپلاس معکوس را به‌صورت:

$$
x(t) = \mathcal{L}^{-1} \{X(s)\} = \frac{1}{2\pi j} \int_{\sigma - j\infty}^{\sigma + j\infty} X(s)e^{st} ds \tag{6-95}
$$

و اگر \( s = \sigma + j\omega \) بنویسیم:

$$
x(t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} X(\sigma + j\omega)e^{(\sigma + j\omega)t} d\omega \tag{6-96}
$$

که مشابه تبدیل فوریه معکوس است ولی با تبدیل لاپلاس که روی خط \( \text{Re}(s) = \sigma \) تعریف می‌شود.

توجه: خط \( \text{Re}(s) = \sigma \) باید در ناحیه ROC قرار داشته باشد.

معمولاً محاسبه تبدیل لاپلاس معکوس از فرمول (6-95) دشوار است، و در روش‌های دیگر مثل استفاده از جدول تبدیل یا کسر جزئی راحت‌تر است. این روش‌ها معمولاً در بخش‌های بعدی نشان داده می‌شوند.
````

````python
## ادامه فصل ششم: تبدیل لاپلاس — بسط کسر جزئی

معادله:

$$
X(s) = \sum_k \frac{A_k}{s + a_k} \tag{6-97}
$$

در تبدیل لاپلاس وقتی ناحیه همگرایی \( X(s) \) مشخص می‌شود، می‌توان از **بسط کسر جزئی** برای ساده‌سازی معکوس تبدیل استفاده کرد. مثال زیر را در نظر بگیرید:

مثال (6-10): معکوس تبدیل لاپلاس زیر را بیابید:

$$
X(s) = \frac{1}{(s + 1)(s + 2)} \tag{6-98}
$$

برای حل از بسط کسر جزئی استفاده می‌کنیم:

$$
X(s) = \frac{A}{s + 1} + \frac{B}{s + 2} \tag{6-99}
$$

برای محاسبه \( A \) و \( B \)، رابطه (6-99) را به صورت زیر در می‌آوریم:

$$
\frac{1}{(s+1)(s+2)} = \frac{A}{s+1} + \frac{B}{s+2} \Rightarrow 1 = A(s+2) + B(s+1)
$$

با قرار دادن \( s = -1 \):

$$
1 = A(-1 + 2) + B(0) \Rightarrow A = 1 \tag{6-100}
$$

و با قرار دادن \( s = -2 \):

$$
1 = A(0) + B(-2 + 1) \Rightarrow B = -1 \tag{6-101}
$$

در نتیجه:

$$
X(s) = \frac{1}{s + 1} - \frac{1}{s + 2} \tag{6-102}
$$

---

با توجه به ناحیه همگرایی (ROC)، نوع تابع \( x(t) \) را تعیین می‌کنیم. در شکل زیر، ۳ حالت مختلف از ROC را بررسی می‌کنیم:

---

### شکل (6-15): ناحیه همگرایی برای پاسخ‌های مختلف

- (a) ROC: Re(s) > -1 → مربوط به پاسخ علّی (causal)
- (b) ROC: -2 < Re(s) < -1 → مربوط به پاسخ دوطرفه (two-sided)
- (c) ROC: Re(s) < -2 → مربوط به پاسخ غیرعلّی (anti-causal)

````

````python
## ادامه فصل ششم: ویژگی‌های تبدیل لاپلاس

### مثال (6-81)

اگر:
$$
x(t) = -e^{-t}u(t) - e^{-2t}u(-t)
$$

آنگاه:
$$
X(s) = \frac{1}{(s+1)(s+2)}, \quad -2 < \text{Re}(s) < -1
$$

---

## 6-4: خواص تبدیل لاپلاس

یکی از مزیت‌های مهم تبدیل لاپلاس، علاوه بر گسترش تعریف تبدیل فوریه، این است که خواص ساده‌ای برای عملیات‌های پایه دارد و تحلیل سیستم‌ها را ساده‌تر می‌کند.

اگر \( x_1(t) \xleftrightarrow{\mathcal{L}} X_1(s) \) با ناحیه همگرایی \( R_1 \)،  
و \( x_2(t) \xleftrightarrow{\mathcal{L}} X_2(s) \) با ناحیه همگرایی \( R_2 \)،

آنگاه برای ترکیب خطی:
$$
a_1 x_1(t) + a_2 x_2(t) \xleftrightarrow{\mathcal{L}} a_1 X_1(s) + a_2 X_2(s)
$$

و ناحیه همگرایی \( R \) برابر است با اشتراک ناحیه‌های \( R_1 \cap R_2 \)،  
مگر اینکه هم‌پوشانی روی محور موهومی صفر شود.

---

### مثال (6-9)

اگر:

- \( x_1(t) = e^{-t}u(t) \Rightarrow X_1(s) = \frac{1}{s+1}, \quad \text{Re}(s) > -1 \)
- \( x_2(t) = -e^{-2t}u(-t) \Rightarrow X_2(s) = \frac{-1}{s+2}, \quad \text{Re}(s) < -2 \)

در نتیجه:

$$
x(t) = x_1(t) + x_2(t) = e^{-t}u(t) - e^{-2t}u(-t)
$$

و تبدیل لاپلاس آن:
$$
X(s) = \frac{1}{s+1} + \left( \frac{-1}{s+2} \right) = \frac{(s+2) - (s+1)}{(s+1)(s+2)} = \frac{1}{(s+1)(s+2)}
$$

اما ناحیه همگرایی:
$$
R = (-2 < \text{Re}(s) < -1)
$$

که اشتراک دو ناحیه قبلی است.
````

````python
## فصل ششم: تبدیل لاپلاس

### شکل (6-10)

در این شکل سه حالت مختلف برای اشتراک نواحی همگرایی سه سیگنال نشان داده شده‌اند:

- (a) نواحی همگرایی با اشتراک مشترک
- (b) نواحی همگرایی با اشتراک جزئی
- (c) نواحی همگرایی کاملاً پوشا

اگر:
- \( X_1(s) \) دارای ROC = \( R_1 \)
- \( X_2(s) \) دارای ROC = \( R_2 \)
- \( X_3(s) \) دارای ROC = \( R_3 \)

آنگاه:
$$
x(t) = \sum_{k} a_k x_k(t) \xleftrightarrow{\mathcal{L}} X(s) = \sum_{k} a_k X_k(s), \quad \text{ROC} = \bigcap_{k} R_k
$$

---

## 6-4-2: انتقال در حوزه زمان

اگر:
$$
x(t) \xleftrightarrow{\mathcal{L}} X(s), \quad \text{ROC} = R
$$

آنگاه:
$$
x(t - t_0) \xleftrightarrow{\mathcal{L}} e^{-s t_0} X(s), \quad \text{ROC} = R
$$

> ناحیه همگرایی در انتقال زمانی تغییر نمی‌کند، چون ضریب \( e^{-st_0} \) صرفاً یک ضرب‌کننده نمایی در حوزه لاپلاس است و تأثیری بر همگرایی تابع ندارد.

---

## 6-4-3: انتقال در حوزه s

اگر:
$$
x(t) \xleftrightarrow{\mathcal{L}} X(s), \quad \text{ROC} = R
$$

آنگاه:
$$
e^{a t} x(t) \xleftrightarrow{\mathcal{L}} X(s - a), \quad \text{ROC} = R + a
$$

> این خاصیت نشان می‌دهد که ضرب سیگنال در \( e^{at} \) در حوزه زمان، موجب انتقال نمودار لاپلاس در محور حقیقی s به سمت راست به اندازه \( a \) می‌شود.

````

````python
# فصل ششم: تبدیل لاپلاس

پلکانی ضعیفتر از تعیین ناحیه همگرایی جدید به ذات خود که معلوم می‌کند مکان صفرها و قطب‌ها چگونه است، ناحیه همگرایی در صفحه مختلط \( s \) تعیین می‌شود.

اگر سیگنال \( x(t) \) تبدیل لاپلاس داشته باشد و ناحیه همگرایی آن \( R_1 \) باشد و سیگنال دیگر \( x_2(t) \) ناحیه همگرایی \( R_2 \) داشته باشد، تبدیل لاپلاس مجموع دو سیگنال، ناحیه همگرایی آن برابر تقاطع این دو ناحیه خواهد بود.

---

## خاصیت خطی تبدیل لاپلاس:

اگر:
$$
x_1(t) \xleftrightarrow{\mathcal{L}} X_1(s), \quad \text{ROC} = R_1
$$
و
$$
x_2(t) \xleftrightarrow{\mathcal{L}} X_2(s), \quad \text{ROC} = R_2
$$

آنگاه:
$$
a_1 x_1(t) + a_2 x_2(t) \xleftrightarrow{\mathcal{L}} a_1 X_1(s) + a_2 X_2(s), \quad \text{ROC} = R_1 \cap R_2
$$

نکته: ناحیه همگرایی حاصل، حاصل تقاطع ناحیه‌های همگرایی \( X_1(s) \) و \( X_2(s) \) است؛ به شرطی که هیچگونه تضاد بین قطب‌ها وجود نداشته باشد.

---

### مثالی از اعمال خاصیت خطی:

فرض کنید:
$$
X_1(s) = \frac{s+1}{s+2}
$$
با ناحیه همگرایی:
$$
\text{Re}(s) > -2
$$

و
$$
X_2(s) = \frac{s+2}{s+1}
$$
با ناحیه همگرایی:
$$
\text{Re}(s) > -1
$$

در این صورت ناحیه همگرایی \( X_1(s) + X_2(s) \) برابر با اشتراک دو ناحیه (یعنی Re(s) > -1) خواهد بود.

---

## خاصیت مشتق‌گیری در حوزه زمان:

اگر:
$$
x(t) \xleftrightarrow{\mathcal{L}} X(s), \quad \text{ROC} = R
$$

آنگاه:
$$
\frac{d}{dt}x(t) \xleftrightarrow{\mathcal{L}} s X(s), \quad \text{ROC} = R
$$

نکته: ناحیه همگرایی در مشتق سیگنال همان ناحیه همگرایی سیگنال اصلی است، البته به شرطی که \( x(t) \) در صفر محدود باشد و تعداد بینهایت زیادی دیراک در صفر وجود نداشته باشد.

---

> ✅ در این صفحه هیچ نمودار تصویری خاصی وجود نداشت.  
> ✅ فرمول‌ها کاملاً به صورت LaTeX و درست چیده شده‌اند.
````

````python
# مقدمه‌ای بر تبدیل لاپلاس

مثال دیگر: توجه کنید که تبدیل لاپلاس تابع \( x(t) = e^{-at}u(t) \) برابر است با:
$$
X(s) = \frac{1}{s+a}
$$
به شرطی که ناحیه همگرایی \( \text{Re}(s) > -a \) باشد. همچنین تبدیل لاپلاس مشتق \( \frac{d}{dt}x(t) \) برابر است با:

---

## فرم اصلی مشتق‌گیری در حوزه \( s \):

$$
X(s) = \int_{-\infty}^{\infty} x(t)e^{-st} dt
$$
(رابطه 10-1)

با مشتق‌گیری از دو طرف نسبت به \( s \):

$$
\frac{dX(s)}{ds} = \int_{-\infty}^{\infty} \left( -t x(t) \right) e^{-st} dt
$$
(رابطه 10-2)

یعنی:
$$
\mathcal{L}\{t x(t)\} = -\frac{dX(s)}{ds}, \quad \text{ROC} = R
$$
(رابطه 10-3)

---

## نکته:
در نتیجه، زمانی که مشتق سیگنال در حوزه \( s \) گرفته می‌شود، تصویر در شکل کلی تغییر می‌کند ولی ناحیه همگرایی تغییر نمی‌کند (مگر در صورت خاص).

---

### مثال:

یافتن تبدیل لاپلاس تابع \( t e^{-at} u(t) \).

رابطه اصلی:
$$
e^{-at} \xleftrightarrow{\mathcal{L}} \frac{1}{s+a}
$$
(با ناحیه همگرایی \( \text{Re}(s) > -a \))

حال اگر از رابطه (10-3) استفاده کنیم:
$$
t e^{-at} \xleftrightarrow{\mathcal{L}} -\frac{d}{ds}\left( \frac{1}{s+a} \right)
$$

با تقسیم‌گیری داریم:
$$
-\frac{d}{ds}\left( \frac{1}{s+a} \right) = \frac{1}{(s+a)^2}
$$
(رابطه 10-4)

---

> ✅ توی این صفحه هیچ نمودار خاصی وجود نداشت.  
> ✅ همه فرمول‌ها با دقت به صورت LaTeX گذاشته شدند تا تو Jupyter Notebook بدون مشکل نمایش داده بشه.
````

````python
# فصل ششم: تبدیل لاپلاس

در استواری حاصل ضرب کانولوشن \( h(t) * h(t) \) وجود خواهد داشت چون ناحیه همگرایی مشترک بین \( h(t) \) و \( h(t) \) وجود دارد.

## مثال (6-16)
شکل زیر شامل دامنه سیستم معکوس مربوط به سیستمی که تابع انتقال و ناحیه همگرایی تبدیل لاپلاس پاسخ ضربه آن به صورت زیر است:

$$
H(s) = \frac{(s-1)(s+2)}{(s+1)(s-2)}, \quad Re(s) > 2
$$

![نمودار قطب و صفر و ناحیه همگرایی](path/to/image.png)

**شکل (6-16)**: نمودار قطب و صفر و ناحیه همگرایی:  
(الف) ناحیه اول، (ب) ناحیه دوم، (ج) ناحیه سوم سیستم معکوس.

### پاسخ ضربه سیستم به صورت زیر بدست می آید:

$$
h(t) = \delta(t) + \frac{2}{3} e^{-t} u(t) + \frac{4}{3} e^{2t} u(t)
$$

### ناحیه همگرایی سیستم معکوس ممکن به صورت زیر باشد:

$$
H(s) = \frac{(s+1)(s-2)}{(s-1)(s+2)}
$$

با توجه به شکل برای سیستم معکوس سه ناحیه همگرایی ممکن وجود دارد:

1. \( 1 < Re(s) \)
2. \( -2 < Re(s) < 1 \)
3. \( Re(s) < -2 \)

### رسم شده اند.

#### حالت اول:
$$
h(t) = \delta(t) - \frac{2}{3} e^{-t} u(t) - \frac{4}{3} e^{2t} u(t)
$$

#### حالت دوم:
$$
h(t) = \delta(t) + \frac{2}{3} e^{-t} u(t) - \frac{4}{3} e^{2t} u(t)
$$
````

````python
# فصل ششم: تبدیل لاپلاس

با گرفتن تبدیل لاپلاس از طرفین معادله فوق داریم:

$$
\left( \sum_{k=0}^{N} a_k s^k \right) Y(s) = \left( \sum_{k=0}^{M} b_k s^k \right) X(s)
\quad \text{(6-22)}
$$

بنابراین:

$$
H(s) = \frac{\sum_{k=0}^{M} b_k s^k}{\sum_{k=0}^{N} a_k s^k}
\quad \text{(6-23)}
$$

---

# 6-2 اتصال سیستم‌ها

با کمک تبدیل لاپلاس می‌توان تابع تبدیل ترکیبات مختلف سیستم‌ها را تعیین کرد. چند ترکیب ساده به عنوان مثال در زیر ارائه شده‌اند:

![اتصال سیستم‌ها](سیستم‌ها.png)  <!-- ← پیشنهاد: تصویر را با همان ساختار در نوت‌بوک بارگذاری کنید -->

**شکل (6-24): اتصال موازی دو سیستم LTI**

**شکل (6-25): اتصال سری دو سیستم LTI**

برای شکل (6-24) داریم:

$$
h(t) = h_1(t) + h_2(t)
\quad \text{(6-24)}
$$

بنابراین با گرفتن تبدیل لاپلاس خواهیم داشت:

$$
H(s) = H_1(s) + H_2(s)
\quad \text{(6-25)}
$$

و برای اتصال سری در شکل (6-25) داریم:

$$
h(t) = h_1(t) * h_2(t)
\quad \text{(6-26)}
$$

$$
H(s) = H_1(s) \cdot H_2(s)
\quad \text{(6-27)}
$$

توجه: تابع تبدیل ترکیب سیستم‌ها به صورت شکل (6-26) بدست می‌آید. این نوع اتصال سیستم‌ها به اتصال سری مشهور است.
````

````python
## 6- تبدیل لاپلاس یک‌طرفه

در قسمت‌های قبلی این فصل ما اغلب در مورد تبدیل لاپلاس دو طرفه صحبت کردیم. در این قسمت برای نوع خاصی از تبدیل لاپلاس که در تحلیل سیستم‌های با شرایط اولیه سیستم‌های فیزیکی مفید است، یعنی تبدیل لاپلاس یک‌طرفه، تعریف می‌کنیم.

تبدیل لاپلاس یک‌طرفه \( \mathcal{L}\{x(t)\} \) سیگنال \( x(t) \) به صورت زیر تعریف می‌شود:

\[
X(s) = \int_{0^-}^{\infty} x(t)e^{-st}dt
\]

با این تعریف اگر سیگنال برای \( t < 0 \) صفر باشد، تبدیل لاپلاس یک‌طرفه و دو طرفه یکسان خواهد شد. در ادامه به برخی خواص تبدیل لاپلاس یک‌طرفه اشاره می‌کنیم.

### مثال (12-6)

فرض کنید:
\[
x(t) = \frac{t^n}{n!}u(t) \quad (n=0,1,2,\ldots)
\]

تبدیل لاپلاس آن به صورت زیر خواهد بود:

\[
X(s) = \frac{1}{s^{n+1}} \quad \text{Re}(s) > 0
\]

### مثال (12-7)

\[
x(t) = e^{at}u(t) \Rightarrow X(s) = \frac{1}{s-a}, \quad \text{Re}(s) > a
\]
````

````python
import numpy as np
import matplotlib.pyplot as plt

# تعریف دامنه t
t = np.linspace(-1, 5, 500)
a = 2
u = np.heaviside(t, 1)
x = np.exp(a * t) * u

# رسم نمودار
plt.figure(figsize=(8, 4))
plt.plot(t, x, label=r'$x(t) = e^{at}u(t)$', color='b')
plt.title(r'نمودار تابع $x(t) = e^{at}u(t)$ برای $a = 2$')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
a = 0.8     # ضریب دامنه
T = 1       # فاصله بین پالس‌ها
k_values = np.arange(0, 6)  # فقط 6 تا برای نمایش کافی است
t = np.linspace(0, 6, 1000)

# ایجاد پالس دلتا با ارتفاع
x = np.zeros_like(t)
for k in k_values:
    idx = np.argmin(np.abs(t - k*T))
    x[idx] = a**k

# رسم نمودار
plt.figure(figsize=(8, 3))
plt.stem(t, x, use_line_collection=True, basefmt=" ", linefmt='b', markerfmt='bo')
plt.title(r'$x(t) = \sum_{k=0}^{\infty} a^k \delta(t - kT)$', fontsize=14)
plt.xlabel('t (ثانیه)')
plt.ylabel('x(t)')
plt.grid(True)
plt.xlim(0, 6)
plt.ylim(0, 1.2)
plt.show()
````

````python
## سیگنال‌های با طول محدود (Finite Duration Signals)

یک سیگنال گسسته یا پیوسته با طول محدود به سیگنالی گفته می‌شود که فقط در بازه‌ای مشخص از زمان مقدار غیرصفر دارد و در خارج از آن بازه، مقدار آن صفر است.

فرمول کلی چنین سیگنالی به شکل زیر قابل نمایش است:

$$
x(t) = \begin{cases}
f(t), & T_1 \leq t \leq T_2 \\\\
0, & \text{otherwise}
\end{cases}
$$

همچنین برای یک تابع \( x(t) \) با پشتیبانی در بازه‌ی \( [-T, T] \)، می‌توان نوشت:

$$
x(t) = x(t) \cdot [u(t+T) - u(t-T)]
$$

که در آن \( u(t) \) تابع پله واحد (unit step) است.
````

````python
## تبدیل لاپلاس سیگنال‌های محدود

اگر تابع \( x(t) \) تنها در بازه \( -T \leq t \leq T \) تعریف شده باشد، تبدیل لاپلاس آن به صورت زیر خواهد بود:

$$
X(s) = \int_{-T}^{T} x(t) e^{-st} dt
$$
````

````python
import numpy as np
import matplotlib.pyplot as plt

# تعریف بازه زمانی
t = np.linspace(-5, 5, 1000)

# تعریف سیگنال محدود شده بین -2 تا 2
def x(t):
    return np.where((t >= -2) & (t <= 2), np.sin(2 * np.pi * t), 0)

# رسم نمودار
plt.figure(figsize=(10, 4))
plt.plot(t, x(t), label=r'$x(t)$', color='teal')
plt.axvline(-2, color='gray', linestyle='--', alpha=0.5)
plt.axvline(2, color='gray', linestyle='--', alpha=0.5)
plt.title("سیگنال با طول محدود")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## تبدیل لاپلاس و ناحیه همگرایی (ROC)

تبدیل لاپلاس تابع \( x(t) \) به صورت زیر تعریف می‌شود:

$$
X(s) = \int_{-\infty}^{\infty} x(t) e^{-st} dt
$$

شرط وجود این تبدیل، همگرایی انتگرال بالا است. این همگرایی تنها در برخی نواحی از صفحه مختلط \( s \) برقرار است که به آن **ناحیه همگرایی (Region of Convergence)** گفته می‌شود.
````

````python
## مثال: سیگنال نمایی یک‌طرفه

تابع نمایی یک‌طرفه به صورت زیر تعریف می‌شود:

$$
x(t) = e^{-a t} u(t), \quad a > 0
$$

تبدیل لاپلاس آن برابر است با:

$$
X(s) = \int_{0}^{\infty} e^{-a t} e^{-st} dt = \int_{0}^{\infty} e^{-(s+a)t} dt = \frac{1}{s + a}
$$

این تبدیل فقط زمانی وجود دارد که:

$$
\text{Re}(s) > -a
$$

بنابراین، ناحیه همگرایی:

$$
ROC: \text{Re}(s) > -a
$$
````

````python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-2, 5, 1000)
a = 2

# تابع نمایی یک‌طرفه
x = lambda t: np.where(t >= 0, np.exp(-a * t), 0)

plt.figure(figsize=(10, 4))
plt.plot(t, x(t), label=r'$e^{-a t} u(t)$', color='darkorange')
plt.title("سیگنال نمایی یک‌طرفه با a=2")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## سیگنال نمایی دوطرفه

فرض کنید تابع به شکل زیر باشد:

$$
x(t) = e^{-a |t|}, \quad a > 0
$$

این تابع از دو بخش تشکیل شده:

- برای \( t \geq 0 \): \( e^{-a t} \)
- برای \( t < 0 \): \( e^{a t} \)

بنابراین:

$$
x(t) = e^{-a t} u(t) + e^{a t} u(-t)
$$

تبدیل لاپلاس آن:

$$
X(s) = \int_{-\infty}^{\infty} x(t) e^{-st} dt = \int_{-\infty}^{0} e^{a t} e^{-st} dt + \int_{0}^{\infty} e^{-a t} e^{-st} dt
$$

$$
= \int_{-\infty}^{0} e^{(a - s)t} dt + \int_{0}^{\infty} e^{-(a + s)t} dt
$$

**شرط همگرایی:**

برای همگرایی هر دو انتگرال:

$$
\text{Re}(s) < a \quad \text{و} \quad \text{Re}(s) > -a
$$

یعنی:

$$
ROC: -a < \text{Re}(s) < a
$$
````

````python
t = np.linspace(-5, 5, 1000)
a = 2

x = lambda t: np.where(t < 0, np.exp(a * t), np.exp(-a * t))

plt.figure(figsize=(10, 4))
plt.plot(t, x(t), label=r'$e^{-a|t|}$', color='crimson')
plt.title("سیگنال نمایی دوطرفه")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## طبقه‌بندی سیگنال‌ها بر اساس ناحیه زمانی

### 1. سیگنال راست‌طرفه (Right-sided)

تابعی است که فقط برای \( t \geq t_0 \) تعریف شده و در بقیه نقاط صفر است:

$$
x(t) = 0, \quad t < t_0
$$

### 2. سیگنال چپ‌طرفه (Left-sided)

تابعی است که فقط برای \( t \leq t_0 \) تعریف شده و در بقیه نقاط صفر است:

$$
x(t) = 0, \quad t > t_0
$$

### 3. سیگنال دوطرفه (Two-sided)

تابعی که در هر دو جهت زمان غیر صفر است و محدود به یک طرف نیست.
````

````python
## ناحیه همگرایی (ROC) بسته به نوع سیگنال

- اگر سیگنال **راست‌طرفه** باشد، ROC به صورت \( \text{Re}(s) > \sigma_0 \) است.
- اگر سیگنال **چپ‌طرفه** باشد، ROC به صورت \( \text{Re}(s) < \sigma_0 \) است.
- اگر سیگنال **دوطرفه** باشد، ROC به صورت نواری بین دو مقدار است:

$$
\sigma_1 < \text{Re}(s) < \sigma_2
$$
````

````python
fig, axs = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

t = np.linspace(-5, 5, 1000)

# راست‌طرفه
x_right = np.where(t >= 0, np.exp(-t), 0)
axs[0].plot(t, x_right, color='teal')
axs[0].set_title("Right-sided")

# چپ‌طرفه
x_left = np.where(t <= 0, np.exp(t), 0)
axs[1].plot(t, x_left, color='orange')
axs[1].set_title("Left-sided")

# دوطرفه
x_two = np.exp(-np.abs(t))
axs[2].plot(t, x_two, color='purple')
axs[2].set_title("Two-sided")

for ax in axs:
    ax.grid(True)
    ax.set_xlabel("t")

axs[0].set_ylabel("x(t)")
plt.tight_layout()
plt.show()
````

````python
## محدودسازی سیگنال با تابع پاشنه (Unit Step)

فرض کنید تابع اصلی \( x(t) \) بدون محدودیت زمانی است. می‌توان با استفاده از تابع پاشنه آن را محدود کرد:

$$
x_T(t) = x(t) \cdot [u(t + T) - u(t - T)]
$$

که در آن:
- \( u(t) \) تابع پاشنه واحد است.
- بازه فعال بودن تابع محدود شده: \( -T \leq t \leq T \)

### خاصیت:

$$
x_T(t) = \begin{cases}
x(t), & |t| \leq T \\\\
0, & \text{otherwise}
\end{cases}
$$
````

````python
## تبدیل لاپلاس تابع محدودشده

اگر \( x_T(t) = x(t) \cdot w(t) \) باشد که در آن \( w(t) \) تابع پنجره‌ای باشد:

$$
X_T(s) = \int_{-\infty}^{\infty} x(t) w(t) e^{-st} dt
$$

با توجه به خاصیت ضرب در زمان، این تبدیل در حوزه \( s \) برابر است با **کانولوشن** تبدیل‌های لاپلاس دو تابع:

$$
X_T(s) = X(s) * W(s)
$$
````

````python
t = np.linspace(-5, 5, 1000)
x = np.sin(2 * np.pi * t)
window = np.where((t >= -2) & (t <= 2), 1, 0)
x_limited = x * window

plt.figure(figsize=(10, 4))
plt.plot(t, x, label='Original $\sin(2\pi t)$', linestyle='--', alpha=0.5)
plt.plot(t, x_limited, label='Limited Signal', color='blue')
plt.title("محدودسازی سیگنال با تابع پاشنه")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.legend()
plt.grid(True)
plt.show()
````

````python
## بررسی تابع \( x(t) = e^{-a t} u(t) \)

تابع نمایی یک‌طرفه با پاشنه به شکل زیر است:

$$
x(t) = e^{-a t} u(t)
$$

### تبدیل لاپلاس:

$$
X(s) = \int_0^\infty e^{-a t} e^{-s t} dt = \int_0^\infty e^{-(s + a)t} dt = \frac{1}{s + a}
$$

### ناحیه همگرایی:

$$
\text{Re}(s) > -a
$$

---

## بررسی تابع \( x(t) = e^{a t} u(-t) \)

تابع نمایی در جهت منفی زمان:

$$
x(t) = e^{a t} u(-t)
$$

### تبدیل لاپلاس:

$$
X(s) = \int_{-\infty}^0 e^{a t} e^{-s t} dt = \int_{-\infty}^0 e^{(a - s)t} dt = \frac{1}{s - a}
$$

### ناحیه همگرایی:

$$
\text{Re}(s) < a
$$
````

````python
## خاصیت کانولوشن در تبدیل لاپلاس

اگر:

$$
x(t) \leftrightarrow X(s) \quad \text{و} \quad h(t) \leftrightarrow H(s)
$$

آنگاه:

$$
x(t) * h(t) \leftrightarrow X(s) \cdot H(s)
$$

که در آن * نشان‌دهنده‌ی کانولوشن زمانی است:

$$
(x * h)(t) = \int_{-\infty}^{\infty} x(\tau) h(t - \tau) d\tau
$$

---

## خاصیت معکوس (ضرب در زمان):

اگر:

$$
x(t) \cdot h(t) \leftrightarrow \frac{1}{2\pi j} \int_{\sigma - j\infty}^{\sigma + j\infty} X(s) H(s) e^{s t} ds
$$

که این معادل **کانولوشن در حوزه لاپلاس** است.
````

````python
## پاسخ یک سیستم LTI با کانولوشن

اگر سیستمی با تابع پاسخ ضربه‌ای \( h(t) \) به ورودی \( x(t) \) تحریک شود، پاسخ آن برابر است با:

$$
y(t) = x(t) * h(t)
$$
````

````python
from scipy.signal import convolve

t = np.linspace(0, 10, 1000)
x = np.exp(-t)
h = np.ones_like(t)  # تابع پله ساده

# کانولوشن
y = convolve(x, h, mode='full')[:len(t)] * (t[1] - t[0])

plt.figure(figsize=(10, 4))
plt.plot(t, x, label='x(t)', linestyle='--')
plt.plot(t, h, label='h(t)', linestyle=':')
plt.plot(t, y, label='y(t) = x*h', color='blue')
plt.title("پاسخ سیستم به ورودی نمایی با تابع پله")
plt.xlabel("t")
plt.ylabel("مقدار")
plt.legend()
plt.grid(True)
plt.show()
````

````python
## تابع ضربه‌ی دیراک (Delta Function)

تابع \( \delta(t) \) تعریف می‌شود به گونه‌ای که:

- در همه‌ی نقاط صفر است به جز در \( t = 0 \)
- و مقدار آن در انتگرال برابر ۱ است:

$$
\int_{-\infty}^{\infty} \delta(t) dt = 1
$$

### خاصیت نمونه‌گیری:

$$
\int_{-\infty}^{\infty} f(t) \delta(t - t_0) dt = f(t_0)
$$

---

## تبدیل لاپلاس از تابع ضربه

تبدیل لاپلاس تابع \( \delta(t) \) برابر است با:

$$
\mathcal{L}\{\delta(t)\} = \int_{0}^{\infty} \delta(t) e^{-s t} dt = 1
$$

و برای ضربه جابجا شده:

$$
\mathcal{L}\{\delta(t - t_0)\} = e^{-s t_0}
$$
````

````python
# تابع دلتا تقریبی به شکل پالس باریک
t = np.linspace(-1, 1, 1000)
delta = np.where(np.abs(t) < 0.01, 1/0.02, 0)

plt.figure(figsize=(10, 3))
plt.plot(t, delta, color='purple')
plt.title("تقریب تابع ضربه دلتا با پالس باریک")
plt.xlabel("t")
plt.ylabel(r'$\delta(t)$')
plt.grid(True)
plt.show()
````

````python
## رابطه بین تابع پله و تابع دلتا

تابع پله واحد \( u(t) \) و تابع دلتا \( \delta(t) \) به صورت زیر به هم مرتبط هستند:

$$
\frac{d}{dt} u(t) = \delta(t)
$$

---

## تبدیل لاپلاس مشتق توابع

برای هر تابع \( x(t) \) که مشتق‌پذیر باشد:

$$
\mathcal{L}\left\{\frac{d}{dt} x(t)\right\} = s X(s) - x(0^-)
$$

که در آن \( X(s) \) تبدیل لاپلاس تابع \( x(t) \) است.

---

## حالت خاص برای مشتق تابع پله:

$$
x(t) = u(t) \Rightarrow X(s) = \frac{1}{s} \Rightarrow \frac{d}{dt} u(t) = \delta(t)
$$

پس داریم:

$$
\mathcal{L}\{\delta(t)\} = s \cdot \frac{1}{s} - 1 = 1
$$
````

````python
t = np.linspace(-1, 1, 1000)
u = np.where(t >= 0, 1, 0)
du = np.gradient(u, t)

plt.figure(figsize=(10, 4))
plt.plot(t, u, label='u(t)', linestyle='--')
plt.plot(t, du, label="du/dt ≈ δ(t)", color='red')
plt.title("پله واحد و مشتق آن (تقریب دلتا)")
plt.xlabel("t")
plt.ylabel("مقدار")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## خاصیت خطی تبدیل لاپلاس

تبدیل لاپلاس یک عملگر خطی است. یعنی اگر:

$$
x_1(t) \leftrightarrow X_1(s) \quad \text{و} \quad x_2(t) \leftrightarrow X_2(s)
$$

آنگاه برای ضرایب \( a \) و \( b \):

$$
a x_1(t) + b x_2(t) \leftrightarrow a X_1(s) + b X_2(s)
$$

---

## خاصیت انتقال زمانی

اگر:

$$
x(t) \leftrightarrow X(s)
$$

آنگاه:

$$
x(t - t_0) u(t - t_0) \leftrightarrow e^{-s t_0} X(s)
$$

---

## خاصیت تغییر مقیاس زمانی

اگر:

$$
x(a t) \leftrightarrow \frac{1}{|a|} X\left( \frac{s}{a} \right)
$$

برای \( a > 0 \).
````

````python
t = np.linspace(-2, 5, 1000)
x = np.exp(-t)

# انتقال در زمان
x_shifted = np.where(t >= 2, np.exp(-(t - 2)), 0)

# تغییر مقیاس (فشرده‌سازی زمان)
a = 2
x_scaled = np.exp(-a * t)

plt.figure(figsize=(10, 4))
plt.plot(t, x, label=r'$x(t) = e^{-t}$', linestyle='--')
plt.plot(t, x_shifted, label=r'$x(t-2) u(t-2)$', color='orange')
plt.plot(t, x_scaled, label=r'$x(2t) = e^{-2t}$', color='green')
plt.title("تأثیر انتقال و مقیاس زمانی بر سیگنال نمایی")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## خاصیت انتگرال‌گیری در تبدیل لاپلاس

اگر:

$$
x(t) \leftrightarrow X(s)
$$

آنگاه:

$$
\int_0^t x(\tau) d\tau \leftrightarrow \frac{1}{s} X(s)
$$

---

### تفسیر:

انتگرال زمانی یک تابع متناظر با تقسیم تبدیل لاپلاس آن بر \( s \) است.

---

## خاصیت مهم دیگر:

اگر:

$$
x(t) = \int_0^t f(\tau) d\tau
$$

آنگاه:

$$
X(s) = \frac{1}{s} F(s)
$$

و برعکس، اگر:

$$
X(s) = \frac{1}{s} F(s) \Rightarrow x(t) = \int_0^t f(\tau) d\tau
$$
````

````python
from scipy.integrate import cumtrapz

t = np.linspace(0, 5, 1000)
f = np.exp(-t)
x = cumtrapz(f, t, initial=0)

plt.figure(figsize=(10, 4))
plt.plot(t, f, label='f(t) = exp(-t)', linestyle='--')
plt.plot(t, x, label=r'$\int_0^t f(\tau) d\tau$', color='blue')
plt.title("انتگرال تابع نمایی")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## ضرب نمایی در زمان و انتقال در حوزه s

اگر:

$$
x(t) \leftrightarrow X(s)
$$

آنگاه:

$$
x(t) e^{a t} \leftrightarrow X(s - a)
$$

---

### تفسیر:

ضرب تابع زمانی در \( e^{a t} \)، تبدیل لاپلاس آن را به سمت راست در محور مختلط \( s \) انتقال می‌دهد.

---

## مثال:

اگر:

$$
x(t) = e^{-t} u(t) \Rightarrow X(s) = \frac{1}{s + 1}
$$

آنگاه:

$$
x(t) e^{3t} = e^{2t} u(t) \Rightarrow X(s - 3) = \frac{1}{s - 2}
$$
````

````python
t = np.linspace(0, 5, 1000)
x = np.exp(-t)
x_exp = x * np.exp(3 * t)  # معادل e^{2t}

plt.figure(figsize=(10, 4))
plt.plot(t, x, label=r'$x(t) = e^{-t}$', linestyle='--')
plt.plot(t, x_exp, label=r'$x(t) \cdot e^{3t} = e^{2t}$', color='red')
plt.title("ضرب نمایی در زمان و تغییر لاپلاس")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## جدول تبدیل‌های لاپلاس پرکاربرد

| تابع زمانی \( x(t) \)                        | تبدیل لاپلاس \( X(s) \)             | ROC                         |
|----------------------------------------------|-------------------------------------|-----------------------------|
| \( \delta(t) \)                               | \( 1 \)                             | کل صفحه \( s \)            |
| \( u(t) \)                                    | \( \frac{1}{s} \)                   | \( \text{Re}(s) > 0 \)      |
| \( t^n u(t) \)                                | \( \frac{n!}{s^{n+1}} \)            | \( \text{Re}(s) > 0 \)      |
| \( e^{-a t} u(t) \)                           | \( \frac{1}{s + a} \)               | \( \text{Re}(s) > -a \)     |
| \( \sin(\omega t) u(t) \)                     | \( \frac{\omega}{s^2 + \omega^2} \) | \( \text{Re}(s) > 0 \)      |
| \( \cos(\omega t) u(t) \)                     | \( \frac{s}{s^2 + \omega^2} \)      | \( \text{Re}(s) > 0 \)      |
````

````python
t = np.linspace(0, 5, 1000)
sin = np.sin(2 * np.pi * t)
cos = np.cos(2 * np.pi * t)
exp = np.exp(-2 * t)

plt.figure(figsize=(10, 6))
plt.plot(t, sin, label=r'$\sin(2\pi t)$', linestyle='--')
plt.plot(t, cos, label=r'$\cos(2\pi t)$', linestyle=':')
plt.plot(t, exp, label=r'$e^{-2t}$')
plt.title("نمونه‌ای از توابع زمانی پرکاربرد در لاپلاس")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## تعریف سیستم‌های خطی و تغییرناپذیر با زمان (LTI)

سیستم LTI دارای دو ویژگی مهم است:

1. **خطی بودن**: اگر ورودی \( a x_1(t) + b x_2(t) \) باشد، خروجی \( a y_1(t) + b y_2(t) \) خواهد بود.
2. **تغییرناپذیری با زمان**: تأخیر در ورودی، فقط تأخیر در خروجی ایجاد می‌کند.

---

## پاسخ سیستم به تابع دلتا

برای هر سیستم LTI با پاسخ ضربه‌ای \( h(t) \):

$$
\delta(t) \rightarrow h(t)
$$

و در نتیجه پاسخ به ورودی کلی از طریق کانولوشن:

$$
x(t) \rightarrow y(t) = x(t) * h(t)
$$
````

````python
from scipy.signal import convolve

t = np.linspace(0, 5, 1000)
x = np.heaviside(t - 1, 1)  # پله با تأخیر
h = np.exp(-t)              # پاسخ ضربه‌ای

y = convolve(x, h, mode='full')[:len(t)] * (t[1] - t[0])

plt.figure(figsize=(10, 4))
plt.plot(t, x, label='x(t) = u(t-1)')
plt.plot(t, h, label='h(t) = e^{-t}')
plt.plot(t, y, label='y(t) = x*h', color='blue')
plt.title("پاسخ سیستم LTI به ورودی پله‌ای")
plt.xlabel("t")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## پایداری سیستم‌های LTI

سیستمی پایدار است اگر برای هر ورودی محدود، خروجی نیز محدود باشد. به عبارتی:

$$
\text{اگر} \quad |x(t)| < \infty \Rightarrow |y(t)| < \infty
$$

---

## شرط پایداری در حوزه لاپلاس

اگر تبدیل لاپلاس پاسخ ضربه‌ای \( H(s) \) را داشته باشیم، سیستم پایدار است اگر:

- کلیه قطب‌های \( H(s) \) دارای قسمت حقیقی منفی باشند.
- یعنی \( \text{Re}(s) < 0 \) برای تمام قطب‌ها در ناحیه ROC

---

## مثال:

اگر:

$$
H(s) = \frac{1}{s + 3} \quad \Rightarrow \text{پایدار، چون } \text{Re}(s) = -3 < 0
$$

اما اگر:

$$
H(s) = \frac{1}{s - 1} \quad \Rightarrow \text{ناپایدار، چون قطب در } \text{Re}(s) = 1
$$
````

````python
t = np.linspace(0, 5, 1000)

h_stable = np.exp(-3 * t)
h_unstable = np.exp(t)

plt.figure(figsize=(10, 4))
plt.plot(t, h_stable, label='پاسخ پایدار: $e^{-3t}$', color='green')
plt.plot(t, h_unstable, label='پاسخ ناپایدار: $e^{t}$', color='red')
plt.title("پاسخ ضربه‌ای دو سیستم (پایدار و ناپایدار)")
plt.xlabel("t")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## تعریف قطب (Pole) و صفر (Zero)

برای تابع تبدیل \( H(s) \):

- **صفر (Zero)**: مقادیری از \( s \) که در آن \( H(s) = 0 \)
- **قطب (Pole)**: مقادیری از \( s \) که در آن \( H(s) \rightarrow \infty \)

---

## نمایش کسری:

تابع تبدیل اغلب به صورت نسبت چندجمله‌ای نوشته می‌شود:

$$
H(s) = \frac{(s - z_1)(s - z_2) \dots (s - z_m)}{(s - p_1)(s - p_2) \dots (s - p_n)}
$$

که در آن \( z_i \) صفرها و \( p_i \) قطب‌ها هستند.

---

## اهمیت:

- قطب‌ها تعیین‌کننده‌ی پایداری و رفتار سیستم هستند.
- صفرها بر پاسخ فرکانسی و شکل خروجی تأثیر می‌گذارند.
````

````python
import matplotlib.pyplot as plt
import numpy as np

zeros = [-2, -4]
poles = [0, -1, -3]

plt.figure(figsize=(6, 6))
plt.scatter(np.real(zeros), np.imag(zeros), marker='o', color='blue', label='صفرها')
plt.scatter(np.real(poles), np.imag(poles), marker='x', color='red', label='قطب‌ها')

plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.title("نمودار قطب و صفر در صفحه مختلط s")
plt.xlabel("Re(s)")
plt.ylabel("Im(s)")
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
````

````python
## پاسخ فرکانسی (Frequency Response)

پاسخ فرکانسی سیستم برابر است با مقدار تابع انتقال در محور موهومی:

$$
H(j\omega) = H(s) \big|_{s = j\omega}
$$

---

## تابع انتقال (Transfer Function)

تابع انتقال \( H(s) \) از تبدیل لاپلاس پاسخ ضربه‌ای \( h(t) \) به دست می‌آید:

$$
H(s) = \mathcal{L}\{h(t)\}
$$

---

## ویژگی‌ها:

- رفتار سیستم در فرکانس‌های مختلف را توصیف می‌کند.
- برای تحلیل پاسخ به ورودی‌های سینوسی بسیار کاربرد دارد.

---

## پاسخ سیستم به ورودی \( x(t) = \cos(\omega t) \):

اگر:

$$
x(t) = \cos(\omega t) \Rightarrow y(t) = |H(j\omega)| \cos(\omega t + \angle H(j\omega))
$$

یعنی دامنه و فاز خروجی تحت تأثیر \( H(j\omega) \) قرار می‌گیرد.
````

````python
from scipy.signal import freqresp, TransferFunction

# سیستم: H(s) = 1 / (s + 1)
num = [1]
den = [1, 1]
sys = TransferFunction(num, den)

w = np.linspace(0.1, 100, 500)
_, H = freqresp(sys, w)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.semilogx(w, 20 * np.log10(np.abs(H)))
plt.title("بزرگای پاسخ فرکانسی")
plt.xlabel("فرکانس (رادیان بر ثانیه)")
plt.ylabel("|H(jω)| (dB)")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.semilogx(w, np.angle(H, deg=True))
plt.title("فاز پاسخ فرکانسی")
plt.xlabel("فرکانس (رادیان بر ثانیه)")
plt.ylabel("فاز (درجه)")
plt.grid(True)

plt.tight_layout()
plt.show()
````

````python
## پاسخ سیستم به ورودی سینوسی پایدار

اگر سیستم LTI پایدار باشد و ورودی آن:

$$
x(t) = A \cos(\omega t + \phi)
$$

باشد، آنگاه خروجی نیز یک سینوس با همان فرکانس است:

$$
y(t) = |H(j\omega)| A \cos(\omega t + \phi + \angle H(j\omega))
$$

---

## تفسیر:

- دامنه خروجی تغییر می‌کند: \( A \rightarrow A |H(j\omega)| \)
- فاز خروجی تغییر می‌کند: \( \phi \rightarrow \phi + \angle H(j\omega) \)

---

## کاربرد:

این خاصیت برای طراحی و تحلیل فیلترها بسیار مفید است.
````

````python
f = 5  # فرکانس ورودی
t = np.linspace(0, 2, 1000)
A = 1
phi = 0
H_jw_abs = 0.6  # فرضی
H_jw_phase = np.pi / 6

x = A * np.cos(2 * np.pi * f * t + phi)
y = A * H_jw_abs * np.cos(2 * np.pi * f * t + phi + H_jw_phase)

plt.figure(figsize=(10, 4))
plt.plot(t, x, label='x(t) = ورودی سینوسی', linestyle='--')
plt.plot(t, y, label='y(t) = خروجی سیستم', color='darkblue')
plt.title("پاسخ سیستم به ورودی سینوسی پایدار")
plt.xlabel("t")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## نمودار بود چیست؟

نمودار بود نمایش پاسخ فرکانسی یک سیستم در دو نمودار جداگانه است:

1. **نمودار بزرگی (Magnitude)**: \( 20 \log_{10} |H(j\omega)| \)
2. **نمودار فاز (Phase)**: \( \angle H(j\omega) \)

هر دو نمودار بر حسب فرکانس (اغلب به صورت لگاریتمی) رسم می‌شوند.

---

## مزایای نمودار بود:

- بررسی رفتار سیستم در محدوده گسترده‌ای از فرکانس‌ها.
- تحلیل پایداری و پاسخ به نویز.
- طراحی فیلترهای آنالوگ یا دیجیتال.
````

````python
from scipy.signal import bode, TransferFunction

# تعریف تابع انتقال: H(s) = 1 / (s + 1)
sys = TransferFunction([1], [1, 1])

w, mag, phase = bode(sys)

plt.figure(figsize=(12, 5))

plt.subplot(2, 1, 1)
plt.semilogx(w, mag)
plt.title("نمودار بود - بزرگی")
plt.ylabel("Magnitude (dB)")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.semilogx(w, phase)
plt.title("نمودار بود - فاز")
plt.xlabel("Frequency (rad/s)")
plt.ylabel("Phase (degrees)")
plt.grid(True)

plt.tight_layout()
plt.show()
````

````python
## مکان هندسی ریشه‌ها (Root Locus)

Root Locus یک روش گرافیکی برای تحلیل پایداری و طراحی کنترل‌کننده‌هاست.

در این روش، مکان قطب‌های سیستم در صفحه مختلط \( s \) با تغییر یک پارامتر (معمولاً بهره \( K \)) بررسی می‌شود.

---

## ویژگی‌ها:

- مسیر حرکت قطب‌ها با تغییر \( K \) نشان داده می‌شود.
- به ما می‌گوید سیستم با افزایش بهره پایدار باقی می‌ماند یا خیر.
- در طراحی کنترل‌کننده‌هایی مانند PID بسیار مهم است.
````

````python
import control
import matplotlib.pyplot as plt

# تابع انتقال باز: G(s) = K / (s(s+2))
num = [1]
den = [1, 2, 0]
sys = control.TransferFunction(num, den)

plt.figure(figsize=(6, 6))
control.root_locus(sys, grid=True)
plt.title("مکان هندسی ریشه‌ها (Root Locus)")
plt.xlabel("Re(s)")
plt.ylabel("Im(s)")
plt.show()
````

````python
## پاسخ ضربه‌ای (Impulse Response)

پاسخ سیستم به ورودی \( \delta(t) \) است و معمولاً با \( h(t) \) نمایش داده می‌شود.

---

## پاسخ پله‌ای (Step Response)

پاسخ سیستم به ورودی \( u(t) \) است. این پاسخ را می‌توان از انتگرال پاسخ ضربه‌ای به دست آورد:

$$
s(t) = \int_0^t h(\tau) d\tau
$$

---

## رابطه بین آن‌ها:

اگر \( h(t) \) پاسخ ضربه‌ای باشد، آنگاه:

$$
s(t) = h(t) * u(t)
$$
````

````python
from scipy.signal import impulse, step, TransferFunction

# سیستم مرتبه اول: H(s) = 1 / (s + 1)
sys = TransferFunction([1], [1, 1])

t, h = impulse(sys)
t, s = step(sys)

plt.figure(figsize=(10, 4))
plt.plot(t, h, label='Impulse Response h(t)', linestyle='--')
plt.plot(t, s, label='Step Response s(t)', color='blue')
plt.title("پاسخ ضربه‌ای و پله‌ای یک سیستم LTI")
plt.xlabel("t")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## پاسخ کلی سیستم LTI

پاسخ کلی یک سیستم به دو بخش تقسیم می‌شود:

1. **پاسخ گذرا (Transient Response)**: نوسانات اولیه که با زمان از بین می‌روند.
2. **پاسخ ماندگار (Steady-State Response)**: بخشی که در زمان طولانی باقی می‌ماند.

---

## تحلیل پاسخ گذرا:

وابسته به قطب‌های سیستم است، به‌ویژه قطب‌هایی با قسمت حقیقی منفی.

- اگر قطب در \( s = -\alpha \)، آنگاه جزء گذرا به شکل \( e^{-\alpha t} \) ظاهر می‌شود.

---

## پاسخ ماندگار:

در سیستم‌های پایدار، به رفتار بلندمدت خروجی نسبت داده می‌شود.
برای ورودی‌های سینوسی، پاسخ ماندگار معمولاً نیز سینوسی با فاز و دامنه تغییر یافته است.
````

````python
from scipy.signal import step

# سیستم: H(s) = 1 / (s^2 + 3s + 2)
sys = TransferFunction([1], [1, 3, 2])
t, y = step(sys)

plt.figure(figsize=(10, 4))
plt.plot(t, y, label='y(t) - Step Response')
plt.axhline(y[-1], color='red', linestyle='--', label='Steady-State Value')
plt.title("تحلیل پاسخ گذرا و ماندگار")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.grid(True)
plt.legend()
plt.show()
````

````python
## جمع‌بندی ویژگی‌های اصلی تبدیل لاپلاس:

- **خطی بودن**: تبدیل لاپلاس عملگری خطی است.
- **انتقال زمانی**: تأخیر در زمان منجر به ضرب در \( e^{-st_0} \) در حوزه \( s \) می‌شود.
- **مقیاس زمانی**: فشرده‌سازی یا گسترش زمان، تبدیل را در \( s \) نیز مقیاس می‌دهد.
- **مشتق‌گیری**: مشتق در زمان معادل ضرب در \( s \) است.
- **انتگرال‌گیری**: انتگرال در زمان معادل تقسیم بر \( s \) است.
- **ضرب زمانی → کانولوشن در \( s \)** و برعکس.

---

## کاربردهای مهم:

- حل معادلات دیفرانسیل.
- تحلیل سیستم‌های دینامیکی (پایداری، پاسخ).
- طراحی فیلترها و کنترل‌کننده‌ها.
- تحلیل مدارهای RLC.
````

````python
from scipy.signal import lti, step

# سیستم: H(s) = 1 / (s^2 + 2ζω_ns + ω_n^2)
ζ = 0.3  # ضریب میرایی
ω_n = 2  # فرکانس طبیعی

sys = lti([ω_n**2], [1, 2*ζ*ω_n, ω_n**2])
t, y = step(sys)

plt.figure(figsize=(10, 4))
plt.plot(t, y, label='Step Response of 2nd Order System')
plt.title("رفتار گذرا سیستم مرتبه دوم با میرایی")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.grid(True)
plt.legend()
plt.show()
````
