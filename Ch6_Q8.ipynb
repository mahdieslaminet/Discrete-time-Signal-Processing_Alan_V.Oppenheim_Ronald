سوال 1:

سیگنال x(t) = e^(-|t|) را در نظر بگیرید.  
1. نمودار سیگنال را در بازه [-5, 5] رسم کنید.  
2. انرژی و توان این سیگنال را به صورت عددی محاسبه کنید.


```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 1000)
x = np.exp(-np.abs(t))

energy = np.trapz(x**2, t)
power = energy / (t[-1] - t[0])

print("Energy:", energy)
print("Power:", power)

plt.plot(t, x)
plt.title("x(t) = e^{- |t|}")
plt.grid(True)
plt.show()
```

سوال 2:

سیگنال x(t) = sinc(t) و نسخه‌ی تأخیری آن x(t - 3) را در نظر بگیرید.  
1. نمودار هر دو را رسم کنید.  
2. تبدیل فوریه آن‌ها را محاسبه کرده و فازهای آن‌ها را مقایسه کنید.


```python
from scipy.fft import fft, fftfreq, fftshift

t = np.linspace(-10, 10, 2048)
dt = t[1] - t[0]
x = np.sinc(t)
x_delay = np.sinc(t - 3)

f = fftshift(fftfreq(len(t), dt))
X = fftshift(fft(x)) * dt
X_delay = fftshift(fft(x_delay)) * dt

plt.plot(f, np.angle(X), label="∠X(f)")
plt.plot(f, np.angle(X_delay), '--', label="∠X(t-3)(f)")
plt.legend()
plt.title("Phase Comparison of Delayed Signal")
plt.grid(True)
plt.show()
```

سوال 3:

یک سیگنال فرد و یک سیگنال زوج به صورت زیر تعریف شده‌اند:  
x_odd(t) = t * rect(t) و x_even(t) = |t| * rect(t).  
طیف فوریه هر دو سیگنال را محاسبه کرده و فاز و دامنه را مقایسه کنید.


```python
x_odd = t * np.where(np.abs(t) <= 1, 1, 0)
x_even = np.abs(t) * np.where(np.abs(t) <= 1, 1, 0)

X_odd = fftshift(fft(x_odd)) * dt
X_even = fftshift(fft(x_even)) * dt

plt.plot(f, np.abs(X_odd), label="|X_odd(f)|")
plt.plot(f, np.abs(X_even), label="|X_even(f)|")
plt.legend()
plt.title("Magnitude Spectra")
plt.grid(True)
plt.show()

plt.plot(f, np.angle(X_odd), label="∠X_odd(f)")
plt.plot(f, np.angle(X_even), label="∠X_even(f)")
plt.legend()
plt.title("Phase Spectra")
plt.grid(True)
plt.show()
```

سوال 4:

سیگنال x(t) = cos(2πt) * cos(4πt) را در نظر بگیرید.  
1. نمودار سیگنال را رسم کنید.  
2. طیف فوریه آن را محاسبه و رسم کنید.


```python
t = np.linspace(0, 5, 1000)
x = np.cos(2*np.pi*t) * np.cos(4*np.pi*t)

dt = t[1] - t[0]
f = fftshift(fftfreq(len(t), dt))
X = fftshift(fft(x)) * dt

plt.plot(t, x)
plt.title("x(t) = cos(2πt) * cos(4πt)")
plt.grid(True)
plt.show()

plt.plot(f, np.abs(X))
plt.title("Magnitude Spectrum")
plt.grid(True)
plt.show()
```

سوال 5:

سیگنال مثلثی تعریف‌شده به صورت tri(t) = max(1 - |t|, 0) را در نظر بگیرید.  
1. آن را در بازه [-2, 2] رسم کنید.  
2. تبدیل فوریه آن را محاسبه و نمایش دهید.


```python
x = np.maximum(1 - np.abs(t), 0)

X = fftshift(fft(x)) * dt
plt.plot(t, x)
plt.title("x(t) = tri(t)")
plt.grid(True)
plt.show()

plt.plot(f, np.abs(X))
plt.title("Fourier Transform of tri(t)")
plt.grid(True)
plt.show()
```

سوال 6:

سیگنال x(t) = e^{-t} * sin(2πt) برای t ≥ 0 را در نظر بگیرید.  
1. سیگنال را رسم کنید.  
2. تبدیل فوریه آن را به صورت عددی نمایش دهید.


```python
t = np.linspace(0, 10, 1000)
x = np.exp(-t) * np.sin(2*np.pi*t)

dt = t[1] - t[0]
f = fftshift(fftfreq(len(t), dt))
X = fftshift(fft(x)) * dt

plt.plot(t, x)
plt.title("x(t) = e^{-t} * sin(2πt)")
plt.grid(True)
plt.show()

plt.plot(f, np.abs(X))
plt.title("Magnitude Spectrum")
plt.grid(True)
plt.show()
```
