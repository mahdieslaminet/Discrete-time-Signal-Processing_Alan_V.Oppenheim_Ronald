سیگنال x(t) = cos(2πt) را در نظر بگیرید.  
۱. نمودار آن را در بازه زمانی [0, 5] رسم کنید.  
۲. توان میانگین این سیگنال را به صورت عددی (با انتگرال‌گیری عددی) محاسبه کنید.




```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 5, 1000)
x = np.cos(2 * np.pi * t)

# محاسبه توان
power = np.trapz(x**2, t) / (t[-1] - t[0])
print("توان سیگنال:", power)

# رسم سیگنال
plt.plot(t, x)
plt.title("x(t) = cos(2πt)")
plt.xlabel("Time (t)")
plt.ylabel("x(t)")
plt.grid(True)
plt.show()
```

سیگنال x(t) = sinc(t) را در نظر بگیرید.  
۱. سیگنال x(2t) را نیز رسم کنید.  
۲. طیف فوریه هر دو سیگنال را محاسبه و رسم نمایید.  
۳. تأثیر فشرده‌سازی زمانی را روی طیف توضیح دهید.


```python
from scipy.fft import fft, fftfreq, fftshift

t = np.linspace(-10, 10, 2048)
dt = t[1] - t[0]

x = np.sinc(t)
x_scaled = np.sinc(2 * t)

f = fftshift(fftfreq(len(t), dt))
X = fftshift(fft(x)) * dt
X_scaled = fftshift(fft(x_scaled)) * dt

plt.plot(t, x, label="x(t) = sinc(t)")
plt.plot(t, x_scaled, '--', label="x(2t) = sinc(2t)")
plt.legend()
plt.title("Time Domain")
plt.grid(True)
plt.show()

plt.plot(f, np.abs(X), label="X(f)")
plt.plot(f, np.abs(X_scaled), '--', label="X_scaled(f)")
plt.legend()
plt.title("Frequency Domain (Magnitude)")
plt.grid(True)
plt.show()
```

سیگنال مختلط x(t) = e^{j4πt} را در نظر بگیرید.  
۱. آن را در دامنه زمان رسم کنید.  
۲. طیف فوریه آن را محاسبه کرده و نمودار دامنه و فاز را رسم کنید.


```python
t = np.linspace(0, 2, 1024)
x = np.exp(1j * 4 * np.pi * t)

dt = t[1] - t[0]
f = fftshift(fftfreq(len(t), dt))
X = fftshift(fft(x)) * dt

plt.plot(t, np.real(x), label="Real")
plt.plot(t, np.imag(x), '--', label="Imaginary")
plt.legend()
plt.title("x(t) = e^{j4πt}")
plt.grid(True)
plt.show()

plt.plot(f, np.abs(X), label="Magnitude")
plt.plot(f, np.angle(X), '--', label="Phase")
plt.legend()
plt.title("Fourier Transform (Magnitude & Phase)")
plt.grid(True)
plt.show()
```

سیگنال x(t) = sinc(t) را در نظر بگیرید.  
۱. سیگنال x(t - 2) را نیز محاسبه و رسم کنید.  
۲. تبدیل فوریه هر دو سیگنال را رسم و مقایسه کنید.  
۳. خاصیت جابجایی زمانی را توضیح دهید.


```python
t = np.linspace(-10, 10, 2048)
x = np.sinc(t)
x_shifted = np.sinc(t - 2)

dt = t[1] - t[0]
f = fftshift(fftfreq(len(t), dt))
X = fftshift(fft(x)) * dt
X_shifted = fftshift(fft(x_shifted)) * dt

plt.plot(t, x, label="x(t) = sinc(t)")
plt.plot(t, x_shifted, '--', label="x(t-2) = sinc(t-2)")
plt.legend()
plt.title("Time Domain Signals")
plt.grid(True)
plt.show()

plt.plot(f, np.abs(X), label="X(f)")
plt.plot(f, np.abs(X_shifted), '--', label="X(t-2)(f)")
plt.legend()
plt.title("Frequency Domain Magnitudes")
plt.grid(True)
plt.show()
```

سیگنال x(t) = rect(t) را در نظر بگیرید.  
۱. آن را در دامنه زمان رسم کنید.  
۲. تبدیل فوریه آن را به کمک FFT رسم کرده و دامنه و فاز آن را نمایش دهید.


```python
t = np.linspace(-5, 5, 2048)
x = np.where(np.abs(t) <= 0.5, 1, 0)  # rect(t)

dt = t[1] - t[0]
f = fftshift(fftfreq(len(t), dt))
X = fftshift(fft(x)) * dt

plt.plot(t, x)
plt.title("x(t) = rect(t)")
plt.grid(True)
plt.show()

plt.plot(f, np.abs(X), label="|X(f)|")
plt.plot(f, np.angle(X), '--', label="∠X(f)")
plt.title("Fourier Transform of rect(t)")
plt.legend()
plt.grid(True)
plt.show()
```
