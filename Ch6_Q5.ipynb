سیگنال x(t) = t * e^(−|t|) را در نظر بگیرید.  
۱. نمودار سیگنال را رسم کنید.  
۲. بررسی کنید که آیا سیگنال زوج، فرد یا هیچ‌کدام نیست.  
۳. مولفه‌های زوج و فرد آن را استخراج و رسم کنید.


```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 1000)
x = t * np.exp(-np.abs(t))

x_even = (x + x[::-1]) / 2
x_odd = (x - x[::-1]) / 2

plt.plot(t, x, label="x(t)")
plt.plot(t, x_even, '--', label="Even part")
plt.plot(t, x_odd, '--', label="Odd part")
plt.title("Even and Odd Decomposition")
plt.legend()
plt.grid(True)
plt.show()
```

تابع پاسخ سیستم h(t) = e^(−t) * u(t) را در نظر بگیرید.  
۱. نمودار h(t) را رسم کنید.  
۲. آیا سیستم علی است؟ چرا؟  
۳. با توجه به نمودار و تعریف تابع واحد، توضیح دهید چرا این سیستم حافظه‌دار است یا خیر.


```python
t = np.linspace(-2, 5, 1000)
u = np.where(t >= 0, 1, 0)
h = np.exp(-t) * u

plt.plot(t, h, label="h(t) = e^(−t)·u(t)")
plt.title("Impulse Response of System")
plt.xlabel("t")
plt.grid(True)
plt.legend()
plt.show()
```

یک سیستم با پاسخ ضربه h(t) = sin(t)/t را در نظر بگیرید.  
۱. نمودار h(t) را رسم کنید.  
۲. بررسی کنید که آیا سیستم پایدار است یا خیر با استفاده از شرط انتگرال مطلق.


```python
t = np.linspace(-50, 50, 10000)
h = np.sinc(t / np.pi)

integral = np.trapezoid(np.abs(h), t)

plt.plot(t, h)
plt.title("Impulse Response h(t) = sin(t)/t")
plt.grid(True)
plt.xlabel("t")
plt.show()

print(f"Integral of |h(t)| = {integral:.4f}")
```

سیستم زیر را بررسی کنید:  
y(t) = t·x(t)  
۱. بررسی کنید آیا سیستم زمان‌ناور است یا نه، با استفاده از تغییر مکان ورودی.  
۲. مثالی عددی با x(t) = cos(2πt) ارائه دهید و پاسخ سیستم را برای x(t) و x(t−1) رسم کنید.


```python
t = np.linspace(0, 5, 1000)
x = np.cos(2 * np.pi * t)
x_shifted = np.cos(2 * np.pi * (t - 1))

y = t * x
y_shifted = t * x_shifted

plt.plot(t, y, label="y(t) = t·x(t)")
plt.plot(t, y_shifted, label="y_shifted = t·x(t−1)")
plt.title("Time Variance Check")
plt.legend()
plt.grid(True)
plt.show()
```
