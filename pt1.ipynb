# فصل ۷: تکنیک‌های طراحی فیلتر

## ۷.۰ مقدمه  
فیلترها یکی از کلاس‌های بسیار مهم در سیستم‌های LTI (خطی و تغییرناپذیر با زمان) هستند.  
به‌طور دقیق، اصطلاح **فیلتر انتخابگر فرکانس** به سیستمی اشاره دارد که برخی از مؤلفه‌های فرکانسی یک سیگنال ورودی را عبور می‌دهد و بقیه را به‌طور کامل حذف می‌کند.  
اما در مفهوم گسترده‌تر، هر سیستمی که برخی فرکانس‌ها را نسبت به بقیه اصلاح کند نیز یک **فیلتر** نامیده می‌شود.  

در این فصل، تمرکز اصلی ما روی **طراحی فیلترهای انتخابگر فرکانس** است، اما برخی از تکنیک‌های مطرح‌شده، کاربردهای گسترده‌تری دارند.  
ما بر طراحی **فیلترهای علی** تمرکز می‌کنیم، اگرچه در بسیاری از موارد، نیازی به این محدودیت نیست.  
در عمل، اغلب **فیلترهای غیرعلی** طراحی می‌شوند و از طریق **اصلاح نسخه‌های علی** پیاده‌سازی می‌گردند.  

---

### **طراحی فیلترهای گسسته‌زمان**
طراحی فیلترهای **گسسته‌زمان** شامل تعیین **پارامترهای تابع تبدیل** یا **معادله تفاضلی** است که یک تقریب مناسب از **پاسخ ضربه‌ای مطلوب** یا **پاسخ فرکانسی مطلوب** را در محدوده تحمل مشخص ایجاد کند.  
همان‌طور که در **فصل ۲** بررسی شد، سیستم‌های گسسته‌زمان که با **معادلات تفاضلی** پیاده‌سازی می‌شوند، به دو دسته کلی تقسیم می‌شوند:

- **فیلترهای IIR** (پاسخ ضربه‌ای نامحدود)
- **فیلترهای FIR** (پاسخ ضربه‌ای محدود)  

طراحی **فیلترهای IIR** به‌معنی یافتن **یک تابع تبدیل که یک تابع گویا از \( z \) باشد**،  
در حالی که طراحی **فیلترهای FIR** بر **تقریب چندجمله‌ای** استوار است.  
روش‌های طراحی این دو نوع فیلتر **متفاوت و متمایز** هستند.  

در روزهای ابتدایی استفاده از فیلترهای گسسته‌زمان، طراحی آن‌ها مبتنی بر **تبدیل فیلترهای آنالوگ به دیجیتال** بود.  
روش‌هایی مانند **تبدیل نامتغیر ضربه‌ای** و **تبدیل دوجهته (Bilinear Transform)** برای این کار استفاده می‌شدند.  
این تکنیک‌ها که در **بخش‌های ۷.۲.۱ و ۷.۲.۲** مورد بررسی قرار می‌گیرند، معمولاً منجر به **فیلترهای IIR** می‌شدند.  
امروزه نیز این روش‌ها **هسته اصلی طراحی فیلترهای IIR گسسته‌زمان** هستند.  

در مقابل، طراحی **فیلترهای FIR** ابتدا در سیستم‌های عملی مورد توجه قرار گرفت،  
زیرا هیچ **معادل پیوسته‌ای** برای این فیلترها وجود نداشت که بتوان به دنیای دیجیتال منتقل کرد.  
روش‌های طراحی **فیلترهای FIR** تنها بعد از اینکه در کاربردهای واقعی مورد نیاز شدند، توسعه یافتند.  

---

### **مراحل کلی طراحی فیلتر**
فرایند طراحی فیلتر شامل مراحل زیر است:

1. **تعیین مشخصات مطلوب سیستم** (بسته به کاربرد موردنظر)  
2. **تقریب مشخصات با استفاده از یک سیستم گسسته‌زمان علی**  
3. **پیاده‌سازی فیلتر با استفاده از سخت‌افزار یا نرم‌افزار مناسب**  

هر سه مرحله **به‌شدت به یکدیگر وابسته‌اند**، اما در این فصل تمرکز اصلی روی **مرحله دوم** است:  
زیرا **مشخصات مطلوب فیلتر (مرحله اول)** وابسته به نوع کاربرد است، و  
**نحوه پیاده‌سازی فیلتر (مرحله سوم)** بستگی به **فناوری سخت‌افزاری یا نرم‌افزاری مورد استفاده دارد**.  

در سیستم‌های عملی، فیلترهای دیجیتال معمولاً با **سخت‌افزار دیجیتال** پیاده‌سازی شده و برای پردازش **سیگنال‌های نمونه‌برداری‌شده از یک سیگنال پیوسته** مورد استفاده قرار می‌گیرند.  
به همین دلیل، در بسیاری از منابع به **فیلترهای گسسته‌زمان، فیلتر دیجیتال نیز گفته می‌شود**،  
هرچند که روش‌های طراحی آن‌ها عمدتاً وابسته به **ماهیت گسسته‌زمان سیگنال‌ها و سیستم‌ها** است.  

**اثرات کوانتیزاسیون ضرایب فیلتر و سیگنال‌ها**، که به‌طور ذاتی در پردازش دیجیتال رخ می‌دهد،  
موضوعی است که قبلاً در **فصل ۶** بررسی شده و در این فصل پوشش داده نمی‌شود.  

---

### **جمع‌بندی**
در این فصل، ما **روش‌های مختلفی** برای طراحی **فیلترهای IIR و FIR** را بررسی خواهیم کرد.  
در هر کاربرد خاص، **انتخاب بین این دو نوع فیلتر وابسته به معیارهای متعددی** است، از جمله:  
- **پیچیدگی محاسباتی**  
- **پایداری و پاسخ فرکانسی**  
- **دقت تقریب و میزان تأخیر فاز**  

 هدف ما این است که **مهم‌ترین تکنیک‌های طراحی فیلتر** را معرفی کرده و مزایا و معایب هر یک را بررسی کنیم.  
 **مسائل و پروژه‌های موجود در وب‌سایت کتاب** به شما امکان می‌دهد تا ویژگی‌های این فیلترها را **بیشتر و عمیق‌تر بررسی کنید**.
